## 1_Introduccion_a_CSS

### 1_1_Que_es_CSS

CSS significa **Cascading Style Sheets** (Hojas de Estilo en Cascada). Es un lenguaje de hojas de estilo utilizado para describir la presentación y el aspecto visual de un documento escrito en un lenguaje de marcado como HTML.

Mientras que HTML se encarga de la **estructura** y el **contenido**, CSS se encarga de la **presentación**: colores, fuentes, espaciado, posicionamiento, diseño, animaciones y la adaptación a diferentes tamaños de pantalla (responsive design).

Separar el contenido (HTML) de la presentación (CSS) ofrece enormes ventajas:
*   **Mantenibilidad:** Puedes cambiar el aspecto de todo un sitio web modificando un solo archivo CSS.
*   **Accesibilidad:** Permite que el contenido sea accesible en diferentes dispositivos y para usuarios con necesidades especiales.
*   **Rendimiento:** Los navegadores pueden cachear los archivos CSS, haciendo que las páginas carguen más rápido en visitas posteriores.
*   **Consistencia:** Asegura una apariencia uniforme en todo el sitio web.

### 1_2_Sintaxis_Basica

Una regla CSS consta de dos partes principales:

1.  **Selector:** Apunta al elemento o elementos HTML a los que quieres aplicar el estilo.
2.  **Bloque de Declaraciones:** Contiene una o más declaraciones, encerradas entre llaves `{}`. Cada declaración incluye:
    *   **Propiedad:** El atributo de estilo que quieres cambiar (ej: `color`, `font-size`, `background-color`).
    *   **Valor:** El valor que quieres asignar a esa propiedad (ej: `blue`, `16px`, `#FFFFFF`).
    *   Las declaraciones se separan por punto y coma `;`.

```css
/* Esto es un comentario CSS */

selector {
  propiedad: valor;
  otra-propiedad: otro-valor;
}

/* Ejemplo concreto */
h1 {
  color: navy; /* Cambia el color del texto de los h1 a azul marino */
  text-align: center; /* Centra el texto */
}

p {
  font-size: 16px;
  line-height: 1.5; /* Espaciado entre líneas */
}
```

### 1_3_Como_Vincular_CSS_con_HTML

Hay tres formas de aplicar CSS a un documento HTML, pero **la forma externa es la más recomendada y utilizada**:

1.  **CSS Externo (Recomendado):**
    *   Escribes tus reglas CSS en un archivo separado con extensión `.css` (ej: `estilos.css`).
    *   Enlazas este archivo desde la sección `<head>` de tu documento HTML usando la etiqueta `<link>`.
    ```html
    <!DOCTYPE html>
    <html>
    <head>
      <title>Mi Página con CSS Externo</title>
      <link rel="stylesheet" href="estilos.css">
    </head>
    <body>
      <h1>Título con Estilo</h1>
      <p>Párrafo con estilo.</p>
    </body>
    </html>
    ```
    *   **Ventajas:** Mantiene el HTML limpio, facilita la reutilización y el mantenimiento, mejora el rendimiento (cacheo).

2.  **CSS Interno:**
    *   Escribes las reglas CSS directamente dentro de una etiqueta `<style>` en la sección `<head>` del documento HTML.
    ```html
    <!DOCTYPE html>
    <html>
    <head>
      <title>Mi Página con CSS Interno</title>
      <style>
        body {
          background-color: lightblue;
        }
        h1 {
          color: white;
          text-align: center;
        }
      </style>
    </head>
    <body>
      <h1>Título con Estilo</h1>
    </body>
    </html>
    ```
    *   **Desventajas:** Los estilos solo aplican a esa página, dificulta la reutilización, mezcla estructura y presentación. Puede ser útil para estilos muy específicos de una sola página o para pruebas rápidas.

3.  **CSS en Línea (Inline):**
    *   Aplicas estilos directamente a un elemento HTML específico usando el atributo `style`.
    ```html
    <!DOCTYPE html>
    <html>
    <body>
      <h1 style="color: blue; text-align: right;">Título con Estilo en Línea</h1>
      <p style="color: green;">Párrafo con estilo en línea.</p>
    </body>
    </html>
    ```
    *   **Desventajas:** Máxima mezcla de estructura y presentación, muy difícil de mantener, anula la especificidad de otras reglas (lo veremos más adelante), no reutilizable. **Evitar en la medida de lo posible**, excepto para casos muy puntuales o manipulación con JavaScript.

## 2_Selectores_CSS

Los selectores son patrones que identifican a qué elementos HTML se aplicarán las reglas CSS. Dominarlos es fundamental.

### 2_1_Selectores_Basicos_Tipo_Clase_ID

*   **Selector de Tipo (Elemento):** Selecciona todos los elementos de un tipo específico.
    ```css
    p { /* Selecciona todos los elementos <p> */
      color: gray;
    }
    h2 { /* Selecciona todos los elementos <h2> */
      font-weight: bold;
    }
    ```
*   **Selector de Clase (`.`):** Selecciona todos los elementos que tienen un atributo `class` específico. Un elemento puede tener múltiples clases (separadas por espacios en el HTML). Es la forma más común y flexible de agrupar elementos para aplicar estilos.
    ```html
    <p class="destacado importante">Texto destacado e importante.</p>
    <span class="destacado">Otro texto destacado.</span>
    ```
    ```css
    .destacado { /* Selecciona elementos con class="destacado" */
      background-color: yellow;
    }
    .importante { /* Selecciona elementos con class="importante" */
      font-weight: bold;
    }
    p.destacado { /* Selecciona solo los <p> que tengan la clase "destacado" */
       border: 1px solid red;
    }
    ```
*   **Selector de ID (`#`):** Selecciona un **único** elemento que tiene un atributo `id` específico. El valor del `id` **debe ser único** en toda la página HTML. Se usa para identificar elementos específicos, a menudo para enlaces de ancla o manipulación con JavaScript. Por su alta especificidad (ver sección 3), se recomienda usar clases para estilado general y reservar los IDs para identificadores únicos.
    ```html
    <div id="menu-principal">...</div>
    ```
    ```css
    #menu-principal { /* Selecciona el elemento con id="menu-principal" */
      background-color: #eee;
      padding: 10px;
    }
    ```
*   **Selector Universal (`*`):** Selecciona **todos** los elementos de la página. Úsalo con precaución, ya que puede afectar al rendimiento y tener efectos no deseados. A menudo se usa en reseteos de estilos.
    ```css
    * { /* Selecciona todos los elementos */
      margin: 0;
      padding: 0;
      box-sizing: border-box; /* Práctica común de reseteo */
    }
    ```
*   **Agrupación de Selectores (`,`):** Permite aplicar las mismas reglas a múltiples selectores, separándolos por comas.
    ```css
    h1, h2, h3 { /* Aplica a h1, h2 y h3 */
      font-family: 'Arial', sans-serif;
      color: darkblue;
    }
    ```

### 2_2_Selectores_de_Atributo

Seleccionan elementos basados en la presencia o el valor de sus atributos HTML.

*   `[atributo]` - Selecciona elementos que tienen el atributo especificado.
    ```css
    a[target] { /* Selecciona enlaces <a> que tengan el atributo target */
      color: green;
    }
    ```
*   `[atributo=valor]` - Selecciona elementos que tienen el atributo con un valor exacto.
    ```css
    input[type="text"] { /* Selecciona inputs de tipo texto */
      border: 1px solid #ccc;
    }
    a[target="_blank"] { /* Selecciona enlaces que abren en nueva pestaña */
       background-image: url('external-link.png');
       background-repeat: no-repeat;
       background-position: right center;
       padding-right: 15px;
    }
    ```
*   `[atributo~=valor]` - Selecciona elementos cuyo atributo contenga una lista de palabras separadas por espacios, y una de ellas sea exactamente `valor`.
    ```html
    <p class="foo bar baz">...</p>
    ```
    ```css
    p[class~="bar"] { /* Selecciona el párrafo anterior */
        font-style: italic;
    }
    ```
*   `[atributo|=valor]` - Selecciona elementos cuyo atributo tenga un valor exacto o que comience por `valor` seguido inmediatamente por un guion (`-`). Útil para códigos de idioma.
    ```html
    <p lang="en-us">Hello</p>
    <p lang="en-gb">Hello</p>
    <p lang="es">Hola</p>
    ```
    ```css
    p[lang|="en"] { /* Selecciona los dos primeros párrafos */
        color: blue;
    }
    ```
*   `[atributo^=valor]` - Selecciona elementos cuyo atributo comience con `valor`.
    ```css
    a[href^="https://"] { /* Selecciona enlaces que empiezan con https:// */
        font-weight: bold;
    }
    ```
*   `[atributo$=valor]` - Selecciona elementos cuyo atributo termine con `valor`.
    ```css
    a[href$=".pdf"] { /* Selecciona enlaces a archivos PDF */
        padding-left: 20px;
        background: url('pdf-icon.png') no-repeat left center;
    }
    ```
*   `[atributo*=valor]` - Selecciona elementos cuyo atributo contenga `valor` en cualquier parte.
    ```css
    a[href*="google"] { /* Selecciona enlaces cuya URL contenga "google" */
        text-decoration: underline wavy red;
    }
    ```

### 2_3_Selectores_Combinadores

Combinan selectores simples para expresar relaciones entre elementos.

*   **Descendiente (` ` espacio):** Selecciona elementos que son descendientes (hijos, nietos, etc.) de otro elemento.
    ```css
    div p { /* Selecciona todos los <p> que están dentro de un <div> */
      margin-bottom: 10px;
    }
    ```
*   **Hijo Directo (`>`):** Selecciona elementos que son hijos directos (primer nivel) de otro elemento.
    ```css
    ul > li { /* Selecciona solo los <li> que son hijos directos de un <ul> */
      list-style-type: square;
    }
    ```
*   **Hermano Adyacente (`+`):** Selecciona un elemento que es inmediatamente precedido por otro elemento (ambos deben compartir el mismo padre).
    ```css
    h2 + p { /* Selecciona el primer <p> que sigue inmediatamente a un <h2> */
      margin-top: 0;
      font-style: italic;
    }
    ```
*   **Hermano General (`~`):** Selecciona todos los elementos que son hermanos posteriores a otro elemento (todos deben compartir el mismo padre).
    ```css
    h2 ~ p { /* Selecciona todos los <p> que son hermanos posteriores a un <h2> */
      color: #555;
    }
    ```

### 2_4_Pseudo_Clases

Añaden un estado especial a un selector. Se indican con dos puntos (`:`). Seleccionan elementos basados en información que no está explícitamente en el árbol del documento (como el estado de un enlace o la posición de un elemento).

*   **Estado de enlaces:**
    *   `:link` - Enlaces no visitados.
    *   `:visited` - Enlaces ya visitados.
    *   `:hover` - Elemento cuando el cursor del ratón está sobre él.
    *   `:active` - Elemento mientras está siendo activado (ej: clic en un enlace).
    *   `:focus` - Elemento que tiene el foco (ej: un campo de formulario seleccionado).
    ```css
    a:link { color: blue; }
    a:visited { color: purple; }
    a:hover { text-decoration: none; color: red; }
    a:active { color: yellow; }
    input:focus { border-color: orange; outline: none; }
    ```
    *(Orden recomendado para enlaces: L-V-H-A - LoVe HAte)*

*   **Pseudo-clases estructurales:**
    *   `:first-child` - Primer elemento hijo de su padre.
    *   `:last-child` - Último elemento hijo de su padre.
    *   `:nth-child(n)` - Elemento hijo n-ésimo. `n` puede ser un número (`2`), `odd` (impar), `even` (par), o una fórmula (`2n+1`).
    *   `:nth-last-child(n)` - Igual que `nth-child` pero contando desde el final.
    *   `:first-of-type` - Primer elemento de su tipo entre los hijos de su padre.
    *   `:last-of-type` - Último elemento de su tipo entre los hijos de su padre.
    *   `:nth-of-type(n)` - Elemento n-ésimo de su tipo.
    *   `:nth-last-of-type(n)` - Igual que `nth-of-type` pero contando desde el final.
    *   `:only-child` - Elemento que es el único hijo de su padre.
    *   `:only-of-type` - Elemento que es el único de su tipo entre los hijos de su padre.
    *   `:empty` - Elemento que no tiene hijos (ni siquiera texto).
    ```css
    li:first-child { font-weight: bold; }
    tr:nth-child(even) { background-color: #f2f2f2; } /* Filas pares de una tabla */
    p:nth-of-type(3) { color: green; } /* Tercer párrafo dentro de su contenedor */
    div:empty { border: 1px dashed red; padding: 5px; }
    ```

*   **Otras pseudo-clases útiles:**
    *   `:not(selector)` - Selecciona elementos que NO coinciden con el `selector` interno.
    *   `:checked` - Para `<input>` de tipo `checkbox` o `radio` que están marcados.
    *   `:disabled` - Elementos de formulario deshabilitados.
    *   `:enabled` - Elementos de formulario habilitados.
    *   `:required` - Elementos de formulario con el atributo `required`.
    *   `:optional` - Elementos de formulario sin el atributo `required`.
    ```css
    input:not([type="submit"]) { padding: 5px; }
    input[type="checkbox"]:checked + label { text-decoration: line-through; }
    button:disabled { opacity: 0.5; cursor: not-allowed; }
    ```

### 2_5_Pseudo_Elementos

Permiten dar estilo a partes específicas de un elemento. Se indican con doble dos puntos (`::`) en CSS3 (aunque la sintaxis de un solo `:` es aceptada por compatibilidad para los más antiguos).

*   `::before` - Inserta contenido cosmético *antes* del contenido real del elemento. Requiere la propiedad `content`.
*   `::after` - Inserta contenido cosmético *después* del contenido real del elemento. Requiere la propiedad `content`.
    ```css
    a.external::after {
      content: " ↗"; /* Añade una flecha después de enlaces externos */
      font-size: 0.8em;
    }
    blockquote::before {
      content: "\“"; /* Comilla de apertura */
      font-size: 2em;
      color: gray;
      margin-right: 5px;
    }
    blockquote::after {
        content: "\”"; /* Comilla de cierre */
        font-size: 2em;
        color: gray;
        margin-left: 5px;
    }
    ```
*   `::first-letter` - Selecciona la primera letra del contenido de texto de un elemento de bloque.
    ```css
    p::first-letter {
      font-size: 2em;
      font-weight: bold;
      color: darkred;
      float: left; /* A menudo se usa con float */
      margin-right: 0.1em;
    }
    ```
*   `::first-line` - Selecciona la primera línea del contenido de texto de un elemento de bloque.
    ```css
    p::first-line {
      font-variant: small-caps;
      letter-spacing: 1px;
    }
    ```
*   `::selection` - Selecciona la parte del documento que ha sido resaltada (seleccionada) por el usuario.
    ```css
    ::selection {
      background-color: lightblue;
      color: black;
    }
    ```
*   `::placeholder` - Selecciona el texto del atributo `placeholder` en elementos `<input>` o `<textarea>`.
    ```css
    input::placeholder {
        color: #aaa;
        font-style: italic;
    }
    ```

## 3_Cascada_Especificidad_e_Herencia

Estos tres conceptos determinan qué reglas CSS se aplican finalmente a un elemento cuando hay múltiples reglas que podrían afectarlo.

### 3_1_La_Cascada

El navegador aplica las reglas CSS siguiendo un orden o cascada:

1.  **Origen:** Las hojas de estilo provienen de diferentes fuentes, y se aplican en este orden (las últimas anulan a las anteriores si hay conflicto):
    *   **User-Agent Stylesheet:** Estilos por defecto del navegador.
    *   **User Stylesheet:** Estilos personalizados por el usuario (menos común).
    *   **Author Stylesheet:** Los estilos definidos por el desarrollador web (tus archivos `.css`, `<style>`, `style=""`).
    *   **Author `!important`:** Reglas del desarrollador marcadas con `!important`.
    *   **User `!important`:** Reglas del usuario marcadas con `!important`.
    *   **User-Agent `!important`:** Reglas del navegador marcadas con `!important` (muy raro).

2.  **Especificidad:** Si varias reglas del mismo origen podrían aplicarse a un elemento, se usa la especificidad para decidir cuál gana (ver 3.2).
3.  **Orden en el Código:** Si varias reglas tienen la misma especificidad, la regla que aparece **más tarde** en el código CSS (o en el HTML si son internos/en línea) es la que se aplica. Los estilos en línea siempre ganan sobre los externos/internos del mismo nivel de `!important` debido a su especificidad y posición.

### 3_2_Especificidad

Es un "peso" o puntuación que el navegador asigna a cada selector CSS para determinar cuál es más específico. El selector con mayor especificidad gana si hay un conflicto.

La especificidad se calcula generalmente como un número de tres o cuatro componentes (simplificado aquí a tres categorías):

1.  **IDs (`#id`):** Cada selector de ID suma puntos a la categoría más alta.
2.  **Clases (`.class`), Pseudo-clases (`:hover`) y Selectores de Atributo (`[type="text"]`):** Cada uno suma puntos a la categoría intermedia.
3.  **Tipos (`p`, `div`) y Pseudo-elementos (`::before`):** Cada uno suma puntos a la categoría más baja.

*   El selector universal (`*`) y los combinadores (`+`, `>`, `~`, ` `) no añaden especificidad por sí mismos.
*   La pseudo-clase `:not()` no añade especificidad, pero sí los selectores que contiene.

**Comparación:**
*   Un selector de ID (`#miId`) siempre será más específico que cualquier número de clases o tipos.
*   Un selector de clase (`.miClase`) es más específico que un selector de tipo (`div`).
*   `div p.destacado` (0 IDs, 1 Clase, 2 Tipos) es más específico que `p.destacado` (0 IDs, 1 Clase, 1 Tipo).
*   `#contenido p` (1 ID, 0 Clases, 1 Tipo) es más específico que `div.principal p.info` (0 IDs, 2 Clases, 2 Tipos).

**Estilos en línea (`style="..."`):** Tienen la máxima especificidad, superando a cualquier combinación de IDs, clases y tipos definidos en hojas de estilo externas o internas (excepto si se usa `!important`).

**`!important`:** (Ver 3.4) Anula cualquier otra regla de especificidad u orden, excepto otras reglas `!important` de mayor origen.

**Consejo:** Intenta mantener la especificidad lo más baja posible. Usa clases para la mayoría de los estilos y evita selectores demasiado complejos o el uso excesivo de IDs para estilado, ya que dificultan la anulación de estilos más adelante.

### 3_3_Herencia

Algunas propiedades CSS aplicadas a un elemento padre son **heredadas** automáticamente por sus elementos hijos. Esto evita tener que declarar la misma propiedad repetidamente.

*   **Propiedades que suelen heredarse:** `color`, `font-family`, `font-size`, `font-weight`, `line-height`, `text-align`, `list-style`.
*   **Propiedades que NO suelen heredarse:** `background-color`, `border`, `padding`, `margin`, `width`, `height`, `position`, `display`.

Ejemplo:
```html
<div style="color: blue; border: 1px solid black;">
  <p>Este párrafo heredará el color azul, pero no el borde.</p>
  <ul>
    <li>Este item de lista también heredará el color azul.</li>
  </ul>
</div>
```

Puedes forzar la herencia de una propiedad que normalmente no se hereda usando el valor `inherit`:
```css
.mi-elemento-hijo {
  border: inherit; /* Hereda el borde del padre */
}
```

También puedes evitar la herencia o restablecer una propiedad a su valor inicial del navegador con `initial` o `unset`.

### 3_4_La_Propiedad_important

La declaración `!important` se añade al final de una declaración CSS (antes del punto y coma) para darle la máxima prioridad, anulando cualquier otra regla de la cascada (especificidad y orden) excepto otras reglas `!important` de mayor origen (usuario > autor).

```css
p {
  color: blue !important; /* Este color azul será muy difícil de anular */
  font-size: 14px;
}

#mi-parrafo {
  color: red; /* Esta regla NO anulará el color azul del párrafo anterior debido al !important */
}
```

**¡Usa `!important` con extrema precaución!**
*   Hace que depurar CSS sea mucho más difícil.
*   Rompe la cascada natural y la especificidad.
*   A menudo indica un problema subyacente en la estructura CSS o en la especificidad de los selectores.
*   Su uso legítimo es muy limitado, quizás para anular estilos de terceros difíciles de modificar o en hojas de estilo de usuario para accesibilidad.

En general, si necesitas anular un estilo, intenta hacerlo aumentando la especificidad de tu selector de forma controlada o reordenando las reglas CSS, en lugar de recurrir a `!important`.

## 4_Colores_Fondos_y_Bordes

Estas propiedades definen aspectos visuales básicos de los elementos.

### 4_1_Formatos_de_Color_Nombre_HEX_RGB_HSL

CSS ofrece varias formas de especificar colores:

*   **Nombres de Color:** Palabras clave predefinidas (ej: `red`, `blue`, `green`, `lightgray`, `transparent`). Hay unos 140 nombres estándar. Son fáciles de leer pero limitados.
    ```css
    p { color: steelblue; }
    ```
*   **Valores Hexadecimales (HEX):** `#RRGGBB` o `#RGB` (forma corta si los pares son iguales). RR, GG y BB son valores hexadecimales (00-FF) para Rojo, Verde y Azul. Es el formato más común en la web.
    ```css
    h1 { color: #FF0000; } /* Rojo puro */
    body { background-color: #333; } /* Gris oscuro (equivalente a #333333) */
    .sutil { color: #f0f0f0; } /* Gris muy claro */
    ```
    También se pueden añadir dos dígitos hexadecimales al final para la transparencia (canal alfa): `#RRGGBBAA`. `00` es totalmente transparente, `FF` es totalmente opaco.
    ```css
    .overlay { background-color: #00000080; } /* Negro semi-transparente (alfa 50%) */
    ```
*   **RGB (Red, Green, Blue):** `rgb(rojo, verde, azul)`. Cada valor es un número entre 0 y 255.
    ```css
    p { color: rgb(50, 150, 200); }
    ```
*   **RGBA (RGB + Alpha):** `rgba(rojo, verde, azul, alfa)`. Alfa es un número entre 0 (transparente) y 1 (opaco), pudiendo ser decimal (ej: `0.5` para 50% de opacidad). Muy útil para transparencias.
    ```css
    .tooltip { background-color: rgba(0, 0, 0, 0.75); color: white; } /* Fondo negro translúcido */
    ```
*   **HSL (Hue, Saturation, Lightness):** `hsl(tono, saturacion%, luminosidad%)`.
    *   `tono`: Ángulo en la rueda de color (0-360). 0 es rojo, 120 es verde, 240 es azul.
    *   `saturacion`: Porcentaje (0% gris, 100% color puro).
    *   `luminosidad`: Porcentaje (0% negro, 50% color normal, 100% blanco).
    Es intuitivo para ajustar variaciones de un color (ej: hacer un color más claro/oscuro o más/menos saturado).
    ```css
    button { background-color: hsl(210, 80%, 50%); } /* Un azul vibrante */
    button:hover { background-color: hsl(210, 80%, 60%); } /* El mismo azul, pero más claro */
    ```
*   **HSLA (HSL + Alpha):** `hsla(tono, saturacion%, luminosidad%, alfa)`. Alfa funciona igual que en RGBA (0 a 1).
    ```css
    .highlight { background-color: hsla(60, 100%, 50%, 0.3); } /* Amarillo (tono 60) semi-transparente */
    ```

### 4_2_Propiedad_color

Establece el color del **contenido de texto** de un elemento. Acepta cualquiera de los formatos de color vistos.

```css
body {
  color: #333; /* Color de texto por defecto para toda la página */
}

h1 {
  color: darkred;
}

a {
  color: rgb(0, 100, 200);
}

.error {
  color: red;
}
```

### 4_3_Propiedades_background

Controlan el fondo de un elemento. Hay varias propiedades relacionadas, que se pueden agrupar en la propiedad abreviada `background`.

*   `background-color`: Establece el color de fondo.
    ```css
    div { background-color: lightcyan; }
    ```
*   `background-image`: Establece una o más imágenes de fondo. Usa `url('ruta/imagen.jpg')`. Puedes usar gradientes también (`linear-gradient()`, `radial-gradient()`).
    ```css
    body { background-image: url('fondo_textura.png'); }
    .gradiente { background-image: linear-gradient(to right, blue, green); }
    ```
*   `background-repeat`: Controla si la imagen de fondo se repite. Valores comunes: `repeat` (defecto), `no-repeat`, `repeat-x` (solo horizontal), `repeat-y` (solo vertical).
    ```css
    body { background-repeat: no-repeat; }
    ```
*   `background-position`: Define la posición inicial de la imagen de fondo. Usa palabras clave (`top`, `center`, `bottom`, `left`, `right`) o valores numéricos/porcentajes (`50% 50%` para centro, `0 0` para arriba-izquierda).
    ```css
    div {
      background-image: url('logo.png');
      background-repeat: no-repeat;
      background-position: top right; /* Posiciona arriba a la derecha */
    }
    ```
*   `background-size`: Define el tamaño de la imagen de fondo. Valores: `auto` (defecto), `cover` (cubre todo el área, puede recortar), `contain` (escala la imagen para que quepa entera), valores numéricos/porcentajes (`100px auto`, `50%`).
    ```css
    body {
      background-image: url('imagen_grande.jpg');
      background-size: cover; /* Cubre toda la ventana */
      background-attachment: fixed; /* Fija la imagen al hacer scroll */
    }
    ```
*   `background-attachment`: Define si la imagen de fondo se desplaza con el contenido (`scroll` - defecto) o permanece fija respecto a la ventana (`fixed`).
*   **Propiedad Abreviada `background`:** Permite establecer varias propiedades de fondo en una sola declaración. El orden suele ser: `color image repeat attachment position / size`. No es necesario incluir todos los valores.
    ```css
    body {
      background: lightblue url('fondo.png') no-repeat fixed center / cover;
    }
    .simple {
      background: #eee url('icono.gif') no-repeat top left;
    }
    ```

### 4_4_Propiedades_border

Definen los bordes alrededor de un elemento (dentro del box model, entre padding y margin).

*   `border-width`: Grosor del borde (ej: `1px`, `medium`, `5px`).
*   `border-style`: Estilo de la línea (ej: `solid`, `dashed`, `dotted`, `double`, `groove`, `ridge`, `inset`, `outset`, `none`). **Es necesario** definir un `border-style` para que el borde sea visible.
*   `border-color`: Color del borde.

Estas propiedades se pueden definir para cada lado individualmente:
*   `border-top-width`, `border-top-style`, `border-top-color`
*   `border-right-width`, `border-right-style`, `border-right-color`
*   `border-bottom-width`, `border-bottom-style`, `border-bottom-color`
*   `border-left-width`, `border-left-style`, `border-left-color`

**Propiedad Abreviada `border`:** Establece `width`, `style` y `color` para los cuatro lados a la vez.
```css
p {
  border: 1px solid black; /* Borde simple negro */
}
div.importante {
  border: 5px dashed red; /* Borde rojo discontinuo grueso */
}
```

**Propiedades Abreviadas por Lado:**
```css
h2 {
  border-bottom: 2px solid blue; /* Solo un borde inferior azul */
  border-left: 5px dotted green; /* Borde izquierdo verde punteado */
}
```

**Abreviaturas para múltiples lados:** Puedes definir los valores para `border-width`, `border-color` (y `margin`, `padding`) usando 1, 2, 3 o 4 valores:
*   `1 valor`: Aplica a los 4 lados. `border-width: 2px;`
*   `2 valores`: El primero para `top/bottom`, el segundo para `left/right`. `border-width: 1px 3px;`
*   `3 valores`: El primero para `top`, el segundo para `left/right`, el tercero para `bottom`. `border-width: 1px 2px 3px;`
*   `4 valores`: `top`, `right`, `bottom`, `left` (sentido horario). `border-width: 1px 2px 3px 4px;`

### 4_5_border_radius

Permite redondear las esquinas de los bordes de un elemento.

*   Un solo valor: Aplica el mismo radio a las 4 esquinas.
    ```css
    .boton {
      border: 1px solid gray;
      padding: 10px 20px;
      border-radius: 5px; /* Esquinas ligeramente redondeadas */
    }
    .circulo {
      width: 100px;
      height: 100px;
      background-color: red;
      border-radius: 50%; /* Crea un círculo perfecto si width=height */
    }
    ```
*   Múltiples valores: Puedes especificar radios diferentes para cada esquina (sentido horario: top-left, top-right, bottom-right, bottom-left).
    ```css
    .pestana {
      border-radius: 10px 10px 0 0; /* Redondea solo las esquinas superiores */
    }
    ```
    Incluso puedes definir radios elípticos (horizontal / vertical) para cada esquina, pero es más avanzado.

## 5_Tipografia_Texto

CSS ofrece un control detallado sobre cómo se muestra el texto.

### 5_1_Propiedad_font_family

Especifica la fuente (tipo de letra) para el texto. Puedes proporcionar una lista de fuentes separadas por comas, en orden de preferencia. El navegador usará la primera fuente de la lista que esté disponible en el sistema del usuario.

**Es crucial incluir una fuente genérica al final de la lista** como fallback:
*   `serif`: Fuentes con remates (ej: Times New Roman, Georgia).
*   `sans-serif`: Fuentes sin remates (ej: Arial, Helvetica, Verdana). La más común para texto en pantalla.
*   `monospace`: Fuentes donde todos los caracteres ocupan el mismo ancho (ej: Courier New, Consolas). Ideal para código.
*   `cursive`: Fuentes que imitan la escritura a mano (ej: Brush Script MT).
*   `fantasy`: Fuentes decorativas (ej: Impact).

```css
body {
  /* Intenta con 'Roboto', si no 'Helvetica Neue', si no 'Arial', si no cualquier sans-serif disponible */
  font-family: 'Roboto', "Helvetica Neue", Arial, sans-serif;
}

h1, h2 {
  font-family: Georgia, Times, serif; /* Usa una fuente serif para encabezados */
}

pre, code {
  font-family: Consolas, "Courier New", monospace; /* Fuente monoespaciada para código */
}
```
*Nota: Si el nombre de la fuente contiene espacios, debe ir entre comillas (simples o dobles).*

### 5_2_font_size_font_weight_font_style

*   `font-size`: Define el tamaño del texto. Se puede especificar en varias unidades (ver sección 7):
    *   Píxeles (`px`): Tamaño fijo. `font-size: 16px;`
    *   `em`: Relativo al tamaño de fuente del elemento padre. `1.2em` es 120% del tamaño del padre.
    *   `rem` (root em): Relativo al tamaño de fuente del elemento raíz (`<html>`). Muy recomendado para escalabilidad. `font-size: 1rem;` (Si el `html` tiene `16px`, `1rem` = `16px`).
    *   Porcentaje (`%`): Similar a `em`. `font-size: 110%;`
    *   Palabras clave: `small`, `medium`, `large`, `x-large`, etc.
*   `font-weight`: Define el grosor (peso) de la fuente.
    *   Valores numéricos: `100` a `900` (en múltiplos de 100). `400` es normal, `700` es bold.
    *   Palabras clave: `normal` (400), `bold` (700), `lighter`, `bolder` (relativos al padre).
    ```css
    p { font-size: 1rem; }
    strong { font-weight: bold; } /* O font-weight: 700; */
    .ligero { font-weight: 300; }
    ```
    *Nota: La fuente debe soportar los pesos especificados para que se vean correctamente.*
*   `font-style`: Define el estilo de la fuente.
    *   `normal`: Estilo normal.
    *   `italic`: Versión cursiva diseñada de la fuente.
    *   `oblique`: Versión inclinada artificialmente de la fuente normal (usar si no hay `italic` disponible).
    ```css
    em { font-style: italic; }
    .inclinado { font-style: oblique; }
    ```

### 5_3_text_align_text_decoration_line_height

*   `text-align`: Alinea el texto horizontalmente dentro de su contenedor.
    *   `left`: Alineado a la izquierda (defecto para idiomas LTR).
    *   `right`: Alineado a la derecha.
    *   `center`: Centrado.
    *   `justify`: Justificado (alineado a ambos márgenes, ajustando el espacio entre palabras). Úsalo con cuidado, puede crear "ríos" de espacio.
    ```css
    h1 { text-align: center; }
    .cita { text-align: right; }
    .articulo { text-align: justify; }
    ```
*   `text-decoration`: Añade o quita decoración al texto.
    *   `none`: Sin decoración (útil para quitar el subrayado de los enlaces).
    *   `underline`: Subrayado.
    *   `overline`: Línea superior.
    *   `line-through`: Tachado.
    *   Se pueden combinar múltiples valores (ej: `underline line-through`). También se puede especificar el color y estilo de la línea (`text-decoration: underline dotted red;`).
    ```css
    a { text-decoration: none; } /* Quita subrayado de enlaces */
    a:hover { text-decoration: underline; } /* Añade subrayado al pasar el ratón */
    .antiguo { text-decoration: line-through; }
    ```
*   `line-height`: Define la altura de la línea de texto (el espacio vertical entre líneas).
    *   Valores sin unidad (recomendado): Se multiplican por el `font-size` del elemento para obtener la altura. `line-height: 1.5;` (Altura de línea es 1.5 veces el tamaño de la fuente). Se hereda mejor.
    *   Valores con unidad (`px`, `em`, `rem`): Definen una altura fija. `line-height: 24px;`
    *   Porcentajes: Relativo al `font-size` del elemento. `line-height: 150%;`
    Un buen `line-height` (ej: 1.4 a 1.7) mejora enormemente la legibilidad del texto largo.

### 5_4_text_transform_letter_spacing_word_spacing

*   `text-transform`: Controla la capitalización del texto.
    *   `none`: Sin transformación.
    *   `capitalize`: Pone en mayúscula la primera letra de cada palabra.
    *   `uppercase`: Convierte todo el texto a mayúsculas.
    *   `lowercase`: Convierte todo el texto a minúsculas.
    ```css
    h1 { text-transform: uppercase; }
    .titulo-seccion { text-transform: capitalize; }
    ```
*   `letter-spacing`: Ajusta el espacio entre caracteres. Puede usar valores como `px`, `em`. Valores negativos juntan las letras.
    ```css
    h1 { letter-spacing: 2px; }
    .condensado { letter-spacing: -1px; }
    ```
*   `word-spacing`: Ajusta el espacio entre palabras.
    ```css
    p { word-spacing: 0.2em; }
    ```

### 5_5_Fuentes_Web_font_face

Permite usar fuentes personalizadas que no están instaladas en el sistema del usuario. Descarga el archivo de la fuente junto con la página web.

Se define usando la regla `@font-face` al principio del CSS:

```css
@font-face {
  font-family: 'MiFuentePersonalizada'; /* Nombre que usarás en font-family */
  src: url('fuentes/mi_fuente.woff2') format('woff2'), /* Formato moderno preferido */
       url('fuentes/mi_fuente.woff') format('woff');   /* Formato WOFF */
       /* Puedes añadir otros formatos como ttf, otf, eot para compatibilidad antigua */
  font-weight: normal; /* Opcional: especificar peso */
  font-style: normal;  /* Opcional: especificar estilo */
}

/* Luego puedes usarla como cualquier otra fuente */
body {
  font-family: 'MiFuentePersonalizada', sans-serif;
}

/* Puedes definir múltiples @font-face para diferentes pesos/estilos */
@font-face {
    font-family: 'MiFuentePersonalizada';
    src: url('fuentes/mi_fuente_bold.woff2') format('woff2'),
         url('fuentes/mi_fuente_bold.woff') format('woff');
    font-weight: bold;
    font-style: normal;
}
```

*   **Formatos de Fuente Web:** `WOFF2` (mejor compresión, preferido), `WOFF` (amplia compatibilidad), `TTF`/`OTF` (menos optimizados para web), `EOT` (para IE antiguo).
*   **Servicios de Fuentes:** Google Fonts, Adobe Fonts, etc., simplifican el proceso proporcionando el código `@font-face` y sirviendo las fuentes desde sus CDNs.
*   **Rendimiento:** Cargar muchas fuentes o pesos puede ralentizar la carga de la página. Sé selectivo.

## 6_El_Modelo_de_Caja_Box_Model

Cada elemento HTML en una página web es tratado por CSS como una caja rectangular. El Modelo de Caja describe cómo se calculan el tamaño y el espaciado de estas cajas.

### 6_1_Content_Padding_Border_Margin

La caja de un elemento se compone de cuatro áreas concéntricas:

1.  **Content (Contenido):** El área donde se muestra el contenido real del elemento (texto, imágenes, etc.). Sus dimensiones son definidas por `width` y `height`.
2.  **Padding (Relleno):** Un espacio transparente *alrededor* del área de contenido, pero *dentro* del borde. El color de fondo del elemento se extiende a través del padding. Controlado por `padding-top`, `padding-right`, `padding-bottom`, `padding-left` o la abreviatura `padding`.
3.  **Border (Borde):** Una línea que rodea el padding y el contenido. Controlado por las propiedades `border-*` (ver sección 4.4).
4.  **Margin (Margen):** Un espacio transparente *fuera* del borde. Separa la caja del elemento de otras cajas adyacentes. Controlado por `margin-top`, `margin-right`, `margin-bottom`, `margin-left` o la abreviatura `margin`.

**Visualización:**

```
+-------------------------------------------+
|                  Margin                   |
|  +-------------------------------------+  |
|  |                Border               |  |
|  |  +-------------------------------+  |  |
|  |  |            Padding            |  |  |
|  |  |  +-------------------------+  |  |  |
|  |  |  |         Content         |  |  |  |
|  |  |  | (width x height)        |  |  |  |
|  |  |  +-------------------------+  |  |  |
|  |  +-------------------------------+  |  |
|  +-------------------------------------+  |
+-------------------------------------------+
```

**Abreviaturas para `padding` y `margin`:** Funcionan igual que para `border-width` (1, 2, 3 o 4 valores para top, right, bottom, left).

```css
.mi-caja {
  width: 200px;
  height: 100px;
  padding: 20px; /* 20px en los 4 lados */
  border: 1px solid black;
  margin: 10px 15px; /* 10px top/bottom, 15px left/right */
  background-color: lightgray;
}
```

### 6_2_Propiedades_width_y_height

Definen las dimensiones del área de **contenido** del elemento.

```css
div {
  width: 300px; /* Ancho del contenido */
  height: 150px; /* Alto del contenido */
  background-color: #eee;
}
```

*   Por defecto (`box-sizing: content-box`), el `padding` y el `border` se *añaden* al `width` y `height` para calcular el tamaño total ocupado por la caja visible. En el ejemplo `.mi-caja` anterior, el ancho total visible sería `200px (width) + 20px (padding-left) + 20px (padding-right) + 1px (border-left) + 1px (border-right) = 242px`. Esto puede ser contraintuitivo.

*   Se pueden usar otras unidades como `%` (relativo al ancho/alto del contenedor padre), `em`, `rem`, `vw`, `vh`.
*   `max-width` / `min-width`: Definen límites máximos y mínimos para el ancho. Útil para responsive design.
*   `max-height` / `min-height`: Definen límites máximos y mínimos para el alto.

### 6_3_box_sizing_border_box

La propiedad `box-sizing` cambia cómo se calcula el tamaño total de la caja.

*   `content-box` (valor por defecto): `width` y `height` definen solo el área de contenido. Padding y border se añaden *fuera* de esas dimensiones.
*   `border-box`: `width` y `height` definen el tamaño total de la caja *incluyendo* el padding y el border. El área de contenido se encoge para acomodar el padding y el border dentro del `width` y `height` especificados.

**Es una práctica muy común y recomendada establecer `box-sizing: border-box;` globalmente** para un comportamiento más intuitivo al definir tamaños:

```css
html {
  box-sizing: border-box;
}
*, *::before, *::after {
  box-sizing: inherit; /* Hereda de html */
}

/* Ahora, si defines: */
.mi-caja-border-box {
  width: 200px;
  padding: 20px;
  border: 1px solid black;
  box-sizing: border-box; /* O heredado globalmente */
}
/* El ancho TOTAL visible de esta caja será exactamente 200px. */
/* El área de contenido se ajustará a: 200 - 20 - 20 - 1 - 1 = 158px */
```
Usar `border-box` simplifica enormemente los cálculos de layout, especialmente con padding y bordes.

### 6_4_Margenes_Colapsables

Cuando dos márgenes verticales *adyacentes* (top y bottom) se encuentran, "colapsan" en uno solo. El tamaño del margen resultante es el del margen *más grande* de los dos que colapsaron (no la suma).

Esto ocurre en varias situaciones:
*   Entre elementos hermanos adyacentes (el margen inferior del primero colapsa con el margen superior del segundo).
*   Entre un elemento padre y su primer o último hijo (si no hay padding o border en el padre que los separe).
*   Entre párrafos vacíos.

**Ejemplo:**
```html
<p style="margin-bottom: 20px;">Párrafo 1</p>
<p style="margin-top: 30px;">Párrafo 2</p>
```
El espacio vertical entre estos dos párrafos **no** será 50px (20+30), sino **30px** (el mayor de los dos márgenes).

*   Los márgenes horizontales (left/right) **nunca** colapsan.
*   El colapso no ocurre para elementos con `display: flex`, `display: grid`, `position: absolute`, `position: fixed`, o elementos flotantes.

Entender el colapso de márgenes ayuda a predecir el espaciado vertical real entre elementos. A veces puede ser útil, otras veces puede requerir añadir un pequeño padding o border para evitarlo si se necesita un espaciado específico.

## 7_Unidades_y_Valores

CSS utiliza diversas unidades para especificar tamaños, longitudes y otros valores. Se dividen principalmente en absolutas y relativas.

### 7_1_Unidades_Absolutas_px_pt

Las unidades absolutas tienen un tamaño fijo y no dependen de otros elementos o del tamaño de la ventana.

*   **`px` (Píxeles):** La unidad absoluta más común en diseño web. Representa un "píxel CSS", que no siempre corresponde a un píxel físico del dispositivo, especialmente en pantallas de alta densidad (Retina). Es útil para tamaños que deben ser exactos, como bordes finos o a veces para `width` y `height` fijos (aunque menos flexible para responsive).
    ```css
    p { font-size: 16px; }
    div { border: 1px solid black; width: 500px; }
    ```
*   **`pt` (Puntos):** Unidad tradicional de tipografía (1pt = 1/72 de pulgada). Más común en medios de impresión (`@media print`). Raramente usada para diseño en pantalla.
*   Otras: `cm` (centímetros), `mm` (milímetros), `in` (pulgadas), `pc` (picas). Aún menos comunes para pantalla.

**Ventaja:** Precisión.
**Desventaja:** Menos flexibles para diseños adaptables y accesibilidad (los usuarios no pueden escalar fácilmente el tamaño base).

### 7_2_Unidades_Relativas_em_rem_porcentaje

Las unidades relativas definen un tamaño en relación a otra longitud. Son fundamentales para crear diseños flexibles y escalables.

*   **`em`:** Relativa al `font-size` del **elemento padre**. Si un elemento tiene `font-size: 16px;`, entonces `1em` dentro de ese elemento o en sus hijos (si heredan) equivale a `16px`. `2em` serían `32px`, `0.5em` serían `8px`.
    *   **Útil para:** Definir `padding`, `margin`, `width`, `height` en relación al tamaño del texto del elemento.
    *   **Desventaja:** El anidamiento puede complicar el cálculo, ya que el valor de `1em` cambia en cada nivel si los padres tienen diferentes `font-size`.
    ```css
    .padre { font-size: 20px; }
    .hijo {
      font-size: 0.8em; /* 0.8 * 20px = 16px */
      padding: 1em; /* 1 * 16px = 16px de padding */
      margin-left: 2em; /* 2 * 16px = 32px de margen */
    }
    ```
*   **`rem` (Root em):** Relativa al `font-size` del **elemento raíz** (`<html>`). Por defecto, los navegadores suelen tener un `font-size` raíz de `16px`, por lo que `1rem` = `16px`, `2rem` = `32px`, etc., *independientemente* del tamaño de fuente del elemento padre.
    *   **Muy recomendado para:** `font-size`, `padding`, `margin`, `width`, `height`. Permite escalar todo el layout de forma consistente simplemente cambiando el `font-size` del elemento `<html>` (útil para accesibilidad y responsive).
    ```css
    html { font-size: 100%; } /* Normalmente 16px por defecto */
    h1 { font-size: 2.5rem; } /* 2.5 * 16px = 40px */
    p { font-size: 1rem; margin-bottom: 1.5rem; } /* 16px y 24px */
    .caja { padding: 1rem; width: 20rem; } /* 16px padding, 320px width */
    ```
*   **`%` (Porcentaje):** Relativa al tamaño del **elemento contenedor padre**.
    *   Para `width`, `%` es relativo al *ancho* del padre.
    *   Para `height`, `%` es relativo a la *altura* del padre (¡pero a menudo requiere que el padre tenga una altura definida explícitamente!).
    *   Para `padding`, `margin` (horizontal y vertical), `%` es relativo al *ancho* del padre.
    *   Para `font-size`, `%` es relativo al `font-size` del padre (similar a `em`).
    *   Para `line-height`, `%` es relativo al `font-size` del *propio* elemento.
    ```css
    .contenedor { width: 800px; height: 400px; border: 1px solid; }
    .hijo {
      width: 50%; /* 50% de 800px = 400px */
      height: 75%; /* 75% de 400px = 300px */
      padding: 5%; /* 5% de 800px = 40px (para los 4 lados) */
      font-size: 120%; /* 120% del font-size del .contenedor */
    }
    ```

### 7_3_Unidades_de_Viewport_vw_vh_vmin_vmax

Estas unidades son relativas al tamaño de la **ventana gráfica (viewport)**, es decir, el área visible de la ventana del navegador. Son muy útiles para responsive design.

*   **`vw` (Viewport Width):** 1vw = 1% del ancho de la ventana gráfica.
*   **`vh` (Viewport Height):** 1vh = 1% de la altura de la ventana gráfica.
*   **`vmin` (Viewport Minimum):** 1vmin = 1% de la dimensión *más pequeña* de la ventana gráfica (ancho o alto).
*   **`vmax` (Viewport Maximum):** 1vmax = 1% de la dimensión *más grande* de la ventana gráfica (ancho o alto).

```css
.hero-section {
  width: 100vw; /* Ocupa todo el ancho de la ventana */
  height: 100vh; /* Ocupa toda la altura de la ventana (sección a pantalla completa) */
  background-color: lightcoral;
}

h1 {
  /* Tamaño de fuente que se adapta al ancho de la ventana */
  font-size: 5vw;
}

.cuadro-responsive {
  /* Un cuadrado cuyo lado es el 50% de la dimensión más pequeña de la ventana */
  width: 50vmin;
  height: 50vmin;
  background-color: mediumpurple;
}
```

**Ventajas:** Permiten crear elementos que se adaptan fluidamente al tamaño de la ventana.
**Consideraciones:** `100vh` puede ser problemático en móviles debido a las barras de navegación del navegador; hay soluciones más avanzadas o considerar `100%` en `html` y `body`. El tamaño de fuente con `vw` puede volverse muy pequeño o muy grande en extremos, a menudo se combina con `calc()` y `min`/`max` o `clamp()`.

## 8_La_Propiedad_Display

La propiedad `display` es una de las más importantes en CSS, ya que define cómo se comporta un elemento en el flujo del documento y cómo interactúa con otros elementos. Determina el tipo de caja de renderizado que genera un elemento.

### 8_1_Valores_block_inline_inline_block

*   **`display: block;`**
    *   El elemento ocupa todo el ancho disponible de su contenedor padre, creando saltos de línea antes y después.
    *   Respeta las propiedades `width`, `height`, `margin` (top/bottom/left/right) y `padding`.
    *   Elementos que son `block` por defecto: `<div>`, `<p>`, `<h1>`-`<h6>`, `<ul>`, `<ol>`, `<li>`, `<form>`, `<header>`, `<footer>`, `<section>`, `<article>`, etc.
    ```css
    div {
      display: block; /* (Ya es el valor por defecto para div) */
      width: 200px;
      height: 100px;
      margin: 10px;
      background-color: lightblue;
    } /* Cada div estará en su propia línea */
    ```

*   **`display: inline;`**
    *   El elemento fluye con el texto circundante, no crea saltos de línea antes ni después.
    *   **Ignora** las propiedades `width` y `height`. El ancho y alto son determinados por el contenido.
    *   Respeta `padding` y `margin` horizontales (`left`/`right`), pero los `padding` y `margin` verticales (`top`/`bottom`) **no** afectan al layout de los elementos circundantes (aunque el fondo y borde sí los muestren).
    *   Elementos que son `inline` por defecto: `<span>`, `<a>`, `<img>`, `<em>`, `<strong>`, `<b>`, `<i>`, `<input>`, `<button>`, `<label>`, etc.
    ```css
    span {
      display: inline; /* (Valor por defecto para span) */
      padding: 5px; /* Padding horizontal OK, vertical no afecta flujo */
      margin: 0 5px; /* Margen horizontal OK, vertical ignorado para flujo */
      background-color: lightcoral;
      /* width: 100px; height: 50px; -- SERÁN IGNORADOS */
    } /* Los spans aparecerán uno tras otro en la misma línea */
    ```

*   **`display: inline-block;`**
    *   Combina características de `block` e `inline`.
    *   El elemento fluye con el texto (como `inline`), no crea saltos de línea completos.
    *   **Respeta** `width`, `height`, `margin` (todos los lados) y `padding` (todos los lados), como `block`.
    *   Es muy útil para crear elementos como botones, insignias o pequeños bloques que deben estar en línea pero necesitan dimensiones y espaciado vertical definidos.
    ```css
    button {
      display: inline-block; /* Permite definir tamaño y margen */
      padding: 10px 15px;
      margin: 5px;
      background-color: lightgreen;
      border: 1px solid darkgreen;
      width: 120px; /* Respetado */
    } /* Los botones aparecerán en línea pero con tamaño y margen */
    ```

### 8_2_Valor_none

*   **`display: none;`**
    *   Oculta completamente el elemento. Es como si no existiera en la página.
    *   El elemento **no ocupa espacio** en el layout.
    *   Es diferente de `visibility: hidden;`, que oculta el elemento pero *sí* reserva su espacio en el layout.
    *   Se usa comúnmente para ocultar/mostrar elementos con JavaScript o en diseño responsive.
    ```css
    .oculto {
      display: none;
    }
    ```

### 8_3_Introduccion_a_Flexbox_display_flex

*   **`display: flex;`**
    *   Convierte un elemento en un **contenedor flex**.
    *   Sus hijos directos se convierten en **items flex**.
    *   Activa un nuevo contexto de formato (flex formatting context) que permite distribuir, alinear y ordenar los items flex de forma muy potente y flexible a lo largo de un eje principal (main axis) y un eje transversal (cross axis).
    *   Es ideal para layouts unidimensionales (filas o columnas).
    *   Lo exploraremos en detalle en la sección 10.
    ```css
    .contenedor-flex {
      display: flex;
      /* Propiedades adicionales para controlar los items irían aquí */
    }
    ```

### 8_4_Introduccion_a_Grid_display_grid

*   **`display: grid;`**
    *   Convierte un elemento en un **contenedor grid**.
    *   Sus hijos directos se convierten en **items grid**.
    *   Permite definir un sistema de rejilla bidimensional (filas y columnas) sobre el que posicionar y alinear los items grid.
    *   Es ideal para layouts bidimensionales complejos.
    *   Lo exploraremos en detalle en la sección 11.
    ```css
    .contenedor-grid {
      display: grid;
      /* Propiedades para definir filas, columnas y ubicar items irían aquí */
    }
    ```

Otros valores de `display` menos comunes incluyen `list-item` (usado por `<li>`), `table`, `table-row`, `table-cell` (para simular tablas, aunque Grid/Flex son preferibles para layout general), `contents` (hace que el contenedor "desaparezca" y sus hijos tomen su lugar en el layout).

## 9_Posicionamiento

La propiedad `position` define cómo se posiciona un elemento en el documento. Junto con las propiedades `top`, `right`, `bottom`, `left` y `z-index`, permite controlar la ubicación exacta de los elementos, sacándolos del flujo normal si es necesario.

### 9_1_Posicionamiento_Estatico_y_Relativo

*   **`position: static;` (Valor por defecto)**
    *   El elemento se posiciona según el flujo normal del documento (como hemos visto hasta ahora con `display: block/inline`).
    *   Las propiedades `top`, `right`, `bottom`, `left` y `z-index` **no tienen ningún efecto**.
*   **`position: relative;`**
    *   El elemento se posiciona inicialmente según el flujo normal (ocupa su espacio original).
    *   **PERO**, las propiedades `top`, `right`, `bottom`, `left` ahora **sí** tienen efecto. Permiten **desplazar** el elemento visualmente *desde su posición original estática*.
    *   El espacio original que ocupaba el elemento **se mantiene reservado** en el layout, los otros elementos no se recolocan para llenar ese hueco.
    *   Crucialmente, un elemento con `position: relative` crea un **contexto de posicionamiento** para sus hijos posicionados de forma `absolute`.
    ```css
    .caja-relativa {
      position: relative;
      left: 30px; /* Se desplaza 30px a la derecha desde su posición original */
      top: -10px; /* Se desplaza 10px hacia arriba desde su posición original */
      background-color: lightyellow;
      border: 1px solid gold;
      width: 150px;
    }
    .caja-normal {
        background-color: lightgray;
        width: 150px;
        border: 1px solid gray;
    }
    /* La caja-normal aparecerá donde le tocaría si caja-relativa no se hubiera movido,
       porque caja-relativa sigue ocupando su espacio original en el flujo. */
    ```

### 9_2_Posicionamiento_Absoluto

*   **`position: absolute;`**
    *   El elemento es **retirado completamente del flujo normal** del documento. Ya no ocupa espacio en el layout, y los demás elementos se comportan como si no existiera.
    *   Se posiciona con respecto a su **ancestro posicionado más cercano**. Un ancestro posicionado es un elemento padre (o abuelo, etc.) que tenga un valor de `position` diferente de `static` (es decir, `relative`, `absolute`, `fixed`, o `sticky`).
    *   Si no hay ningún ancestro posicionado, se posiciona con respecto al bloque contenedor inicial (normalmente el elemento `<html>` o el `<body>`).
    *   Las propiedades `top`, `right`, `bottom`, `left` determinan la distancia entre los bordes del elemento absoluto y los bordes de su ancestro posicionado (o del bloque inicial).
    ```html
    <div style="position: relative; width: 300px; height: 200px; border: 1px solid blue;">
      Contenedor Relativo
      <div style="position: absolute; top: 20px; right: 10px; width: 50px; height: 50px; background-color: red;">
        Absoluto
      </div>
    </div>
    ```
    *En este ejemplo, la caja roja se posicionará 20px desde arriba y 10px desde la derecha *dentro* de la caja azul (su ancestro posicionado relativo).*

### 9_3_Posicionamiento_Fijo_Fixed

*   **`position: fixed;`**
    *   Similar a `absolute`, el elemento es **retirado del flujo normal** y no ocupa espacio.
    *   Se posiciona con respecto a la **ventana gráfica (viewport)**, no a un ancestro posicionado.
    *   Las propiedades `top`, `right`, `bottom`, `left` determinan la distancia a los bordes de la ventana.
    *   El elemento **permanece fijo** en su posición incluso cuando el usuario hace scroll en la página.
    *   Ideal para elementos como cabeceras/pies de página fijos, botones de "volver arriba", menús persistentes, etc.
    ```css
    .cabecera-fija {
      position: fixed;
      top: 0;
      left: 0;
      width: 100%; /* Ocupa todo el ancho */
      height: 50px;
      background-color: rgba(50, 50, 50, 0.9);
      color: white;
      z-index: 1000; /* Asegura que esté por encima de otros elementos */
    }
    /* ¡Importante! Añadir padding/margin al body o al contenido principal
       para evitar que la cabecera fija oculte el inicio del contenido */
    body {
        padding-top: 50px; /* Mismo alto que la cabecera fija */
    }
    ```

### 9_4_Posicionamiento_Adhesivo_Sticky

*   **`position: sticky;`**
    *   Es un híbrido entre `relative` y `fixed`.
    *   El elemento se comporta como `relative` dentro de su contenedor padre hasta que alcanza un umbral especificado por `top`, `right`, `bottom`, o `left` al hacer scroll.
    *   Una vez que alcanza ese umbral, se comporta como `fixed` (se "pega" a esa posición) *dentro de los límites de su contenedor padre*.
    *   Cuando el final del contenedor padre sale de la pantalla, el elemento sticky deja de pegarse y se desplaza junto con el padre.
    *   Requiere que se especifique al menos uno de los umbrales (`top`, `right`, `bottom`, `left`).
    *   Útil para cabeceras de sección que se quedan fijas mientras se recorre esa sección, índices laterales, etc.
    ```css
    .titulo-seccion-pegajoso {
      position: sticky;
      top: 0; /* Se pegará a la parte superior de la ventana cuando el scroll lo alcance */
      background-color: white;
      padding: 10px;
      border-bottom: 1px solid #ccc;
      z-index: 100;
    }
    /* El elemento se comportará como relative hasta que su borde superior
       esté a 0px del borde superior del viewport, momento en que se fijará ahí,
       pero solo mientras su contenedor padre sea visible. */
    ```
    *Nota: El posicionamiento `sticky` puede tener comportamientos inesperados si algún contenedor padre tiene `overflow: hidden/scroll/auto`.*

### 9_5_Apilamiento_z_index

Cuando los elementos se posicionan fuera del flujo normal (`relative`, `absolute`, `fixed`, `sticky`), pueden superponerse. La propiedad `z-index` controla el orden de apilamiento a lo largo del eje Z (profundidad, hacia/desde el espectador).

*   Solo funciona en **elementos posicionados** (cualquier `position` excepto `static`).
*   Acepta un valor entero (positivo, negativo o cero).
*   Un elemento con un `z-index` mayor se muestra **delante** de un elemento con un `z-index` menor.
*   Si dos elementos posicionados tienen el mismo `z-index`, su orden de apilamiento lo determina su orden en el código HTML (el que aparece más tarde está delante).
*   Los elementos posicionados crean **contextos de apilamiento**. El `z-index` de un elemento hijo solo tiene significado dentro del contexto de apilamiento de su padre. Un elemento con `z-index: 1` puede estar detrás de otro con `z-index: 1000` si este último pertenece a un contexto de apilamiento diferente que, en conjunto, está detrás.

```css
.elemento-detras {
  position: absolute;
  top: 10px; left: 10px;
  width: 100px; height: 100px;
  background-color: lightcoral;
  z-index: 1;
}
.elemento-delante {
  position: absolute;
  top: 30px; left: 30px;
  width: 100px; height: 100px;
  background-color: lightblue;
  z-index: 2; /* Valor mayor, aparecerá delante */
}
```

## 10_Layout_con_Flexbox

Flexbox (Flexible Box Layout) es un modelo de layout unidimensional diseñado para distribuir el espacio entre los items de un contenedor y alinearlos, incluso cuando su tamaño es desconocido o dinámico. Es extremadamente potente para componentes y layouts más simples.

Para usar Flexbox:
1.  Establece `display: flex` o `display: inline-flex` en el elemento contenedor.
2.  Los hijos directos de ese contenedor se convierten automáticamente en items flex.
3.  Usa propiedades en el **contenedor flex** para controlar la dirección, espaciado, alineación y envoltura de los items.
4.  Usa propiedades en los **items flex** individuales para controlar su tamaño, orden y alineación específica.

### 10_1_Contenedor_Flex_Direccion_y_Envoltura

Propiedades aplicadas al **contenedor** (`display: flex`):

*   **`flex-direction`**: Establece el eje principal (main axis) donde se colocarán los items.
    *   `row` (defecto): Eje principal horizontal (izquierda a derecha).
    *   `row-reverse`: Eje principal horizontal (derecha a izquierda).
    *   `column`: Eje principal vertical (arriba a abajo).
    *   `column-reverse`: Eje principal vertical (abajo a arriba).
*   **`flex-wrap`**: Controla si los items deben envolverse en múltiples líneas si no caben en una sola.
    *   `nowrap` (defecto): Los items se encogen/desbordan para caber en una línea.
    *   `wrap`: Los items pasan a la siguiente línea si no caben.
    *   `wrap-reverse`: Igual que `wrap`, pero las nuevas líneas se apilan en orden inverso.
*   **`flex-flow`**: Abreviatura para `flex-direction` y `flex-wrap`.
    ```css
    .contenedor {
      display: flex;
      flex-flow: row wrap; /* Coloca en fila, permite envolver */
    }
    ```

### 10_2_Alineacion_Justificacion_Contenido_y_Items

Propiedades aplicadas al **contenedor** (`display: flex`):

*   **`justify-content`**: Alinea los items a lo largo del **eje principal (main axis)**. Controla cómo se distribuye el espacio sobrante.
    *   `flex-start` (defecto): Items agrupados al inicio del eje principal.
    *   `flex-end`: Items agrupados al final del eje principal.
    *   `center`: Items agrupados en el centro del eje principal.
    *   `space-between`: Espacio distribuido equitativamente *entre* los items (el primero al inicio, el último al final).
    *   `space-around`: Espacio distribuido equitativamente *alrededor* de los items (los items tienen medio espacio en los extremos).
    *   `space-evenly`: Espacio distribuido equitativamente entre cada item y en los extremos.
*   **`align-items`**: Alinea los items a lo largo del **eje transversal (cross axis)** dentro de su línea actual.
    *   `stretch` (defecto): Items se estiran para llenar la altura/anchura del contenedor en el eje transversal (si no tienen tamaño fijo en ese eje).
    *   `flex-start`: Items alineados al inicio del eje transversal.
    *   `flex-end`: Items alineados al final del eje transversal.
    *   `center`: Items alineados en el centro del eje transversal.
    *   `baseline`: Items alineados según su línea base de texto.
*   **`align-content`**: Similar a `justify-content`, pero alinea las **líneas** (cuando hay `flex-wrap: wrap`) a lo largo del **eje transversal**. Solo tiene efecto si hay múltiples líneas. Acepta valores como `flex-start`, `flex-end`, `center`, `space-between`, `space-around`, `stretch`.

```css
.contenedor {
  display: flex;
  height: 300px; /* Necesario para ver align-items/content */
  flex-wrap: wrap;
  justify-content: center; /* Centra items horizontalmente */
  align-items: center; /* Centra items verticalmente en su línea */
  align-content: space-around; /* Distribuye las líneas verticalmente */
  border: 1px solid;
}
.item { width: 80px; height: 80px; background: lightblue; margin: 5px; }
```

### 10_3_Propiedades_de_Items_Flex_order_grow_shrink_basis

Propiedades aplicadas a los **items flex** (hijos directos del contenedor):

*   **`order`**: Cambia el orden visual de los items. Por defecto es `0`. Valores enteros (positivos o negativos). Items con `order` menor aparecen antes.
    ```css
    .item-1 { order: 2; }
    .item-2 { order: 1; } /* Aparecerá antes que item-1 */
    ```
*   **`flex-grow`**: Define la capacidad de un item para **crecer** y ocupar el espacio disponible en el eje principal. Valor numérico no negativo (defecto `0`). Si todos son `0`, no crecen. Si un item tiene `flex-grow: 1` y los demás `0`, ocupará todo el espacio sobrante. Si varios tienen `flex-grow`, el espacio se reparte proporcionalmente a sus valores.
    ```css
    .item-principal { flex-grow: 2; } /* Crecerá el doble que otros con flex-grow: 1 */
    .item-secundario { flex-grow: 1; }
    ```
*   **`flex-shrink`**: Define la capacidad de un item para **encogerse** si no hay suficiente espacio en el eje principal. Valor numérico no negativo (defecto `1`). Un valor de `0` evita que el item se encoja. Items con valores mayores se encogen más.
    ```css
    .no-encoger { flex-shrink: 0; }
    ```
*   **`flex-basis`**: Define el tamaño inicial del item a lo largo del eje principal *antes* de que se distribuya el espacio sobrante (con `flex-grow` o `flex-shrink`). Puede ser un valor de longitud (`px`, `%`, `em`) o `auto` (defecto, el tamaño se basa en el contenido o `width`/`height`).
    ```css
    .item { flex-basis: 150px; } /* Tamaño inicial de 150px */
    ```
*   **`flex`**: Propiedad abreviada para `flex-grow`, `flex-shrink` y `flex-basis`.
    *   `flex: 0 1 auto;` (Valor por defecto: no crece, puede encogerse, tamaño base automático).
    *   `flex: 1;` (Equivale a `flex: 1 1 0;` - crece y encoge, base cero -> ocupa espacio proporcional).
    *   `flex: auto;` (Equivale a `flex: 1 1 auto;` - crece y encoge, base automática).
    *   `flex: none;` (Equivale a `flex: 0 0 auto;` - no crece, no encoge, tamaño base automático).
*   **`align-self`**: Permite anular el `align-items` del contenedor para un **item flex individual**. Acepta los mismos valores que `align-items` (`auto`, `flex-start`, `flex-end`, `center`, `baseline`, `stretch`).
    ```css
    .item-especial { align-self: flex-end; } /* Este item se alinea abajo aunque los demás estén arriba */
    ```

### 10_4_Ejemplos_Practicos_Flexbox

*   **Centrar un elemento perfectamente:**
    ```css
    .contenedor-centrado {
      display: flex;
      justify-content: center; /* Centrado horizontal */
      align-items: center; /* Centrado vertical */
      min-height: 100vh; /* Ocupa al menos la altura de la ventana */
      border: 1px dashed gray;
    }
    .elemento-centrado {
      padding: 20px;
      background: lightgreen;
    }
    ```
*   **Navegación simple:**
    ```css
    nav ul {
      display: flex;
      list-style: none;
      padding: 0;
      margin: 0;
      background-color: #eee;
    }
    nav li {
      margin: 0 5px; /* Espacio entre items */
    }
    nav a {
      display: block; /* Para que el padding funcione bien */
      padding: 10px 15px;
      text-decoration: none;
      color: #333;
    }
    nav a:hover { background-color: #ddd; }

    /* Para alinear un item (ej: login) a la derecha */
    nav li.login {
        margin-left: auto; /* ¡Truco útil! Empuja el item a la derecha */
    }
    ```
*   **Columnas de igual altura (implícito):** Si los items flex están en una fila, `align-items: stretch` (defecto) hará que todos tengan la misma altura (la del más alto).

Flexbox es increíblemente versátil para muchos patrones de layout comunes.

## 11_Layout_con_Grid

CSS Grid Layout es un sistema de maquetación bidimensional basado en rejillas. Permite dividir el espacio en filas y columnas, y posicionar los elementos hijos (items grid) dentro de esas celdas, áreas o líneas definidas. Es ideal para layouts complejos y estructuras de página completas.

Para usar Grid:
1.  Establece `display: grid` o `display: inline-grid` en el elemento contenedor.
2.  Los hijos directos se convierten en items grid.
3.  Define la estructura de la rejilla (filas y columnas) en el **contenedor grid**.
4.  Opcionalmente, posiciona los **items grid** explícitamente en la rejilla.

### 11_1_Contenedor_Grid_Definicion_Columnas_y_Filas

Propiedades aplicadas al **contenedor** (`display: grid`):

*   **`grid-template-columns`**: Define las columnas de la rejilla. Especifica el tamaño de cada columna.
    *   Puedes usar unidades de longitud (`px`, `rem`, `%`).
    *   La unidad `fr` (fracción): Representa una fracción del espacio *disponible* en el contenedor. Es muy potente para layouts flexibles. `1fr 2fr` significa que la segunda columna será el doble de ancha que la primera, ocupando el espacio sobrante.
    *   La función `repeat(n, tamaño)`: Para repetir un patrón. `repeat(3, 1fr)` crea 3 columnas de igual ancho. `repeat(auto-fit, minmax(100px, 1fr))` crea columnas flexibles que se ajustan automáticamente.
    *   `minmax(min, max)`: Define un rango de tamaño para una columna o fila (mínimo y máximo).
*   **`grid-template-rows`**: Define las filas de la rejilla. Funciona de manera similar a `grid-template-columns`.
    ```css
    .contenedor-grid {
      display: grid;
      /* 3 columnas: 100px fija, auto (contenido), 1 fracción de espacio */
      grid-template-columns: 100px auto 1fr;
      /* 2 filas: 50px fija, la segunda ocupa el espacio restante */
      grid-template-rows: 50px 1fr;
      /* 3 columnas de igual ancho flexible */
      /* grid-template-columns: repeat(3, 1fr); */
      /* Filas de altura mínima 100px, máxima automática */
      /* grid-template-rows: repeat(2, minmax(100px, auto)); */
      border: 1px solid blue;
      height: 300px; /* Para visualizar filas */
    }
    .item-grid { background: lightcoral; border: 1px solid red; padding: 5px; }
    ```
*   **`grid-template-areas`**: Permite nombrar áreas de la rejilla y asignar items a esas áreas. Define la estructura visualmente usando nombres.
    ```css
    .contenedor-areas {
      display: grid;
      grid-template-columns: 1fr 1fr 1fr;
      grid-template-rows: auto 1fr auto;
      grid-template-areas:
        "header header header"
        "sidebar main main"
        "footer footer footer";
      height: 400px; border: 1px solid green;
    }
    /* Luego asignas los items a las áreas */
    .header { grid-area: header; background: lightblue; }
    .sidebar { grid-area: sidebar; background: lightgreen; }
    .main { grid-area: main; background: lightyellow; }
    .footer { grid-area: footer; background: lightgray; }
    ```

### 11_2_Espaciado_gap

Propiedades aplicadas al **contenedor** (`display: grid`):

*   **`gap` (o `grid-gap` en sintaxis antigua):** Define el tamaño de los espacios (canalones o gutters) *entre* las filas y columnas de la rejilla. No añade espacio en los bordes exteriores del contenedor.
    *   `row-gap`: Espacio entre filas.
    *   `column-gap`: Espacio entre columnas.
    *   `gap`: Abreviatura para `row-gap column-gap`. Si solo se da un valor, se aplica a ambos.
    ```css
    .contenedor-con-gap {
      display: grid;
      grid-template-columns: repeat(3, 1fr);
      grid-template-rows: repeat(2, 100px);
      gap: 10px 20px; /* 10px entre filas, 20px entre columnas */
      /* gap: 15px; -- 15px entre filas y columnas */
      border: 1px solid;
    }
    .item { background: lightpink; border: 1px solid; }
    ```

### 11_3_Ubicacion_de_Items_Lineas_y_Areas

Por defecto, los items grid se colocan automáticamente en las celdas disponibles, una tras otra. Puedes controlar explícitamente dónde se ubica cada item usando las líneas de la rejilla o las áreas nombradas.

Las líneas de la rejilla se numeran automáticamente. Para `N` columnas, hay `N+1` líneas verticales (1 a N+1). Para `M` filas, hay `M+1` líneas horizontales (1 a M+1). También se pueden usar números negativos para contar desde el final (`-1` es la última línea).

Propiedades aplicadas a los **items grid**:

*   **`grid-column-start`, `grid-column-end`**: Especifican la línea vertical de inicio y fin donde se coloca el item.
*   **`grid-row-start`, `grid-row-end`**: Especifican la línea horizontal de inicio y fin donde se coloca el item.
*   **`grid-column`**: Abreviatura para `grid-column-start / grid-column-end`. Se puede usar `span N` para indicar que ocupe N columnas/filas.
*   **`grid-row`**: Abreviatura para `grid-row-start / grid-row-end`.
*   **`grid-area`**:
    *   Si se usa con `grid-template-areas` en el contenedor: Asigna el item al área nombrada. `grid-area: nombre-area;`
    *   Como abreviatura para las 4 propiedades de línea: `grid-area: row-start / column-start / row-end / column-end;`

```css
.contenedor-ubicacion {
  display: grid;
  grid-template-columns: repeat(4, 1fr); /* 5 líneas verticales */
  grid-template-rows: repeat(3, 100px); /* 4 líneas horizontales */
  gap: 5px; border: 1px solid;
}
.item-a {
  background: lightcoral;
  grid-column-start: 1;
  grid-column-end: 3; /* Ocupa columnas 1 y 2 */
  grid-row-start: 1;
  grid-row-end: 2;
}
.item-b {
  background: lightblue;
  /* Abreviatura: empieza en col 3, ocupa 2 columnas; empieza en fila 1, ocupa 3 filas */
  grid-column: 3 / span 2;
  grid-row: 1 / span 3;
}
.item-c {
    background: lightgreen;
    /* Abreviatura grid-area: fila 2 / col 1 / fila 4 (-1) / col 3 */
    grid-area: 2 / 1 / 4 / 3;
}
```

### 11_4_Alineacion_en_Grid

Grid ofrece propiedades de alineación similares a Flexbox, pero adaptadas a un contexto bidimensional.

*   **Alineación del Contenido (Distribución del espacio dentro del contenedor grid):**
    *   `justify-content`: Alinea la rejilla completa horizontalmente dentro del contenedor (si la rejilla es más pequeña que el contenedor). Valores: `start`, `end`, `center`, `space-between`, `space-around`, `space-evenly`, `stretch`.
    *   `align-content`: Alinea la rejilla completa verticalmente dentro del contenedor. Mismos valores.
*   **Alineación de Items (Dentro de su celda/área):**
    *   `justify-items`: Alinea los items horizontalmente dentro de su celda. Valores: `start`, `end`, `center`, `stretch` (defecto).
    *   `align-items`: Alinea los items verticalmente dentro de su celda. Mismos valores.
    *   `place-items`: Abreviatura para `align-items justify-items`.
*   **Alineación Individual de Item:**
    *   `justify-self`: Anula `justify-items` para un item específico.
    *   `align-self`: Anula `align-items` para un item específico.
    *   `place-self`: Abreviatura para `align-self justify-self`.

```css
.contenedor-alineacion {
    display: grid;
    grid-template-columns: repeat(2, 150px);
    grid-template-rows: repeat(2, 150px);
    gap: 10px;
    height: 400px; width: 400px; /* Contenedor más grande que la rejilla */
    border: 1px solid;

    /* Centra la rejilla dentro del contenedor */
    justify-content: center;
    align-content: center;

    /* Centra los items dentro de sus celdas por defecto */
    /* justify-items: center; */
    /* align-items: center; */
    place-items: center; /* Abreviatura para los dos anteriores */
}
.item { background: lightcyan; border: 1px solid; width: 80px; height: 80px;}
.item-especial {
    background: lightgoldenrodyellow;
    /* Este item se alinea al inicio horizontal y final vertical */
    /* justify-self: start; */
    /* align-self: end; */
    place-self: end start; /* Abreviatura: align / justify */
}
```

### 11_5_Ejemplos_Practicos_Grid

*   **Layout "Santo Grial" (Header, Footer, 3 columnas):**
    ```css
    .santo-grial {
      display: grid;
      min-height: 100vh;
      grid-template-columns: auto 1fr auto; /* Sidebar izq, main flexible, sidebar der */
      grid-template-rows: auto 1fr auto; /* Header, contenido, footer */
      grid-template-areas:
        "header header header"
        "nav main aside"
        "footer footer footer";
      gap: 5px;
    }
    header { grid-area: header; background: #eee; }
    nav { grid-area: nav; background: #ddd; width: 150px; } /* Ancho fijo para sidebar */
    main { grid-area: main; background: #fff; }
    aside { grid-area: aside; background: #ddd; width: 120px; }
    footer { grid-area: footer; background: #eee; }
    ```
*   **Galería de Imágenes Adaptable:**
    ```css
    .galeria {
      display: grid;
      /* Columnas que tienen mínimo 150px, pero crecen para llenar espacio.
         El navegador crea tantas columnas como quepan. */
      grid-template-columns: repeat(auto-fit, minmax(150px, 1fr));
      gap: 10px;
    }
    .galeria img {
      width: 100%; /* Imagen ocupa el ancho de la celda grid */
      height: auto; /* Mantiene proporción */
      display: block; /* Evita espacio extra debajo de la imagen */
    }
    ```

Grid Layout es la herramienta más potente para maquetación web compleja y bidimensional.

## 12_Diseno_Web_Adaptable_Responsive

El Diseño Web Adaptable (Responsive Web Design - RWD) es un enfoque del diseño y desarrollo web que busca crear sitios que se vean y funcionen bien en una amplia variedad de dispositivos y tamaños de pantalla (móviles, tablets, portátiles, escritorios).

Los pilares del RWD son:
*   **Fluid Grids:** Usar unidades relativas como porcentajes o `fr` para que los layouts se adapten al ancho disponible.
*   **Flexible Images:** Asegurarse de que las imágenes (y otros medios) no excedan el ancho de su contenedor.
*   **Media Queries:** Aplicar diferentes estilos CSS según las características del dispositivo, principalmente el ancho de la pantalla.

### 12_1_Concepto_Mobile_First

Es una estrategia de RWD donde se diseñan y desarrollan primero los estilos para pantallas pequeñas (móviles) y luego se añaden estilos más complejos para pantallas más grandes usando media queries con `min-width`.

**Ventajas:**
*   Fuerza a priorizar el contenido esencial para móviles.
*   Generalmente resulta en un CSS más limpio y eficiente (menos anulaciones).
*   Mejor rendimiento en móviles, ya que no cargan estilos innecesarios de escritorio.

**Alternativa:** Desktop-first (diseñar para escritorio y usar media queries con `max-width` para simplificar en pantallas pequeñas). Mobile-first es generalmente la práctica recomendada actualmente.

### 12_2_Meta_Etiqueta_Viewport

Es **fundamental** incluir esta etiqueta `<meta>` en el `<head>` de tu HTML para que el RWD funcione correctamente en dispositivos móviles:

```html
<meta name="viewport" content="width=device-width, initial-scale=1.0">
```
*   `width=device-width`: Le dice al navegador que ajuste el ancho de la ventana gráfica (viewport) al ancho real del dispositivo. Sin esto, los móviles a menudo simulan un viewport de escritorio más ancho y escalan la página hacia abajo, haciendo que el texto sea ilegible.
*   `initial-scale=1.0`: Establece el nivel de zoom inicial al 100%, evitando que la página aparezca alejada.

### 12_3_Media_Queries

Permiten aplicar bloques de reglas CSS condicionalmente, basándose en las características del dispositivo o del viewport. La característica más usada es el ancho del viewport (`width`, `min-width`, `max-width`).

La sintaxis básica es `@media`:

```css
/* --- Estilos Base (Mobile First) --- */
body { font-size: 1rem; }
.contenedor { width: 95%; margin: 0 auto; }
nav ul { display: block; /* Menú vertical por defecto */ }
nav li { margin-bottom: 5px; }
.columna { width: 100%; margin-bottom: 15px; }

/* --- Estilos para Tablets y superiores (ej: > 600px) --- */
@media screen and (min-width: 600px) {
  body { font-size: 1.1rem; }
  .contenedor { width: 90%; max-width: 1200px; } /* Limita el ancho máximo */
  nav ul { display: flex; justify-content: space-around; } /* Menú horizontal */
  nav li { margin-bottom: 0; }
  .layout-2-columnas { display: grid; grid-template-columns: 1fr 1fr; gap: 20px; }
  .columna { width: auto; margin-bottom: 0; }
}

/* --- Estilos para Escritorios y superiores (ej: > 992px) --- */
@media screen and (min-width: 992px) {
  body { font-size: 1.2rem; }
  .layout-3-columnas { display: grid; grid-template-columns: 1fr 2fr 1fr; gap: 30px; }
}
```

*   `@media`: Inicia la regla.
*   `screen`: Tipo de medio (otros: `print`, `speech`, `all`). `screen` es para pantallas.
*   `and`: Operador lógico para combinar condiciones.
*   `(min-width: 600px)`: La condición (característica y valor). `min-width` aplica los estilos si el ancho es 600px *o mayor*. `max-width` aplica si es 600px *o menor*. `width` aplica solo si es *exactamente* 600px.
*   Las reglas CSS dentro de las llaves `{}` de la media query solo se aplicarán si la condición se cumple.

Puedes usar otras características como `orientation: portrait/landscape`, `resolution`, `hover`, etc.

### 12_4_Imagenes_Flexibles

Para que las imágenes se adapten a sus contenedores y no los desborden en pantallas pequeñas, una técnica común es:

```css
img, video, embed, object { /* Aplicar a todos los medios incrustados */
  max-width: 100%; /* La imagen nunca será más ancha que su contenedor */
  height: auto;   /* Mantiene la proporción al escalar el ancho */
  display: block; /* Opcional, para eliminar espacio extra debajo de imágenes inline */
}
```
Esto asegura que si el contenedor se encoge, la imagen se encogerá con él, hasta su tamaño máximo intrínseco.

### 12_5_Unidades_Relativas_en_Responsive

Usar unidades relativas como `rem`, `%`, `vw`, `vh`, `fr` es clave para RWD, ya que permiten que los tamaños y espaciados se adapten proporcionalmente al tamaño de fuente base o al tamaño del viewport, en lugar de tener valores fijos en píxeles que pueden no funcionar bien en todos los tamaños.

Combinar `rem` para tipografía y espaciado general, con `%` o `fr` (en Grid/Flex) para anchos de layout, suele ser un buen enfoque.

## 13_Transiciones_y_Animaciones

CSS permite crear efectos visuales suaves y animaciones sin necesidad de JavaScript (para efectos más simples).

### 13_1_Propiedad_transition

Aplica una transición suave entre el estado inicial y final de una o más propiedades CSS cuando su valor cambia (normalmente debido a una pseudo-clase como `:hover` o `:focus`, o por un cambio de clase con JavaScript).

La sintaxis abreviada es `transition: property duration timing-function delay;`

*   `transition-property`: La propiedad CSS a la que se aplicará la transición (ej: `background-color`, `opacity`, `transform`, o `all` para todas las propiedades aplicables).
*   `transition-duration`: La duración de la transición (ej: `0.3s`, `500ms`).
*   `transition-timing-function`: La curva de aceleración de la transición (cómo varía la velocidad). Valores comunes:
    *   `ease` (defecto): Inicio lento, rápido en medio, final lento.
    *   `linear`: Velocidad constante.
    *   `ease-in`: Inicio lento.
    *   `ease-out`: Final lento.
    *   `ease-in-out`: Inicio y final lentos.
    *   `cubic-bezier(n,n,n,n)`: Permite definir curvas personalizadas.
*   `transition-delay`: Un retardo antes de que comience la transición (ej: `0.1s`).

```css
button {
  background-color: steelblue;
  color: white;
  padding: 10px 20px;
  border: none;
  cursor: pointer;
  /* Aplicar transición a background-color y transform durante 0.3 segundos con ease-out */
  transition: background-color 0.3s ease-out, transform 0.2s ease;
  /* O transition: all 0.3s ease; */
}

button:hover {
  background-color: darkblue;
  transform: scale(1.05); /* Agranda ligeramente el botón */
}

button:active {
    transform: scale(0.95); /* Encoge al hacer clic */
}
```
*Importante:* La propiedad `transition` se declara en el **estado base** del elemento, no en el estado `:hover` o `:active`. Esto asegura que la transición ocurra tanto al entrar como al salir del estado.

No todas las propiedades CSS son "animables". Propiedades como `color`, `background-color`, `opacity`, `width`, `height`, `transform`, `font-size` sí lo son. Propiedades como `display` no lo son (el cambio es instantáneo).

### 13_2_Funciones_de_Temporizacion

Controlan la aceleración de las transiciones y animaciones. Las más comunes (`ease`, `linear`, `ease-in`, `ease-out`, `ease-in-out`) cubren la mayoría de los casos. `cubic-bezier()` ofrece control total para efectos personalizados. Herramientas online como [cubic-bezier.com](https://cubic-bezier.com/) ayudan a visualizar y crear estas curvas.

### 13_3_Animaciones_con_keyframes

Para animaciones más complejas que simples transiciones entre dos estados, se usan las reglas `@keyframes`. Permiten definir múltiples etapas (fotogramas clave) de una animación.

Se define un nombre para la animación y luego se especifican los estilos para diferentes puntos porcentuales (o usando `from` para 0% y `to` para 100%) a lo largo de la duración de la animación.

```css
/* Definición de la animación llamada 'fadeInScaleUp' */
@keyframes fadeInScaleUp {
  from { /* Estado inicial (0%) */
    opacity: 0;
    transform: scale(0.8);
  }
  to { /* Estado final (100%) */
    opacity: 1;
    transform: scale(1);
  }
}

/* Definición de una animación más compleja llamada 'pulse' */
@keyframes pulse {
  0% { transform: scale(1); box-shadow: 0 0 0 0 rgba(0, 123, 255, 0.7); }
  70% { transform: scale(1.1); box-shadow: 0 0 10px 20px rgba(0, 123, 255, 0); }
  100% { transform: scale(1); box-shadow: 0 0 0 0 rgba(0, 123, 255, 0); }
}
```

### 13_4_Propiedad_animation

Una vez definida una animación con `@keyframes`, se aplica a un elemento usando la propiedad `animation` (o sus sub-propiedades).

La sintaxis abreviada es `animation: name duration timing-function delay iteration-count direction fill-mode play-state;`

*   `animation-name`: El nombre de la animación definida con `@keyframes`.
*   `animation-duration`: Duración de un ciclo de la animación (ej: `2s`, `500ms`).
*   `animation-timing-function`: Curva de aceleración (igual que `transition-timing-function`).
*   `animation-delay`: Retardo antes de iniciar la animación.
*   `animation-iteration-count`: Cuántas veces se repite la animación. `infinite` para repetición infinita.
*   `animation-direction`: Dirección de la animación en ciclos alternos. `normal` (defecto), `reverse`, `alternate` (ida y vuelta), `alternate-reverse`.
*   `animation-fill-mode`: Define qué estilos aplica la animación antes de empezar y después de terminar. `none` (defecto), `forwards` (mantiene los estilos del último fotograma), `backwards` (aplica los estilos del primer fotograma antes del delay), `both`.
*   `animation-play-state`: Permite pausar (`paused`) o reanudar (`running`) la animación.

```css
.elemento-animado-1 {
  /* Aplicar la animación 'fadeInScaleUp' */
  animation-name: fadeInScaleUp;
  animation-duration: 0.5s;
  animation-timing-function: ease-out;
  animation-fill-mode: backwards; /* Aplica estado inicial antes de empezar */
}

.elemento-animado-2 {
  width: 50px; height: 50px; background: dodgerblue; border-radius: 50%;
  /* Aplicar la animación 'pulse' */
  animation: pulse 2s ease-in-out infinite; /* Repetición infinita */
}

/* Pausar la animación al pasar el ratón */
.elemento-animado-2:hover {
    animation-play-state: paused;
}
```

Las animaciones CSS son excelentes para microinteracciones, efectos de carga, llamadas a la acción y añadir dinamismo visual sin sobrecargar el rendimiento como podría hacerlo JavaScript para animaciones simples.

## 14_Variables_CSS_Propiedades_Personalizadas

Las Variables CSS, formalmente conocidas como **Propiedades Personalizadas CSS (CSS Custom Properties)**, permiten definir valores reutilizables directamente en CSS. Son extremadamente útiles para crear temas, mantener la consistencia y facilitar cambios globales.

### 14_1_Declaracion_y_Uso

*   **Declaración:** Las variables se declaran usando un nombre que comienza con dos guiones (`--`) y asignándoles un valor. Generalmente se declaran en el selector `:root` (que representa al elemento `<html>`) para que estén disponibles globalmente, o en un selector específico para limitar su alcance.
*   **Uso:** Se accede al valor de una variable usando la función `var(--nombre-variable)`. Se puede proporcionar un valor de fallback opcional: `var(--nombre-variable, valor-fallback)`.

```css
/* Declaración global de variables */
:root {
  --color-primario: steelblue;
  --color-secundario: orange;
  --fuente-principal: 'Arial', sans-serif;
  --padding-base: 15px;
  --radio-borde: 4px;
}

/* Uso de las variables */
body {
  font-family: var(--fuente-principal);
  color: #333;
}

h1 {
  color: var(--color-primario);
}

button {
  background-color: var(--color-primario);
  color: white;
  padding: var(--padding-base);
  border: none;
  border-radius: var(--radio-borde);
  /* Usando fallback */
  margin-bottom: var(--margen-botones, 10px); /* Si --margen-botones no está definida, usa 10px */
}

.alerta {
  background-color: var(--color-secundario);
  padding: var(--padding-base);
  border-radius: var(--radio-borde);
}
```

### 14_2_Alcance_Scope_y_Herencia

*   Las variables CSS siguen las reglas normales de la cascada y la herencia.
*   Una variable declarada en `:root` está disponible para todos los elementos del documento.
*   Puedes **anular** una variable global declarándola de nuevo dentro de un selector más específico. Los descendientes de ese selector usarán el valor anulado.

```css
:root {
  --color-texto: black;
}

body {
  color: var(--color-texto); /* Será negro */
}

.tema-oscuro {
  --color-texto: white; /* Anula la variable para elementos dentro de .tema-oscuro */
  background-color: #222;
}

.tema-oscuro p {
    /* Heredará --color-texto: white del padre .tema-oscuro */
    /* No es necesario redeclarar color aquí si hereda de body */
}

/* Este h2 usará el --color-texto global (negro) */
h2 { color: var(--color-texto); }

/* Este h2 dentro del tema oscuro usará el --color-texto anulado (blanco) */
.tema-oscuro h2 { color: var(--color-texto); }
```

### 14_3_Ventajas_y_Casos_de_Uso

*   **Reutilización (DRY - Don't Repeat Yourself):** Define un valor (como un color o tamaño) una vez y úsalo en múltiples lugares.
*   **Mantenimiento Fácil:** Cambia un valor en un solo lugar (`:root`) y se actualizará en todo el sitio. Ideal para colores de marca, espaciados consistentes, etc.
*   **Tematización (Theming):** Fácil creación de temas (claro/oscuro, diferentes paletas de colores) anulando las variables en un selector de clase aplicado al `<body>` o a un contenedor principal.
*   **Legibilidad:** Nombres de variables descriptivos (ej: `--color-primario-marca`) hacen el código más fácil de entender que valores hexadecimales repetidos.
*   **Manipulación con JavaScript:** Las variables CSS pueden ser leídas y modificadas dinámicamente usando JavaScript, permitiendo interactividad avanzada basada en las variables.

Las variables CSS son una característica moderna y muy poderosa que ha simplificado enormemente la gestión de valores en CSS.

## 15_Buenas_Practicas_y_Organizacion

Escribir CSS mantenible, escalable y eficiente es tan importante como conocer las propiedades.

### 15_1_Organizacion_de_Archivos_CSS

Para proyectos medianos o grandes, evita poner todo tu CSS en un solo archivo monolítico. Considera dividirlo en archivos más pequeños y lógicos:

*   **Base/Reset:** Estilos básicos, reseteos (como `box-sizing`), estilos para `html`/`body`.
*   **Layout:** Estilos para la estructura principal (header, footer, grid, contenedores principales).
*   **Components:** Estilos para elementos de UI reutilizables (botones, tarjetas, modales, formularios). Cada componente puede tener su propio archivo (ej: `_button.scss`, `_card.scss` si usas preprocesadores).
*   **Pages:** Estilos específicos para páginas particulares que anulan o complementan los estilos generales.
*   **Themes:** (Si aplica) Variables y estilos para diferentes temas visuales.
*   **Utils/Helpers:** Clases de utilidad (ej: `.text-center`, `.margin-top-small`).

Luego, puedes importar estos archivos parciales en un archivo principal usando `@import` de CSS (menos recomendado por rendimiento) o, preferiblemente, usando las capacidades de importación de **preprocesadores CSS** como Sass o Less.

### 15_2_Comentarios_en_CSS

Usa comentarios (`/* Comentario aquí */`) para explicar tu código, especialmente las partes complejas o decisiones de diseño no obvias. Organiza tu CSS en secciones lógicas usando comentarios como títulos.

```css
/*------------------------------------*\
    #BOTONES
\*------------------------------------*/
.button {
  /* Estilos base del botón */
}

/* Modificador de tamaño */
.button--large {
  /* Estilos para botón grande */
}

/*------------------------------------*\
    #TARJETAS
\*------------------------------------*/
.card {
  /* Estilos de la tarjeta */
}
```

### 15_3_Metodologias_BEM_OOCSS_SMACSS_Introduccion

Son convenciones y metodologías para escribir CSS de forma estructurada, escalable y mantenible, especialmente en equipos grandes o proyectos complejos. Ayudan a evitar colisiones de nombres, reducir la especificidad y hacer el código más predecible.

*   **BEM (Block, Element, Modifier):** Una de las más populares. Nombra las clases siguiendo una estructura estricta:
    *   `bloque`: Componente independiente (`.card`, `.menu`, `.form`).
    *   `bloque__elemento`: Parte de un bloque (`.card__title`, `.menu__item`, `.form__input`).
    *   `bloque--modificador` o `bloque__elemento--modificador`: Una variante o estado del bloque o elemento (`.card--dark`, `.menu__item--active`, `.form__input--error`).
    ```css
    /* Bloque: .search-form */
    .search-form { /* ... */ }
    /* Elemento: .search-form__input */
    .search-form__input { /* ... */ }
    /* Elemento: .search-form__button */
    .search-form__button { /* ... */ }
    /* Modificador del botón */
    .search-form__button--disabled { /* ... */ }
    ```
*   **OOCSS (Object-Oriented CSS):** Se enfoca en separar la estructura del skin (apariencia) y el contenedor del contenido. Promueve la creación de "objetos" CSS reutilizables.
*   **SMACSS (Scalable and Modular Architecture for CSS):** Categoriza las reglas CSS en Base, Layout, Module, State y Theme para organizar el código.

No es necesario seguir una metodología al pie de la letra, pero entender sus principios (modularidad, reutilización, bajo acoplamiento, convenciones de nombres) ayuda a escribir mejor CSS.

### 15_4_Preprocesadores_CSS_Sass_Less_Concepto

Son herramientas que extienden CSS con funcionalidades que no existen de forma nativa (o que son más nuevas), como variables (antes de las variables nativas), anidamiento de selectores, mixins (funciones reutilizables), herencia, operaciones matemáticas, etc.

Escribes código en la sintaxis del preprocesador (ej: `.scss` para Sass o `.less` para Less) y luego una herramienta lo **compila** a código CSS estándar que los navegadores pueden entender.

**Ventajas:**
*   Código más organizado y DRY (Don't Repeat Yourself).
*   Facilitan la creación de sistemas de diseño complejos.
*   Permiten usar características avanzadas que mejoran la productividad.

**Ejemplo (Sass - SCSS):**
```scss
// Variables
$color-primario: steelblue;
$padding-base: 10px;

// Anidamiento
nav {
  ul {
    list-style: none;
    padding: 0;
  }
  li {
    display: inline-block;
    margin-right: $padding-base;
  }
  a {
    text-decoration: none;
    color: $color-primario;
    &:hover { // Anidamiento del estado hover
      text-decoration: underline;
    }
  }
}

// Mixin (función reutilizable)
@mixin border-radius($radius) {
  -webkit-border-radius: $radius;
          border-radius: $radius;
}

.box {
  @include border-radius(5px); // Usa el mixin
}
```

Sass y Less son los preprocesadores más populares. Su uso es muy extendido en el desarrollo profesional.

Al igual que con HTML, es importante validar tu CSS para asegurar que sigue los estándares y evitar errores de sintaxis que puedan causar problemas.

**Herramientas de Validación:**

*   **W3C CSS Validation Service:** La herramienta oficial del W3C.
    *   [https://jigsaw.w3.org/css-validator/](https://jigsaw.w3.org/css-validator/)

Valida tu código regularmente para detectar errores tipográficos, propiedades desconocidas o valores incorrectos.
