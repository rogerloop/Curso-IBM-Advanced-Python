# Introducción a Bootstrap 4

Bootstrap es el <mark>framework</mark> de <mark>frontend</mark> (HTML, CSS y JavaScript) más popular del mundo para desarrollar <mark>sitios</mark> web y aplicaciones **responsive** (adaptables a diferentes tamaños de pantalla) y **mobile-first** (diseñadas pensando primero en dispositivos móviles).

Fue creado originalmente por Mark Otto y Jacob Thornton en Twitter como una herramienta interna para fomentar la consistencia entre sus herramientas, y fue liberado como proyecto open-source en agosto de 2011. Bootstrap 4, lanzado en enero de 2018, fue una reescritura importante que introdujo cambios significativos respecto a Bootstrap 3, como el paso a Flexbox para el sistema de rejilla, nuevas utilidades y componentes, y el abandono de LESS en favor de Sass.

Este curso se centrará específicamente en **Bootstrap 4**. Aunque Bootstrap 5 es la versión más reciente, Bootstrap 4 sigue siendo ampliamente utilizado y comprender sus fundamentos es esencial, además de facilitar la transición a la versión 5 si fuera necesario.

## 1_1
## ¿Qué es Bootstrap 4?

Bootstrap 4 es esencialmente una vasta <mark>colección</mark> de:

1.  **CSS Predefinido:** Proporciona <mark>estilos</mark> listos para usar para elementos HTML comunes (tipografía, tablas, formularios, botones, imágenes), un potente <mark>sistema de layout responsive</mark> basado en Flexbox (el Grid System), y docenas de componentes reutilizables (alertas, barras de navegación, modales, carruseles, etc.).
2.  **Componentes JavaScript (Plugins):** Ofrece <mark>plugins de JavaScript</mark> (que dependen de jQuery y Popper.js en la versión 4) para añadir interactividad a ciertos componentes (desplegables, modales, carruseles, tooltips, etc.) sin necesidad de escribir mucho código JavaScript personalizado.
3.  **Sistema de Rejilla (Grid System):** Un sistema de layout <mark>responsive</mark> de 12 columnas basado en Flexbox, que facilita enormemente la creación de diseños que se adaptan a diferentes anchos de pantalla (móviles, tablets, escritorios).
4.  **Utilidades (Utilities):** Clases CSS de propósito único que permiten aplicar <mark>estilos</mark> comunes rápidamente (márgenes, paddings, colores, bordes, alineación de texto, display, etc.) directamente en el HTML sin escribir CSS personalizado.

Bootstrap te da una base sólida y consistente para empezar a construir interfaces de usuario de forma rápida y eficiente.

## 1_2
## ¿Por qué usar Bootstrap? (Ventajas)

*   **Desarrollo Rápido:** <mark>Acelera</mark> enormemente el proceso de maquetación y diseño frontend al proporcionar componentes y estilos listos para usar. Puedes crear interfaces funcionales y atractivas en mucho menos tiempo.
*   **Responsive Design Fácil:** Su sistema de rejilla y utilidades responsive hacen que crear <mark>sitios</mark> que se adaptan a móviles, tablets y escritorios sea relativamente sencillo.
*   **Consistencia:** Ayuda a mantener una <mark>apariencia y comportamiento</mark> consistentes en toda la aplicación, especialmente útil en equipos.
*   **Compatibilidad entre Navegadores:** El equipo de Bootstrap se esfuerza por asegurar que los componentes funcionen correctamente en los <mark>navegadores</mark> web modernos más populares.
*   **Personalizable:** Aunque viene con un estilo por defecto, es altamente <mark>personalizable</mark>, especialmente si utilizas Sass para modificar variables y recompilar.
*   **Gran Comunidad y Documentación:** Cuenta con una <mark>documentación</mark> excelente y una comunidad enorme, lo que facilita encontrar ayuda, ejemplos y extensiones.
*   **Popularidad:** Al ser tan <mark>popular</mark>, muchos desarrolladores están familiarizados con él, facilitando la colaboración. Es una habilidad demandada en el mercado laboral.
*   **Mobile-First:** Su enfoque de diseño prioriza la experiencia en <mark>dispositivos móviles</mark>.

**¿Desventajas?**
*   **Aspecto "Bootstrap":** Si no se personaliza, los sitios pueden tener una <mark>apariencia similar</mark> ("genérica").
*   **Peso:** Incluir todo el CSS y JS puede añadir un <mark>peso</mark> considerable a la página si no se utilizan muchas de sus funcionalidades (aunque se puede personalizar qué incluir).
*   **Dependencia de jQuery (en v4):** Bootstrap 4 requiere <mark>jQuery</mark> para sus componentes JavaScript, lo cual algunos desarrolladores modernos intentan evitar. (Bootstrap 5 eliminó la dependencia de jQuery).
*   **Curva de Aprendizaje (Inicial):** Aunque fácil para empezar, dominar todas sus clases, componentes y el sistema de rejilla requiere un <mark>aprendizaje</mark> inicial.

## 1_3
## Novedades Principales (BS4 vs BS3)

Bootstrap 4 fue una reescritura importante. Si vienes de Bootstrap 3, estas son algunas diferencias clave:

*   **Flexbox por Defecto:** El sistema de rejilla y muchos componentes ahora usan <mark>Flexbox</mark>, ofreciendo capacidades de alineación y distribución mucho más potentes y flexibles que los floats usados en BS3.
*   **Sass en lugar de Less:** El CSS fuente se reescribió usando <mark>Sass</mark> (SCSS), facilitando la personalización a través de variables y mixins.
*   **Nueva Unidad de Medida:** Se prioriza el uso de <mark>`rem` sobre `px`</mark> para una mejor escalabilidad y responsiveness tipográfica.
*   **Sistema de Rejilla Mejorado:** Se añadió un <mark>nuevo breakpoint (`xl`)</mark> para pantallas extra grandes. Clases de rejilla más intuitivas.
*   **Reboot:** Un nuevo módulo <mark>(`reboot.css`)</mark> basado en Normalize.css para establecer estilos base consistentes entre navegadores.
*   **Componentes Actualizados:**
    *   **Tarjetas (Cards):** Reemplazan a los antiguos Panels, Wells y Thumbnails, ofreciendo un componente mucho <mark>más flexible</mark>.
    *   **Navbar:** Reescriba completa con más opciones de <mark>personalización</mark> y mejor <mark>soporte responsive</mark>.
    *   **Formularios:** Mejoras en la estructura y introducción de <mark>formularios</mark> "custom" para una apariencia más consistente entre navegadores.
*   **Clases de Utilidad:** Se expandió enormemente el <mark>número de clases</mark> de utilidad (spacing, display, flex, borders, etc.), permitiendo construir layouts y aplicar estilos rápidamente sin CSS personalizado.
*   **Sin Glyphicons:** Los iconos <mark>Glyphicons ya no</mark> se incluyen por defecto. Se recomienda usar librerías externas como Font Awesome, Material Design Icons, o SVGs.
*   **JavaScript:** Todos los plugins JS se reescribieron en ES6 y requieren jQuery y Popper.js.

En resumen, BS4 es más moderno, potente y flexible que BS3, especialmente gracias a Flexbox y las utilidades.

### 1_4_1
### Opciones de Inclusión (CDN, Descarga, Gestor Paquetes)

Tienes varias formas de <mark>añadir</mark> Bootstrap a tu proyecto:

1.  **CDN (Content Delivery Network) - La forma <mark>más rápida y fácil</mark> para empezar:**
    *   Simplemente <mark>enlaza</mark> los archivos CSS y JS de Bootstrap directamente desde una CDN pública en tu HTML. Bootstrap y otras CDNs (como jsDelivr, cdnjs) alojan los archivos compilados.
    *   **Ventajas:** No necesitas descargar ni gestionar archivos. Los usuarios pueden tener los archivos cacheados en su navegador si han visitado otros sitios que usan la misma CDN.
    *   **Desventajas:** <mark>Requiere conexión a internet</mark> para funcionar. Menos control sobre la versión exacta si no especificas una.
    *   **Ejemplo (usando la CDN oficial de Bootstrap 4.6):**
        ```html
        <!-- CSS -->
        <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css" integrity="sha384-xOolHFLEh07PJGoPkLv1IbcEPTNtaed2xpHsD9ESMhqIYd0nLMwNLD69Npy4HI+N" crossorigin="anonymous">

        <!-- JS, Popper.js, y jQuery (al final del body) -->
        <script src="https://cdn.jsdelivr.net/npm/jquery@3.5.1/dist/jquery.slim.min.js" integrity="sha384-DfXdz2htPH0lsSSs5nCTpuj/zy4C+OGpamoFVy38MVBnE+IbbVYUew+OrCXaRkfj" crossorigin="anonymous"></script>
        <script src="https://cdn.jsdelivr.net/npm/popper.js@1.16.1/dist/umd/popper.min.js" integrity="sha384-9/reFTGAW83EW2RDu2S0VKaIzap3H66lZH81PoYlFhbGU+6BZp6G7niu735Sk7lN" crossorigin="anonymous"></script>
        <script src="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/js/bootstrap.min.js" integrity="sha384-+sLIOodYLS7CIrQpBjl+C7nPvqq+FbNUBDunl/OZv93DB7Ln/533i8e/mZXLi/P+" crossorigin="anonymous"></script>
        ```
        *(Siempre verifica las URLs y las sumas de integridad (integrity) en la [documentación oficial de Bootstrap](https://getbootstrap.com/docs/4.6/getting-started/introduction/) para la versión que uses).*

2.  **Descarga de Archivos Compilados:**
    *   Descarga el <mark>paquete `.zip` compilado</mark> desde el sitio oficial de Bootstrap.
    *   Descomprime y copia los directorios `css/` y `js/` a tu proyecto.
    *   Enlaza los archivos locales en tu HTML.
    *   **Ventajas:** <mark>Funciona offline</mark>. Tienes los archivos específicos.
    *   **Desventajas:** Necesitas gestionar los archivos y actualizarlos manualmente.
    *   **Ejemplo:**
        ```html
        <!-- Asumiendo que copiaste los archivos a una carpeta 'vendor/bootstrap' -->
        <link rel="stylesheet" href="vendor/bootstrap/css/bootstrap.min.css">
        <!-- ... -->
        <script src="vendor/jquery/jquery.slim.min.js"></script> {# Necesitas descargar jQuery y Popper también #}
        <script src="vendor/popper/popper.min.js"></script>
        <script src="vendor/bootstrap/js/bootstrap.min.js"></script>
        ```

3.  **Gestor de Paquetes (npm, Yarn):**
    *   Si estás trabajando en un proyecto con Node.js (aunque sea solo para herramientas de frontend), puedes <mark>instalar Bootstrap usando npm o Yarn</mark>.
    *   `npm install bootstrap@4.6 jquery popper.js`
    *   Luego necesitarás un proceso de *build* (como Webpack, Parcel) para incluir el CSS y JS en tu proyecto final, o enlazar directamente desde `node_modules/` (menos común para producción).
    *   **Ventajas:** Gestión de dependencias fácil, integración con herramientas de build, permite usar los fuentes Sass para personalizar.
    *   **Desventajas:** Requiere un entorno Node.js y herramientas de build. Curva de aprendizaje mayor.

**Para este curso, generalmente usaremos la opción CDN por simplicidad**, pero es bueno conocer las otras.

### 1_4_2
### Plantilla HTML Inicial

Bootstrap recomienda una <mark>plantilla HTML5 básica</mark> para empezar, asegurando que incluyas el viewport meta tag para el diseño responsive y los enlaces a los archivos CSS y JS.

```html
<!doctype html>
<html lang="es">
  <head>
    <!-- Metatags requeridos -->
    <meta charset="utf-8">
    <meta name="viewport" content="width=device-width, initial-scale=1, shrink-to-fit=no">

    <!-- CSS de Bootstrap (vía CDN ejemplo) -->
    <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css" integrity="sha384-xOolHFLEh07PJGoPkLv1IbcEPTNtaed2xpHsD9ESMhqIYd0nLMwNLD69Npy4HI+N" crossorigin="anonymous">

    <title>¡Hola, mundo!</title>
  </head>
  <body>
    <h1>¡Hola, mundo con Bootstrap 4!</h1>

    <!-- Contenido de tu página aquí -->


    <!-- JavaScript Opcional; elige uno de los dos! -->

    <!-- Opción 1: jQuery y Bootstrap Bundle (incluye Popper) -->
    <script src="https://cdn.jsdelivr.net/npm/jquery@3.5.1/dist/jquery.slim.min.js" integrity="sha384-DfXdz2htPH0lsSSs5nCTpuj/zy4C+OGpamoFVy38MVBnE+IbbVYUew+OrCXaRkfj" crossorigin="anonymous"></script>
    <script src="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/js/bootstrap.bundle.min.js" integrity="sha384-Fy6S3B9q64WdZWQUiU+q4/2Lc9npb8tCaSX9FK7E8HnRr0Jz8D6OP9dO5Vg3Q9ct" crossorigin="anonymous"></script>

    <!-- Opción 2: jQuery, Popper.js, y Bootstrap JS por separado
    <script src="https://cdn.jsdelivr.net/npm/jquery@3.5.1/dist/jquery.slim.min.js" integrity="sha384-DfXdz2htPH0lsSSs5nCTpuj/zy4C+OGpamoFVy38MVBnE+IbbVYUew+OrCXaRkfj" crossorigin="anonymous"></script>
    <script src="https://cdn.jsdelivr.net/npm/popper.js@1.16.1/dist/umd/popper.min.js" integrity="sha384-9/reFTGAW83EW2RDu2S0VKaIzap3H66lZH81PoYlFhbGU+6BZp6G7niu735Sk7lN" crossorigin="anonymous"></script>
    <script src="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/js/bootstrap.min.js" integrity="sha384-+sLIOodYLS7CIrQpBjl+C7nPvqq+FbNUBDunl/OZv93DB7Ln/533i8e/mZXLi/P+" crossorigin="anonymous"></script>
    -->
  </body>
</html>
```
*   `<!doctype html>` y `<html lang="es">`: Estándar HTML5.
*   `<meta charset="utf-8">`: Esencial para la codificación de caracteres.
*   `<meta name="viewport" ...>`: **Crucial** para el diseño responsive. `width=device-width` ajusta el ancho al del dispositivo, `initial-scale=1` establece el zoom inicial, `shrink-to-fit=no` evita comportamientos extraños de escalado en algunos navegadores móviles (especialmente iOS).
*   Enlace al CSS de Bootstrap en el `<head>`.
*   Scripts JS (jQuery, Popper, Bootstrap) **al final del `<body>`**. Esto permite que el HTML se parsee y renderice antes de cargar/ejecutar los scripts, mejorando la percepción de velocidad de carga. La opción del "Bundle" (`bootstrap.bundle.min.js`) es conveniente ya que incluye Popper.js.

### 1_4_3
### Dependencias JS (jQuery y Popper.js)

Muchos <mark>componentes interactivos</mark> de Bootstrap 4 (desplegables, modales, tooltips, popovers, carruseles, colapsables) **requieren <mark>JavaScript</mark> para funcionar**. Específicamente, dependen de dos librerías externas:

1.  **jQuery:** Una librería JavaScript muy popular (especialmente en la época de BS4) que simplifica la manipulación del <mark>DOM</mark>, el manejo de eventos y las peticiones AJAX. Bootstrap 4 usa jQuery para sus plugins. Necesitas incluir **jQuery antes** que el archivo `bootstrap.js` o `bootstrap.bundle.js`. Generalmente se usa la versión "slim" que excluye módulos AJAX y efectos (no necesarios para los plugins de Bootstrap).
2.  **Popper.js:** Una librería para gestionar el posicionamiento de <mark>elementos emergentes</mark> (tooltips, popovers, desplegables). Es necesaria para que estos componentes se ubiquen correctamente en relación a su elemento disparador. Debe incluirse **antes** que `bootstrap.js`, pero **después** de jQuery. Si usas `bootstrap.bundle.js`, Popper.js ya está incluido ahí.

**Orden de Inclusión de Scripts (al final del `<body>`):**

*   **Opción A (Bundle):**
    1.  jQuery (slim.min.js)
    2.  Bootstrap Bundle (bootstrap.bundle.min.js)
*   **Opción B (Separados):**
    1.  jQuery (slim.min.js)
    2.  Popper.js (popper.min.js)
    3.  Bootstrap JS (bootstrap.min.js)

Si **no** planeas usar componentes que requieran JavaScript (como desplegables, modales, etc.), *técnicamente* podrías omitir los archivos JS y sus dependencias, usando Bootstrap solo por su CSS (layout y estilos). Sin embargo, la mayoría de los sitios necesitarán al menos alguna de estas funcionalidades interactivas.

*(Nota: Bootstrap 5 eliminó la dependencia de jQuery y usa Popper.js v2, simplificando las dependencias).*

# 2
# Sistema de Layout (Grid y Contenedores)

El sistema de <mark>layout</mark> es uno de los pilares de Bootstrap. Permite organizar el contenido de tu página en <mark>filas y columnas</mark> de manera responsive, asegurando que tu diseño se vea bien en cualquier tamaño de pantalla. En Bootstrap 4, este sistema se basa principalmente en **Flexbox** y utiliza **contenedores**, **filas** y **columnas**.

[Volver a la tabla de contenido](#Contenido)
## 2_1
## Contenedores (`container`, `container-fluid`)

Los <mark>contenedores</mark> son el elemento **fundamental y necesario** para envolver el contenido de tu sitio y el sistema de rejilla. Bootstrap ofrece dos tipos principales:

1.  **`.container`:**
    *   Proporciona un contenedor de **ancho fijo y responsive**.
    *   Su ancho máximo (`max-width`) cambia en cada breakpoint (sm, md, lg, xl), manteniendo márgenes a los lados en pantallas más grandes.
    *   Centra el contenido horizontalmente.
    *   Es el tipo de contenedor más común para layouts estándar.

    ```html
    <div class="container">
      <!-- Tu contenido y sistema de rejilla (filas/columnas) van aquí -->
      <p>Este contenido está dentro de un contenedor de ancho fijo.</p>
    </div>
    ```

2.  **`.container-fluid`:**
    *   Proporciona un contenedor de **ancho completo**.
    *   Ocupa siempre el 100% del ancho de la ventana gráfica (viewport), sin importar el tamaño de la pantalla.
    *   Útil para secciones que deben abarcar todo el ancho, como cabeceras, pies de página o secciones de fondo completo.

    ```html
    <div class="container-fluid">
      <!-- Contenido que ocupa todo el ancho -->
      <p>Este contenido está dentro de un contenedor fluido.</p>
    </div>
    ```

**¿Cuándo usar cuál?**
*   Usa <mark>`.container`</mark> para la mayoría del contenido principal de tu página, para darle <mark>márgenes</mark> y un ancho máximo controlado en pantallas grandes.
*   Usa <mark>`.container-fluid`</mark> para elementos que necesiten extenderse <mark>de borde a borde</mark> de la pantalla.
*   Puedes anidar contenedores o usar ambos tipos en diferentes secciones de la misma página.

**Importante:** El <mark>sistema de rejilla</mark> de Bootstrap (`<div class="row">` y `<div class="col-...">`) **debe** colocarse <mark>dentro</mark> de un `.container` o `.container-fluid`. No pongas filas directamente en el `<body>`.

[Volver a la tabla de contenido](#Contenido)
## 2_2
## Sistema de Rejilla (Grid) basado en Flexbox

El corazón del layout en Bootstrap 4 es su potente <mark>sistema de rejilla</mark> de **12 columnas**, construido con **Flexbox**. Permite crear layouts complejos y adaptables de forma sencilla y semántica.

[Volver a la tabla de contenido](#Contenido)
### 2_2_1
### ¿Cómo funciona? (Filas `row` y Columnas `col`)

La estructura básica del grid es:

1.  <mark>**Contenedor (`.container` o `.container-fluid`):**</mark> Necesario para envolver y alinear el grid.
2.  <mark>**Fila (`<div class="row">`):**</mark> Actúa como un contenedor *Flexbox* horizontal para las columnas. Las filas eliminan los márgenes negativos del contenedor y proporcionan *gutters* (espaciado) entre las columnas a través de padding en las propias columnas. **Las columnas deben ser hijas directas de una fila.**
3.  <mark>**Columna (`<div class="col...">`):**</mark> Son los elementos que contienen tu contenido real. Deben sumar hasta 12 (o menos) por cada fila para un layout estándar. Bootstrap proporciona clases para definir el ancho de las columnas en diferentes tamaños de pantalla.

**Ejemplo Simple (3 columnas iguales):**
```html
<div class="container">
  <div class="row">
    <div class="col"> {# Columna 1 (ocupa 1/3 del espacio disponible) #}
      Contenido Columna 1
    </div>
    <div class="col"> {# Columna 2 (ocupa 1/3 del espacio disponible) #}
      Contenido Columna 2
    </div>
    <div class="col"> {# Columna 3 (ocupa 1/3 del espacio disponible) #}
      Contenido Columna 3
    </div>
  </div>
  <div class="row">
    <div class="col-8"> {# Columna que ocupa 8 de 12 unidades #}
       Contenido 8/12
    </div>
    <div class="col-4"> {# Columna que ocupa 4 de 12 unidades #}
       Contenido 4/12
    </div>
  </div>
</div>
```
*   Usar solo `class="col"` dentro de una fila hace que Bootstrap distribuya el ancho <mark>equitativamente</mark> entre las columnas presentes en esa fila (gracias a Flexbox).
*   Usar `class="col-N"` (donde N es un número de 1 a 12) asigna explícitamente N unidades de las 12 disponibles a esa columna.

[Volver a la tabla de contenido](#Contenido)
### 2_2_2
### Opciones de Rejilla (Breakpoints: xs, sm, md, lg, xl)

<mark>Bootstrap es *mobile-first*.</mark> Los estilos y layouts definidos sin un breakpoint específico (`col-*`, `col-N`) aplican a **todos** los tamaños de pantalla, desde el más pequeño (`xs`) hacia arriba.

Para crear layouts *responsive*, Bootstrap define **breakpoints** (puntos de quiebre) basados en el ancho del viewport, y proporciona infijos de clase para aplicar estilos o anchos de columna específicos *a partir* de ese breakpoint:

*   **`xs` (Extra small):** `< 576px` (Implícito, no hay infijo `xs`, ej: `col-6` aplica a `xs` y superiores)
*   **`sm` (Small):** `≥ 576px` (Infijo: `sm`, ej: `col-sm-6`)
*   **`md` (Medium):** `≥ 768px` (Infijo: `md`, ej: `col-md-4`)
*   **`lg` (Large):** `≥ 992px` (Infijo: `lg`, ej: `col-lg-3`)
*   **`xl` (Extra large):** `≥ 1200px` (Infijo: `xl`, ej: `col-xl-2`)

**¿Cómo se usan?**
Combinas estas clases en tus columnas para definir cómo deben comportarse en diferentes tamaños de pantalla.

**Ejemplo:** Una barra lateral y contenido principal.
*   En pantallas pequeñas (`xs`), queremos que la barra lateral y el contenido ocupen todo el ancho (12 columnas), apilados verticalmente.
*   En pantallas medianas (`md`) y superiores, queremos que la barra lateral ocupe 4 columnas y el contenido principal 8 columnas.

```html
<div class="container">
  <div class="row">
    {!-- Columna de Contenido Principal --}
    {# Ocupa 12 columnas en xs (implícito), 8 columnas desde md hacia arriba #}
    <div class="col-12 col-md-8">
      <h2>Contenido Principal</h2>
      <p>Este es el contenido principal...</p>
    </div>

    {!-- Columna de Barra Lateral --}
    {# Ocupa 12 columnas en xs (implícito), 4 columnas desde md hacia arriba #}
    <div class="col-12 col-md-4">
      <h3>Barra Lateral</h3>
      <p>Widgets, enlaces, etc.</p>
    </div>
  </div>
</div>
```
*   En pantallas `< 768px`, ambas `div` tendrán `col-12` (ancho completo), apilándose.
*   En pantallas `≥ 768px`, la primera `div` tomará `col-md-8` (8/12) y la segunda `col-md-4` (4/12), colocándose una al lado de la otra.

Puedes combinar múltiples clases de breakpoint para un control más fino (ej: `col-12 col-sm-6 col-lg-4`).

[Volver a la tabla de contenido](#Contenido)
### 2_2_3
### Columnas de Ancho Automático

Si usas solo la clase `col` (sin número o breakpoint), Bootstrap distribuirá el espacio horizontal disponible equitativamente entre todas las columnas `col` dentro de esa fila.

```html
<div class="container">
  <div class="row">
    <div class="col">Columna 1 (Ancho Automático)</div>
    <div class="col">Columna 2 (Ancho Automático)</div>
  </div>
  <div class="row">
    <div class="col">Columna 1 (Ancho Automático)</div>
    <div class="col">Columna 2 (Ancho Automático)</div>
    <div class="col">Columna 3 (Ancho Automático)</div>
  </div>
</div>
```
En la primera fila, cada columna ocupará el 50%. En la segunda, cada una ocupará el 33.33%.

También puedes usar <mark>`col-auto`</mark>. Esta clase hace que la columna ocupe solo el espacio necesario para su contenido (basado en el ancho intrínseco del contenido).

```html
<div class="container">
  <div class="row">
    <div class="col">Columna Principal (ocupa espacio restante)</div>
    <div class="col-auto">Contenido Corto</div> {# Ocupa solo lo necesario #}
    <div class="col-auto"><button class="btn btn-sm btn-info">Botón</button></div> {# Ocupa solo lo necesario #}
  </div>
</div>
```
Esto es útil para alinear elementos de ancho variable junto a otros que deben ocupar el espacio restante.

Puedes combinar `col` y `col-auto` con clases de ancho específico (`col-N`, `col-BP-N`) en la misma fila.

[Volver a la tabla de contenido](#Contenido)
### 2_2_4
### Estableciendo Ancho de Columna (`col-*`, `col-BP-*`)

Como vimos, puedes <mark>definir explícitamente</mark> cuántas de las 12 unidades de la rejilla debe ocupar una columna usando las clases `col-N` (para `xs` y superiores) o `col-BP-N` (para un breakpoint específico y superiores).

*   `col-6`: Ocupa 6/12 (la mitad) en todas las pantallas.
*   `col-sm-4`: Ocupa 4/12 (un tercio) en pantallas `sm` (≥576px) y superiores. En `xs` (<576px), ocupará el 100% por defecto.
*   `col-12 col-md-6 col-lg-4`:
    *   `xs` (<768px): Ocupa 12/12 (ancho completo).
    *   `md` (≥768px): Ocupa 6/12 (la mitad).
    *   `lg` (≥992px): Ocupa 4/12 (un tercio).

**Ejemplo de Galería:**
```html
<div class="container">
  <div class="row">
    {# En pantallas pequeñas, 1 columna. Medias, 2 columnas. Grandes, 3 columnas. #}
    <div class="col-12 col-md-6 col-lg-4 mb-3"> {# mb-3 añade margen inferior #}
      <img src="..." class="img-fluid" alt="...">
    </div>
    <div class="col-12 col-md-6 col-lg-4 mb-3">
      <img src="..." class="img-fluid" alt="...">
    </div>
    <div class="col-12 col-md-6 col-lg-4 mb-3">
      <img src="..." class="img-fluid" alt="...">
    </div>
    {# ... más columnas ... #}
  </div>
</div>
```
Recuerda que las <mark>clases de breakpoint</mark>aplican desde ese tamaño **hacia arriba**, a menos que sean anuladas por una clase de un breakpoint mayor.

### 2_2_5
### Anidamiento de Columnas

Puedes colocar una nueva `<div class="row">` **dentro** de una `<div class="col-...">` existente para crear <mark>rejillas anidadas</mark>. Las columnas dentro de la fila anidada volverán a sumar 12 unidades relativas al ancho de la columna padre que las contiene.

```html
<div class="container">
  <div class="row">
    <div class="col-sm-9">
      Contenido Principal (9 columnas)
      {# Fila anidada dentro de la columna principal #}
      <div class="row">
        <div class="col-8 col-sm-6"> {# Ocupa 6/12 de las 9 columnas padre en sm+ #}
          Sub-columna A
        </div>
        <div class="col-4 col-sm-6"> {# Ocupa 6/12 de las 9 columnas padre en sm+ #}
          Sub-columna B
        </div>
      </div>
    </div>
    <div class="col-sm-3">
      Barra Lateral (3 columnas)
    </div>
  </div>
</div>
```
El anidamiento permite crear layouts más complejos y estructurados.

## 2_3
## Alineación de Columnas

Dado que las filas (`.row`) son contenedores Flexbox, puedes usar las <mark>utilidades de alineación</mark> de Flexbox para controlar cómo se posicionan las columnas dentro de la fila, tanto vertical como horizontalmente. Estas clases se aplican a la **fila (`.row`)** o a las **columnas individuales (`.col-`)**.

[Volver a la tabla de contenido](#Contenido)
### 2_3_1
### Alineación Vertical (`align-items-*`)

Controla cómo se <mark>alinean</mark> **todas** las <mark>columnas</mark> <mark>dentro</mark> de una <mark>fila</mark> en el eje vertical (cuando las columnas tienen diferente altura). Se aplica a la **fila (`.row`)**.

*   `align-items-start`: Alinea las columnas en la parte superior de la fila.
*   `align-items-center`: Centra las columnas verticalmente en la fila.
*   `align-items-end`: Alinea las columnas en la parte inferior de la fila.
*   `align-items-baseline`: Alinea las columnas según su línea base de texto.
*   `align-items-stretch` (Comportamiento por defecto): Estira las columnas para que tengan la misma altura (la de la más alta).

```html
<div class="container mt-4">
  <h4>Alineación Vertical de Columnas</h4>

  <p><strong>align-items-start:</strong></p>
  <div class="row align-items-start bg-light mb-3" style="min-height: 150px; border: 1px solid #ccc;">
    <div class="col pt-2 pb-2 border">Columna con poco contenido.</div>
    <div class="col pt-2 pb-2 border">Columna con más contenido que la anterior y un poco más para ver la diferencia.</div>
    <div class="col pt-2 pb-2 border">Otra columna corta.</div>
  </div>

  <p><strong>align-items-center:</strong></p>
  <div class="row align-items-center bg-light mb-3" style="min-height: 150px; border: 1px solid #ccc;">
    <div class="col pt-2 pb-2 border">Columna con poco contenido.</div>
    <div class="col pt-2 pb-2 border">Columna con más contenido que la anterior y un poco más para ver la diferencia.</div>
    <div class="col pt-2 pb-2 border">Otra columna corta.</div>
  </div>

  <p><strong>align-items-end:</strong></p>
  <div class="row align-items-end bg-light mb-3" style="min-height: 150px; border: 1px solid #ccc;">
    <div class="col pt-2 pb-2 border">Columna con poco contenido.</div>
    <div class="col pt-2 pb-2 border">Columna con más contenido que la anterior y un poco más para ver la diferencia.</div>
    <div class="col pt-2 pb-2 border">Otra columna corta.</div>
  </div>

  <p><strong>align-items-stretch (comportamiento por defecto):</strong></p>
  <div class="row bg-light mb-3" style="min-height: 100px; border: 1px solid #ccc;"> {# No se necesita align-items-stretch explícitamente si es el default #}
    <div class="col pt-2 pb-2 border">Columna A</div>
    <div class="col pt-2 pb-2 border">Columna B con <br> más altura</div>
    <div class="col pt-2 pb-2 border">Columna C</div>
  </div>

  <p><strong>align-items-baseline (útil con diferentes tamaños de texto):</strong></p>
  <div class="row align-items-baseline bg-light" style="border: 1px solid #ccc;">
    <div class="col pt-2 pb-2 border"><h3>Título</h3></div>
    <div class="col pt-2 pb-2 border">Texto normal</div>
    <div class="col pt-2 pb-2 border"><small>Texto pequeño</small></div>
  </div>
</div>
```

Al visualizar este HTML en un navegador, observarás cómo las columnas se alinean verticalmente dentro de cada fila según la clase `align-items-*` aplicada a la `div.row`. La clase `bg-light` y `border` son solo para ayudar a visualizar las cajas.

### 2_3_2
### Alineación Horizontal (`justify-content-*`)

Controla cómo se distribuyen las <mark>columnas</mark> horizontalmente <mark>dentro</mark> de una <mark>fila</mark> cuando el ancho total de las columnas es menor que el ancho de la fila. Se aplica a la **fila (`.row`)**.

*   `justify-content-start` (Comportamiento por defecto): Agrupa las columnas al inicio (izquierda en LTR).
*   `justify-content-end`: Agrupa las columnas al final (derecha en LTR).
*   `justify-content-center`: Centra las columnas en la fila.
*   `justify-content-between`: Distribuye el espacio equitativamente *entre* las columnas (primera al inicio, última al final).
*   `justify-content-around`: Distribuye el espacio equitativamente *alrededor* de las columnas (mitad de espacio en los extremos).

```html
<div class="container">
  {# Fila con 2 columnas de 4 unidades cada una (total 8 < 12) #}
  <div class="row justify-content-center" style="background-color: #e9ecef; padding: 10px 0;"> {# Centrar #}
    <div class="col-4">Columna 1 (4)</div>
    <div class="col-4">Columna 2 (4)</div>
  </div>
  <div class="row justify-content-between mt-3" style="background-color: #e9ecef; padding: 10px 0;"> {# Espacio entre #}
    <div class="col-4">Columna 1 (4)</div>
    <div class="col-4">Columna 2 (4)</div>
  </div>
</div>
```

### 2_3_3
### Alineación Individual (`align-self-*`)

Permite <mark>anular la alineación vertical</mark> (`align-items`) definida en la fila para una **columna específica**. Se aplica a la **columna (`.col-`)**.

Acepta los mismos valores que `align-items`: `start`, `center`, `end`, `baseline`, `stretch`.

```html
<div class="container">
  <div class="row align-items-start" style="height: 100px; background-color: #f8f9fa;"> {# Default arriba #}
    <div class="col">Alineada Arriba</div>
    <div class="col align-self-center">Alineada al Centro (Individual)</div> {# Anula align-items-start #}
    <div class="col align-self-end">Alineada Abajo (Individual)</div> {# Anula align-items-start #}
  </div>
</div>
```

[Volver a la tabla de contenido](#Contenido)
### 2_3_4
### Sin Gutters (`no-gutters`)

Por defecto, las filas (`.row`) tienen márgenes negativos y las columnas (`.col-`) tienen padding horizontal para crear el espaciado (gutter) entre columnas.

Si quieres <mark>eliminar</mark> este <mark>espaciado</mark> (hacer que el contenido de las columnas se toque), añade la clase `.no-gutters` a la **fila (`.row`)**.

```html
<div class="container">
  <div class="row no-gutters"> {# Eliminar espaciado entre estas columnas #}
    <div class="col-sm-6" style="border: 1px solid red;">Contenido A (sin gutter)</div>
    <div class="col-sm-6" style="border: 1px solid blue;">Contenido B (sin gutter)</div>
  </div>
  <div class="row mt-3"> {# Fila normal con gutters #}
     <div class="col-sm-6" style="border: 1px solid red;">Contenido C (con gutter)</div>
     <div class="col-sm-6" style="border: 1px solid blue;">Contenido D (con gutter)</div>
  </div>
</div>
```
Verás que el borde entre A y B se toca, mientras que entre C y D hay un espacio.

## 2_4
## Reordenamiento_y_Desplazamiento

Flexbox (y por tanto el grid de Bootstrap) permite <mark>cambiar el orden visual</mark> de las columnas y crear espacios vacíos (offsets) sin alterar el orden del HTML.

[Volver a la tabla de contenido](#Contenido)
### 2_4_1
### Reordenamiento (`order-*`)

Puedes cambiar el orden visual de las columnas usando las clases `order-*`. Por defecto, todas las columnas tienen `order: 0`.
*   `order-N`: (N de 1 a 12) Establece un orden específico. Números menores aparecen antes.
*   `order-first`: Equivalente a un orden muy bajo (aparece primero).
*   `order-last`: Equivalente a un orden alto (aparece último).
*   También existen versiones responsivas: `order-BP-N`, `order-BP-first`, `order-BP-last` (ej: `order-md-1`, `order-sm-last`).

```html
<div class="container">
  <div class="row">
    <div class="col order-3">Columna A (HTML) - Orden 3 (Visual)</div>
    <div class="col order-1">Columna B (HTML) - Orden 1 (Visual)</div>
    <div class="col order-2">Columna C (HTML) - Orden 2 (Visual)</div>
  </div>

  <div class="row mt-3">
    <div class="col order-last order-md-first">Aparece último en móvil, primero en MD+</div>
    <div class="col">Orden normal</div>
    <div class="col order-first order-md-last">Aparece primero en móvil, último en MD+</div>
  </div>
</div>
```

[Volver a la tabla de contenido](#Contenido)
### 2_4_2
### Desplazamiento (Offsetting) (`offset-*`)

Puedes "<mark>empujar</mark>" <mark>columnas</mark> hacia la <mark>derecha</mark> creando un espacio vacío a su izquierda usando las clases `offset-BP-N`. Esto mueve la columna N unidades de grid hacia la derecha en el breakpoint BP y superiores.

```html
<div class="container">
  <div class="row">
    <div class="col-md-4">.col-md-4</div>
    {# Empuja esta columna 4 unidades a la derecha en MD y superiores #}
    <div class="col-md-4 offset-md-4">.col-md-4 .offset-md-4</div>
  </div>
  <div class="row">
    {# Mueve 3 a la derecha en MD #}
    <div class="col-md-3 offset-md-3">.col-md-3 .offset-md-3</div>
    {# Mueve 3 más a la derecha (total 6 desde inicio) en MD #}
    <div class="col-md-3 offset-md-3">.col-md-3 .offset-md-3</div>
  </div>
   <div class="row">
     {# Centrar una columna de 6: 12 - 6 = 6. Mitad a cada lado = offset 3 #}
    <div class="col-md-6 offset-md-3">.col-md-6 .offset-md-3 (Centrada)</div>
  </div>
</div>
```
Los <mark>offsets</mark> son útiles para <mark>crear espaciado entre columnas sin usar columnas vacías.</mark>

1.  **Layout Básico (Barra Lateral + Contenido):**
    *   Crea una estructura dentro de un `.container`.
    *   En pantallas pequeñas (`xs`), el contenido principal debe ocupar todo el ancho, y la barra lateral debe aparecer debajo, también ocupando todo el ancho.
    *   En pantallas medianas (`md`) y superiores, el contenido principal debe ocupar 8 columnas y la barra lateral 4 columnas, una al lado de la otra.
    *   Añade algo de texto placeholder en cada columna.

2.  **Layout de 3 Columnas Iguales:**
    *   Dentro de un `.container`, crea una fila.
    *   Haz que en pantallas pequeñas (`xs`), cada una de las 3 columnas ocupe la mitad del ancho (se apilarán 2 y 1).
    *   En pantallas `md` y superiores, haz que las 3 columnas ocupen un tercio del ancho cada una (quedando en una sola fila).

    *   Crea una fila principal con 2 columnas (ej: `col-md-8` y `col-md-4`).

# 3
# Contenido y Tipografía

Bootstrap no solo proporciona un sistema de layout, sino que también aplica <mark>estilos base</mark> y ofrece clases adicionales para dar formato al contenido textual común de una página web, como encabezados, párrafos, listas, código, imágenes y tablas. Estos estilos base se encuentran en el <mark>módulo "Reboot"</mark> de Bootstrap, que normaliza el renderizado de HTML entre diferentes navegadores.

[Volver a la tabla de contenido](#Contenido)
## 3_1
## Encabezados (h1-h6, Clases `display-*`)

*   **Encabezados Estándar:** Bootstrap aplica estilos básicos (tamaño de fuente, margen inferior) a los encabezados HTML estándar (`<h1>` a `<h6>`). Puedes usarlos directamente.

    ```html
    <h1>Encabezado h1 Bootstrap</h1>
    <h2>Encabezado h2 Bootstrap</h2>
    <h3>Encabezado h3 Bootstrap</h3>
    <h4>Encabezado h4 Bootstrap</h4>
    <h5>Encabezado h5 Bootstrap</h5>
    <h6>Encabezado h6 Bootstrap</h6>
    ```
*   **Clases de Encabezado:** Si necesitas que un elemento que *no* es un encabezado semántico *parezca* un encabezado, puedes usar las clases `.h1` a `.h6`.
    ```html
    <p class="h1">Párrafo que parece un h1</p>
    <p class="h4">Párrafo que parece un h4</p>
    ```
*   **Personalización de Encabezados:** Puedes añadir texto secundario más apagado dentro de un encabezado usando `<small>` o la clase `.text-muted`.
    ```html
    <h3>Encabezado con <small class="text-muted">texto secundario</small></h3>
    ```
*   **Clases Display (`display-*`):** Para encabezados mucho más grandes y destacados (típicos en secciones "hero" o títulos principales), Bootstrap ofrece las clases `display-1`, `display-2`, `display-3`, y `display-4`. Tienen un `font-weight` más ligero por defecto.
    ```html
    <h1 class="display-1">Display 1</h1>
    <h1 class="display-4">Display 4</h1>
    ```

[Volver a la tabla de contenido](#Contenido)
## 3_2
## Párrafos (`lead`, `blockquote`)

*   **Párrafos Estándar:** Los elementos `<p>` tienen un margen inferior por defecto (`margin-bottom: 1rem`).
*   **Clase `lead`:** Para destacar un párrafo introductorio, aplica la clase `.lead`. Esto le da un tamaño de fuente ligeramente mayor y un peso más ligero.
    ```html
    <p class="lead">
      Este es un párrafo destacado (lead). Ideal para introducciones cortas. Vivamus sagittis lacus vel augue laoreet rutrum faucibus dolor auctor.
    </p>
    <p>Este es un párrafo normal. Sigue al párrafo lead.</p>
    ```
*   **Citas en Bloque (`blockquote`):** Para marcar citas de otras fuentes.
    *   Envuelve la cita en `<blockquote class="blockquote">`.
    *   Usa `<p class="mb-0">` (margen inferior cero) para el texto de la cita si es un párrafo.
    *   Añade la fuente de la cita dentro de `<footer class="blockquote-footer">`. Puedes usar `<cite>` dentro del footer para el título de la fuente.
    ```html
    <blockquote class="blockquote">
      <p class="mb-0">Una cita memorable e inspiradora sobre diseño web.</p>
      <footer class="blockquote-footer">Alguien famoso en <cite title="Título de la Fuente">Título de la Fuente</cite></footer>
    </blockquote>
    ```
    *   Puedes alinear la cita a la derecha añadiendo `text-center` o `text-right` a la `<blockquote>`.

## 3_3
## Estilos de Texto Inline (`mark`, `small`, `strong`, `em`, `abbr`)

Bootstrap aplica estilos o permite usar etiquetas HTML estándar para modificar texto en línea:

*   **`<mark>` o `.mark`:** Resalta texto como si se hubiera usado un marcador.
    ```html
    Puedes usar la etiqueta <mark>o la clase .mark</mark> para resaltar.
    ```
*   **`<small>` o `.small`:** Crea texto secundario más pequeño (generalmente 85% del tamaño del padre). Útil para textos legales, leyendas, etc.
    ```html
    <p>Texto principal <small>y algo de texto pequeño.</small></p>
    ```
*   **`<strong>` o `.font-weight-bold`:** Texto en negrita (semánticamente importante).
*   **`<b>` o `.font-weight-bolder`:** Texto en negrita (sin énfasis semántico adicional respecto a strong).
*   **`<em>` o `.font-italic`:** Texto en cursiva (semánticamente enfatizado).
*   **`<i>` o `.font-italic`:** Texto en cursiva (sin énfasis semántico adicional).
*   **`<abbr>`:** Para abreviaturas y acrónimos. Añade un atributo `title` con la expansión completa, y Bootstrap añadirá un borde punteado sutil y un cursor de ayuda al pasar el ratón.
    ```html
    La <abbr title="Organización Mundial de la Salud">OMS</abbr> declaró... Necesitamos conocer el <abbr title="HyperText Markup Language">HTML</abbr>.
    ```
*   **`.text-decoration-none`:** Utilidad para quitar el subrayado (útil en enlaces).
*   **`.text-monospace`:** Aplica la pila de fuentes monoespaciadas de Bootstrap.

## 3_4
## Listas (`list-unstyled`, `list-inline`, `dl-horizontal`)

Bootstrap estiliza las listas ordenadas (`<ol>`), desordenadas (`<ul>`) y de definición (`<dl>`) de forma básica.

*   **Listas Estándar:** Tienen `margin-bottom` y los `<li>` tienen `margin-left` (o `padding-left` dependiendo de la configuración).
*   **`.list-unstyled`:** Elimina el `list-style` por defecto (viñetas/números) y el margen/padding izquierdo de los items de la lista (`<li>`). Útil para menús o listas que no necesitan marcadores.
    ```html
    <ul class="list-unstyled">
      <li>Item sin estilo</li>
      <li>Otro item
        <ul>
          <li>Item anidado (también sin estilo)</li>
        </ul>
      </li>
    </ul>
    ```
*   **`.list-inline` y `.list-inline-item`:** Muestra los elementos de la lista (`<li>`) en una sola línea horizontal (usando `display: inline-block;` y algo de padding). Aplica `.list-inline` al `<ul>` u `<ol>` y `.list-inline-item` a cada `<li>`.
    ```html
    <ul class="list-inline">
      <li class="list-inline-item">Item 1</li>
      <li class="list-inline-item">Item 2</li>
      <li class="list-inline-item">Item 3</li>
    </ul>
    ```
*   **Listas de Definición (`<dl>`) y Layout Horizontal:**
    *   Bootstrap estiliza `<dt>` (término) en negrita.
    *   Puedes usar las clases del sistema de rejilla (`row`, `col-sm-3`, `col-sm-9`) dentro de un `<dl>` para crear un layout horizontal donde los términos (`<dt>`) estén en una columna y las descripciones (`<dd>`) en otra.
    ```html
    <dl class="row">
      <dt class="col-sm-3">Título A</dt>
      <dd class="col-sm-9">Descripción para el título A.</dd>

      <dt class="col-sm-3">Título B</dt>
      <dd class="col-sm-9">
        <p>Descripción para B (puede tener párrafos).</p>
        <p>Otra línea de descripción.</p>
      </dd>

      <dt class="col-sm-3 text-truncate">Término Muy Largo Que Se Trunca</dt> {# Utilidad de Texto #}
      <dd class="col-sm-9">Descripción...</dd>
    </dl>
    ```

## 3_5
## Código (`code`, `kbd`, `pre`, `samp`)

Bootstrap proporciona estilos básicos para mostrar fragmentos de código.

*   **Código Inline (`<code>`):** Para fragmentos cortos de código dentro de texto normal. Bootstrap lo muestra en fuente monoespaciada y un color ligeramente diferente.
    ```html
    Usa la etiqueta <code>&lt;section&gt;</code> para contenido semántico.
    ```
*   **Entrada de Teclado (`<kbd>`):** Para indicar entrada de usuario típicamente introducida por teclado. Estilo ligeramente diferente a `<code>`.
    ```html
    Para guardar, presiona <kbd><kbd>Ctrl</kbd> + <kbd>S</kbd></kbd>.
    ```
    *Nota: Anidar `<kbd>` es una forma de representar combinaciones de teclas.*
*   **Bloques de Código (`<pre>`):** Para mostrar bloques de código de varias líneas, preservando espacios en blanco y saltos de línea. Bootstrap le aplica estilos básicos. Es recomendable envolver el código dentro de `<pre>` con una etiqueta `<code>` para una representación semántica correcta. Puedes añadir la clase `.pre-scrollable` a `<pre>` para darle un alto máximo y una barra de scroll vertical si el contenido es muy largo.
    ```html
    <pre><code>&lt;p&gt;Párrafo de ejemplo.&lt;/p&gt;
    &lt;div class="container"&gt;
      Contenido...
    &lt;/div&gt;</code></pre>

    <pre class="pre-scrollable"><code>// Código largo con scroll...
    // línea 1
    // línea 2
    // ...
    // línea N</code></pre>
    ```
*   **Variables (`<var>`):** Para indicar variables en expresiones matemáticas o de programación. (Bootstrap no aplica un estilo muy distintivo por defecto, a menudo es cursiva).
    ```html
    La ecuación es <var>y</var> = <var>m</var><var>x</var> + <var>b</var>.
    ```
*   **Salida de Programa (`<samp>`):** Para indicar salida de ejemplo de un programa o script. Estilo similar a `<code>`.
    ```html
    La salida fue: <samp>Proceso completado.</samp>
    ```

## 3_6
## Imágenes (`img-fluid`, `img-thumbnail`, alineación)

Bootstrap facilita el manejo de imágenes, especialmente para hacerlas responsive.

*   **Imágenes Responsive (`.img-fluid`):** **La clase más importante para imágenes.** Aplica `max-width: 100%;` y `height: auto;` a la imagen. Esto asegura que la imagen escale hacia abajo para ajustarse a su contenedor padre, pero nunca se haga más grande que su tamaño original. Esencial para que las imágenes no rompan el layout en pantallas pequeñas.
    ```html
    <img src="mi-imagen.jpg" class="img-fluid" alt="Imagen responsive">
    ```
*   **Miniaturas (`.img-thumbnail`):** Añade un borde redondeado sutil y un poco de padding a la imagen, dándole apariencia de miniatura.
    ```html
    <img src="avatar.jpg" class="img-thumbnail" alt="Avatar miniatura">
    ```
*   **Alineación de Imágenes:** Puedes alinear imágenes fácilmente usando las utilidades de **float** o las de **block y centrado con márgenes automáticos**.
    *   **Float:** Usa `.float-left` o `.float-right` para que la imagen flote a un lado y el texto la rodee. **Recuerda usar un clearfix** (ver sección 6.2) en el contenedor padre si todos sus hijos son flotantes.
        ```html
        <img src="..." class="float-left mr-3" alt="..."> {# mr-3 añade margen derecho #}
        <p>Texto que rodea la imagen flotante...</p>
        <div class="clearfix"></div> {# Si es necesario #}
        ```
    *   **Centrado (como bloque):** Para centrar una imagen, conviértela en un elemento de bloque y usa márgenes automáticos horizontales. Necesita `display: block`. Las clases son `.d-block` y `.mx-auto` (margen x automático).
        ```html
        <img src="..." class="d-block mx-auto" alt="Imagen centrada">
        ```
*   **Figuras:** (Ver sección 3.8) Usa `<figure>` y `<figcaption>` para agrupar imágenes con leyendas.

## 3_7
## Tablas (`table`, clases contextuales, responsive)

Bootstrap proporciona estilos base y clases adicionales para mejorar la apariencia y funcionalidad de las tablas HTML (`<table>`).

*   **Estilo Base (`.table`):** Añade esta clase a cualquier `<table>` para aplicar el formato básico de Bootstrap (padding ligero, bordes horizontales sutiles).
    ```html
    <table class="table">
      <thead>...</thead>
      <tbody>...</tbody>
    </table>
    ```
*   **Variantes de Estilo:**
    *   `.table-striped`: Añade franjas de cebra (fondo alterno) a las filas `<tbody>`.
    *   `.table-bordered`: Añade bordes a todos los lados de la tabla y las celdas.
    *   `.table-borderless`: Elimina todos los bordes.
    *   `.table-hover`: Habilita un estado hover (cambio de fondo) en las filas `<tbody>`.
    *   `.table-sm`: Hace la tabla más compacta reduciendo el padding de las celdas.
    ```html
    <table class="table table-striped table-hover table-bordered">
      ...
    </table>
    ```
*   **Estilos de Cabecera:**
    *   `.thead-dark`: Aplica un fondo oscuro a la cabecera (`<thead>`).
    *   `.thead-light`: Aplica un fondo claro a la cabecera.
*   **Clases Contextuales:** Puedes aplicar clases a las filas (`<tr>`) o a celdas individuales (`<td>`, `<th>`) para darles un color de fondo contextual (basado en los colores temáticos de Bootstrap).
    *   `.table-primary`, `.table-secondary`, `.table-success`, `.table-danger`, `.table-warning`, `.table-info`, `.table-light`, `.table-dark`, `.table-active`.
    ```html
    <tbody>
      <tr class="table-success">
        <td>Éxito</td>
        <td>Datos correctos</td>
      </tr>
      <tr class="table-danger">
        <td>Peligro</td>
        <td>Error encontrado</td>
      </tr>
      <tr>
        <td>Normal</td>
        <td class="table-warning">Advertencia Celda</td>
      </tr>
    </tbody>
    ```
*   **Tablas Responsive (`.table-responsive`):** Para tablas anchas que podrían desbordar contenedores pequeños, envuelve la `<table>` (con su clase `.table`) dentro de un `<div class="table-responsive">`. Esto añadirá una barra de scroll horizontal a la tabla *solo* cuando su ancho exceda el del contenedor, evitando que rompa el layout general. ## 4_1
## Alertas (Alerts) (colores, contenido, descarte)

Las alertas se usan para mostrar mensajes importantes o feedback al usuario (éxito, error, advertencia, información).

*   **Estructura Básica:** Un `<div>` con la clase base `.alert` y una clase contextual de color (`.alert-success`, `.alert-danger`, `.alert-warning`, `.alert-info`, `.alert-primary`, `.alert-secondary`, `.alert-light`, `.alert-dark`).
*   **Contenido:** Puedes incluir encabezados (`.alert-heading`), párrafos y enlaces (`.alert-link` para que el color del enlace coincida).
*   **Descarte (Requiere JS):** Para hacer una alerta descartable, añade la clase `.alert-dismissible` al `<div>` y un botón `<button type="button" class="close" data-dismiss="alert" aria-label="Close">` dentro. El atributo `data-dismiss="alert"` activa el comportamiento JavaScript.

```html
<div class="container mt-4">

  {# Alerta de éxito #}
  <div class="alert alert-success" role="alert">
    ¡Operación realizada con éxito!
  </div>

  {# Alerta de peligro con más contenido y enlace #}
  <div class="alert alert-danger" role="alert">
    <h4 class="alert-heading">¡Error Grave!</h4>
    <p>Ha ocurrido un problema inesperado al procesar tu solicitud. Por favor, inténtalo de nuevo más tarde.</p>
    <hr>
    <p class="mb-0">Si el problema persiste, <a href="#" class="alert-link">contacta con soporte</a>.</p>
  </div>

  {# Alerta de advertencia descartable #}
  <div class="alert alert-warning alert-dismissible fade show" role="alert">
    <strong>¡Atención!</strong> Deberías revisar algunos campos de tu formulario.
    <button type="button" class="close" data-dismiss="alert" aria-label="Close">
      <span aria-hidden="true">&times;</span> {# El símbolo 'x' #}
    </button>
  </div>

</div>
```
*   `role="alert"` es importante para la accesibilidad.
*   `.fade` y `.show` se usan junto con `.alert-dismissible` para añadir una animación de desvanecimiento al cerrar la alerta.

## Insignias (Badges) (contextuales, pills, en botones)

## 4_3
## Migas de Pan (Breadcrumb)

Los breadcrumbs (migas de pan) son una ayuda de navegación secundaria que indica la posición del usuario dentro de la jerarquía del sitio.

*   **Estructura:** Usa una lista ordenada (`<ol>`) o no ordenada (`<ul>`) con la clase `.breadcrumb`.
*   **Items:** Cada elemento de la ruta es un `<li>` con la clase `.breadcrumb-item`.
*   **Item Activo:** El último `<li>`, que representa la página actual, debe tener también la clase `.active` y el atributo `aria-current="page"`. No suele ser un enlace.
*   **Separador:** Bootstrap añade automáticamente el separador (`/` por defecto) usando CSS (`::before`). Puedes personalizarlo modificando la variable Sass `$breadcrumb-divider`.

```html
<div class="container mt-4">
  <nav aria-label="breadcrumb">
    <ol class="breadcrumb">
      <li class="breadcrumb-item"><a href="/">Inicio</a></li>
      <li class="breadcrumb-item"><a href="/productos/">Productos</a></li>
      <li class="breadcrumb-item active" aria-current="page">Electrónica</li> {# Página actual #}
    </ol>
  </nav>
</div>
```
*   La etiqueta `<nav>` con `aria-label="breadcrumb"` mejora la accesibilidad.

## 4_4
## Botones (Buttons) (estilos, outline, tamaños, estados, block)

Bootstrap ofrece una amplia variedad de estilos y opciones para los botones (`<button>`, `<input type="button/submit/reset">`, `<a>`).

*   **Clase Base:** `.btn`. Es necesaria para aplicar los estilos base.
*   **Clases de Estilo Contextual:** Determinan el color de fondo y borde.
    *   `.btn-primary` (Azul)
    *   `.btn-secondary` (Gris)
    *   `.btn-success` (Verde)
    *   `.btn-danger` (Rojo)
    *   `.btn-warning` (Amarillo)
    *   `.btn-info` (Celeste)
    *   `.btn-light` (Blanco/Gris claro)
    *   `.btn-dark` (Gris oscuro)
    *   `.btn-link` (Apariencia de enlace)
*   **Estilos Outline:** Botones con fondo transparente y borde coloreado. Añade `-outline` al estilo contextual: `.btn-outline-primary`, `.btn-outline-success`, etc. Cambian de color al hacer hover.
*   **Tamaños:**
    *   `.btn-lg`: Botón grande.
    *   `.btn-sm`: Botón pequeño.
    *   (Sin clase de tamaño para el tamaño normal).
*   **Estado Deshabilitado:**
    *   Añade el atributo `disabled` al elemento `<button>`.
    *   Para elementos `<a>` usados como botones, añade la clase `.disabled` y `aria-disabled="true"` (además de `tabindex="-1"` para quitarlo del foco).
*   **Botones de Ancho Completo (Block):** Añade la clase `.btn-block`. El botón ocupará todo el ancho de su contenedor padre.
*   **Estado Activo:** Añade la clase `.active` para que el botón parezca presionado (útil para simular estados toggle).

```html
<div class="container mt-4">
  <button type="button" class="btn btn-primary">Primario</button>
  <button type="button" class="btn btn-outline-success">Éxito Outline</button>
  <a href="#" class="btn btn-info btn-lg" role="button">Enlace Info Grande</a>
  <input type="submit" class="btn btn-warning btn-sm" value="Enviar Pequeño">
  <button type="button" class="btn btn-danger" disabled>Deshabilitado</button>
  <a href="#" class="btn btn-secondary disabled" tabindex="-1" aria-disabled="true" role="button">Enlace Deshabilitado</a>
  <button type="button" class="btn btn-dark btn-block mt-3">Botón Block</button>
</div>
```

## Grupos de Botones (Button Group, Toolbar, Anidados)

Permiten agrupar varios botones en una sola línea, visualmente conectados.

*   **Grupo Básico:** Envuelve los botones (`.btn`) en un `<div class="btn-group" role="group" aria-label="...">`.
*   **Toolbar:** Agrupa varios `btn-group` en una barra de herramientas más compleja usando `<div class="btn-toolbar" role="toolbar" aria-label="...">`. Puedes añadir espaciado entre grupos con utilidades de margen (`mr-2`, `mx-1`). También puedes incluir `<div class="input-group">` dentro.
*   **Tamaños:** Añade `.btn-group-lg` o `.btn-group-sm` al `.btn-group` para cambiar el tamaño de todos los botones dentro.
*   **Anidamiento (Dropdowns):** Puedes colocar un botón con un desplegable dentro de un `.btn-group`.
*   **Vertical:** Usa `.btn-group-vertical` para apilar los botones verticalmente.

```html
<div class="container mt-4">
  <h4>Grupo Básico</h4>
  <div class="btn-group" role="group" aria-label="Grupo básico">
    <button type="button" class="btn btn-secondary">Izquierda</button>
    <button type="button" class="btn btn-secondary">Medio</button>
    <button type="button" class="btn btn-secondary">Derecha</button>
  </div>

  <h4 class="mt-3">Toolbar</h4>
  <div class="btn-toolbar" role="toolbar" aria-label="Toolbar con grupos">
    <div class="btn-group mr-2" role="group" aria-label="Primer grupo">
      <button type="button" class="btn btn-info">1</button>
      <button type="button" class="btn btn-info">2</button>
    </div>
    <div class="btn-group mr-2" role="group" aria-label="Segundo grupo">
      <button type="button" class="btn btn-info">3</button>
      <button type="button" class="btn btn-info">4</button>
    </div>
  </div>

  <h4 class="mt-3">Grupo con Dropdown</h4>
   <div class="btn-group">
    <button type="button" class="btn btn-danger">Acción</button>
    <button type="button" class="btn btn-danger dropdown-toggle dropdown-toggle-split" data-toggle="dropdown" aria-expanded="false">
      <span class="sr-only">Desplegar menú</span>
    </button>
    <div class="dropdown-menu">
      <a class="dropdown-item" href="#">Opción 1</a>
      <a class="dropdown-item" href="#">Opción 2</a>
    </div>
  </div>

   <h4 class="mt-3">Grupo Vertical</h4>
    <div class="btn-group-vertical">
      <button type="button" class="btn btn-secondary">Botón 1</button>
      <button type="button" class="btn btn-secondary">Botón 2</button>
    </div>
</div>
```

## 4_6
## Tarjetas (Cards) (anatomía, contenido, layouts, grupos)

Las tarjetas son contenedores de contenido flexibles y extensibles. Reemplazaron a los antiguos panels, wells y thumbnails de Bootstrap 3. Son ideales para mostrar contenido relacionado como una unidad (ej: un producto, un post de blog, un perfil).

*   **Anatomía Básica:**
    *   Contenedor principal: `<div class="card">`.
    *   **(Opcional) Cabecera:** `<div class="card-header">`.
    *   **Cuerpo:** `<div class="card-body">` (aquí va el contenido principal).
    *   **(Opcional) Título:** `<h5 class="card-title">` (o cualquier `h*`).
    *   **(Opcional) Subtítulo:** `<h6 class="card-subtitle mb-2 text-muted">`.
    *   **(Opcional) Texto:** `<p class="card-text">`.
    *   **(Opcional) Enlaces:** `<a href="#" class="card-link">`.
    *   **(Opcional) Imagen:** `<img class="card-img-top" src="...">` o `<img class="card-img-bottom" src="...">`. La imagen se puede colocar arriba o abajo.
    *   **(Opcional) Pie de página:** `<div class="card-footer">`.
    *   **(Opcional) Lista dentro:** `<ul class="list-group list-group-flush">`.

```html
<div class="container mt-4">
<div class="card" style="width: 18rem;"> {# Ancho fijo como ejemplo #}
  <img src="https://via.placeholder.com/286x180" class="card-img-top" alt="...">
  <div class="card-body">
    <h5 class="card-title">Título de la Tarjeta</h5>
    <p class="card-text">Texto de ejemplo rápido para construir sobre el título de la tarjeta y componer el grueso del contenido de la tarjeta.</p>
    <a href="#" class="btn btn-primary">Ir a algún lugar</a>
    <a href="#" class="card-link">Otro enlace</a>
  </div>
  <div class="card-footer text-muted">
    Actualizado hace 2 días
  </div>
</div>
</div>
```

*   **Contenido:** Puedes mezclar y combinar cabeceras, cuerpo, pies, imágenes, listas dentro de la tarjeta.
*   **Colores de Fondo y Borde:** Usa utilidades de borde (`border-*`) y fondo (`bg-*`) en el `.card` para cambiar su apariencia.
*   **Layouts de Tarjeta:**
    *   **Grupos de Tarjetas (`.card-group`):** Para mostrar tarjetas una al lado de la otra con el mismo alto y ancho (sin márgenes entre ellas). Envuelve las `.card` en `<div class="card-group">`.
    *   **Mazos de Tarjetas (`.card-deck`):** Similar a `card-group`, pero añade margen entre las tarjetas. Envuelve las `.card` en `<div class="card-deck">`.
    *   **Columnas de Tarjetas (`.card-columns`):** Organiza las tarjetas en columnas tipo Masonry (como Pinterest), ajustando la altura. Envuelve las `.card` en `<div class="card-columns">`. (Puede tener limitaciones).
*   **Navegación en Tarjetas:** Puedes usar `.card-header-tabs` o `.card-header-pills` en un `.nav` dentro del `.card-header` para crear navegación tipo tabs/pills.
*   **Superposición de Imágenes:** Puedes superponer texto sobre imágenes usando `.card-img-overlay`.

Las tarjetas son extremadamente flexibles para presentar contenido diverso.

## Carrusel (Carousel) (slides, controles, indicadores)

Un componente tipo slideshow para ciclar a través de elementos (imágenes o diapositivas con texto). **Requiere JavaScript de Bootstrap y sus dependencias (jQuery, Popper).**

*   **Estructura Básica:**
    *   Contenedor principal: `<div id="miCarrusel" class="carousel slide" data-ride="carousel">`. El `id` es importante para los controles. `data-ride="carousel"` inicia el carrusel automáticamente al cargar la página.
    *   **Indicadores (Opcional):** `<ol class="carousel-indicators">`. Contiene `<li>` con atributos `data-target="#miCarrusel"` y `data-slide-to="N"` (índice N empezando en 0). El item activo necesita la clase `.active`.
    *   **Contenedor de Slides:** `<div class="carousel-inner">`.
    *   **Cada Slide:** `<div class="carousel-item">`. El **primer** item debe tener la clase `.active`. Puedes poner imágenes (`<img>` con `.d-block .w-100`) o cualquier otro HTML dentro.
    *   **Leyendas (Opcional):** Dentro de un `.carousel-item`, puedes añadir `<div class="carousel-caption d-none d-md-block">` (clases de display responsive opcionales) para superponer texto.
    *   **Controles Anterior/Siguiente (Opcional):** Botones (`<button>` o `<a>`) con clase `.carousel-control-prev` y `.carousel-control-next`, atributos `href` o `data-target` apuntando al ID del carrusel, y `data-slide="prev"` o `data-slide="next"`. Incluyen iconos (`.carousel-control-prev-icon`, `.carousel-control-next-icon`) y texto para screen readers (`.sr-only`).

```html
<div class="container mt-4">
<div id="carouselExampleCaptions" class="carousel slide" data-ride="carousel">
  <ol class="carousel-indicators">
    <li data-target="#carouselExampleCaptions" data-slide-to="0" class="active"></li>
    <li data-target="#carouselExampleCaptions" data-slide-to="1"></li>
    <li data-target="#carouselExampleCaptions" data-slide-to="2"></li>
  </ol>
  <div class="carousel-inner">
    <div class="carousel-item active"> {# Primer item activo #}
      <img src="https://via.placeholder.com/800x400?text=Primera+Slide" class="d-block w-100" alt="...">
      <div class="carousel-caption d-none d-md-block">
        <h5>Título Primera Slide</h5>
        <p>Descripción opcional de la primera slide.</p>
      </div>
    </div>
    <div class="carousel-item">
      <img src="https://via.placeholder.com/800x400?text=Segunda+Slide" class="d-block w-100" alt="...">
      <div class="carousel-caption d-none d-md-block">
        <h5>Título Segunda Slide</h5>
        <p>Descripción opcional.</p>
      </div>
    </div>
    <div class="carousel-item">
      <img src="https://via.placeholder.com/800x400?text=Tercera+Slide" class="d-block w-100" alt="...">
    </div>
  </div>
  <button class="carousel-control-prev" type="button" data-target="#carouselExampleCaptions" data-slide="prev">
    <span class="carousel-control-prev-icon" aria-hidden="true"></span>
    <span class="sr-only">Anterior</span>
  </button>
  <button class="carousel-control-next" type="button" data-target="#carouselExampleCaptions" data-slide="next">
    <span class="carousel-control-next-icon" aria-hidden="true"></span>
    <span class="sr-only">Siguiente</span>
  </button>
</div>
</div>
```

*   **Opciones (vía Atributos `data-*` o JavaScript):** `data-interval="5000"` (tiempo entre slides, `false` para detener), `data-keyboard="true"` (navegar con teclado), `data-pause="hover"` (pausar al poner ratón encima), `data-wrap="true"` (ciclar continuamente).

## Acordeon_Collapse_Acordeon

El componente Collapse permite mostrar u ocultar contenido de forma dinámica. **Requiere JavaScript.** Se usa comúnmente para crear acordeones (donde solo un panel está visible a la vez) o secciones desplegables.

*   **Activador:** Un botón (`<button>`) o enlace (`<a>`) con:
    *   `data-toggle="collapse"`
    *   `href="#idDelContenido"` o `data-target="#idDelContenido"` apuntando al ID del elemento a colapsar.
    *   `aria-expanded="true|false"` indicando el estado inicial.
    *   `aria-controls="idDelContenido"` vinculando el control al contenido.
*   **Contenido Colapsable:** Un `<div>` (u otro elemento) con:
    *   La clase base `.collapse`.
    *   La clase `.show` si quieres que esté visible inicialmente.
    *   Un `id` único que coincida con el `href`/`data-target` del activador.

**Ejemplo Simple:**
```html
<p>
  <a class="btn btn-primary" data-toggle="collapse" href="#contenidoColapso1" role="button" aria-expanded="false" aria-controls="contenidoColapso1">
    Mostrar/Ocultar Contenido
  </a>
</p>
<div class="collapse" id="contenidoColapso1">
  <div class="card card-body">
    Contenido secreto que aparece y desaparece. Anim pariatur cliche reprehenderit, enim eiusmod high life accusamus terry richardson ad squid.
  </div>
</div>
```

**Acordeón:** Para crear un acordeón (varios colapsables donde solo uno puede estar abierto), envuelve todo en un `<div>` con un `id` (ej: `id="miAcordeon"`). Luego, en cada botón/enlace activador, añade el atributo `data-parent="#miAcordeon"`.

```html
<div class="accordion" id="accordionExample">
  <div class="card">
    <div class="card-header" id="headingOne">
      <h2 class="mb-0">
        <button class="btn btn-link btn-block text-left" type="button" data-toggle="collapse" data-target="#collapseOne" aria-expanded="true" aria-controls="collapseOne">
          Elemento Acordeón #1
        </button>
      </h2>
    </div>
    {# El primer item tiene 'show' para estar abierto por defecto #}
    <div id="collapseOne" class="collapse show" aria-labelledby="headingOne" data-parent="#accordionExample">
      <div class="card-body">
        Contenido del primer elemento.
      </div>
    </div>
  </div>
  <div class="card">
    <div class="card-header" id="headingTwo">
      <h2 class="mb-0">
        <button class="btn btn-link btn-block text-left collapsed" type="button" data-toggle="collapse" data-target="#collapseTwo" aria-expanded="false" aria-controls="collapseTwo">
          Elemento Acordeón #2
        </button>
      </h2>
    </div>
    <div id="collapseTwo" class="collapse" aria-labelledby="headingTwo" data-parent="#accordionExample">
      <div class="card-body">
        Contenido del segundo elemento.
      </div>
    </div>
  </div>
  {# ... más elementos .card ... #}
</div>
```
*   Usar la estructura de `card` es común para los items del acordeón.
*   La clase `.collapsed` en el botón y `aria-expanded="false"` indican visual y semánticamente que el panel está cerrado inicialmente.

## 4_9
## Desplegables (Dropdowns) (botones, split, direcciones)

Los desplegables (dropdowns) permiten mostrar una lista de enlaces u opciones al hacer clic en un botón o enlace activador. **Requiere JavaScript (jQuery y Popper.js).**

*   **Contenedor:** Envuelve el activador y el menú en un `<div>` (u otro elemento) con la clase `.dropdown`.
*   **Activador:** Un `<button>` o `<a>` con:
    *   `id` único (ej: `dropdownMenuButton`).
    *   `class="dropdown-toggle"` (añade el indicador de flecha).
    *   `data-toggle="dropdown"`.
    *   `aria-haspopup="true"` y `aria-expanded="false"`.
*   **Menú Desplegable:** Un `<div>` con:
    *   `class="dropdown-menu"`.
    *   `aria-labelledby="idDelActivador"`.
    *   Dentro, usa elementos `<a>` o `<button>` con clase `.dropdown-item`. Puedes añadir `.dropdown-divider` para separadores y `.dropdown-header` para títulos. El item activo puede tener clase `.active`. Un item deshabilitado tiene clase `.disabled`.

```html
<div class="container mt-4">
<div class="dropdown">
  <button class="btn btn-secondary dropdown-toggle" type="button" id="dropdownMenuButton" data-toggle="dropdown" aria-expanded="false">
    Menú Desplegable
  </button>
  <div class="dropdown-menu" aria-labelledby="dropdownMenuButton">
    <a class="dropdown-item" href="#">Acción</a>
    <a class="dropdown-item active" href="#">Acción Activa</a> {# Item activo #}
    <a class="dropdown-item disabled" href="#" tabindex="-1" aria-disabled="true">Acción Deshabilitada</a> {# Item deshabilitado #}
    <div class="dropdown-divider"></div> {# Separador #}
    <h6 class="dropdown-header">Cabecera</h6> {# Cabecera #}
    <a class="dropdown-item" href="#">Enlace Separado</a>
  </div>
</div>
</div>
```

*   **Botones Divididos (Split Buttons):** Para tener una acción principal en el botón y el desplegable en una flecha separada. Usa un `.btn-group`, un botón normal y otro botón con `.dropdown-toggle` y `.dropdown-toggle-split`.
    ```html
    <div class="btn-group mt-3">
      <button type="button" class="btn btn-success">Acción Principal</button>
      <button type="button" class="btn btn-success dropdown-toggle dropdown-toggle-split" data-toggle="dropdown" aria-expanded="false">
        <span class="sr-only">Desplegar Menú</span>
      </button>
      <div class="dropdown-menu">
        <a class="dropdown-item" href="#">Opción A</a>
        <a class="dropdown-item" href="#">Opción B</a>
      </div>
    </div>
    ```
*   **Direcciones:** Por defecto, el menú se despliega hacia abajo. Puedes usar clases en el contenedor `.dropdown` para cambiarlo: `.dropup`, `.dropright`, `.dropleft`.
*   **Alineación del Menú:** Añade `.dropdown-menu-right` al `.dropdown-menu` para alinearlo a la derecha del activador. También hay alineación responsive (`.dropdown-menu-md-right`).

### 5_1_7
### Validación de Formularios (Estados y Feedback)

Bootstrap utiliza las **pseudo-clases CSS `:valid` e `:invalid`** de HTML5 para aplicar estilos a los campos de formulario según su estado de validación (basado en atributos como `required`, `pattern`, `type="email"`, etc., o validación vía JavaScript). También proporciona clases para mostrar mensajes de feedback.

*   **Estilos de Estado:** Los campos válidos suelen tener un borde verde y un icono de check, mientras que los inválidos tienen un borde rojo y un icono de exclamación. Esto se aplica automáticamente a los inputs dentro de formularios que tengan la clase `.was-validated` (validación al intentar enviar) o que se validen con JavaScript.
*   **Mensajes de Feedback:**
    *   `.valid-feedback` o `.valid-tooltip`: Mensaje a mostrar si el campo es válido.
    *   `.invalid-feedback` o `.invalid-tooltip`: Mensaje a mostrar si el campo es inválido.
    *   Coloca estos `<div>` justo después del `<input>` o `<select>` al que se refieren. Son `display: none` por defecto y se muestran automáticamente cuando el input hermano es `:valid` o `:invalid` (dentro de un form `.was-validated` o validado por JS). La versión `-tooltip` los muestra de forma emergente.

```html
<form class="was-validated"> {# Activa estilos de validación del navegador/Bootstrap #}
  <div class="form-group">
    <label for="validationCustom01">Nombre</label>
    <input type="text" class="form-control" id="validationCustom01" placeholder="Nombre" required>
    <div class="valid-feedback">
      ¡Correcto!
    </div>
     <div class="invalid-feedback">
      Por favor, introduce tu nombre.
    </div>
  </div>

  <div class="form-group">
    <label for="validationServerUsername">Usuario</label>
    <div class="input-group">
      <div class="input-group-prepend">
        <span class="input-group-text">@</span>
      </div>
      <input type="text" class="form-control" id="validationServerUsername" required>
      <div class="invalid-feedback">
        Elige un nombre de usuario.
      </div>
    </div>
  </div>

   <div class="form-group">
      <div class="form-check">
        <input class="form-check-input" type="checkbox" value="" id="invalidCheck3" required>
        <label class="form-check-label" for="invalidCheck3">
          Acepto los términos y condiciones
        </label>
        <div class="invalid-feedback">
          Debes aceptar los términos.
        </div>
      </div>
    </div>

  <button class="btn btn-primary" type="submit">Enviar</button>
</form>
```
*   La clase `.was-validated` en el `<form>` fuerza la muestra de los estilos de validación. Normalmente, esta clase se añade dinámicamente con JavaScript después de que el usuario intente enviar el formulario, para no mostrar errores antes de tiempo. Bootstrap también soporta la validación del lado del servidor añadiendo clases `.is-valid` o `.is-invalid` a los inputs y mostrando los feedback correspondientes.

## Ventanas Modales (Modal) (activación, opciones, eventos)

Las ventanas modales son diálogos emergentes que aparecen sobre el contenido de la página. Se usan para notificaciones importantes, formularios de login/registro, confirmaciones, galerías de imágenes, etc. **Requieren JavaScript.**

*   **Activador (Trigger):** Un botón o enlace que abre el modal. Necesita:
    *   `data-toggle="modal"`
    *   `data-target="#idDelModal"` (o `href="#idDelModal"` para enlaces) apuntando al ID del contenedor del modal.
*   **Estructura del Modal:**
    *   Contenedor principal: `<div class="modal fade" id="idDelModal" tabindex="-1" aria-labelledby="tituloModal" aria-hidden="true">`.
        *   `.modal`: Clase base.
        *   `.fade`: Añade animación de desvanecimiento.
        *   `id`: Coincide con `data-target` del activador.
        *   `tabindex="-1"`: Para poder cerrarlo con la tecla Esc.
        *   `aria-labelledby`, `aria-hidden`: Atributos de accesibilidad.
    *   Diálogo: `<div class="modal-dialog">` (controla el tamaño y posición).
        *   Puedes añadir clases de tamaño: `.modal-sm`, `.modal-lg`, `.modal-xl`.
        *   Puedes añadir `.modal-dialog-centered` para centrarlo verticalmente.
        *   Puedes añadir `.modal-dialog-scrollable` para que el *cuerpo* del modal tenga scroll si el contenido es muy largo.
    *   Contenido: `<div class="modal-content">`.
        *   **Cabecera:** `<div class="modal-header">`. Contiene:
            *   Título: `<h5 class="modal-title" id="tituloModal">`. El `id` coincide con `aria-labelledby`.
            *   Botón de cierre: `<button type="button" class="close" data-dismiss="modal" aria-label="Close">`.
        *   **Cuerpo:** `<div class="modal-body">`. Aquí va el contenido principal.
        *   **Pie (Opcional):** `<div class="modal-footer">`. Normalmente contiene botones de acción (Guardar, Cerrar). Puedes usar botones con `data-dismiss="modal"` para cerrarlo.

```html
<div class="container mt-4">
  <!-- Botón Activador -->
  <button type="button" class="btn btn-primary" data-toggle="modal" data-target="#exampleModal">
    Abrir Modal de Ejemplo
  </button>

  <!-- El Modal -->
  <div class="modal fade" id="exampleModal" tabindex="-1" aria-labelledby="exampleModalLabel" aria-hidden="true">
    <div class="modal-dialog modal-dialog-centered"> {# Centrado vertical #}
      <div class="modal-content">
        <div class="modal-header">
          <h5 class="modal-title" id="exampleModalLabel">Título del Modal</h5>
          <button type="button" class="close" data-dismiss="modal" aria-label="Close">
            <span aria-hidden="true">&times;</span>
          </button>
        </div>
        <div class="modal-body">
          <p>Woohoo, ¡estás leyendo este texto en un modal!</p>
          {# Aquí podría ir un formulario, más texto, una imagen... #}
        </div>
        <div class="modal-footer">
          <button type="button" class="btn btn-secondary" data-dismiss="modal">Cerrar</button>
          <button type="button" class="btn btn-primary">Guardar Cambios</button>
        </div>
      </div>
    </div>
  </div>
</div>
```

*   **Opciones (vía Atributos `data-*` en el activador o JS):** `data-backdrop="static"` (no cerrar al hacer clic fuera), `data-keyboard="false"` (no cerrar con Esc).
*   **Eventos JS:** Puedes escuchar eventos como `show.bs.modal`, `shown.bs.modal`, `hide.bs.modal`, `hidden.bs.modal` para ejecutar código cuando el modal se abre o se cierra.

## Navegación (Navs) (tabs, pills, fill, justify)

Bootstrap proporciona el componente `.nav` para crear bloques de navegación reutilizables. Es la base para barras de navegación, pestañas (tabs) y píldoras (pills).

*   **Estructura Base:** Un `<ul>` con clase `.nav`. Cada item es un `<li>` con clase `.nav-item`, conteniendo un `<a>` con clase `.nav-link`.
*   **Estado Activo/Deshabilitado:** Añade `.active` o `.disabled` al `.nav-link` (o al `.nav-item` padre).
*   **Apariencia:**
    *   `.nav-tabs`: Estilo de pestañas.
    *   `.nav-pills`: Estilo de píldoras (botones redondeados).
*   **Alineación y Distribución (Flexbox):**
    *   Alineación horizontal: `.justify-content-center`, `.justify-content-end`.
    *   Orientación vertical: `.flex-column`.
    *   **Rellenar (`.nav-fill`):** Hace que los `.nav-item` ocupen el mismo ancho, rellenando todo el espacio horizontal disponible.
    *   **Justificar (`.nav-justify`):** Similar a `fill`, pero el ancho de cada item puede variar, distribuyendo el espacio extra.

```html
<div class="container mt-4">
  <h4>Nav Base</h4>
  <ul class="nav">
    <li class="nav-item">
      <a class="nav-link active" href="#">Activo</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Enlace</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Otro Enlace</a>
    </li>
    <li class="nav-item">
      <a class="nav-link disabled" href="#" tabindex="-1" aria-disabled="true">Deshabilitado</a>
    </li>
  </ul>

  <h4 class="mt-4">Nav como Pestañas (Tabs)</h4>
  <ul class="nav nav-tabs">
    <li class="nav-item">
      <a class="nav-link active" href="#">Activo</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Enlace</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Enlace</a>
    </li>
  </ul>

  <h4 class="mt-4">Nav como Píldoras (Pills) con Fill</h4>
  <ul class="nav nav-pills nav-fill">
    <li class="nav-item">
      <a class="nav-link active" href="#">Activo</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Enlace Mucho Más Largo</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Enlace</a>
    </li>
  </ul>

   <h4 class="mt-4">Nav Vertical</h4>
   <ul class="nav flex-column">
    <li class="nav-item">
      <a class="nav-link active" href="#">Activo</a>
    </li>
    <li class="nav-item">
      <a class="nav-link" href="#">Enlace</a>
    </li>
   </ul>
</div>
```
*   **Funcionalidad Tabs/Pills (Requiere JS):** Para hacer que las pestañas o píldoras muestren contenido diferente al hacer clic, necesitas usar el plugin JavaScript de Tab de Bootstrap. Esto implica añadir atributos `data-toggle="tab"` o `data-toggle="pill"` a los enlaces, `href` apuntando al ID del panel de contenido, y estructurar los paneles de contenido con clases `.tab-content` y `.tab-pane`.

## Barra de Navegación (Navbar) (marca, links, formularios, responsive)

El componente `.navbar` es una cabecera de navegación potente y responsive que puede incluir la marca del sitio, enlaces de navegación, formularios y se colapsa automáticamente en pantallas pequeñas. **Requiere el plugin JS Collapse y sus dependencias.**

*   **Estructura Básica:**
    *   Contenedor principal: `<nav class="navbar navbar-expand-BP [otras clases...]">`.
        *   `navbar`: Clase base.
        *   `navbar-expand-BP`: **Crucial para responsive.** Define en qué breakpoint (`sm`, `md`, `lg`, `xl`) la barra de navegación dejará de estar colapsada y se expandirá horizontalmente. Sin esto, siempre estará colapsada verticalmente.
        *   Clases de color de fondo: `bg-light`, `bg-dark`, `bg-primary`, etc.
        *   Clases de esquema de color (para el texto): `navbar-light` (texto oscuro para fondos claros), `navbar-dark` (texto claro para fondos oscuros).
    *   **Marca/Logo (`.navbar-brand`):** Un `<a>` o `<span>` para el logo o nombre del sitio.
    *   **Botón Toggler (`.navbar-toggler`):** El botón (generalmente con icono de hamburguesa `.navbar-toggler-icon`) que aparece en pantallas pequeñas para mostrar/ocultar el contenido colapsable. Necesita `data-toggle="collapse"` y `data-target="#idDelContenidoColapsable"`.
    *   **Contenido Colapsable:** Un `<div>` con clases `.collapse` y `.navbar-collapse`, y un `id` que coincida con el `data-target` del toggler. Aquí dentro va la navegación principal.
    *   **Lista de Navegación:** Típicamente un `<ul>` con clase `.navbar-nav`. Los items son `<li>` con `.nav-item` y los enlaces `<a>` con `.nav-link`. Puedes usar `.active` en el item actual. Utilidades de margen como `mr-auto` (margen derecho automático) o `ml-auto` (margen izquierdo automático) en `.navbar-nav` ayudan a posicionar los bloques de enlaces.
    *   **Formularios Inline (`.form-inline`):** Puedes incluir formularios de búsqueda dentro de la navbar.
    *   **Texto (`.navbar-text`):** Para texto simple no enlazable.
    *   **Dropdowns:** Se pueden integrar dentro de los `.nav-item`.

```html
<nav class="navbar navbar-expand-lg navbar-dark bg-dark"> {# Expande en LG, fondo oscuro, texto claro #}
  <a class="navbar-brand" href="#">MiTienda</a>
  <button class="navbar-toggler" type="button" data-toggle="collapse" data-target="#navbarNavDropdown" aria-controls="navbarNavDropdown" aria-expanded="false" aria-label="Toggle navigation">
    <span class="navbar-toggler-icon"></span>
  </button>
  <div class="collapse navbar-collapse" id="navbarNavDropdown">
    {# Bloque de enlaces a la izquierda #}
    <ul class="navbar-nav mr-auto"> {# mr-auto empuja lo siguiente a la derecha #}
      <li class="nav-item active">
        <a class="nav-link" href="#">Inicio <span class="sr-only">(current)</span></a>
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#">Productos</a>
      </li>
      <li class="nav-item dropdown">
        <a class="nav-link dropdown-toggle" href="#" id="navbarDropdownMenuLink" role="button" data-toggle="dropdown" aria-expanded="false">
          Categorías
        </a>
        <div class="dropdown-menu" aria-labelledby="navbarDropdownMenuLink">
          <a class="dropdown-item" href="#">Electrónica</a>
          <a class="dropdown-item" href="#">Hogar</a>
          <a class="dropdown-item" href="#">Ropa</a>
        </div>
      </li>
    </ul>
    {# Bloque a la derecha #}
    <span class="navbar-text mr-3"> {# Texto simple #}
      ¡Bienvenido!
    </span>
    <form class="form-inline my-2 my-lg-0"> {# Formulario inline #}
      <input class="form-control mr-sm-2" type="search" placeholder="Buscar" aria-label="Buscar">
      <button class="btn btn-outline-success my-2 my-sm-0" type="submit">Buscar</button>
    </form>
  </div>
</nav>
```

*   **Posicionamiento:** Puedes hacer la navbar fija (`.fixed-top`, `.fixed-bottom`) o adhesiva (`.sticky-top`). Recuerda añadir padding al `<body>` si usas fixed-top para evitar que el contenido quede oculto debajo.

## 5_9
## Popovers (Emergentes)

Similares a los Tooltips, pero diseñados para mostrar contenido más extenso al hacer clic (o hover) en un elemento. **Requieren JavaScript (jQuery y Popper.js) y deben inicializarse explícitamente.**

*   **Activador:** Un elemento (normalmente `<button>`) con:
    *   `data-toggle="popover"`
    *   `title="..."`: Título del popover.
    *   `data-content="..."`: Contenido principal del popover (puede incluir HTML simple).
    *   `data-placement="top|bottom|left|right"`: Dirección en la que aparece (defecto `right`).
    *   `data-trigger="click|hover|focus"`: Cómo se activa (defecto `click`).
*   **Inicialización JS:** Debes seleccionar los elementos con `data-toggle="popover"` y llamar al método `.popover()` sobre ellos usando jQuery.

```html
<div class="container mt-4 text-center">
  <button type="button" class="btn btn-lg btn-danger"
          data-toggle="popover"
          title="Título del Popover"
          data-content="Y aquí hay un contenido asombroso. Es muy interesante. ¿Verdad?">
    Clic para Popover
  </button>

  <button type="button" class="btn btn-secondary ml-2"
          data-container="body" data-toggle="popover" data-placement="top"
          data-content="Este aparece arriba y al hacer hover."
          data-trigger="hover">
    Hover para Popover
  </button>
</div>

<!-- Script de inicialización (al final del body) -->
<script>
  $(function () {
    // Inicializar todos los popovers en la página
    $('[data-toggle="popover"]').popover()
  })
</script>
```

## Scrollspy (Espiando Scroll)

Actualiza automáticamente los enlaces de una navegación (ej: un menú lateral) para resaltar cuál sección de la página está visible actualmente mientras el usuario hace scroll. **Requiere JavaScript.**

*   **Estructura:**
    *   El elemento "espiado" (normalmente `<body>` o un contenedor con scroll) necesita `position: relative;` y el atributo `data-spy="scroll"`.
    *   Necesita un `data-target="#idDeLaNavegacion"` apuntando al ID del elemento de navegación (`<nav>` o `<ul>`) que contiene los enlaces.
    *   (Opcional) `data-offset="N"` para ajustar el punto de activación del resaltado (en píxeles).
    *   La navegación (`<nav>`, `<ul>` con `.nav`) debe contener enlaces `<a>` cuyo `href` coincida **exactamente** con el `id` de las secciones correspondientes en el contenido.
    *   Las secciones del contenido deben tener esos `id`s.

```html
{# Estructura HTML (simplificada) #}
<body data-spy="scroll" data-target="#navbar-scrollspy" data-offset="70">

  {# Barra de Navegación (puede ser fixed-top) #}
  <nav id="navbar-scrollspy" class="navbar navbar-light bg-light">
    <a class="navbar-brand" href="#">Navegación</a>
    <ul class="nav nav-pills">
      <li class="nav-item">
        <a class="nav-link" href="#seccion1">Sección 1</a> {# href coincide con id de sección #}
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#seccion2">Sección 2</a>
      </li>
      {# ... más enlaces ... #}
    </ul>
  </nav>

  {# Contenido Espiado #}
  <div data-spy="scroll" data-target="#navbar-scrollspy" data-offset="0" class="scrollspy-example" style="position: relative; height: 200px; overflow-y: scroll;">
      <h4 id="seccion1">Sección 1</h4> {# id coincide con href del enlace #}
      <p>Contenido de la sección 1...</p>
      <h4 id="seccion2">Sección 2</h4>
      <p>Contenido de la sección 2...</p>
      {# ... más secciones con ids ... #}
  </div>

</body>
```
Al hacer scroll en el `<body>` (o en el `div` con scroll), Bootstrap añadirá automáticamente la clase `.active` al enlace correspondiente en `#navbar-scrollspy` cuando la sección asociada entre en el viewport. Necesitarás CSS para estilizar el estado `.active` del `.nav-link`.

## Toasts (Notificaciones Emergentes)

Notificaciones ligeras, no intrusivas, similares a las notificaciones push. **Requieren JavaScript y deben inicializarse.**

*   **Estructura:**
    *   Contenedor: `<div class="toast" role="alert" aria-live="assertive" aria-atomic="true">`. Opciones con `data-*`: `data-autohide="false"` (no ocultar automáticamente), `data-delay="5000"` (tiempo en ms antes de ocultar).
    *   Cabecera (Opcional): `<div class="toast-header">`. Contiene:
        *   Imagen/Icono: `<img src="..." class="rounded mr-2" alt="...">`
        *   Título: `<strong class="mr-auto">`
        *   Tiempo: `<small>` (ej: "hace 11 mins")
        *   Botón de cierre: `<button type="button" class="ml-2 mb-1 close" data-dismiss="toast" aria-label="Close">`
    *   Cuerpo: `<div class="toast-body">`. Contenido del mensaje.
*   **Posicionamiento:** Los toasts deben colocarse dentro de un contenedor (ej: un `<div>` con `position: fixed` o `absolute` y estilos `top`, `right`, etc.) para que aparezcan correctamente superpuestos. Bootstrap no los posiciona por defecto.
*   **Inicialización JS:** Selecciona los elementos `.toast` y llama a `.toast('show')` para mostrarlos o `.toast({opciones})` para inicializarlos.

```html
{# Contenedor para posicionar los toasts (ej: esquina superior derecha) #}
<div aria-live="polite" aria-atomic="true" style="position: relative; min-height: 200px;">
  <div style="position: absolute; top: 0; right: 0; z-index: 1050;">

    {# Toast Ejemplo #}
    <div class="toast" role="alert" aria-live="assertive" aria-atomic="true" data-delay="10000"> {# Oculta después de 10s #}
      <div class="toast-header">
        <img src="..." class="rounded mr-2" alt="..." style="width:20px; height:20px;">
        <strong class="mr-auto">Bootstrap</strong>
        <small>Ahora mismo</small>
        <button type="button" class="ml-2 mb-1 close" data-dismiss="toast" aria-label="Close">
          <span aria-hidden="true">&times;</span>
        </button>
      </div>
      <div class="toast-body">
        ¡Hola! Este es un mensaje toast.
      </div>
    </div>

    {# Otro toast #}
    {# ... #}

  </div>
</div>

{# Botón para mostrar el toast (ejemplo, requiere JS) #}
<button type="button" class="btn btn-primary" id="showToast">Mostrar Toast</button>

<!-- Script de inicialización -->
<script>
  $(document).ready(function(){
    // Inicializar y mostrar al hacer clic en el botón
    $('#showToast').click(function(){
      $('.toast').toast('show'); // Muestra todos los toasts en la página
    });
    // O inicializar sin mostrar: $('.toast').toast({ autohide: false });
  });
</script>
```

# Utilidades (Utilities)

Además de los componentes estructurados, Bootstrap 4 ofrece un conjunto extremadamente útil de **clases de utilidad (utility classes)**.     *   `.rounded-0`: Elimina cualquier radio de borde.
    *   `.rounded-sm`, `.rounded-lg`: Tamaños de radio pequeño y grande.


## 6_5
## Display (`d-*`, responsive, print)

Controla la propiedad `display` de CSS, permitiendo cambiar cómo se renderiza un elemento (bloque, en línea, flex, grid, etc.) y su visibilidad. Son **fundamentales para el diseño responsive**.

*   **Valores Comunes:**
    *   `d-none`: Equivalente a `display: none;` (oculto y no ocupa espacio).
    *   `d-inline`: `display: inline;`
    *   `d-inline-block`: `display: inline-block;`
    *   `d-block`: `display: block;`
    *   `d-table`, `d-table-row`, `d-table-cell`: Para layout tabular.
    *   `d-flex`: `display: flex;` (Activa Flexbox).
    *   `d-inline-flex`: `display: inline-flex;`
    *   `d-grid` (Aunque el soporte completo de Grid llegó más tarde, BS4 tiene algunas utilidades).
*   **Variantes Responsive:** Puedes aplicar estas utilidades condicionalmente a partir de un breakpoint específico usando `d-BP-VALOR` (ej: `d-md-none`, `d-lg-block`, `d-sm-flex`).
    *   Las clases sin breakpoint (`d-block`) aplican a todos los tamaños.
    *   Las clases con breakpoint (`d-md-block`) aplican a ese breakpoint **y superiores**.
    *   Puedes combinar clases para crear comportamientos complejos (ej: oculto en móvil, bloque en escritorio).
*   **Utilidades para Impresión:** Controlan cómo se muestra un elemento al *imprimir* la página.
    *   `d-print-none`: Oculto al imprimir.
    *   `d-print-inline`, `d-print-block`, etc.: Aplica el display solo al imprimir.

```html
<div class="d-inline p-2 bg-primary text-white">d-inline</div>
<div class="d-inline p-2 bg-dark text-white">d-inline</div>

<span class="d-block bg-success">d-block (ocupa toda la línea)</span>
<span class="d-block bg-danger">d-block (ocupa toda la línea)</span>

{# Visible solo en pantallas medianas y superiores #}
<div class="d-none d-md-block">Visible desde MD</div>

{# Oculto en pantallas grandes y superiores #}
<div class="d-lg-none">Oculto desde LG</div>

{# Oculto al imprimir #}
<div class="d-print-none">No imprimir esto</div>

{# Visible solo al imprimir #}
<div class="d-none d-print-block">URL: https://... (solo para impresión)</div>
```
Las utilidades de display responsive son esenciales para adaptar qué elementos se muestran u ocultan en diferentes dispositivos.

## Embed (Responsive Video/Iframes)

Para hacer que elementos incrustados como `<iframe>` (ej: videos de YouTube/Vimeo, Google Maps) o `<video>` mantengan una **proporción (aspect ratio)** específica (ej: 16:9, 4:3) a medida que su contenedor cambia de tamaño.

## Flexbox Utilities (`d-flex`, `flex-*`, `justify-content-*`, `align-items-*`, etc.)

Bootstrap 4 adoptó Flexbox ampliamente. Además del sistema de rejilla (`.row` es `display: flex`), proporciona un conjunto completo de **utilidades Flexbox** para aplicar comportamientos flex directamente a cualquier elemento sin escribir CSS personalizado.

*   **Habilitar Flexbox:**
    *   `.d-flex`: Convierte el elemento en un contenedor flex de nivel de bloque.
    *   `.d-inline-flex`: Lo convierte en contenedor flex en línea.
    *   Versiones responsive: `.d-BP-flex`, `.d-BP-inline-flex`.
*   **Dirección (`flex-direction`):**
    *   `.flex-row` (defecto), `.flex-row-reverse`
    *   `.flex-column`, `.flex-column-reverse`
    *   Versiones responsive: `.flex-BP-row`, `.flex-BP-column`, etc.
*   **Justificación (Eje Principal - `justify-content`):**
    *   `.justify-content-start`, `.justify-content-end`, `.justify-content-center`, `.justify-content-between`, `.justify-content-around`.
    *   Versiones responsive: `.justify-content-BP-start`, etc.
*   **Alineación Items (Eje Transversal - `align-items`):**
    *   `.align-items-start`, `.align-items-end`, `.align-items-center`, `.align-items-baseline`, `.align-items-stretch`.
    *   Versiones responsive: `.align-items-BP-start`, etc.
*   **Alineación Item Individual (`align-self`):** (Se aplican al item flex hijo)
    *   `.align-self-start`, `.align-self-end`, `.align-self-center`, `.align-self-baseline`, `.align-self-stretch`.
    *   Versiones responsive: `.align-self-BP-start`, etc.
*   **Relleno (`flex-fill`):** (Se aplica a items flex hijos) Hace que el item ocupe el espacio disponible en el eje principal.
*   **Crecimiento/Encogimiento (`flex-grow-*`, `flex-shrink-*`):** (Se aplican a items flex hijos)
    *   `.flex-grow-0`, `.flex-grow-1`: Controla si el item puede crecer.
    *   `.flex-shrink-0`, `.flex-shrink-1`: Controla si el item puede encogerse.
    *   Versiones responsive: `.flex-BP-grow-1`, etc.
*   **Envoltura (`flex-wrap`):**
    *   `.flex-nowrap` (defecto), `.flex-wrap`, `.flex-wrap-reverse`.
    *   Versiones responsive: `.flex-BP-wrap`, etc.
*   **Orden (`order`):** (Se aplican a items flex hijos)
    *   `.order-N` (0 a 12), `.order-first`, `.order-last`.
    *   Versiones responsive: `.order-BP-N`, etc.
*   **Alineación Contenido (`align-content`):** (Para contenedores flex multilínea)
    *   `.align-content-start`, `.align-content-end`, `.align-content-center`, `.align-content-between`, `.align-content-around`, `.align-content-stretch`.
    *   Versiones responsive: `.align-content-BP-start`, etc.

```html
{# Centrar contenido vertical y horizontalmente #}
<div class="d-flex justify-content-center align-items-center" style="height: 100px; background-color: rgba(0,0,0,.1);">
  Contenido Centrado
</div>

{# Barra de navegación simple con logo a la izquierda y links a la derecha #}
<nav class="d-flex justify-content-between align-items-center p-2 bg-light">
  <span>Logo</span>
  <div>
    <a href="#" class="p-2">Link 1</a>
    <a href="#" class="p-2">Link 2</a>
  </div>
</nav>

{# Items que ocupan espacio igual #}
<div class="d-flex mt-3">
  <div class="p-2 flex-fill text-center" style="border:1px solid">Fill 1</div>
  <div class="p-2 flex-fill text-center" style="border:1px solid">Fill 2</div>
  <div class="p-2 flex-fill text-center" style="border:1px solid">Fill 3</div>
</div>

{# Item que crece para ocupar espacio #}
<div class="d-flex mt-3">
  <div class="p-2">Fijo</div>
  <div class="p-2 flex-grow-1" style="border:1px solid">Crece</div> {# Ocupa el resto #}
  <div class="p-2">Fijo</div>
</div>
```
Las utilidades Flexbox son extremadamente potentes para crear layouts complejos y alineaciones sin CSS personalizado.

## Overflow

Controla cómo se maneja el contenido que desborda las dimensiones de un elemento.

## Position (`position-*`, `fixed-top`, `sticky-top`)

Utilidades para controlar la propiedad `position` de CSS y algunas posiciones comunes.

    *   `.position-relative`
    *   `.position-absolute`
    *   `.position-fixed`
    *   `.position-sticky`
*   **Posicionamiento de Esquinas (para elementos `absolute` o `fixed`):** Bootstrap no tiene utilidades directas para `top: 0; left: 0;` etc. Necesitarías CSS personalizado o combinar con utilidades de espaciado/transform si aplica.

## Spacing (Margin y Padding) (`m-*`, `p-*`, `mt-*`, `pb-*`, `mx-*`, `py-*`, `auto`, responsive)

Las utilidades de espaciado son **extremadamente poderosas y usadas constantemente**. Permiten añadir `margin` o `padding` a un elemento de forma rápida y consistente, usando una escala predefinida basada en la variable Sass `$spacer` (que por defecto es `1rem`, usualmente `16px`).

*   **Formato:** `propiedadlados-tamaño`
    *   **Propiedad:** `m` (para `margin`) o `p` (para `padding`).
    *   **Lados:**
        *   `(ninguno)`: Aplica a los 4 lados (ej: `m-3`, `p-2`).
        *   `t`: Solo `margin-top` o `padding-top` (ej: `mt-1`, `pt-4`).
        *   `b`: Solo `margin-bottom` o `padding-bottom` (ej: `mb-5`, `pb-0`).
        *   `l`: Solo `margin-left` o `padding-left` (ej: `ml-2`, `pl-1`).
        *   `r`: Solo `margin-right` o `padding-right` (ej: `mr-auto`, `pr-3`).
        *   `x`: Eje horizontal (`left` y `right`) (ej: `mx-auto`, `px-5`).
        *   `y`: Eje vertical (`top` y `bottom`) (ej: `my-4`, `py-2`).
    *   **Tamaño:** Un número de `0` a `5` que representa múltiplos del valor de `$spacer`.
        *   `0`: Elimina el margin/padding (`0`).
        *   `1`: `$spacer * .25` (por defecto `4px`).
        *   `2`: `$spacer * .5` (por defecto `8px`).
        *   `3`: `$spacer * 1` (por defecto `16px`).
        *   `4`: `$spacer * 1.5` (por defecto `24px`).
        *   `5`: `$spacer * 3` (por defecto `48px`).
        *   `auto`: Establece el margen a `auto` (útil para centrar elementos de bloque con ancho fijo: `mx-auto`).
*   **Márgenes Negativos:** Para márgenes, también existen clases con `n` antes del tamaño (ej: `mt-n1`, `mx-n3`) para márgenes negativos.
*   **Variantes Responsive:** Puedes aplicar espaciado condicionalmente a partir de un breakpoint añadiendo el infijo del breakpoint: `propiedadlados-BP-tamaño` (ej: `mt-md-5`, `px-lg-4`, `mb-0 mb-sm-3`).

```html
{# Margen inferior de 3 unidades ($spacer * 1) #}
<div class="mb-3">...</div>

{# Padding horizontal de 5 unidades ($spacer * 3) #}
<div class="px-5">...</div>

{# Margen vertical de 2 ($spacer * .5), Padding total de 4 ($spacer * 1.5) #}
<div class="my-2 py-4 bg-light">...</div>

{# Centrar un div de ancho fijo #}
<div class="w-75 bg-dark text-white p-3 mx-auto">Centrado con mx-auto</div>

{# Margen inferior 0 en móvil, margen inferior 3 desde SM #}
<div class="mb-0 mb-sm-3">...</div>

{# Padding top 5 solo en pantallas LG y superiores #}
<div class="pt-lg-5">...</div>
```
Estas utilidades cubren la gran mayoría de las necesidades de espaciado, reduciendo la necesidad de CSS personalizado.

## Visibility (`visibility`)

Controla la visibilidad de los elementos, afectando si ocupan espacio o no.

    ```html
    </div>
    <div>
      <span>Info 1</span>
      <span>Info 2</span>
      <span>Info 3</span>
    </div>
    ```
2.  **Aplica Utilidades:** Modifica el HTML usando clases de utilidad para lograr lo siguiente:
    *   Al primer `<div>`: Añádele un padding vertical (`py-4`) y un borde inferior (`border-bottom`).
    *   Al `<h2>`: Céntralo (`text-center`) y dale un margen inferior (`mb-3`).
    *   Al `<span>` dentro del primer `<p>`: Dale un color de fondo de advertencia (`bg-warning`) y un ligero padding (`p-1`).

    *   Al segundo `<div>`: Conviértelo en un contenedor flex (`d-flex`) y justifica su contenido para que haya espacio entre los spans (`justify-content-between`). Añádele un margen superior (`mt-4`) y un padding (`p-2`) con un fondo claro (`bg-light`).
    *   A los `<span>` dentro del segundo `<div>`: Dales un color de texto apagado (`text-muted`).
3.  **Visualiza:** Abre el archivo en el navegador y comprueba que los estilos de espaciado, color, alineación y texto se han aplicado correctamente gracias a las clases de utilidad.

**Solución Propuesta:**
```html
<div class="container mt-4">

  <div class="py-4 border-bottom"> {# Padding Y, Borde inferior #}
    <h2 class="text-center mb-3">Título Sección</h2> {# Centrado, Margen B #}
    <p>Texto normal. <span class="bg-warning p-1">Algo destacado</span>.</p> {# Fondo y Padding #}
    <button type="button" class="btn btn-primary mr-2">Botón 1</button> {# Clases botón y Margen R #}
    <button type="button" class="btn btn-secondary">Botón 2</button> {# Clases botón #}
  </div>

  <div class="d-flex justify-content-between mt-4 p-2 bg-light"> {# Flex, Justify, Margen T, Padding, Fondo #}
    <span class="text-muted">Info 1</span> {# Color texto #}
    <span class="text-muted">Info 2</span>
    <span class="text-muted">Info 3</span>
  </div>

</div>
```
Este ejercicio demuestra cómo puedes construir y refinar interfaces rápidamente combinando componentes y utilidades sin escribir apenas CSS personalizado.

# 7
# Personalización y Theming con Sass

Si bien las clases y componentes por defecto de Bootstrap son muy útiles, a menudo querrás personalizar la apariencia (colores, fuentes, espaciado, radios de borde, etc.) para que coincida con la identidad visual de tu proyecto. Bootstrap 4 está construido con **Sass (Syntactically Awesome Style Sheets)**, un preprocesador CSS muy popular, lo que hace que la personalización sea potente y sistemática.

Usar Sass te permite:
*   Modificar las variables globales de Bootstrap (colores, fuentes, breakpoints, etc.).
*   Seleccionar qué componentes de Bootstrap incluir en tu compilación final para reducir el tamaño del archivo CSS.

## 7_1
## Introducción a Sass en Bootstrap

**Sass (SCSS):** Es un preprocesador CSS. Escribes código en archivos `.scss` usando una sintaxis que es un superconjunto de CSS, pero con características adicionales como:
*   **Variables:** Para almacenar valores reutilizables (ej: `$color-primario: #007bff;`).
*   **Anidamiento:** Anidar selectores CSS de forma similar a la estructura HTML, haciendo el código más legible.
*   **Mixins:** Bloques de código CSS reutilizables que pueden aceptar argumentos (similares a funciones).
*   **Herencia (`@extend`):** Compartir conjuntos de propiedades CSS entre selectores.
*   **Parciales e Importación:** Dividir tu CSS en múltiples archivos (`_partial.scss`) e importarlos en uno principal (`@import 'partial';`) para una mejor organización.

Bootstrap 4 utiliza la sintaxis **SCSS** (Sassy CSS), que es la más parecida al CSS estándar.

**¿Por qué usar Sass para personalizar Bootstrap?**
En lugar de sobrescribir los estilos de Bootstrap con tu propio CSS (lo cual puede aumentar la especificidad y el tamaño del archivo final), puedes modificar directamente las *variables* y *mapas* Sass que Bootstrap usa para generar su CSS. Al recompilar Bootstrap con tus variables modificadas, obtienes una versión personalizada con tus colores, fuentes, etc., integrados de forma nativa y eficiente.

## 7_2
## Estructura de Archivos Sass de Bootstrap

Si descargas el **código fuente** de Bootstrap (no los archivos compilados), encontrarás una estructura de directorios Sass (`scss/`) bien organizada:

*   **`bootstrap.scss`:** El archivo principal que importa todos los demás parciales necesarios. Al compilar, compilas este archivo.
*   **`_variables.scss`:** **Archivo clave.** Contiene todas las variables Sass globales de Bootstrap con sus valores por defecto (colores, espaciadores, tipografía, breakpoints, opciones, etc.). ¡Aquí es donde harás la mayoría de las personalizaciones!
*   **`_functions.scss`:** Funciones Sass usadas internamente por Bootstrap.
*   **`_mixins.scss`:** Contiene los mixins Sass reutilizables de Bootstrap (ej: para crear media queries, gradientes, etc.).
*   **`_reboot.scss`, `_type.scss`, `_images.scss`, `_code.scss`, `_grid.scss`, `_tables.scss`, `_forms.scss`, `_buttons.scss`, etc.:** Archivos parciales que definen los estilos para cada componente o sistema base. Son importados por `bootstrap.scss`.
*   **`utilities/_*.scss`:** Archivos que definen las clases de utilidad.

**Flujo de Personalización Típico:**

1.  **Obtén los fuentes Sass:** Descarga el código fuente de Bootstrap o instálalo vía npm/Yarn.
2.  **Crea tu propio archivo SCSS:** (ej: `custom.scss`).
3.  **Importa las funciones y variables de Bootstrap:** Necesitas importar *al menos* `functions` y `variables` al principio de tu archivo `custom.scss`.
4.  **Sobrescribe Variables:** **Antes** de importar el resto de Bootstrap, redefine las variables de `_variables.scss` que quieras cambiar en tu `custom.scss`.
5.  **Importa el resto de Bootstrap:** Añade `@import "ruta/a/bootstrap/scss/bootstrap";` después de tus sobrescrituras de variables.
6.  **(Opcional) Añade tu CSS personalizado:** Escribe tus propias reglas SCSS/CSS después de importar Bootstrap.
7.  **Compila `custom.scss`:** Usa tu compilador Sass para generar el archivo `custom.css` final que enlazarás en tu HTML.

**Ejemplo `custom.scss` (simplificado):**
```scss
// 1. Importar funciones (necesario para que las variables funcionen)
@import "node_modules/bootstrap/scss/functions"; // Ajusta la ruta según tu proyecto

// 2. Sobrescribir variables (¡ANTES de importar variables!)
// Cambiar colores primario y secundario
$primary: #900; // Un rojo oscuro
$secondary: #6c757d; // Mantener el gris secundario por defecto
$success: #198754; // Un verde diferente

// Cambiar fuente base
$font-family-sans-serif: "Lato", sans-serif;

// Habilitar esquinas redondeadas por defecto (Bootstrap 4 las tiene desactivadas por defecto)
// $enable-rounded: true; // Si quieres redondear todo
$border-radius: .5rem; // Radio de borde más grande
$border-radius-lg: .8rem;
$border-radius-sm: .3rem;

// 3. Importar el archivo de variables original de Bootstrap
// (Incluye los defaults y tus sobrescrituras anteriores)
@import "node_modules/bootstrap/scss/variables";

// 4. (Opcional) Importar solo los componentes que necesites
// En lugar de importar todo bootstrap.scss, puedes importar partes:
// @import "node_modules/bootstrap/scss/root";
// @import "node_modules/bootstrap/scss/reboot";
// @import "node_modules/bootstrap/scss/type";
// @import "node_modules/bootstrap/scss/grid";
// @import "node_modules/bootstrap/scss/buttons";
// @import "node_modules/bootstrap/scss/card";
// @import "node_modules/bootstrap/scss/nav";
// @import "node_modules/bootstrap/scss/navbar";
// @import "node_modules/bootstrap/scss/utilities"; // Importante para utilidades

// O importar todo Bootstrap (más simple para empezar)
@import "node_modules/bootstrap/scss/bootstrap";

// 5. Añadir tus propias reglas CSS/SCSS personalizadas aquí
body {
  // Tu estilo personalizado para el body
  background-color: $light; // Usa una variable Bootstrap
}

.mi-componente-custom {
  padding: $spacer * 2; // Usa la variable $spacer de Bootstrap
  border: 1px solid $primary; // Usa tu color primario personalizado
}
```

### 7_3_1
### Colores (`$theme-colors`)

Bootstrap define sus colores principales en un **mapa Sass** llamado `$theme-colors`. Puedes añadir, eliminar o modificar colores en este mapa.

```scss
// --- En tu custom.scss (ANTES de importar variables/bootstrap) ---

// Obtener el mapa de colores por defecto
$theme-colors: map-get($theme-colors,()); // No necesario si importas variables, pero como ejemplo

// Modificar un color existente
$theme-colors: map-merge(
  $theme-colors,
  (
    "primary": #525ddc, // Cambiar el azul primario
    "danger": #d9534f // Cambiar el rojo peligro
  )
);

// Añadir un nuevo color temático
$theme-colors: map-merge(
  $theme-colors,
  (
    "tertiary": #ffc107 // Añadir un color terciario (amarillo)
  )
);

// Ahora, clases como .btn-primary, .bg-danger, .text-tertiary usarán estos colores.

// También puedes sobrescribir variables individuales de color
// $blue: #0d6efd;
// $red: #dc3545;
// $mi-color-custom: #db3e8a;
```
Modificar `$theme-colors` actualizará automáticamente todos los componentes y utilidades que dependen de esos colores (botones, alertas, fondos, texto, etc.).

[Volver a la tabla de contenido](#Contenido)
### 7_3_2
### Espaciado (`$spacers`)

Bootstrap genera las utilidades de espaciado (`m-*`, `p-*`) a partir de un mapa Sass llamado `$spacers`, que se basa en la variable `$spacer` (defecto `1rem`). Puedes modificar `$spacer` o añadir/modificar claves en el mapa `$spacers`.

```scss
// --- En tu custom.scss (ANTES de importar variables/bootstrap) ---

// Cambiar el espaciador base (afecta a todos los múltiplos)
$spacer: 1.25rem; // Hacer todos los espacios un 25% más grandes

// Modificar o añadir tamaños específicos al mapa $spacers
// (Obtener el mapa por defecto primero)
$spacers: map-get($spacers,());

// Añadir un tamaño 6
$spacers: map-merge(
  $spacers,
  (
    6: ($spacer * 4) // Añade clases .m-6, .p-6, .mt-6, etc. = 4 * $spacer
  )
);

// Modificar el tamaño 3 (hacerlo más pequeño)
$spacers: map-merge(
  $spacers,
  (
    3: ($spacer * 0.75) // Modifica .m-3, .p-3, etc.
  )
);
```

## 7_4
## Compilación de Sass a CSS (Herramientas)

Una vez que has creado tu archivo `custom.scss` (o como lo llames) con tus importaciones y personalizaciones, necesitas **compilarlo** a un archivo CSS estándar que el navegador pueda entender.

**Opciones Populares:**

1.  **Compilador Sass de Línea de Comandos (Dart Sass):**
    *   Instalación: Sigue las instrucciones en [sass-lang.com](https://sass-lang.com/install) (npm, descarga directa, etc.).
    *   Uso Básico:
        ```bash
        # Compila custom.scss a public/css/custom.css y observa cambios
        sass --watch custom.scss public/css/custom.css
        # Compilar una vez para producción (comprimido)
        sass --style=compressed custom.scss public/css/custom.min.css
        ```
2.  **Node Sass (vía npm/Node.js):**
    *   Instalación: `npm install node-sass --save-dev`
    *   Uso (a menudo a través de scripts npm en `package.json`):
        ```json
        // package.json
        "scripts": {
          "compile:sass": "node-sass custom.scss public/css/custom.css -w", // -w para watch
          "build:sass": "node-sass custom.scss public/css/custom.min.css --output-style compressed"
        }
        ```
        Luego ejecutas `npm run compile:sass` o `npm run build:sass`.
3.  **Herramientas de Build (Webpack, Parcel, Gulp):**
    *   Si ya usas un empaquetador de módulos o un task runner, estos suelen tener plugins o loaders para compilar Sass como parte de tu proceso de build (ej: `sass-loader` para Webpack). Esta es la opción más común en proyectos frontend complejos.
4.  **Extensiones de Editor (VS Code, etc.):**
    *   Extensiones como "Live Sass Compiler" para VS Code pueden compilar automáticamente tus archivos `.scss` a `.css` cada vez que guardas. Muy conveniente para desarrollo rápido.

**Resultado:** Sea cual sea la herramienta, el objetivo es generar un archivo `custom.css` (o `custom.min.css` para producción) que contenga todos los estilos de Bootstrap *con tus personalizaciones aplicadas*, más tus propias reglas CSS. Luego, enlazas **solo este archivo CSS compilado** en tu HTML, en lugar del CSS original de Bootstrap.

```html
<!-- En tu HTML (reemplaza el enlace CDN o el bootstrap.min.css descargado) -->
<link rel="stylesheet" href="public/css/custom.min.css"> {# Tu CSS compilado y personalizado #}
```

## 7_5
## Añadir Reglas CSS Personalizadas sobre Bootstrap

Incluso si personalizas Bootstrap extensamente con Sass, casi siempre necesitarás añadir tus propias reglas CSS para estilos específicos de tu diseño que no cubre Bootstrap o para anular algún estilo de Bootstrap de forma muy puntual.

**Dónde añadirlas:**

*   **En tu archivo `custom.scss`:** Escribe tus reglas CSS/SCSS **después** de la línea `@import ".../bootstrap";`. Esto asegura que tus reglas tengan mayor especificidad o aparezcan más tarde en la cascada que las de Bootstrap, permitiéndote anularlas si es necesario. Puedes usar variables y mixins de Bootstrap en tu CSS personalizado si las importaste correctamente.
    ```scss
    // custom.scss
    // ... (importaciones y variables sobrescritas) ...

    // Importar todo Bootstrap (o los componentes necesarios)
    @import "node_modules/bootstrap/scss/bootstrap";

    // --- Tus estilos personalizados AQUÍ ---
    body {
      font-size: 1.1rem; // Anula el font-size base si es necesario
    }

    .mi-clase-especifica {
      background-color: $warning; // Usa un color de Bootstrap
      border: 2px solid darken($warning, 20%); // Usa función Sass y color Bootstrap
      padding: $spacer * 1.5; // Usa variable Bootstrap

      &:hover { // Anidamiento Sass
        background-color: darken($warning, 10%);
      }
    }
    ```
*   **En un Archivo CSS Separado:** Puedes mantener un archivo CSS completamente separado (ej: `extra-styles.css`) y enlazarlo en tu HTML **después** del CSS de Bootstrap (o tu `custom.css` compilado). Las reglas en este archivo anularán las de Bootstrap si tienen la misma o mayor especificidad debido al orden de aparición.
    ```html
    <head>
      ...
      <link rel="stylesheet" href="public/css/custom.min.css"> {# Bootstrap personalizado #}
      <link rel="stylesheet" href="public/css/extra-styles.css"> {# Tus anulaciones/adiciones #}
      ...
    </head>
    ```

**Recomendación:** Si estás usando Sass para personalizar Bootstrap, lo más limpio y potente suele ser incluir tus estilos personalizados en el mismo archivo `.scss` después de importar Bootstrap, aprovechando así las variables y mixins. Reserva un archivo CSS separado solo si tienes una razón específica o si no estás usando el flujo de compilación Sass.

# JavaScript en Bootstrap 4

Muchos de los componentes interactivos de Bootstrap (como desplegables, modales, carruseles, tooltips, etc.) requieren JavaScript para funcionar. Bootstrap 4 proporciona un conjunto de **plugins JavaScript** construidos sobre la popular librería **jQuery**. Además, estos plugins dependen de **Popper.js** para el posicionamiento preciso de elementos emergentes como tooltips, popovers y menús desplegables.

[Volver a la tabla de contenido](#Contenido)
## 8_1
## Visión General (Dependencias jQuery/Popper, Data Attributes, API JS)

**Dependencias:**
*   **jQuery:** Bootstrap 4 **requiere jQuery**. Debes incluir jQuery (la versión slim es suficiente) en tu página **antes** de incluir el archivo JavaScript de Bootstrap.
*   **Popper.js:** Necesario para el posicionamiento de tooltips, popovers y dropdowns. Debe incluirse **después** de jQuery pero **antes** de `bootstrap.js`. Si usas el archivo `bootstrap.bundle.js` (o `bootstrap.bundle.min.js`), Popper.js ya viene incluido y no necesitas añadirlo por separado.

**Formas de Uso:**

Bootstrap ofrece dos maneras principales de activar y usar sus componentes JavaScript:

1.  **Mediante Atributos `data-*` (Data Attributes):**
    *   Es la forma **más fácil y común** de usar los plugins. Simplemente añades atributos especiales `data-*` a tus elementos HTML (como `data-toggle`, `data-target`, `data-slide`, `data-dismiss`, `data-ride`).
    *   Bootstrap detecta automáticamente estos atributos al cargar la página y "activa" la funcionalidad del componente correspondiente sin que necesites escribir ni una línea de JavaScript.
    *   Es declarativo y mantiene la lógica fuera del JS.

2.  **Mediante la API JavaScript:**
    *   Puedes inicializar, controlar y responder a eventos de los componentes Bootstrap mediante su API programática JavaScript.
    *   Esto te da **más control** sobre el comportamiento, permite activar componentes dinámicamente o integrar Bootstrap con tu propio código JavaScript.
    *   Implica seleccionar el elemento con JavaScript/jQuery y llamar a los métodos del plugin correspondiente (ej: `$('#miModal').modal('show')`, `$('.miTooltip').tooltip({ delay: 500 })`).

**¿Cuál usar?**
Empieza con los atributos `data-*`. Son suficientes para la mayoría de los casos y más sencillos. Recurre a la API JavaScript cuando necesites:
*   Inicializar componentes en contenido cargado dinámicamente (AJAX).
*   Controlar los componentes de forma programática (ej: abrir un modal desde tu código JS).
*   Personalizar opciones avanzadas que no están disponibles vía `data-*`.
*   Escuchar y reaccionar a los eventos disparados por los componentes Bootstrap.

## 8_2
## Uso mediante Atributos Data (`data-*`)

Esta es la forma preferida para la mayoría de los casos. Bootstrap utiliza atributos HTML `data-*` para configurar y activar sus plugins JS.

**Atributos Clave Comunes:**

*   **`data-toggle="componente"`:** El atributo principal que le dice a Bootstrap qué componente activar en este elemento. Ejemplos: `data-toggle="modal"`, `data-toggle="dropdown"`, `data-toggle="collapse"`, `data-toggle="tooltip"`, `data-toggle="popover"`, `data-toggle="tab"`, `data-toggle="pill"`.
*   **`data-target="#id" | .selector`:** Especifica el elemento objetivo sobre el cual actuará el componente (ej: el `<div>` del modal a abrir, el `<div>` del contenido a colapsar). Usa un selector CSS válido.
*   **`href="#id"`:** A menudo, en elementos `<a>`, el `href` puede usarse en lugar de `data-target` para apuntar al elemento objetivo.
*   **`data-dismiss="componente"`:** Usado en botones dentro de componentes como Alertas (`alert`) o Modales (`modal`) para indicar que ese botón debe cerrar el componente.
*   **`data-ride="carousel"`:** Usado en el contenedor principal del carrusel para iniciarlo automáticamente al cargar la página.
*   **`data-slide="prev|next"`:** Usado en los controles de navegación del carrusel.
*   **`data-slide-to="N"`:** Usado en los indicadores del carrusel para ir a una slide específica (N es el índice).
*   **`data-placement="top|bottom|left|right"`:** Para Tooltips y Popovers, define dónde aparecen.
*   **`data-content="..."`:** Para Popovers, define el contenido.
*   **`data-trigger="click|hover|focus|manual"`:** Para Tooltips y Popovers, define cómo se activan.
*   **`data-delay='{"show":500, "hide":100}'`:** Para Tooltips/Popovers, configura retardos (en ms).
*   **`data-offset="N"`:** Para Scrollspy y otros, ajusta el cálculo de posición.
*   **`data-autohide="false"` / `data-delay="N"`:** Para Toasts.

**Ejemplos (ya vistos en unidades anteriores):**

*   **Modal:** El botón activador usa `data-toggle="modal"` y `data-target="#miModal"`. El botón de cierre dentro del modal usa `data-dismiss="modal"`.
*   **Dropdown:** El botón activador usa `data-toggle="dropdown"`.
*   **Collapse/Acordeón:** El activador usa `data-toggle="collapse"` y `data-target="#miContenido"`. Los items del acordeón usan `data-parent="#idAcordeon"`.
*   **Alert Descartable:** El botón 'x' usa `data-dismiss="alert"`.
*   **Carousel:** El contenedor tiene `data-ride="carousel"`. Los controles usan `data-slide="prev/next"`. Los indicadores usan `data-slide-to="N"`.
*   **Tooltip/Popover:** El elemento tiene `data-toggle="tooltip/popover"`, `title="..."`, `data-content="..."`, `data-placement="..."`.

Simplemente añadiendo estos atributos al HTML correcto, Bootstrap se encarga de la funcionalidad JavaScript subyacente.

    *   **Métodos de Acción:** Después de inicializar, puedes llamar a métodos específicos para controlar el componente: `$('#miModal').modal('show')`, `$('#miModal').modal('hide')`, `$('#miCarrusel').carousel('next')`, `$('#miAlerta').alert('close')`.

### 8_3_1
### Constructores y Métodos

Cada plugin JS de Bootstrap se puede invocar como una función en un selector jQuery.

**Inicialización con Opciones:**
```javascript
$(function () { // Asegura que el DOM esté listo

  // Inicializar todos los tooltips con opciones personalizadas
  $('[data-toggle="tooltip"]').tooltip({
    delay: { "show": 500, "hide": 100 }, // Retraso al mostrar/ocultar
    html: true // Permitir HTML en el título
  });

  // Inicializar un modal específico con opciones
  $('#miModalEspecifico').modal({
    keyboard: false, // No cerrar con Esc
    backdrop: 'static' // No cerrar al hacer clic fuera
  });

  // Inicializar un carrusel para que no empiece automáticamente
  $('#miCarruselManual').carousel({
    interval: false // No ciclar automáticamente
  });

});
```

**Llamada a Métodos:**
Después de la inicialización (o si usas `data-` attributes, Bootstrap lo inicializa por ti), puedes llamar a métodos específicos pasando el nombre del método como string.

```javascript
// Abrir un modal desde JS
$('#botonAbrirModal').on('click', function() {
  $('#modalLogin').modal('show');
});

// Cerrar un modal desde JS
$('#botonCerrarModal').on('click', function() {
  $('#modalLogin').modal('hide');
});

// Ir a la siguiente slide de un carrusel
$('#botonSiguienteSlide').on('click', function() {
  $('#miCarruselPrincipal').carousel('next');
});

// Ir a la slide número 2 (índice 1)
$('#botonIrSlide2').on('click', function() {
  $('#miCarruselPrincipal').carousel(1); // Pasar el índice
});

// Cerrar una alerta específica
$('#botonCerrarAlerta').on('click', function() {
  $('#alertaImportante').alert('close');
});

// Mostrar/Ocultar un colapsable
$('#botonToggleColapso').on('click', function() {
  $('#miContenidoColapsable').collapse('toggle');
});

// Activar una pestaña específica
$('#botonActivarTab2').on('click', function(e){
  e.preventDefault(); // Prevenir acción por defecto si es un enlace
  $('#tab-2-link').tab('show'); // Asumiendo que el enlace/botón de la tab tiene id="tab-2-link"
});
```
Consulta la documentación de Bootstrap 4 para cada componente para ver la lista completa de opciones y métodos disponibles.

### 8_3_2
### Eventos

Los componentes JavaScript de Bootstrap disparan eventos en ciertos puntos de su ciclo de vida. Puedes escuchar estos eventos usando el método `.on()` de jQuery para ejecutar tu propio código en respuesta.

Los eventos suelen tener dos versiones:
*   Una que se dispara **antes** de que la acción comience (ej: `show.bs.modal`). Permite cancelar la acción llamando a `event.preventDefault()`.
*   Otra que se dispara **después** de que la acción se ha completado (ej: `shown.bs.modal`).

**Eventos Comunes:**

*   **Modal:** `show.bs.modal`, `shown.bs.modal`, `hide.bs.modal`, `hidden.bs.modal`
*   **Dropdown:** `show.bs.dropdown`, `shown.bs.dropdown`, `hide.bs.dropdown`, `hidden.bs.dropdown`
*   **Tab/Pill:** `show.bs.tab`, `shown.bs.tab`, `hide.bs.tab`, `hidden.bs.tab`
*   **Alert:** `close.bs.alert`, `closed.bs.alert`
*   **Collapse:** `show.bs.collapse`, `shown.bs.collapse`, `hide.bs.collapse`, `hidden.bs.collapse`
*   **Carousel:** `slide.bs.carousel` (antes de transición), `slid.bs.carousel` (después de transición)

**Ejemplo: Ejecutar código cuando un modal se muestra completamente:**
```javascript
$(function () {
  $('#miModalInfo').on('shown.bs.modal', function (event) {
    // El modal ya está visible
    console.log('Modal de información mostrado!');
    // Podrías hacer una petición AJAX aquí para cargar contenido, por ejemplo
    const botonActivador = $(event.relatedTarget); // Botón que abrió el modal (si lo hubo)
    console.log('Abierto por:', botonActivador.text());
  });

  $('#miModalConfirmacion').on('hide.bs.modal', function (event) {
    // Justo antes de que el modal empiece a ocultarse
    // const confirmar = confirm("¿Estás seguro de que quieres cerrar?");
    // if (!confirmar) {
    //   event.preventDefault(); // Previene que el modal se cierre
    // }
    console.log("Modal de confirmación ocultándose...");
  });
});
```
Los eventos te permiten integrar la lógica de los componentes Bootstrap con el resto de tu aplicación JavaScript.

## Revisión de Componentes JS (Alert, Button, Carousel, Collapse, Dropdown, Modal, Popover, Scrollspy, Tab, Toast, Tooltip)

Resumen rápido de los componentes que dependen de JavaScript:

*   **Alert:** Para el descarte (método `close`, eventos `close/closed`).
*   **Button:** Para estados toggle (`data-toggle="buttons"`) o comportamiento tipo radio/checkbox en grupos de botones, y estado de carga (método `button('toggle')`, `button('loading')` - aunque 'loading' fue menos robusto).
*   **Carousel:** Funcionalidad completa (ciclo, controles, indicadores, pausa). Métodos: `cycle`, `pause`, `prev`, `next`, número de slide, `dispose`. Eventos: `slide`, `slid`.
*   **Collapse:** Mostrar/ocultar contenido, funcionalidad de acordeón. Métodos: `toggle`, `show`, `hide`, `dispose`. Eventos: `show`, `shown`, `hide`, `hidden`.
*   **Dropdown:** Mostrar/ocultar menús desplegables. Métodos: `toggle`, `show`, `hide`, `update`, `dispose`. Eventos: `show`, `shown`, `hide`, `hidden`.
*   **Modal:** Mostrar/ocultar diálogos modales. Métodos: `toggle`, `show`, `hide`, `handleUpdate`, `dispose`. Eventos: `show`, `shown`, `hide`, `hidden`.
*   **Popover:** Mostrar/ocultar popovers. Requiere inicialización. Métodos: `show`, `hide`, `toggle`, `dispose`, `enable`, `disable`, `toggleEnabled`, `update`. Eventos: `show`, `shown`, `hide`, `hidden`, `inserted`. Depende de Popper.js.
*   **Scrollspy:** Resaltar navegación basada en la posición de scroll. Requiere inicialización si no usas `data-spy`. Métodos: `refresh`, `dispose`. Evento: `activate.bs.scrollspy`.
*   **Tab/Pill:** Cambiar entre paneles de contenido asociados a pestañas/píldoras de navegación. Métodos: `show`, `dispose`. Eventos: `show`, `shown`, `hide`, `hidden`.
*   **Toast:** Mostrar/ocultar notificaciones toast. Requiere inicialización. Métodos: `show`, `hide`, `dispose`. Eventos: `show`, `shown`, `hide`, `hidden`.
*   **Tooltip:** Mostrar/ocultar tooltips. Requiere inicialización. Métodos: `show`, `hide`, `toggle`, `dispose`, `enable`, `disable`, `toggleEnabled`, `update`. Eventos: `show`, `shown`, `hide`, `hidden`, `inserted`. Depende de Popper.js.

**Objetivo:** Usar la API JavaScript para inicializar tooltips y mostrar/ocultar un modal programáticamente.

4.  **Dentro del script:**

    *   **Inicializa el tooltip:** Selecciona el botón con ID `btnTooltip` usando jQuery (`$('#btnTooltip')`) y llama al método `.tooltip()`.
    *   **Abre el modal:** Añade un event listener de clic al botón `#btnAbrirModal`. Cuando se haga clic, llama a `$('#miModalJS').modal('show');`.
    *   **Cierra el modal:** Añade un event listener de clic al botón `#btnCerrarModalDesdeJS` (que está *dentro* del modal). Cuando se haga clic, llama a `$('#miModalJS').modal('hide');`.

5.  **Prueba:**

**Solución Propuesta (HTML relevante y Script):**
```html
<!-- En el body -->
<div class="container mt-5 text-center">
    <button type="button" id="btnTooltip" class="btn btn-secondary"
            data-toggle="tooltip" data-placement="top"
            title="Este es mi tooltip JS">
      Hover para Tooltip (Init con JS)
    </button>

    <button type="button" id="btnAbrirModal" class="btn btn-primary ml-3">
      Abrir Modal con JS
    </button>
</div>

<!-- Modal -->
<div class="modal fade" id="miModalJS" tabindex="-1" aria-labelledby="miModalJSLabel" aria-hidden="true">
  <div class="modal-dialog">
    <div class="modal-content">
      <div class="modal-header">
        <h5 class="modal-title" id="miModalJSLabel">Modal JS</h5>
        <button type="button" class="close" data-dismiss="modal" aria-label="Close">
          <span aria-hidden="true">&times;</span>
        </button>
      </div>
      <div class="modal-body">
        Contenido del modal controlado por JavaScript.
      </div>
      <div class="modal-footer">
        <button type="button" class="btn btn-secondary" data-dismiss="modal">Cerrar (data-dismiss)</button>
        <button type="button" id="btnCerrarModalDesdeJS" class="btn btn-warning">Cerrar con JS</button>
      </div>
    </div>
  </div>
</div>

<!-- Scripts (jQuery y Bootstrap Bundle ya incluidos antes) -->
<script>
  $(function () { // Document ready

    // 1. Inicializar Tooltip
    $('#btnTooltip').tooltip();
    console.log('Tooltip inicializado');

    // 2. Abrir Modal al hacer clic en el botón
    $('#btnAbrirModal').on('click', function () {
      console.log('Abriendo modal...');
      $('#miModalJS').modal('show');
    });

    // 3. Cerrar Modal al hacer clic en el botón interno
    $('#btnCerrarModalDesdeJS').on('click', function () {
      console.log('Cerrando modal desde botón JS...');
      $('#miModalJS').modal('hide');
    });

    // 4. (Opcional) Escuchar evento 'shown' del modal
    $('#miModalJS').on('shown.bs.modal', function (event) {
      console.log('¡El modal se ha mostrado completamente!');
      // console.log('Evento:', event);
    });

     // 5. (Opcional) Escuchar evento 'hidden' del modal
     $('#miModalJS').on('hidden.bs.modal', function (event) {
      console.log('El modal se ha ocultado completamente.');
    });

  });
</script>
```
