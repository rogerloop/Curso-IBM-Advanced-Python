## 1_Introduccion_a_Javascript

### 1_1_Que_es_Javascript

JavaScript (a menudo abreviado como JS) es un lenguaje de programación **interpretado**, **orientado a objetos**, **basado en prototipos**, **débilmente tipado** y **dinámico**. Es uno de los tres pilares fundamentales de la tecnología web, junto con HTML y CSS.

*   **HTML:** Define la estructura y el contenido de la página web.
*   **CSS:** Define la presentación y el estilo visual de la página web.
*   **JavaScript:** Añade **interactividad**, comportamiento dinámico y lógica de programación a la página web. Permite manipular el contenido HTML (DOM), reaccionar a eventos del usuario (clics, teclado), realizar peticiones a servidores (AJAX/Fetch), crear animaciones complejas y mucho más.

Aunque inicialmente se diseñó para ejecutarse en el navegador web (lado del cliente), ahora también se utiliza ampliamente en el lado del servidor (Node.js), desarrollo de aplicaciones móviles, escritorio y otras áreas.

### 1_2_Historia_y_Evolucion_ECMAScript

*   JavaScript fue creado por Brendan Eich en Netscape en 1995, originalmente llamado Mocha, luego LiveScript y finalmente JavaScript.
*   Para estandarizar el lenguaje, Netscape lo presentó a ECMA International, lo que llevó a la creación del estándar **ECMAScript**. JavaScript es la implementación más popular de ECMAScript.
*   Versiones clave de ECMAScript:
    *   ES1 (1997), ES2 (1998), ES3 (1999): Bases iniciales.
    *   ES5 (2009): Importante actualización con `strict mode`, soporte JSON, nuevos métodos de Array, etc.
    *   **ES6 / ECMAScript 2015:** La mayor actualización hasta la fecha. Introdujo `let`, `const`, funciones flecha, clases, módulos, promesas, desestructuración, template literals, y mucho más. Marcó un antes y un después.
    *   ES2016, ES2017, ..., ESNext: A partir de ES6, las actualizaciones son anuales, añadiendo nuevas características de forma incremental.

Es importante conocer las características de ES6+ ya que son el estándar moderno para escribir JavaScript.

### 1_3_Javascript_en_el_Navegador_y_Fuera_Node_js

*   **En el Navegador (Client-Side):** Es su entorno original. Interactúa con el DOM (Document Object Model) y el BOM (Browser Object Model) para manipular la página, responder a eventos y comunicarse con APIs del navegador (Fetch, Geolocation, Web Storage, etc.).
*   **Fuera del Navegador (Server-Side, etc.):** Gracias a entornos como **Node.js**, JavaScript puede ejecutarse fuera del navegador. Node.js permite construir servidores web, APIs, herramientas de línea de comandos, aplicaciones de escritorio (con Electron), etc., usando JavaScript. Tiene acceso al sistema de archivos, redes y otras funcionalidades del sistema operativo que no están disponibles en el navegador.

Este curso se centrará principalmente en JavaScript para el navegador, pero muchos de los conceptos fundamentales del lenguaje son aplicables a cualquier entorno.

### 1_4_Como_Incluir_Javascript_en_HTML

Hay tres formas principales, similar a CSS:

1.  **Script Externo (Recomendado):**
    *   Escribes tu código JS en un archivo separado con extensión `.js` (ej: `script.js`).
    *   Lo enlazas desde tu HTML usando la etiqueta `<script>` con el atributo `src`. Generalmente se coloca justo antes de la etiqueta de cierre `</body>` para asegurar que el HTML se haya cargado antes de que el script intente manipularlo.
    ```html
    <!DOCTYPE html>
    <html>
    <head>
      <title>JS Externo</title>
      <link rel="stylesheet" href="style.css">
    </head>
    <body>
      <h1>Contenido</h1>
      <button id="miBoton">Haz Clic</button>

      <script src="script.js"></script> <!-- Enlace al final del body -->
    </body>
    </html>
    ```
    *   **Atributos `async` y `defer`:** Para scripts en el `<head>`:
        *   `<script async src="script.js"></script>`: Descarga el script asíncronamente y lo ejecuta tan pronto como se descarga, sin esperar al HTML y sin garantizar orden respecto a otros scripts `async`. Puede interrumpir el parseo del HTML.
        *   `<script defer src="script.js"></script>`: Descarga el script asíncronamente pero **garantiza** que se ejecutará *después* de que el HTML haya sido parseado completamente, y en el orden en que aparecen los scripts `defer`. **Generalmente preferido si pones scripts en el `<head>`**.

2.  **Script Interno:**
    *   Escribes el código JS directamente dentro de etiquetas `<script>` en el HTML (ya sea en `<head>` o `<body>`).
    ```html
    <!DOCTYPE html>
    <html>
    <head><title>JS Interno</title></head>
    <body>
      <button onclick="alert('¡Hola desde interno!')">Clic Interno</button>

      <script>
        console.log("Hola desde script interno en el body");
        function miFuncionInterna() {
          // ...
        }
      </script>
    </body>
    </html>
    ```
    *   Útil para scripts muy pequeños o específicos de una página, pero menos mantenible y reutilizable.

3.  **JavaScript en Línea (Inline - Atributos de Evento):**
    *   Se usa código JS directamente en atributos HTML como `onclick`, `onmouseover`, etc.
    ```html
    <button onclick="console.log('¡Clic!'); alert('¡Hola!');">Clic Inline</button>
    ```
    *   **Práctica muy desaconsejada:** Mezcla comportamiento con estructura, difícil de mantener, problemas de scope, seguridad (CSP). Es mucho mejor usar `addEventListener` desde un script externo o interno.

### 1_5_La_Consola_del_Navegador

Es una herramienta **fundamental** para cualquier desarrollador JavaScript. Se accede a través de las Herramientas de Desarrollador del navegador (normalmente presionando F12).

Usos principales:
*   **Mostrar Mensajes:** Usando `console.log()`, `console.warn()`, `console.error()`, `console.info()`. Esencial para depuración y ver el estado de las variables.
*   **Ejecutar Código JS:** Puedes escribir y ejecutar código JavaScript directamente en la consola.
*   **Ver Errores:** Muestra los errores de sintaxis o de ejecución que ocurren en tus scripts.
*   **Interactuar con el DOM:** Puedes usar comandos como `document.getElementById(...)` directamente en la consola.

```javascript
console.log("Este mensaje aparecerá en la consola.");
let miVariable = 10;
console.log("El valor de miVariable es:", miVariable);
console.warn("Esto es una advertencia.");
console.error("Esto simula un error.");
```
¡Acostúmbrate a usar la consola constantemente!

### 1_6_Comentarios_en_Javascript

Los comentarios se usan para explicar el código o para desactivar temporalmente líneas. Son ignorados por el intérprete de JavaScript.

*   **Comentario de una línea:** Comienza con `//`. Todo lo que sigue en esa línea es un comentario.
    ```javascript
    // Esto es un comentario de una línea
    let x = 5; // Asigna 5 a x
    ```
*   **Comentario multilínea:** Comienza con `/*` y termina con `*/`. Puede abarcar varias líneas.
    ```javascript
    /*
      Este es un comentario
      que ocupa varias
      líneas. Útil para explicaciones
      más largas.
    */
    let y = 10;
    ```

## 2_Fundamentos_Variables_Tipos_y_Operadores

### 2_1_Variables_var_let_const

Las variables son contenedores para almacenar datos. En JavaScript moderno (ES6+), se usan principalmente `let` y `const`.

*   **`let`:**
    *   Declara una variable cuyo valor **puede ser reasignado** posteriormente.
    *   Tiene **ámbito de bloque** (`{...}`). Solo existe dentro del bloque donde se declaró (incluyendo bucles `for`, bloques `if`, etc.).
    *   No permite redeclarar la misma variable en el mismo ámbito.
    *   No sufre de *hoisting* de la misma manera que `var` (no se puede usar antes de su declaración).
    ```javascript
    let edad = 30;
    edad = 31; // OK: Reasignación permitida

    if (true) {
      let mensajeBloque = "Hola";
      console.log(mensajeBloque); // OK
    }
    // console.log(mensajeBloque); // Error: mensajeBloque no está definido aquí
    ```
*   **`const`:**
    *   Declara una variable cuyo valor **NO puede ser reasignado** una vez inicializado. Es una constante.
    *   También tiene **ámbito de bloque**.
    *   Debe ser inicializada en el momento de la declaración.
    *   **Importante:** Si la constante almacena un objeto o un array, el *contenido* del objeto/array *sí* puede modificarse, pero la variable no puede apuntar a un objeto/array diferente.
    ```javascript
    const PI = 3.14159;
    // PI = 3.14; // Error: Assignment to constant variable.

    const persona = { nombre: "Ana" };
    persona.nombre = "Elena"; // OK: Modificar propiedad del objeto
    persona.edad = 25;      // OK: Añadir propiedad
    // persona = { nombre: "Luis" }; // Error: No se puede reasignar la referencia

    const colores = ["rojo", "verde"];
    colores.push("azul"); // OK: Modificar el array
    // colores = ["amarillo"]; // Error: No se puede reasignar la referencia
    ```
    *   **Usa `const` por defecto** a menos que sepas que necesitarás reasignar la variable; en ese caso, usa `let`.
*   **`var` (Forma antigua - Evitar en código moderno):**
    *   Forma original de declarar variables.
    *   Tiene **ámbito de función** o ámbito global, *no* de bloque. Esto puede llevar a comportamientos inesperados en bucles o condicionales.
    *   Permite redeclarar la misma variable en el mismo ámbito.
    *   Sufre de *hoisting* (la declaración se "mueve" al inicio de su ámbito, pero no la inicialización).
    ```javascript
    function ejemploVar() {
      if (true) {
        var mensaje = "Hola Var";
      }
      console.log(mensaje); // "Hola Var" - ¡Accesible fuera del bloque if!
      var mensaje = "Adiós Var"; // Redeclaración permitida
      console.log(mensaje); // "Adiós Var"
    }
    ejemploVar();

    console.log(variableHoist); // undefined (declaración elevada, pero no asignación)
    var variableHoist = "Hoisted";
    ```
    **Conclusión:** Usa `let` y `const` en tu código nuevo. Olvida `var` a menos que trabajes con código muy antiguo.

### 2_2_Tipos_de_Datos_Primitivos

Son los tipos de datos más básicos e inmutables (su valor no puede ser alterado directamente).

*   **`string` (Cadena de texto):** Secuencia de caracteres. Se definen con comillas simples (`'...'`), dobles (`"..."`) o backticks (`` `...` `` - template literals, ver ES6).
    ```javascript
    let saludo = "Hola Mundo";
    let nombre = 'Ana';
    let frase = `Ella dijo: "${saludo}"`; // Template literals permiten interpolación y multilínea
    ```
*   **`number` (Número):** Representa tanto números enteros como de punto flotante (decimales). También incluye valores especiales como `Infinity`, `-Infinity` y `NaN` (Not a Number).
    ```javascript
    let cantidad = 10;
    let precio = 99.95;
    let noEsNumero = "abc" / 2; // Resultado es NaN
    ```
*   **`boolean` (Booleano):** Representa un valor lógico: `true` (verdadero) o `false` (falso). Usado extensamente en condicionales.
    ```javascript
    let activo = true;
    let mayorDeEdad = false;
    ```
*   **`undefined`:** Tipo y valor especial que indica que una variable ha sido declarada pero aún no se le ha asignado un valor. También es el valor devuelto por funciones que no tienen `return`.
    ```javascript
    let sinValor;
    console.log(sinValor); // undefined
    ```
*   **`null`:** Tipo y valor especial que representa la ausencia intencional de un valor de objeto. Se asigna explícitamente.
    ```javascript
    let usuario = null; // Indica que no hay usuario seleccionado (intencionalmente)
    ```
    *(Diferencia clave: `undefined` es "aún no definido", `null` es "definido como vacío/nada")*
*   **`symbol` (ES6):** Tipo de dato único e inmutable, usado a menudo como clave de propiedad en objetos para evitar colisiones de nombres.
    ```javascript
    const idUnico = Symbol('descripcionOpcional');
    ```
*   **`bigint` (ES2020):** Para representar números enteros de precisión arbitraria (más grandes de lo que `number` puede manejar de forma segura). Se crean añadiendo una `n` al final del número o con la función `BigInt()`.
    ```javascript
    const numeroMuyGrande = 900719925474099112345n;
    ```

### 2_3_Tipo_de_Dato_Object

Cualquier valor que no sea primitivo es un objeto. Los objetos son colecciones de pares clave-valor (propiedades y métodos). Son mutables.

Incluye:
*   **Objetos literales:** `{ clave: valor, ... }`
*   **Arrays:** Colecciones ordenadas de valores `[valor1, valor2, ...]` (técnicamente, son un tipo especial de objeto).
*   **Funciones:** Bloques de código reutilizables (también son objetos en JS).
*   **Date, RegExp, Error:** Objetos incorporados para fechas, expresiones regulares y errores.
*   Y muchos más...

```javascript
let coche = { marca: "Toyota", modelo: "Corolla", anio: 2022 }; // Objeto literal
let numeros = [1, 2, 3, 4, 5]; // Array
function saludar() { console.log("Hola"); } // Función
```

### 2_4_typeof_Operator

El operador `typeof` devuelve una cadena indicando el tipo de dato de un operando.

```javascript
console.log(typeof "Hola");      // "string"
console.log(typeof 100);         // "number"
console.log(typeof NaN);         // "number" (¡Cuidado!)
console.log(typeof true);        // "boolean"
console.log(typeof undefined);   // "undefined"
console.log(typeof null);        // "object" (¡Este es un error histórico de JS!)
console.log(typeof Symbol('id'));// "symbol"
console.log(typeof 123n);        // "bigint"

console.log(typeof { a: 1 });    // "object"
console.log(typeof [1, 2]);      // "object" (Arrays son objetos)
console.log(typeof function(){}); // "function" (Aunque las funciones son objetos, typeof tiene un resultado especial para ellas)
```
*Nota: Para diferenciar `null` de otros objetos, usa `valor === null`. Para diferenciar arrays de otros objetos, usa `Array.isArray(valor)`.*

### 2_5_Operadores_Aritmeticos_Asignacion_Comparacion_Logicos

*   **Operadores Aritméticos:**
    *   `+` (Suma, también concatenación de strings)
    *   `-` (Resta)
    *   `*` (Multiplicación)
    *   `/` (División)
    *   `%` (Módulo - Resto de la división)
    *   `**` (Exponenciación - ES7)
    *   `++` (Incremento)
    *   `--` (Decremento)
*   **Operadores de Asignación:**
    *   `=` (Asignación simple)
    *   `+=`, `-=`, `*=`, `/=`, `%=`, `**=` (Asignación compuesta, ej: `x += 5` es `x = x + 5`)
*   **Operadores de Comparación:** Devuelven `true` o `false`.
    *   `==` (Igualdad - **con coerción de tipo**, ¡evitar!)
    *   `!=` (Desigualdad - **con coerción de tipo**, ¡evitar!)
    *   `===` (Igualdad Estricta - **sin coerción**, compara valor Y tipo. **Usar siempre este**)
    *   `!==` (Desigualdad Estricta - **sin coerción**. **Usar siempre este**)
    *   `>` (Mayor que)
    *   `<` (Menor que)
    *   `>=` (Mayor o igual que)
    *   `<=` (Menor o igual que)
    ```javascript
    console.log(5 == "5");  // true (coerción) - MAL
    console.log(5 === "5"); // false (estricto) - BIEN
    console.log(0 == false); // true (coerción) - MAL
    console.log(0 === false);// false (estricto) - BIEN
    console.log(null == undefined); // true (coerción) - Excepción histórica
    console.log(null === undefined);// false (estricto) - BIEN
    ```
*   **Operadores Lógicos:** Combinan expresiones booleanas.
    *   `&&` (AND Lógico): Devuelve `true` si *ambos* operandos son verdaderos. Usa cortocircuito (si el primero es falso, no evalúa el segundo).
    *   `||` (OR Lógico): Devuelve `true` si *al menos uno* de los operandos es verdadero. Usa cortocircuito (si el primero es verdadero, no evalúa el segundo).
    *   `!` (NOT Lógico): Invierte el valor booleano de un operando.
    ```javascript
    let a = true;
    let b = false;
    console.log(a && b); // false
    console.log(a || b); // true
    console.log(!a);     // false
    ```
*   **Otros Operadores:** Ternario (`?:`), Coma (`,`), `instanceof`, `in`, etc.

### 2_6_Conversion_de_Tipos_Coercion

JavaScript es un lenguaje débilmente tipado, lo que significa que a menudo intenta **convertir automáticamente** tipos de datos (coerción) cuando se usan operadores con tipos diferentes. Esto puede llevar a resultados inesperados, especialmente con `==`.

*   **Coerción Implícita:** Ocurre automáticamente.
    ```javascript
    console.log("5" + 3); // "53" (El número 3 se convierte a string)
    console.log("5" - 3); // 2 (El string "5" se convierte a número)
    console.log("5" * 3); // 15 (El string "5" se convierte a número)
    console.log(5 + true); // 6 (true se convierte a 1)
    console.log(5 + false); // 5 (false se convierte a 0)
    console.log(5 + null); // 5 (null se convierte a 0)
    console.log(5 + undefined); // NaN (undefined se convierte a NaN)
    ```
*   **Conversión Explícita:** Usando funciones incorporadas. Es la forma **preferida** para controlar las conversiones.
    *   `String(valor)` o `valor.toString()`: Convierte a cadena.
    *   `Number(valor)`: Convierte a número (puede dar `NaN`).
    *   `parseInt(valor, base)`: Convierte cadena a entero (ignora caracteres no numéricos al final, `base` es opcional pero recomendado, ej: 10 para decimal).
    *   `parseFloat(valor)`: Convierte cadena a número de punto flotante.
    *   `Boolean(valor)`: Convierte a booleano. Valores "falsy" (`false`, `0`, `""` (string vacío), `null`, `undefined`, `NaN`) se convierten a `false`, todos los demás ("truthy") a `true`.
    ```javascript
    let numStr = "10.5";
    let num = Number(numStr); // 10.5 (número)
    let intNum = parseInt("42px", 10); // 42 (entero)
    let floatNum = parseFloat("3.1416 es pi"); // 3.1416 (número)
    let boolVal = Boolean(0); // false
    let boolVal2 = Boolean("Hola"); // true
    ```

**Conclusión:** Sé consciente de la coerción implícita, especialmente con `==`. Prefiere la igualdad estricta (`===`, `!==`) y realiza conversiones explícitas cuando necesites cambiar de tipo.

## 3_Estructuras_de_Control

Permiten controlar el flujo de ejecución del código, tomando decisiones o repitiendo bloques de código.

### 3_1_Condicionales_if_else_if_else

Ejecutan bloques de código basados en si una condición es verdadera (`true`) o falsa (`false`).

*   **`if`:** Ejecuta un bloque si la condición es verdadera.
    ```javascript
    let temperatura = 25;
    if (temperatura > 20) {
      console.log("Hace calor.");
    }
    ```
*   **`if...else`:** Ejecuta un bloque si la condición es verdadera, y otro bloque si es falsa.
    ```javascript
    let edad = 17;
    if (edad >= 18) {
      console.log("Es mayor de edad.");
    } else {
      console.log("Es menor de edad.");
    }
    ```
*   **`if...else if...else`:** Permite encadenar múltiples condiciones. Se evalúan en orden, y se ejecuta el bloque de la *primera* condición verdadera. El `else` final es opcional y se ejecuta si ninguna condición anterior fue verdadera.
    ```javascript
    let nota = 75;
    if (nota >= 90) {
      console.log("Sobresaliente (A)");
    } else if (nota >= 80) {
      console.log("Notable (B)");
    } else if (nota >= 70) {
      console.log("Bien (C)");
    } else if (nota >= 60) {
      console.log("Suficiente (D)");
    } else {
      console.log("Insuficiente (F)");
    }
    ```
*Recuerda que cualquier valor "truthy" o "falsy" puede usarse como condición.*

### 3_2_Operador_Ternario

Es una forma abreviada de escribir una sentencia `if...else` simple que asigna un valor a una variable o devuelve un valor.

Sintaxis: `condicion ? valorSiVerdadero : valorSiFalso`

```javascript
let edad = 20;
let mensaje = (edad >= 18) ? "Adulto" : "Menor";
console.log(mensaje); // "Adulto"

let precio = 100;
let descuento = (precio > 50) ? 10 : 0;
let precioFinal = precio - descuento;
console.log(precioFinal); // 90
```
Útil para asignaciones condicionales concisas, pero evita anidarlos excesivamente, ya que puede volverse ilegible.

### 3_3_Switch_Statement

Permite evaluar una expresión y ejecutar diferentes bloques de código basados en el valor resultante (`case`). Es una alternativa a `if...else if` cuando se comparan múltiples valores exactos de una misma expresión.

```javascript
let diaSemana = 3; // Domingo=0, Lunes=1, ..., Sábado=6
let nombreDia;

switch (diaSemana) {
  case 0:
    nombreDia = "Domingo";
    break; // ¡Importante! Sale del switch
  case 1:
    nombreDia = "Lunes";
    break;
  case 2:
    nombreDia = "Martes";
    break;
  case 3:
    nombreDia = "Miércoles";
    break; // Sin break, continuaría ejecutando los siguientes cases ("fall-through")
  case 4:
    nombreDia = "Jueves";
    break;
  case 5:
    nombreDia = "Viernes";
    break;
  case 6:
    nombreDia = "Sábado";
    break;
  default: // Opcional: se ejecuta si ningún case coincide
    nombreDia = "Día inválido";
    break;
}

console.log("Hoy es:", nombreDia); // Hoy es: Miércoles
```
*   La comparación en `case` usa igualdad estricta (`===`).
*   La sentencia `break` es crucial para salir del `switch` después de que un `case` coincida. Sin ella, la ejecución continúa en los siguientes `case` (fall-through), lo cual raramente es deseado.
*   El bloque `default` es opcional.

### 3_4_Bucles_for

Repiten un bloque de código un número determinado de veces o mientras se cumpla una condición.

*   **Bucle `for` clásico:** El más común. Tiene tres partes opcionales separadas por `;`:
    1.  Inicialización: Se ejecuta una vez antes de empezar (ej: `let i = 0`).
    2.  Condición: Se evalúa antes de cada iteración. Si es `true`, el bucle continúa.
    3.  Expresión final: Se ejecuta al final de cada iteración (ej: `i++`).
    ```javascript
    // Imprimir números del 0 al 4
    for (let i = 0; i < 5; i++) {
      console.log("Iteración número:", i);
    }

    // Recorrer un array
    let frutas = ["manzana", "plátano", "cereza"];
    for (let i = 0; i < frutas.length; i++) {
      console.log("Fruta:", frutas[i]);
    }
    ```
*   **Bucle `for...of` (ES6):** Itera sobre los **valores** de objetos iterables (como Arrays, Strings, Maps, Sets). Más simple y legible para recorrer colecciones.
    ```javascript
    let colores = ["rojo", "verde", "azul"];
    for (let color of colores) {
      console.log("Color:", color);
    }

    let saludo = "Hola";
    for (let letra of saludo) {
      console.log("Letra:", letra);
    }
    ```
*   **Bucle `for...in`:** Itera sobre las **claves (propiedades enumerables)** de un objeto. **No se recomienda para iterar arrays** porque puede incluir índices inesperados o propiedades del prototipo, y no garantiza el orden. Úsalo principalmente para objetos.
    ```javascript
    let coche = { marca: "Ford", modelo: "Mustang", anio: 1969 };
    for (let propiedad in coche) {
      // propiedad contendrá "marca", "modelo", "anio"
      console.log(`${propiedad}: ${coche[propiedad]}`);
    }
    ```

### 3_5_Bucles_while_y_do_while

*   **Bucle `while`:** Repite un bloque de código *mientras* una condición especificada sea verdadera. La condición se evalúa *antes* de cada iteración. Si la condición es falsa desde el principio, el bloque nunca se ejecuta.
    ```javascript
    let contador = 0;
    while (contador < 3) {
      console.log("Contador while:", contador);
      contador++;
    }
    ```
    *¡Cuidado con crear bucles infinitos si la condición nunca se vuelve falsa!*
*   **Bucle `do...while`:** Similar a `while`, pero la condición se evalúa *después* de ejecutar el bloque. Esto garantiza que el bloque de código se ejecute **al menos una vez**, incluso si la condición es inicialmente falsa.
    ```javascript
    let intentos = 0;
    let pinCorrecto = false;
    do {
      console.log("Intento de conexión número:", intentos + 1);
      // ...lógica para intentar conectar...
      // Supongamos que pinCorrecto se vuelve true si la conexión es exitosa
      intentos++;
      // pinCorrecto = (Math.random() > 0.7); // Simular éxito aleatorio
    } while (!pinCorrecto && intentos < 3);

    if (pinCorrecto) {
      console.log("Conectado!");
    } else {
      console.log("Fallaron los intentos.");
    }
    ```

### 3_6_Break_y_Continue

Se usan dentro de los bucles para alterar su flujo normal:

*   **`break`:** Termina inmediatamente la ejecución del bucle actual (`for`, `while`, `do...while`, o `switch`) y continúa con la siguiente instrucción después del bucle.
    ```javascript
    for (let i = 0; i < 10; i++) {
      if (i === 5) {
        console.log("Encontrado el 5, saliendo del bucle.");
        break; // Termina el bucle aquí
      }
      console.log("Número:", i);
    }
    // Salida: Números 0 a 4, y el mensaje de "Encontrado el 5..."
    ```
*   **`continue`:** Salta el resto de la iteración actual del bucle y pasa directamente a la siguiente iteración (evaluando la condición/expresión final).
    ```javascript
    for (let i = 0; i < 5; i++) {
      if (i === 2) {
        console.log("Saltando la iteración 2.");
        continue; // Salta el console.log de abajo para i=2
      }
      console.log("Procesando número:", i);
    }
    // Salida: Procesando 0, 1, Saltando 2, Procesando 3, Procesando 4
    ```

## 4_Funciones

Las funciones son bloques de código reutilizables que realizan una tarea específica. Permiten organizar el código, hacerlo más legible y evitar la repetición.

### 4_1_Declaracion_de_Funciones

Se definen usando la palabra clave `function`, seguida del nombre de la función, una lista de parámetros entre paréntesis `()`, y el bloque de código entre llaves `{}`.

```javascript
// Declaración de una función llamada 'sumar' que toma dos parámetros
function sumar(a, b) {
  let resultado = a + b;
  console.log("La suma es:", resultado);
  return resultado; // Opcional: devuelve un valor
}

// Llamada (invocación) de la función
let miSuma = sumar(5, 3); // Ejecuta el código de la función con a=5, b=3
console.log("Valor devuelto:", miSuma); // Valor devuelto: 8
```
*   Las declaraciones de funciones sufren *hoisting* (se pueden llamar antes de su definición en el código).

### 4_2_Expresiones_de_Funcion

Se crea una función y se asigna a una variable. La función puede ser anónima (sin nombre) o tener un nombre (útil para depuración).

```javascript
// Expresión de función anónima asignada a la variable 'restar'
const restar = function(a, b) {
  return a - b;
};

// Expresión de función con nombre (menos común)
const multiplicar = function fnMultiplicar(a, b) {
    return a * b;
};

// Llamada
let diferencia = restar(10, 4); // 6
let producto = multiplicar(3, 6); // 18
```
*   Las expresiones de función **no** sufren hoisting de la misma manera que las declaraciones. La variable (`restar`, `multiplicar`) sí se eleva si se usó `var`, pero su valor (la función) no está disponible hasta la línea de asignación. Si usas `let` o `const`, no puedes llamarla antes de la asignación.

### 4_3_Funciones_Flecha_Arrow_Functions

Introducidas en ES6, proporcionan una sintaxis más concisa para escribir funciones, especialmente útil para funciones cortas o callbacks. Tienen un comportamiento diferente con la palabra clave `this`.

Sintaxis:
*   ` (param1, param2) => { /* bloque de código */ return valor; }`
*   Si solo hay un parámetro, los paréntesis son opcionales: `param => { ... }`
*   Si el cuerpo de la función es una sola expresión que se quiere retornar, las llaves y `return` son implícitos: `(param1, param2) => expresionARetornar`

```javascript
// Declaración tradicional
function sumarTradicional(a, b) {
  return a + b;
}

// Expresión de función tradicional
const restarTradicional = function(a, b) {
  return a - b;
};

// Función Flecha (equivalente a sumarTradicional)
const sumarFlecha = (a, b) => {
  return a + b;
};

// Función Flecha concisa (retorno implícito)
const restarFlecha = (a, b) => a - b;

// Función Flecha con un solo parámetro
const cuadradoFlecha = numero => numero * numero;

// Función Flecha sin parámetros
const saludarFlecha = () => console.log("¡Hola desde flecha!");

console.log(sumarFlecha(2, 3)); // 5
console.log(restarFlecha(5, 1)); // 4
console.log(cuadradoFlecha(4)); // 16
saludarFlecha(); // ¡Hola desde flecha!
```
*   Las funciones flecha **no** tienen su propio `this`. Heredan el `this` del contexto léxico (el ámbito donde fueron creadas). Esto es muy útil en callbacks y manejo de eventos.
*   No tienen el objeto `arguments`. Se usa el parámetro rest (`...args`) en su lugar.
*   No se pueden usar como constructores (`new`).

### 4_4_Parametros_y_Argumentos

*   **Parámetros:** Variables listadas en la definición de la función (ej: `a` y `b` en `function sumar(a, b)`).
*   **Argumentos:** Valores reales que se pasan a la función cuando se la llama (ej: `5` y `3` en `sumar(5, 3)`).

*   **Parámetros por Defecto (ES6):** Puedes asignar valores por defecto a los parámetros, que se usarán si no se pasa un argumento para ellos.
    ```javascript
    function saludar(nombre = "Invitado") {
      console.log(`Hola, ${nombre}!`);
    }
    saludar("Ana"); // Hola, Ana!
    saludar();      // Hola, Invitado!
    ```
*   **Parámetro Rest (`...`) (ES6):** Permite agrupar un número indefinido de argumentos restantes en un array. Debe ser el último parámetro.
    ```javascript
    function sumarTodo(...numeros) { // numeros será un array
      let total = 0;
      for (let num of numeros) {
        total += num;
      }
      return total;
    }
    console.log(sumarTodo(1, 2, 3)); // 6
    console.log(sumarTodo(10, 20, 30, 40, 50)); // 150
    ```

### 4_5_Valor_de_Retorno_return

La sentencia `return` especifica el valor que la función devolverá al código que la llamó.
*   Una función puede tener múltiples sentencias `return` (ej: dentro de un `if...else`), pero solo se ejecutará una.
*   Cuando se ejecuta `return`, la función termina inmediatamente.
*   Si una función no tiene una sentencia `return`, o tiene `return;` sin un valor, devuelve `undefined` por defecto.

```javascript
function esMayorDeEdad(edad) {
  if (edad >= 18) {
    return true;
  } else {
    return false;
  }
  // Código aquí nunca se ejecutaría si edad >= 18
}

function mostrarMensaje(msg) {
  console.log(msg);
  // No hay return, devuelve undefined implícitamente
}

let mayor = esMayorDeEdad(25); // true
let resultadoMsg = mostrarMensaje("Hola"); // Imprime "Hola", resultadoMsg es undefined
```

### 4_6_Ambito_Scope_Global_Funcion_Bloque

El ámbito (scope) determina la accesibilidad (visibilidad) de las variables y funciones en diferentes partes del código.

*   **Ámbito Global:** Variables declaradas fuera de cualquier función o bloque (`{}`). Son accesibles desde cualquier lugar del código. Es mala práctica abusar del ámbito global, ya que puede llevar a colisiones de nombres y dificultar el mantenimiento.
*   **Ámbito de Función:** Variables declaradas con `var` (o `let`/`const`) *dentro* de una función. Solo son accesibles *dentro* de esa función.
*   **Ámbito de Bloque (ES6):** Variables declaradas con `let` o `const` *dentro* de un bloque (`{...}`), como en un `if`, `for`, o simplemente llaves. Solo son accesibles *dentro* de ese bloque.

```javascript
const variableGlobal = "Soy global";

function miFuncion() {
  const variableFuncion = "Soy de función";
  var variableVarFuncion = "Soy var de función";
  console.log(variableGlobal); // OK

  if (true) {
    const variableBloque = "Soy de bloque (const)";
    let variableLetBloque = "Soy de bloque (let)";
    var variableVarBloque = "Soy var en bloque"; // ¡Ámbito de función!
    console.log(variableFuncion); // OK
    console.log(variableBloque); // OK
  }

  console.log(variableVarBloque); // OK (var tiene ámbito de función)
  // console.log(variableBloque); // Error: variableBloque no está definida aquí
  // console.log(variableLetBloque); // Error: variableLetBloque no está definida aquí
}

miFuncion();
// console.log(variableFuncion); // Error: variableFuncion no está definida aquí
// console.log(variableVarFuncion); // Error
```

JavaScript también tiene **clausuras (closures)**, que es cuando una función interna "recuerda" y tiene acceso al ámbito de su función externa contenedora, incluso después de que la función externa haya terminado de ejecutarse. Es un concepto más avanzado pero fundamental.

### 4_7_Hoisting

Es un comportamiento de JavaScript donde las **declaraciones** de variables (con `var`) y funciones (declaraciones, no expresiones) son "movidas" conceptualmente al inicio de su ámbito (global o de función) durante la fase de compilación, *antes* de que el código se ejecute.

*   **Declaraciones de `var`:** Solo se eleva la declaración, no la inicialización. La variable existe pero su valor es `undefined` hasta la línea donde se asigna.
*   **Declaraciones de `function`:** Se eleva tanto la declaración como el cuerpo de la función. Se pueden llamar antes de su definición.
*   **`let` y `const`:** También se elevan técnicamente, pero entran en una "Zona Temporal Muerta" (Temporal Dead Zone - TDZ). No se pueden acceder de ninguna manera (ni siquiera para obtener `undefined`) antes de su línea de declaración en el código. Esto ayuda a prevenir errores.
*   **Expresiones de Función:** No se elevan como las declaraciones. Si se asignan a una variable `var`, la variable se eleva como `undefined`. Si se asignan a `let` o `const`, aplica la TDZ.

```javascript
console.log(miVar); // undefined (Hoisting de var)
var miVar = 5;

saludar(); // "Hola desde declaración" (Hoisting de declaración de función)
function saludar() {
  console.log("Hola desde declaración");
}

// console.log(miLet); // Error: Cannot access 'miLet' before initialization (TDZ)
let miLet = 10;

// miExpresion(); // Error: miExpresion is not a function (si es var) o ReferenceError (si es let/const)
const miExpresion = function() {
  console.log("Hola desde expresión");
};
```
**Mejor práctica:** Declara siempre tus variables (`let`, `const`) y funciones al principio de su ámbito respectivo para evitar confusiones relacionadas con el hoisting.

## 5_Arrays

Los arrays son estructuras de datos tipo objeto que permiten almacenar una colección ordenada de elementos (pueden ser de diferentes tipos). Se accede a los elementos mediante un índice numérico (empezando en 0).

### 5_1_Creacion_y_Acceso

*   **Literal de Array (preferido):** Usando corchetes `[]`.
    ```javascript
    let frutas = ["Manzana", "Plátano", "Naranja"];
    let mixto = [1, "Hola", true, null, { id: 1 }];
    let vacio = [];
    ```
*   **Constructor `Array()`:** Menos común.
    ```javascript
    let numeros = new Array(1, 2, 3);
    let predefinido = new Array(5); // Crea un array con 5 posiciones vacías
    ```
*   **Acceso a Elementos:** Usando el índice entre corchetes. El primer elemento es el índice `0`.
    ```javascript
    console.log(frutas[0]); // "Manzana"
    console.log(frutas[2]); // "Naranja"
    console.log(mixto[1]); // "Hola"
    console.log(frutas[3]); // undefined (índice fuera de rango)

    // Modificar un elemento
    frutas[1] = "Banana";
    console.log(frutas); // ["Manzana", "Banana", "Naranja"]
    ```

### 5_2_Propiedad_length

Devuelve el número de elementos en el array. Es dinámica.

```javascript
let colores = ["rojo", "verde", "azul"];
console.log(colores.length); // 3

// Añadir elemento al final usando length
colores[colores.length] = "amarillo";
console.log(colores);      // ["rojo", "verde", "azul", "amarillo"]
console.log(colores.length); // 4

// Acortar el array modificando length
colores.length = 2;
console.log(colores);      // ["rojo", "verde"]
```

### 5_3_Metodos_Comunes_push_pop_shift_unshift

Estos métodos modifican el array original (son mutadores).

*   **`push(elemento1, ..., elementoN)`:** Añade uno o más elementos al **final** del array. Devuelve la nueva longitud del array.
    ```javascript
    let letras = ["a", "b"];
    let nuevaLongitud = letras.push("c", "d");
    console.log(letras); // ["a", "b", "c", "d"]
    console.log(nuevaLongitud); // 4
    ```
*   **`pop()`:** Elimina y devuelve el **último** elemento del array.
    ```javascript
    let numeros = [1, 2, 3];
    let ultimo = numeros.pop();
    console.log(numeros); // [1, 2]
    console.log(ultimo); // 3
    ```
*   **`unshift(elemento1, ..., elementoN)`:** Añade uno o más elementos al **principio** del array. Devuelve la nueva longitud. (Puede ser menos eficiente que `push` en arrays grandes).
    ```javascript
    let items = [3, 4];
    items.unshift(1, 2);
    console.log(items); // [1, 2, 3, 4]
    ```
*   **`shift()`:** Elimina y devuelve el **primer** elemento del array. (Puede ser menos eficiente que `pop`).
    ```javascript
    let cola = ["A", "B", "C"];
    let primero = cola.shift();
    console.log(cola); // ["B", "C"]
    console.log(primero); // "A"
    ```

### 5_4_Metodos_slice_y_splice

*   **`slice(inicio, fin)`:** Devuelve una **copia superficial (shallow copy)** de una porción del array, desde el índice `inicio` (incluido) hasta el índice `fin` (excluido). **No modifica el array original.**
    *   Si `fin` se omite, extrae hasta el final del array.
    *   Si `inicio` es negativo, cuenta desde el final.
    *   Si se llama sin argumentos (`slice()`), crea una copia superficial de todo el array.
    ```javascript
    let animales = ["perro", "gato", "conejo", "hamster", "pájaro"];
    let mamiferos = animales.slice(0, 4); // Desde índice 0 hasta 4 (excluido)
    console.log(mamiferos); // ["perro", "gato", "conejo", "hamster"]
    let ultimosDos = animales.slice(-2); // Los últimos 2 elementos
    console.log(ultimosDos); // ["hamster", "pájaro"]
    let copiaAnimales = animales.slice(); // Copia todo el array
    console.log(animales); // El original no cambia
    ```
*   **`splice(inicio, cantidadAEliminar, item1, item2, ...)`:** **Modifica el array original** eliminando elementos y/o añadiendo nuevos.
    *   `inicio`: Índice donde empezar a cambiar el array.
    *   `cantidadAEliminar`: Número de elementos a eliminar (0 si solo quieres añadir).
    *   `item1, item2, ...` (Opcional): Elementos a añadir en la posición `inicio` después de eliminar.
    *   Devuelve un array con los elementos eliminados.
    ```javascript
    let meses = ["Ene", "Mar", "Abr", "Jun"];

    // Insertar "Feb" en índice 1 (sin eliminar nada)
    meses.splice(1, 0, "Feb");
    console.log(meses); // ["Ene", "Feb", "Mar", "Abr", "Jun"]

    // Reemplazar "Abr" (índice 3) por "May"
    let eliminados = meses.splice(3, 1, "May");
    console.log(meses); // ["Ene", "Feb", "Mar", "May", "Jun"]
    console.log(eliminados); // ["Abr"]

    // Eliminar "Jun" (índice 4)
    meses.splice(4, 1);
    console.log(meses); // ["Ene", "Feb", "Mar", "May"]
    ```

### 5_5_Iteracion_forEach_map_filter_reduce

Métodos modernos y funcionales para trabajar con arrays. **No modifican el array original** (excepto `forEach` si la función callback lo hace).

*   **`forEach(callback(elemento, indice, array))`:** Ejecuta una función `callback` una vez por cada elemento del array. No devuelve nada (`undefined`). Útil para realizar acciones con cada elemento.
    ```javascript
    let nombres = ["Ana", "Luis", "Eva"];
    nombres.forEach(function(nombre, index) {
      console.log(`Índice ${index}: ${nombre}`);
    });
    // Usando función flecha:
    nombres.forEach((nombre, index) => console.log(`Índice ${index}: ${nombre}`));
    ```
*   **`map(callback(elemento, indice, array))`:** Crea un **nuevo array** con los resultados de llamar a la función `callback` para cada elemento. La longitud del nuevo array es igual a la del original. Ideal para transformar datos.
    ```javascript
    let numeros = [1, 2, 3, 4];
    let cuadrados = numeros.map(num => num * num);
    console.log(cuadrados); // [1, 4, 9, 16]
    console.log(numeros);   // [1, 2, 3, 4] (original sin cambios)
    ```
*   **`filter(callback(elemento, indice, array))`:** Crea un **nuevo array** con todos los elementos que pasen la prueba implementada por la función `callback` (es decir, para los que el callback devuelva `true`).
    ```javascript
    let edades = [15, 22, 18, 30, 12];
    let mayoresDeEdad = edades.filter(edad => edad >= 18);
    console.log(mayoresDeEdad); // [22, 18, 30]
    console.log(edades);       // [15, 22, 18, 30, 12] (original sin cambios)
    ```
*   **`reduce(callback(acumulador, elementoActual, indice, array), valorInicial)`:** Aplica una función `callback` a un `acumulador` y a cada elemento del array (de izquierda a derecha) para reducirlo a un único valor.
    *   `acumulador`: El valor resultante de la iteración anterior (o `valorInicial` en la primera).
    *   `elementoActual`: El elemento que se está procesando.
    *   `valorInicial` (Opcional): Valor para iniciar el `acumulador`. Si no se provee, se usa el primer elemento del array como inicial y la iteración empieza desde el segundo.
    ```javascript
    let nums = [10, 5, 8, 20];
    let sumaTotal = nums.reduce((total, num) => total + num, 0); // 0 es el valor inicial
    console.log(sumaTotal); // 43

    // Encontrar el máximo
    let maximo = nums.reduce((max, current) => (current > max ? current : max), nums[0]);
    console.log(maximo); // 20
    ```

### 5_6_Busqueda_indexOf_includes_find_findIndex

*   **`indexOf(elemento, desdeIndice)`:** Devuelve el **primer índice** en el que se encuentra un `elemento` dado, o `-1` si no está presente. Busca usando igualdad estricta (`===`). `desdeIndice` es opcional.
    ```javascript
    let frutas = ["Manzana", "Plátano", "Naranja", "Plátano"];
    console.log(frutas.indexOf("Plátano")); // 1 (encuentra la primera ocurrencia)
    console.log(frutas.indexOf("Plátano", 2)); // 3 (busca a partir del índice 2)
    console.log(frutas.indexOf("Pera")); // -1
    ```
*   **`includes(elemento, desdeIndice)` (ES7):** Devuelve `true` si el array contiene el `elemento`, y `false` si no. Más legible que `indexOf !== -1`. También usa `===`, pero maneja `NaN` correctamente (a diferencia de `indexOf`).
    ```javascript
    let mascotas = ["perro", "gato", "pez"];
    console.log(mascotas.includes("gato")); // true
    console.log(mascotas.includes("loro")); // false
    ```
*   **`find(callback(elemento, indice, array))`:** Devuelve el **valor del primer elemento** en el array que cumple la condición proporcionada por la función `callback`. Si no encuentra ninguno, devuelve `undefined`.
    ```javascript
    let numeros = [5, 12, 8, 130, 44];
    let primerMayorQue10 = numeros.find(num => num > 10);
    console.log(primerMayorQue10); // 12
    ```
*   **`findIndex(callback(elemento, indice, array))`:** Devuelve el **índice del primer elemento** que cumple la condición del `callback`. Si no encuentra ninguno, devuelve `-1`.
    ```javascript
    let edades = [3, 10, 18, 20];
    let indicePrimerAdulto = edades.findIndex(edad => edad >= 18);
    console.log(indicePrimerAdulto); // 2
    ```

## 6_Objetos

Los objetos son la estructura de datos fundamental en JavaScript para agrupar datos relacionados y funcionalidades. Son colecciones dinámicas de pares clave-valor (propiedades). Las claves suelen ser strings (o Symbols), y los valores pueden ser cualquier tipo de dato, incluyendo otros objetos o funciones (métodos).

### 6_1_Objetos_Literales

La forma más común de crear objetos es usando la sintaxis literal con llaves `{}`.

```javascript
// Objeto vacío
let coche = {};

// Objeto con propiedades
let persona = {
  nombre: "Carlos", // Clave: nombre, Valor: "Carlos" (string)
  edad: 35,         // Clave: edad, Valor: 35 (number)
  esEstudiante: false, // Clave: esEstudiante, Valor: false (boolean)
  direccion: {      // Clave: direccion, Valor: otro objeto
    calle: "Av. Siempre Viva",
    numero: 742
  },
  saludar: function() { // Clave: saludar, Valor: una función (método)
    console.log("Hola, soy " + this.nombre);
  },
  // ES6: Método abreviado
  despedir() {
    console.log("Adiós!");
  }
};
```
*   Las claves (nombres de propiedad) que no son identificadores válidos (ej: contienen espacios o guiones) deben ir entre comillas: `let obj = { "mi-propiedad": 1 };`

### 6_2_Acceso_a_Propiedades_Notacion_Punto_y_Corchetes

Se puede acceder (leer o escribir) a los valores de las propiedades de dos maneras:

*   **Notación de Punto (`objeto.propiedad`):** Más común y legible. Funciona si el nombre de la propiedad es un identificador válido de JavaScript.
    ```javascript
    console.log(persona.nombre); // "Carlos"
    console.log(persona.direccion.calle); // "Av. Siempre Viva"

    persona.edad = 36; // Modificar valor
    console.log(persona.edad); // 36

    persona.profesion = "Ingeniero"; // Añadir nueva propiedad
    console.log(persona.profesion); // "Ingeniero"
    ```
*   **Notación de Corchetes (`objeto['propiedad']`):** Necesaria si:
    *   El nombre de la propiedad es una cadena que no es un identificador válido (contiene espacios, guiones, etc.).
    *   El nombre de la propiedad está almacenado en una variable.
    *   Se usan Symbols como claves.
    ```javascript
    console.log(persona['edad']); // 35 (igual que persona.edad)

    let clave = "nombre";
    console.log(persona[clave]); // "Carlos" (acceso usando variable)

    let objEspecial = { "espacio-prop": 100 };
    console.log(objEspecial["espacio-prop"]); // 100 (necesario corchetes)
    // console.log(objEspecial.espacio-prop); // Error de sintaxis
    ```
*   Si intentas acceder a una propiedad que no existe, devuelve `undefined`.
    ```javascript
    console.log(persona.apellido); // undefined
    ```
*   Para eliminar una propiedad: `delete persona.profesion;`

### 6_3_Metodos_de_Objeto

Cuando el valor de una propiedad es una función, se le llama **método**. Los métodos definen el comportamiento o las acciones que un objeto puede realizar.

```javascript
let calculadora = {
  sumar: function(a, b) {
    return a + b;
  },
  restar(a, b) { // Sintaxis ES6
    return a - b;
  }
};

let resultadoSuma = calculadora.sumar(5, 3); // 8
let resultadoResta = calculadora.restar(10, 4); // 6
```
Dentro de un método, a menudo se necesita acceder a otras propiedades del *mismo* objeto. Para esto se usa la palabra clave `this`.

### 6_4_La_Palabra_Clave_this

`this` es una palabra clave especial cuyo valor se determina por **cómo se llama a la función** (contexto de invocación). Es uno de los conceptos más confusos pero importantes de JS.

*   **En un Método de Objeto:** `this` generalmente se refiere al **objeto que llamó al método**.
    ```javascript
    let perro = {
      nombre: "Fido",
      ladrar() {
        // Aquí, 'this' se refiere al objeto 'perro'
        console.log(`¡Guau! Soy ${this.nombre}`);
      }
    };
    perro.ladrar(); // ¡Guau! Soy Fido
    ```
*   **En una Función Simple (Modo no estricto):** `this` se refiere al **objeto global** (`window` en navegadores, `global` en Node.js). ¡Esto es fuente común de errores!
*   **En una Función Simple (Modo Estricto `'use strict';`):** `this` es `undefined`. Es la forma recomendada de evitar errores con `this` global.
    ```javascript
    'use strict';
    function mostrarThis() {
      console.log(this);
    }
    mostrarThis(); // undefined
    ```
*   **En Funciones Flecha:** Las funciones flecha **NO tienen su propio `this`**. Heredan el valor de `this` del **contexto léxico circundante** (el `this` del lugar donde se *definió* la función flecha). Esto las hace muy útiles para callbacks dentro de métodos.
    ```javascript
    const miObjeto = {
      valor: 42,
      obtenerValorDespuesDe1Seg() {
        // 'this' aquí es miObjeto
        setTimeout(() => {
          // Dentro de la función flecha, 'this' sigue siendo miObjeto
          console.log("El valor es:", this.valor);
        }, 1000);
      },
      obtenerValorErroneo() {
        // 'this' aquí es miObjeto
        setTimeout(function() {
          // Dentro de esta función normal, 'this' es 'window' o 'undefined' (strict mode)
          // ¡No es miObjeto!
          // console.log("El valor es:", this.valor); // Error o undefined
        }, 1000);
      }
    };
    miObjeto.obtenerValorDespuesDe1Seg(); // El valor es: 42
    ```
*   **En Constructores (`new`):** `this` se refiere a la nueva instancia del objeto que se está creando.
*   **Con `call()`, `apply()`, `bind()`:** Permiten establecer explícitamente el valor de `this` al llamar a una función.

### 6_5_Iteracion_for_in_Object_keys_values_entries

*   **`for...in`:** (Visto antes) Itera sobre las claves enumerables de un objeto. No recomendado para arrays.
    ```javascript
    for (let clave in persona) {
      // Es buena práctica verificar si la propiedad es propia del objeto y no heredada
      if (persona.hasOwnProperty(clave)) {
        console.log(`${clave}: ${persona[clave]}`);
      }
    }
    ```
*   **`Object.keys(obj)`:** Devuelve un array con los nombres de las propiedades (claves) enumerables propias del objeto.
    ```javascript
    let claves = Object.keys(persona);
    console.log(claves); // ["nombre", "edad", "esEstudiante", "direccion", "saludar", "despedir"]
    claves.forEach(clave => console.log(persona[clave]));
    ```
*   **`Object.values(obj)`:** Devuelve un array con los valores de las propiedades enumerables propias del objeto.
    ```javascript
    let valores = Object.values(persona);
    console.log(valores); // ["Carlos", 35, false, {calle: "...", num: ...}, function(){...}, function(){...}]
    ```
*   **`Object.entries(obj)`:** Devuelve un array de arrays, donde cada subarray contiene un par `[clave, valor]` de las propiedades enumerables propias. Muy útil para iterar con `for...of`.
    ```javascript
    let entradas = Object.entries(persona);
    console.log(entradas);
    /* Salida:
    [
      ["nombre", "Carlos"],
      ["edad", 35],
      ...
      ["despedir", function despedir(){...}]
    ] */

    for (let [clave, valor] of Object.entries(persona)) {
      console.log(`Par -> ${clave}: ${valor}`);
    }
    ```

## 7_JSON_Javascript_Object_Notation

JSON (JavaScript Object Notation) es un formato **ligero** de intercambio de datos, basado en un subconjunto de la sintaxis de objetos literales de JavaScript. Es muy fácil de leer y escribir para humanos, y fácil de interpretar y generar para las máquinas. Es el formato estándar para intercambiar datos en la web (especialmente con APIs REST).

### 7_1_Sintaxis_JSON

*   Los datos se representan en pares **clave/valor**.
*   Las **claves** DEBEN ser **cadenas de texto** entre comillas dobles (`"`).
*   Los **valores** pueden ser:
    *   Cadenas de texto (entre comillas dobles `"`).
    *   Números.
    *   Booleanos (`true` o `false`).
    *   `null`.
    *   Arrays (usando corchetes `[]`, los valores dentro siguen las reglas JSON).
    *   Otros objetos JSON (usando llaves `{}`).
*   **NO** se permiten: Funciones, `undefined`, comentarios, comillas simples para claves o strings, comas al final del último elemento.

**Ejemplo de JSON:**
```json
{
  "id": 101,
  "nombre": "Producto Ejemplo",
  "activo": true,
  "precio": 49.99,
  "tags": ["web", "ejemplo", "datos"],
  "descripcion": null,
  "dimensiones": {
    "alto": 10.5,
    "ancho": 20,
    "unidad": "cm"
  }
}
```

### 7_2_JSON_parse_Conversion_JSON_a_Objeto_JS

La función `JSON.parse()` toma una **cadena de texto** que contiene datos en formato JSON válido y la convierte en un objeto o valor de JavaScript correspondiente.

```javascript
const jsonString = `{
  "titulo": "Libro de JS",
  "paginas": 350,
  "publicado": true,
  "autores": ["Autor A", "Autor B"]
}`;

try {
  const libroObjeto = JSON.parse(jsonString);

  console.log(libroObjeto.titulo); // "Libro de JS"
  console.log(libroObjeto.paginas); // 350
  console.log(libroObjeto.autores[0]); // "Autor A"
  console.log(typeof libroObjeto); // "object"
  console.log(Array.isArray(libroObjeto.autores)); // true
} catch (error) {
  // Se produce un error si la cadena jsonString no es JSON válido
  console.error("Error al parsear JSON:", error);
}
```
*   Si la cadena JSON no es válida, `JSON.parse()` lanzará un error `SyntaxError`. Es buena práctica usar `try...catch`.

### 7_3_JSON_stringify_Conversion_Objeto_JS_a_JSON

La función `JSON.stringify()` toma un objeto o valor de JavaScript y lo convierte en una **cadena de texto** en formato JSON.

```javascript
const productoObjeto = {
  id: 205,
  nombre: "Widget Pro",
  stock: 150,
  activo: true,
  // Las funciones y undefined son ignorados por stringify
  metodo: function() { console.log("Hola"); },
  color: undefined
};

const jsonProducto = JSON.stringify(productoObjeto);

console.log(jsonProducto);
// Salida: {"id":205,"nombre":"Widget Pro","stock":150,"activo":true}
console.log(typeof jsonProducto); // "string"

// stringify con formato (indentación)
const jsonProductoFormateado = JSON.stringify(productoObjeto, null, 2); // Indenta con 2 espacios
console.log(jsonProductoFormateado);
/* Salida:
{
  "id": 205,
  "nombre": "Widget Pro",
  "stock": 150,
  "activo": true
}
*/
```
*   `JSON.stringify()` ignora propiedades cuyo valor sea `undefined`, una `function` o un `Symbol`.
*   Si un objeto tiene referencias circulares (un objeto se refiere a sí mismo directa o indirectamente), `JSON.stringify()` lanzará un error `TypeError`.
*   Admite dos argumentos opcionales:
    1.  `replacer`: Una función o un array para transformar los valores antes de convertirlos.
    2.  `space`: Un número (espacios de indentación) o una cadena para formatear la salida JSON y hacerla más legible.

JSON es fundamental para trabajar con APIs web, ya que es el formato estándar para enviar y recibir datos entre el cliente (JavaScript en el navegador) y el servidor.

## 8_El_Modelo_de_Objetos_del_Documento_DOM

El DOM (Document Object Model) es una **interfaz de programación** (API) para documentos HTML y XML. Representa la estructura del documento como un **árbol de objetos** (nodos), donde cada etiqueta HTML, atributo y texto es un nodo. JavaScript puede usar el DOM para **acceder y manipular** dinámicamente el contenido, la estructura y los estilos de una página web.

### 8_1_Que_es_el_DOM

Piensa en el DOM como una representación en memoria de tu archivo HTML. El navegador crea este árbol DOM cuando carga la página.

*   El nodo raíz es `document`.
*   El elemento `<html>` es hijo de `document`.
*   `<body>` y `<head>` son hijos de `<html>`.
*   Todos los demás elementos HTML son nodos dentro de esta jerarquía.
*   El texto dentro de las etiquetas también son nodos (nodos de texto).
*   Los atributos HTML son nodos de atributo asociados a los nodos de elemento.

JavaScript interactúa con esta estructura de árbol para modificar la página en tiempo real sin necesidad de recargarla.

### 8_2_Seleccion_de_Elementos_Singulares

Métodos para obtener una referencia a un **único** elemento del DOM:

*   **`document.getElementById('idDelElemento')`:** Método clásico y muy rápido. Devuelve el elemento que tiene el `id` especificado, o `null` si no se encuentra. Recuerda que los IDs deben ser únicos.
    ```html
    <div id="principal">Contenido Principal</div>
    ```
    ```javascript
    const divPrincipal = document.getElementById('principal');
    if (divPrincipal) {
      console.log(divPrincipal);
    }
    ```
*   **`document.querySelector('selectorCSS')`:** Más moderno y versátil. Devuelve el **primer** elemento del documento que coincide con el `selectorCSS` especificado (cualquier selector CSS válido: tipo, clase, ID, atributo, combinador, etc.). Devuelve `null` si no hay coincidencias.
    ```html
    <p class="info">Primer párrafo.</p>
    <div id="menu"><ul><li class="item">Item 1</li></ul></div>
    ```
    ```javascript
    const primerParrafoInfo = document.querySelector('p.info'); // Selecciona por etiqueta y clase
    const menuPorId = document.querySelector('#menu');        // Selecciona por ID
    const primerItemLista = document.querySelector('#menu li.item'); // Selector complejo
    ```

### 8_3_Seleccion_de_Elementos_Multiples

Métodos para obtener una colección de **varios** elementos del DOM:

*   **`document.getElementsByClassName('nombreClase')`:** Devuelve una `HTMLCollection` (un objeto similar a un array, pero **vivo**) de todos los elementos que tienen la clase especificada.
    ```html
    <p class="destacado">Texto 1</p>
    <span class="destacado">Texto 2</span>
    ```
    ```javascript
    const destacados = document.getElementsByClassName('destacado');
    console.log(destacados.length); // 2
    console.log(destacados[0]); // El elemento <p>
    // Ojo: HTMLCollection es viva, si añades/quitas elementos con esa clase, la colección se actualiza.
    // No tiene métodos de array modernos como forEach, map directamente (hay que convertirla).
    for (let i = 0; i < destacados.length; i++) {
        console.log(destacados[i].tagName);
    }
    ```
*   **`document.getElementsByTagName('nombreEtiqueta')`:** Devuelve una `HTMLCollection` viva de todos los elementos con el nombre de etiqueta especificado (ej: `'p'`, `'li'`, `'div'`).
    ```javascript
    const todosLosParrafos = document.getElementsByTagName('p');
    console.log(`Hay ${todosLosParrafos.length} párrafos.`);
    ```
*   **`document.querySelectorAll('selectorCSS')`:** El método moderno y más versátil. Devuelve una `NodeList` (un objeto similar a un array, pero **estático**) de todos los elementos que coinciden con el `selectorCSS` especificado.
    ```html
    <ul>
      <li class="fruta">Manzana</li>
      <li class="verdura">Lechuga</li>
      <li class="fruta">Plátano</li>
    </ul>
    ```
    ```javascript
    const todasLasFrutas = document.querySelectorAll('li.fruta');
    console.log(todasLasFrutas.length); // 2
    // NodeList SÍ tiene forEach (en navegadores modernos)
    todasLasFrutas.forEach(fruta => {
      console.log(fruta.textContent);
    });
    // NodeList es estática: si añades más li.fruta después, esta lista no se actualiza.
    ```
**Recomendación:** Usa `querySelector` y `querySelectorAll` siempre que puedas, ya que utilizan la sintaxis de selectores CSS que ya conoces y `querySelectorAll` devuelve una `NodeList` más fácil de iterar que una `HTMLCollection`. Usa `getElementById` si necesitas seleccionar por ID específicamente (suele ser más rápido).

### 8_4_Navegacion_por_el_DOM_Padres_Hijos_Hermanos

Una vez que tienes una referencia a un nodo (elemento), puedes navegar por el árbol DOM relativo a ese nodo:

*   **Padre:**
    *   `elemento.parentNode`: Devuelve el nodo padre (puede ser otro elemento o el nodo `document`).
    *   `elemento.parentElement`: Similar, pero siempre devuelve un elemento padre (o `null` si el padre no es un elemento, ej: `<html>`). Generalmente más útil.
*   **Hijos:**
    *   `elemento.childNodes`: Devuelve una `NodeList` viva de **todos** los nodos hijos (incluyendo elementos, texto, comentarios). A menudo inconveniente por incluir nodos de texto (espacios en blanco).
    *   `elemento.children`: Devuelve una `HTMLCollection` viva de solo los nodos hijos que son **elementos HTML**. ¡Mucho más útil generalmente!
    *   `elemento.firstChild`: Primer nodo hijo (puede ser texto).
    *   `elemento.lastChild`: Último nodo hijo (puede ser texto).
    *   `elemento.firstElementChild`: Primer nodo hijo que es un elemento.
    *   `elemento.lastElementChild`: Último nodo hijo que es un elemento.
*   **Hermanos:**
    *   `elemento.previousSibling`: Nodo hermano inmediatamente anterior (puede ser texto).
    *   `elemento.nextSibling`: Nodo hermano inmediatamente posterior (puede ser texto).
    *   `elemento.previousElementSibling`: Elemento hermano inmediatamente anterior.
    *   `elemento.nextElementSibling`: Elemento hermano inmediatamente posterior.

```html
<div id="contenedor">
  Texto antes
  <h2>Título</h2>
  <p>Párrafo 1</p>
  <!-- Comentario -->
  <p>Párrafo 2</p>
  Texto después
</div>
```
```javascript
const contenedor = document.getElementById('contenedor');
const titulo = contenedor.querySelector('h2');
const parrafo1 = contenedor.querySelector('p'); // El primer p

console.log(titulo.parentElement === contenedor); // true
console.log(contenedor.children.length); // 3 (h2, p, p)
console.log(contenedor.childNodes.length); // 7 (incluye nodos de texto y comentario)
console.log(titulo.nextElementSibling); // El primer <p>
console.log(parrafo1.nextElementSibling); // El segundo <p>
console.log(parrafo1.previousElementSibling); // El <h2>
```

### 8_5_Modificacion_Contenido_innerHTML_textContent

*   **`elemento.innerHTML`:** Obtiene o establece el **contenido HTML completo** (incluyendo etiquetas) dentro de un elemento.
    *   **Lectura:** `let contenidoHtml = elemento.innerHTML;`
    *   **Escritura:** `elemento.innerHTML = '<p>Nuevo <strong>contenido</strong> HTML</p>';`
    *   **¡Peligro!** Al establecer `innerHTML` con contenido proveniente del usuario sin sanitizarlo, puedes crear vulnerabilidades de seguridad **XSS (Cross-Site Scripting)**. El navegador interpretará cualquier etiqueta HTML o `<script>` incluida. ¡Úsalo con precaución!
*   **`elemento.textContent`:** Obtiene o establece solo el **contenido de texto** dentro de un elemento y todos sus descendientes, ignorando las etiquetas HTML. Es más seguro y a menudo más eficiente que `innerHTML` si solo necesitas trabajar con texto.
    *   **Lectura:** `let soloTexto = elemento.textContent;`
    *   **Escritura:** `elemento.textContent = 'Este es texto plano, sin HTML.';` (Cualquier HTML aquí se mostrará literalmente, no se interpretará).

```html
<div id="miDiv">Hola <b>Mundo</b>!</div>
```
```javascript
const miDiv = document.getElementById('miDiv');
console.log(miDiv.innerHTML);   // "Hola <b>Mundo</b>!"
console.log(miDiv.textContent); // "Hola Mundo!"

// miDiv.innerHTML = "<em>Importante</em>"; // Reemplaza con texto en cursiva
// miDiv.textContent = "<em>Importante</em>"; // Reemplaza con el texto literal "<em>Importante</em>"
```

### 8_6_Modificacion_Atributos_getAttribute_setAttribute

Permiten leer, establecer y eliminar atributos HTML estándar y personalizados (`data-*`).

*   **`elemento.getAttribute('nombreAtributo')`:** Obtiene el valor del atributo especificado. Devuelve `null` si el atributo no existe.
*   **`elemento.setAttribute('nombreAtributo', 'nuevoValor')`:** Establece el valor de un atributo. Si el atributo no existe, lo crea.
*   **`elemento.removeAttribute('nombreAtributo')`:** Elimina el atributo especificado.
*   **`elemento.hasAttribute('nombreAtributo')`:** Devuelve `true` si el elemento tiene el atributo, `false` si no.

```html
<img id="miImagen" src="logo.png" alt="Logo" width="100">
<a id="miEnlace" href="#" data-info="dato-importante">Enlace</a>
```
```javascript
const img = document.getElementById('miImagen');
const link = document.getElementById('miEnlace');

console.log(img.getAttribute('src')); // "logo.png"
console.log(img.getAttribute('width')); // "100" (¡Ojo! getAttribute devuelve string)

img.setAttribute('alt', 'Nuevo Texto Alternativo');
img.setAttribute('height', '150'); // Añade el atributo height

console.log(link.getAttribute('data-info')); // "dato-importante"
link.removeAttribute('data-info');
console.log(link.hasAttribute('href')); // true
```
*   Para atributos comunes como `id`, `src`, `href`, `value`, `className`, a menudo puedes acceder a ellos directamente como propiedades del objeto elemento (ej: `img.src`, `link.href`). Esto suele ser más conveniente, pero `getAttribute/setAttribute` son más generales y necesarios para atributos personalizados o cuando el nombre del atributo coincide con una palabra reservada.

### 8_7_Modificacion_Estilos_style_property

Permite acceder y modificar los estilos CSS **en línea** (inline) de un elemento a través de la propiedad `style`.

*   Se accede a las propiedades CSS usando **camelCase** en lugar de guiones (ej: `backgroundColor` en lugar de `background-color`, `fontSize` en lugar de `font-size`).
*   Los valores se asignan como cadenas de texto (`string`).

```html
<p id="parrafoEstilo">Texto para estilizar.</p>
```
```javascript
const parrafo = document.getElementById('parrafoEstilo');

// Establecer estilos
parrafo.style.color = 'blue';
parrafo.style.backgroundColor = '#f0f0f0'; // Corresponde a background-color
parrafo.style.fontSize = '20px';       // Corresponde a font-size
parrafo.style.marginTop = '15px';

// Leer estilos (solo lee estilos INLINE, no los de hojas CSS externas/internas)
console.log(parrafo.style.color); // "blue"
// console.log(parrafo.style.fontFamily); // "" (vacío si no se estableció inline)
```
*   **Importante:** La propiedad `style` solo refleja los estilos aplicados directamente al atributo `style` del elemento. **NO** lee los estilos aplicados a través de hojas de estilo CSS externas o internas. Para obtener el estilo *calculado* final (considerando todas las reglas CSS), se usa `window.getComputedStyle(elemento)`.
    ```javascript
    const estilosCalculados = window.getComputedStyle(parrafo);
    console.log(estilosCalculados.getPropertyValue('font-size')); // Devolverá el tamaño final (ej: '20px')
    console.log(estilosCalculados.fontFamily); // Devolverá la fuente final aplicada
    ```
*   Aunque puedes modificar estilos con `elemento.style`, generalmente es **mejor práctica** modificar las **clases** del elemento (`classList`) y definir los estilos correspondientes en tu archivo CSS. Esto mantiene la separación de intereses (HTML estructura, CSS presentación, JS comportamiento). Usa `elemento.style` para cambios muy dinámicos o específicos que serían difíciles de manejar solo con clases.

### 8_8_Modificacion_Clases_classList_add_remove_toggle

La propiedad `elemento.classList` proporciona métodos convenientes para manipular las clases CSS de un elemento sin tener que trabajar directamente con el string del atributo `class`. Es la **forma recomendada** de añadir/quitar estilos dinámicamente.

*   **`elemento.classList.add('nombreClase1', 'nombreClase2', ...)`:** Añade una o más clases al elemento.
*   **`elemento.classList.remove('nombreClase1', 'nombreClase2', ...)`:** Elimina una o más clases del elemento.
*   **`elemento.classList.toggle('nombreClase', force?)`:** Añade la clase si no existe, y la elimina si existe. Devuelve `true` si la clase fue añadida, `false` si fue eliminada. El segundo argumento opcional `force` (booleano) permite forzar la adición (`true`) o eliminación (`false`).
*   **`elemento.classList.contains('nombreClase')`:** Devuelve `true` si el elemento tiene la clase especificada, `false` si no.
*   **`elemento.classList.replace('claseAntigua', 'claseNueva')`:** Reemplaza una clase existente por una nueva.

```html
<div id="miCaja" class="caja visible">Contenido</div>
```
```css
/* En tu CSS */
.caja { border: 1px solid black; padding: 10px; }
.visible { background-color: lightgreen; }
.oculto { display: none; }
.destacado { font-weight: bold; border-color: red; }
```
```javascript
const caja = document.getElementById('miCaja');

// Añadir una clase
caja.classList.add('destacado');
// Ahora el div tiene class="caja visible destacado"

// Eliminar una clase
caja.classList.remove('visible');
// Ahora tiene class="caja destacado"

// Comprobar si tiene una clase
console.log(caja.classList.contains('destacado')); // true
console.log(caja.classList.contains('visible')); // false

// Alternar una clase (como no tiene 'oculto', la añade)
caja.classList.toggle('oculto');
console.log(caja.classList.contains('oculto')); // true

// Alternar de nuevo (como tiene 'oculto', la elimina)
caja.classList.toggle('oculto');
console.log(caja.classList.contains('oculto')); // false
```

### 8_9_Creacion_y_Eliminacion_Elementos

Puedes crear nuevos nodos DOM y añadirlos o eliminarlos de la estructura.

*   **`document.createElement('nombreEtiqueta')`:** Crea un nuevo nodo de elemento HTML (ej: `'p'`, `'div'`, `'li'`). El elemento se crea en memoria, aún no está en la página.
*   **`document.createTextNode('texto')`:** Crea un nuevo nodo de texto.
*   **`elementoPadre.appendChild(nuevoHijo)`:** Añade `nuevoHijo` como el **último** hijo de `elementoPadre`.
*   **`elementoPadre.insertBefore(nuevoNodo, nodoReferencia)`:** Inserta `nuevoNodo` antes del `nodoReferencia` existente dentro de `elementoPadre`. Si `nodoReferencia` es `null`, funciona como `appendChild`.
*   **`elemento.remove()`:** Elimina el `elemento` del DOM (método moderno).
*   **`elementoPadre.removeChild(hijoAeliminar)`:** Elimina `hijoAeliminar` de `elementoPadre` (método más antiguo).

```html
<ul id="miLista">
  <li>Item 1</li>
</ul>
<button id="btnAdd">Añadir Item</button>
<button id="btnRemove">Eliminar Último</button>
```
```javascript
const lista = document.getElementById('miLista');
const btnAdd = document.getElementById('btnAdd');
const btnRemove = document.getElementById('btnRemove');
let contador = 2;

btnAdd.addEventListener('click', () => {
  // 1. Crear nuevo elemento li
  const nuevoLi = document.createElement('li');

  // 2. Crear nodo de texto para el contenido
  const textoLi = document.createTextNode(`Item ${contador}`);

  // 3. Añadir el texto al li
  nuevoLi.appendChild(textoLi);

  // 4. Añadir el nuevo li a la lista ul
  lista.appendChild(nuevoLi);

  contador++;
});

btnRemove.addEventListener('click', () => {
  // Obtener el último elemento hijo de la lista
  const ultimoItem = lista.lastElementChild;
  // Si existe, eliminarlo
  if (ultimoItem) {
    // ultimoItem.remove(); // Forma moderna
    lista.removeChild(ultimoItem); // Forma antigua
  }
});
```

## 9_Eventos_del_DOM

Los eventos son acciones o sucesos que ocurren en la página web, como un clic del ratón, la pulsación de una tecla, el envío de un formulario, la carga completa de la página, etc. JavaScript permite **escuchar** estos eventos y **reaccionar** a ellos ejecutando código.

### 9_1_Introduccion_a_Eventos

El navegador dispara eventos constantemente. Podemos "enganchar" funciones (llamadas **manejadores de eventos** o **event listeners**) a elementos específicos para que se ejecuten cuando ocurra un evento determinado en ese elemento.

### 9_2_Manejadores_de_Eventos_addEventListener

La forma moderna y recomendada de registrar manejadores de eventos es usando el método `elemento.addEventListener()`.

Sintaxis: `elemento.addEventListener('tipoDeEvento', funcionManejadora, opciones?)`

*   `tipoDeEvento`: Un string con el nombre del evento (sin el prefijo "on"), ej: `'click'`, `'mouseover'`, `'keydown'`, `'submit'`, `'load'`.
*   `funcionManejadora`: La función que se ejecutará cuando ocurra el evento. Esta función recibe automáticamente un objeto `event` como argumento con información sobre el evento.
*   `opciones` (Opcional): Un objeto con opciones, como:
    *   `capture: true/false` (Define si el evento se captura en la fase de captura o bubbling, ver 9.5. Por defecto es `false`, bubbling).
    *   `once: true/false` (Si es `true`, el listener se elimina automáticamente después de ejecutarse una vez. Por defecto `false`).
    *   `passive: true/false` (Indica al navegador que el listener no llamará a `preventDefault()`. Puede mejorar el rendimiento del scroll en eventos táctiles. Por defecto `false`).

```html
<button id="miBoton">Haz Clic Aquí</button>
<input type="text" id="miInput">
```
```javascript
const boton = document.getElementById('miBoton');
const input = document.getElementById('miInput');

// Función manejadora nombrada
function manejarClicBoton(event) {
  console.log("¡Botón clickeado!");
  console.log("Tipo de evento:", event.type); // 'click'
  console.log("Elemento origen (target):", event.target); // El botón
  // 'this' dentro del manejador (si no es flecha) suele referirse al elemento (boton)
  console.log("Referencia con this:", this);
  this.textContent = "¡Clickeado!";
}

// Función manejadora anónima (flecha)
const manejarInput = (evento) => {
  console.log("Tecla presionada en el input:", evento.key);
  console.log("Valor actual:", evento.target.value);
  // 'this' en función flecha es el del contexto donde se definió addEventListener
};

// Registrar los listeners
boton.addEventListener('click', manejarClicBoton);
input.addEventListener('keydown', manejarInput);

// Para eliminar un listener, necesitas la referencia a la MISMA función
// boton.removeEventListener('click', manejarClicBoton);
```
*   Se pueden añadir múltiples listeners para el mismo evento en el mismo elemento.
*   `removeEventListener` requiere exactamente los mismos argumentos (incluyendo la referencia a la función y la opción `capture` si se usó) para eliminar un listener. Por eso es útil usar funciones nombradas si necesitas eliminarlos después.

### 9_3_El_Objeto_Event

Cuando se dispara un evento y se llama a la función manejadora, esta recibe automáticamente un objeto `event` (a menudo abreviado como `e` o `evt`) como primer argumento. Este objeto contiene información valiosa sobre el evento ocurrido.

Propiedades comunes del objeto `event`:

*   `event.type`: El tipo de evento (ej: `'click'`).
*   `event.target`: El elemento **original** sobre el que ocurrió el evento (el origen más profundo).
*   `event.currentTarget`: El elemento al que está **actualmente asociado** el listener (útil en bubbling o delegación). A menudo `this === event.currentTarget` en funciones manejadoras normales.
*   `event.preventDefault()`: Método para **cancelar la acción por defecto** del navegador para ese evento (ej: evitar que un clic en un enlace navegue, evitar que un formulario se envíe).
*   `event.stopPropagation()`: Método para **detener la propagación** del evento hacia elementos ancestros (detiene el bubbling/capturing).
*   `event.key` (eventos de teclado): La tecla presionada.
*   `event.code` (eventos de teclado): El código físico de la tecla.
*   `event.clientX`, `event.clientY` (eventos de ratón): Coordenadas X/Y relativas al viewport.
*   `event.pageX`, `event.pageY` (eventos de ratón): Coordenadas X/Y relativas al documento completo.
*   Y muchas más dependiendo del tipo de evento...

### 9_4_Eventos_Comunes_click_mouseover_keydown_submit_load

Algunos de los eventos más utilizados:

*   **Eventos de Ratón:**
    *   `click`: Clic izquierdo.
    *   `dblclick`: Doble clic.
    *   `mousedown`: Botón del ratón presionado.
    *   `mouseup`: Botón del ratón soltado.
    *   `mouseover`: El cursor entra sobre el elemento o uno de sus hijos.
    *   `mouseout`: El cursor sale del elemento o uno de sus hijos.
    *   `mouseenter`: El cursor entra sobre el elemento (no burbujea, más simple que `mouseover`).
    *   `mouseleave`: El cursor sale del elemento (no burbujea).
    *   `mousemove`: El cursor se mueve sobre el elemento.
*   **Eventos de Teclado:** (Normalmente se escuchan en `document`, `window` o campos de input).
    *   `keydown`: Tecla presionada.
    *   `keyup`: Tecla soltada.
    *   `keypress`: (Deprecado, evitar) Tecla que produce un carácter presionada.
*   **Eventos de Formulario:**
    *   `submit`: Se intenta enviar un formulario (normalmente se escucha en el elemento `<form>`). Se usa `event.preventDefault()` para manejar el envío con JS (AJAX).
    *   `change`: El valor de un elemento de formulario (`<input>`, `<select>`, `<textarea>`) cambia y el elemento pierde el foco.
    *   `input`: El valor de un `<input>` o `<textarea>` cambia (se dispara inmediatamente, más útil que `change` para feedback en tiempo real).
    *   `focus`: Un elemento recibe el foco.
    *   `blur`: Un elemento pierde el foco.
*   **Eventos del DOM / Ventana:**
    *   `DOMContentLoaded`: El HTML ha sido completamente cargado y parseado, el árbol DOM está listo (los scripts `defer` se ejecutan después de este). No espera a CSS, imágenes. **Muy útil** para empezar a ejecutar tu JS que manipula el DOM.
    *   `load`: La página completa (HTML, CSS, imágenes, etc.) se ha cargado. Se escucha en `window`.
    *   `resize`: La ventana del navegador cambia de tamaño. Se escucha en `window`.
    *   `scroll`: El documento o un elemento se desplaza (scroll). Se escucha en `window` o en el elemento con scroll.

### 9_5_Propagacion_de_Eventos_Bubbling_y_Capturing

Cuando un evento ocurre en un elemento, puede viajar a través del árbol DOM en dos fases:

1.  **Fase de Captura (Capturing Phase):** El evento viaja desde el ancestro más externo (`window`) hacia abajo hasta el elemento `target` donde ocurrió el evento. Los listeners registrados con `capture: true` se ejecutan en esta fase. Es menos común usarla.
2.  **Fase de Burbujeo (Bubbling Phase):** El evento viaja desde el `target` hacia arriba, a través de sus ancestros, hasta `window`. Los listeners registrados por defecto (`capture: false`) se ejecutan en esta fase. Es la fase más utilizada.

Imagina hacer clic en un botón dentro de un div, dentro del body:
*   **Captura:** Evento viaja: `window` -> `document` -> `<html>` -> `<body>` -> `div` -> `button`. Listeners `capture:true` se disparan en este orden.
*   **Burbujeo:** Evento viaja: `button` -> `div` -> `<body>` -> `<html>` -> `document` -> `window`. Listeners por defecto (`capture:false`) se disparan en este orden.

Puedes detener esta propagación en cualquier punto usando `event.stopPropagation()`.

### 9_6_Delegacion_de_Eventos

Es una técnica poderosa y eficiente que aprovecha el burbujeo (bubbling). En lugar de añadir un listener a cada uno de muchos elementos hijos (ej: a cada `<li>` de una lista larga), añades **un solo listener al contenedor padre**.

Dentro del manejador del padre, usas `event.target` para identificar qué elemento hijo específico originó el evento y actuar en consecuencia.

```html
<ul id="listaItems">
  <li>Item 1</li>
  <li>Item 2</li>
  <li>Item 3</li>
  <!-- Se añadirán más items dinámicamente -->
</ul>
```
```javascript
const lista = document.getElementById('listaItems');

lista.addEventListener('click', function(event) {
  // Comprobar si el elemento clickeado (event.target) es un LI
  if (event.target && event.target.tagName === 'LI') {
    console.log("Clickeaste en:", event.target.textContent);
    event.target.style.textDecoration = 'line-through'; // Tachar el item clickeado
  }
});

// Puedes añadir más LIs a la lista dinámicamente,
// y el listener en el UL seguirá funcionando para ellos.
```
**Ventajas de la Delegación:**
*   **Eficiencia:** Menos listeners que añadir y gestionar, mejor rendimiento, especialmente con muchos elementos.
*   **Dinamismo:** Funciona automáticamente para elementos añadidos al contenedor *después* de que el listener se haya registrado. No necesitas añadir listeners a los nuevos elementos.

### 9_7_preventDefault_y_stopPropagation

*   **`event.preventDefault()`:** Cancela la acción por defecto del navegador asociada al evento.
    *   Ej: Evitar que un enlace `<a>` navegue.
    *   Ej: Evitar que un formulario `<form>` se envíe y recargue la página (crucial para AJAX).
    *   Ej: Evitar que marcar un checkbox lo marque visualmente (si quieres manejarlo tú).
*   **`event.stopPropagation()`:** Detiene la propagación del evento (captura o burbujeo) a los elementos ancestros. Útil en casos específicos para evitar que listeners en elementos padres se disparen inesperadamente, pero úsalo con cuidado, ya que puede romper funcionalidades esperadas.

## 10_Javascript_Asincrono

JavaScript es, por naturaleza, **síncrono** y de **un solo hilo (single-threaded)**. Esto significa que ejecuta una instrucción a la vez, en orden. Si una operación tarda mucho (ej: una petición de red, una lectura de archivo grande en Node.js), bloquearía todo el hilo, haciendo que la interfaz de usuario (en el navegador) se congele.

El JavaScript Asíncrono permite realizar operaciones de larga duración sin bloquear el hilo principal, ejecutando otras tareas mientras espera que la operación asíncrona termine. Cuando termina, el resultado se maneja posteriormente.

### 10_1_Naturaleza_Asincrona_Event_Loop_Call_Stack_Queue

El navegador (o Node.js) gestiona la asincronía usando varios componentes:

*   **Call Stack (Pila de Llamadas):** Donde se apilan las funciones que se están ejecutando actualmente. JS toma la función de arriba, la ejecuta, y la saca. Es LIFO (Last-In, First-Out).
*   **Web APIs / Node APIs:** Interfaces proporcionadas por el entorno (navegador/Node) para operaciones asíncronas (ej: `setTimeout`, `fetch`, manipulación del DOM, acceso a archivos). Cuando llamas a una de estas, la operación se inicia fuera del hilo principal de JS.
*   **Callback Queue (Cola de Callbacks) / Microtask Queue:** Cuando una operación asíncrona termina (ej: `setTimeout` expira, `fetch` recibe respuesta), la función *callback* asociada (o la resolución de la Promesa) no se ejecuta inmediatamente. Se coloca en una cola de espera.
    *   **Callback Queue (o Task Queue):** Para callbacks de `setTimeout`, `setInterval`, eventos del DOM.
    *   **Microtask Queue:** Para callbacks de Promesas (`.then`, `.catch`, `.finally`) y `queueMicrotask`. Tiene **mayor prioridad** que la Callback Queue.
*   **Event Loop (Bucle de Eventos):** Es el corazón del modelo de concurrencia de JS. Constantemente revisa si el **Call Stack está vacío**. Si lo está, toma la primera tarea de la **Microtask Queue** y la empuja al Call Stack para su ejecución. Si la Microtask Queue está vacía, toma la primera tarea de la **Callback Queue** y la empuja al Call Stack.

Este ciclo permite que JS maneje operaciones largas sin bloquearse, procesando los resultados cuando el hilo principal está libre.

### 10_2_Callbacks

Una de las formas más antiguas de manejar la asincronía. Pasas una función (el callback) como argumento a una función asíncrona. Esta función callback será llamada cuando la operación asíncrona termine.

```javascript
console.log("Inicio");

// setTimeout es una Web API asíncrona
setTimeout(function() { // Esta función es el callback
  console.log("Timeout de 2 segundos terminado");
}, 2000); // Espera 2000ms (2s)

console.log("Fin (antes del timeout)");

// Función simulando una petición con callback
function obtenerDatos(url, callbackExito, callbackError) {
  console.log(`Pidiendo datos de ${url}...`);
  setTimeout(() => { // Simula retardo de red
    let exito = Math.random() > 0.3; // Simula éxito/error aleatorio
    if (exito) {
      let datos = { id: 1, contenido: "Datos recibidos!" };
      callbackExito(datos);
    } else {
      callbackError("Error al obtener datos");
    }
  }, 1500);
}

obtenerDatos(
  "/api/data",
  (datosRecibidos) => { // Callback de éxito
    console.log("Éxito:", datosRecibidos);
  },
  (error) => { // Callback de error
    console.error("Error:", error);
  }
);
```
**Problema:** Anidar múltiples operaciones asíncronas dependientes usando callbacks puede llevar al "Callback Hell" o "Pyramid of Doom", haciendo el código difícil de leer y mantener.

```javascript
// Callback Hell (Ejemplo conceptual)
operacion1(arg1, (res1) => {
  operacion2(res1, (res2) => {
    operacion3(res2, (res3) => {
      operacion4(res3, (res4) => {
        // ...muchos niveles de anidación
      }, errorCallback);
    }, errorCallback);
  }, errorCallback);
}, errorCallback);
```

### 10_3_Promesas_Promises_then_catch_finally

Introducidas en ES6 para solucionar los problemas de los callbacks. Una `Promise` es un objeto que representa el resultado eventual (éxito o fallo) de una operación asíncrona.

Una Promesa puede estar en uno de estos estados:
*   **`pending` (pendiente):** Estado inicial, la operación aún no ha terminado.
*   **`fulfilled` (resuelta/cumplida):** La operación terminó con éxito. La promesa tiene un valor resultante.
*   **`rejected` (rechazada):** La operación falló. La promesa tiene una razón (error).

Las promesas se consumen usando los métodos `.then()`, `.catch()` y `.finally()`:

*   **`.then(onFulfilled, onRejected?)`:** Se ejecuta cuando la promesa se cumple (`fulfilled`). Recibe el valor resultante como argumento en `onFulfilled`. Opcionalmente, puede recibir un segundo argumento `onRejected` para manejar errores (aunque `.catch()` es preferible para esto). `.then()` devuelve una *nueva* promesa, permitiendo **encadenar** operaciones asíncronas de forma legible.
*   **`.catch(onRejected)`:** Se ejecuta si la promesa es rechazada (`rejected`). Recibe el error como argumento. Es la forma preferida de manejar errores en la cadena de promesas. También devuelve una promesa.
*   **`.finally(onFinally)`:** Se ejecuta *siempre* cuando la promesa se resuelve (ya sea `fulfilled` o `rejected`). No recibe argumentos. Útil para limpieza (ej: ocultar un spinner de carga). También devuelve una promesa.

```javascript
console.log("Inicio Promesa");

// Función que devuelve una promesa
function obtenerDatosPromesa(url) {
  return new Promise((resolve, reject) => {
    // El código dentro del constructor de Promise se ejecuta inmediatamente
    console.log(`(Promesa) Pidiendo datos de ${url}...`);
    setTimeout(() => { // Simular operación asíncrona
      let exito = Math.random() > 0.3;
      if (exito) {
        let datos = { id: 2, contenido: "Datos con Promesa!" };
        resolve(datos); // Resuelve la promesa con los datos (pasa a fulfilled)
      } else {
        reject(new Error("Error en la Promesa")); // Rechaza la promesa con un error
      }
    }, 1800);
  });
}

obtenerDatosPromesa("/api/data-promise")
  .then(datos => { // Se ejecuta si la promesa se resuelve (fulfilled)
    console.log("Promesa resuelta:", datos);
    // Puedes encadenar más operaciones asíncronas devolviendo otra promesa
    // return otraOperacionAsincrona(datos.id);
    return datos.contenido.toUpperCase(); // O devolver un valor síncrono
  })
  .then(contenidoMayusculas => { // Se ejecuta con el resultado del .then anterior
      console.log("Contenido transformado:", contenidoMayusculas);
  })
  .catch(error => { // Se ejecuta si la promesa original o alguna anterior en la cadena es rechazada
    console.error("Promesa rechazada:", error.message);
  })
  .finally(() => { // Se ejecuta siempre al final
    console.log("Promesa terminada (finally)");
  });

console.log("Fin Promesa (antes de que termine)");
```
Las promesas mejoran enormemente la legibilidad y el manejo de errores en código asíncrono comparado con los callbacks.

### 10_4_Async_Await_Sintaxis_Simplificada_Promesas

Introducido en ES2017 (ES8), `async/await` es "azúcar sintáctico" construido sobre las Promesas. Permite escribir código asíncrono que **parece síncrono**, haciéndolo aún más legible y fácil de razonar.

*   **`async`:** Palabra clave que se pone **antes** de la declaración de una función (`async function miFuncion() {...}` o `const miFunc = async () => {...}`). Indica que la función *siempre* devolverá una Promesa. Si la función devuelve un valor normal, `async` lo envolverá automáticamente en una Promesa resuelta.
*   **`await`:** Palabra clave que solo se puede usar **dentro** de una función `async`. Pausa la ejecución de la función `async` hasta que la Promesa a la derecha de `await` se resuelva (se cumpla o se rechace).
    *   Si la Promesa se cumple (`fulfilled`), `await` devuelve el valor resuelto.
    *   Si la Promesa se rechaza (`rejected`), `await` **lanza** el error (como si fuera un `throw`).

```javascript
console.log("Inicio Async/Await");

// Usando la misma función obtenerDatosPromesa de antes

async function procesarDatos() {
  console.log("Dentro de async function - Inicio");
  try {
    // Pausa aquí hasta que la promesa se resuelva
    const datos = await obtenerDatosPromesa("/api/data-async");
    // Si la promesa se cumple, 'datos' contiene el resultado
    console.log("Async/Await - Datos recibidos:", datos);

    // Puedes hacer más awaits si necesitas encadenar
    // const resultado2 = await otraOperacionAsincrona(datos.id);
    // console.log("Resultado 2:", resultado2);

    console.log("Dentro de async function - Fin");
    return "Proceso completado con éxito"; // Se envolverá en una Promesa resuelta

  } catch (error) {
    // Si cualquier 'await' lanza un error (promesa rechazada), se captura aquí
    console.error("Async/Await - Error capturado:", error.message);
    return "Proceso fallido"; // Se envolverá en una Promesa resuelta con este valor
  } finally {
      console.log("Async/Await - Bloque finally ejecutado");
  }
}

// Llamar a la función async devuelve una promesa
const promesaResultado = procesarDatos();

promesaResultado.then(mensajeFinal => {
    console.log("Resultado final de procesarDatos:", mensajeFinal);
});

console.log("Fin Async/Await (antes de que termine la función async)");
```
*   `async/await` simplifica enormemente el código asíncrono, especialmente cuando hay múltiples pasos dependientes.
*   El manejo de errores se realiza con bloques `try...catch` estándar, lo cual es más familiar que `.catch()`.

### 10_5_Fetch_API_Peticiones_HTTP

Es la API moderna incorporada en los navegadores (y disponible en Node.js con librerías) para realizar peticiones de red (HTTP), por ejemplo, para obtener datos de una API. `fetch()` está **basado en Promesas**.

Sintaxis básica: `fetch(url, opciones?)`

*   `url`: La URL del recurso al que quieres acceder.
*   `opciones` (Opcional): Un objeto para configurar la petición (método HTTP, cabeceras, cuerpo de la petición, etc.). Por defecto, `fetch` hace una petición `GET`.

`fetch()` devuelve una Promesa que se resuelve con un objeto `Response` tan pronto como se reciben las **cabeceras** de la respuesta del servidor. El cuerpo de la respuesta aún no está disponible.

El objeto `Response` tiene métodos para leer el cuerpo de la respuesta de diferentes formas, y estos métodos también devuelven Promesas:
*   `response.json()`: Parsea el cuerpo como JSON.
*   `response.text()`: Lee el cuerpo como texto.
*   `response.blob()`: Lee el cuerpo como un objeto Blob (datos binarios).
*   `response.formData()`: Lee el cuerpo como FormData.

```javascript
const apiUrl = 'https://jsonplaceholder.typicode.com/posts/1'; // API pública de ejemplo

console.log("Iniciando fetch...");

fetch(apiUrl)
  .then(response => {
    console.log("Respuesta recibida (cabeceras)");
    // fetch NO rechaza la promesa por errores HTTP (404, 500).
    // Debemos comprobar el estado nosotros mismos.
    if (!response.ok) { // ok es true si el estado es 200-299
      throw new Error(`Error HTTP: ${response.status} ${response.statusText}`);
    }
    // Devolver la promesa de leer el cuerpo como JSON
    return response.json();
  })
  .then(data => {
    // Aquí 'data' es el objeto JavaScript parseado del JSON
    console.log("Datos recibidos (JSON parseado):", data);
    console.log("Título:", data.title);
  })
  .catch(error => {
    // Captura errores de red (ej: sin conexión) o los lanzados por nosotros (ej: !response.ok)
    console.error("Error en fetch:", error);
  });

console.log("Fetch iniciado (código síncrono continúa)");

// Ejemplo con async/await
async function obtenerPost() {
  try {
    console.log("Iniciando fetch (async)...");
    const response = await fetch(apiUrl);
    console.log("Respuesta recibida (async)");

    if (!response.ok) {
      throw new Error(`Error HTTP: ${response.status} ${response.statusText}`);
    }

    const data = await response.json(); // Espera a que el cuerpo se parseé
    console.log("Datos recibidos (async):", data.title);
    return data;

  } catch (error) {
    console.error("Error en fetch (async):", error);
  }
}

obtenerPost();
```
*   **Para enviar datos (POST, PUT, etc.):** Se usa el segundo argumento `opciones` de `fetch`.
    ```javascript
    fetch('/api/users', {
      method: 'POST', // o 'PUT', 'DELETE', etc.
      headers: {
        'Content-Type': 'application/json' // Indicar que enviamos JSON
        // Otras cabeceras si son necesarias (ej: 'Authorization': 'Bearer TOKEN')
      },
      body: JSON.stringify({ nombre: "Nuevo Usuario", email: "user@example.com" }) // Convertir objeto JS a JSON string
    })
    .then(response => response.json()) // Asumiendo que la API devuelve JSON
    .then(data => console.log("Respuesta del POST:", data))
    .catch(error => console.error("Error en POST:", error));
    ```

### 10_6_Manejo_de_Errores_Asincronos

*   **Callbacks:** Usar un parámetro separado para el callback de error.
*   **Promesas:** Usar el método `.catch()` al final de la cadena `.then()`. Capturará cualquier rechazo ocurrido en la promesa original o en cualquier `.then()` anterior. También puedes pasar un segundo callback a `.then()`, pero es menos común.
*   **Async/Await:** Usar bloques `try...catch` estándar para envolver las llamadas `await`. Cualquier rechazo de una promesa esperada con `await` será capturado por el `catch`.

Es crucial manejar los errores en operaciones asíncronas para evitar que la aplicación falle silenciosamente o muestre comportamientos inesperados.

## 11_Caracteristicas_ES6_y_Posteriores

ECMAScript 2015 (ES6) y las versiones posteriores han añadido muchas características que hacen JavaScript más moderno, legible y potente. Ya hemos visto algunas (`let`, `const`, funciones flecha, promesas, `async/await`). Aquí otras importantes:

### 11_1_let_y_const_vs_var

Revisión: `let` y `const` tienen ámbito de bloque, no se elevan (hoisting) de forma problemática como `var` (tienen TDZ), y `const` previene la reasignación. Son la forma estándar de declarar variables hoy en día.

### 11_2_Funciones_Flecha_Revisado

Revisión: Sintaxis más concisa, `this` léxico (heredado del entorno), sin objeto `arguments` (usar `...rest`), no se pueden usar como constructores. Ideales para callbacks y funciones cortas.

### 11_3_Template_Literals_Plantillas_de_Texto

Permiten incrustar expresiones y crear strings multilínea de forma más sencilla usando backticks (`` ` ``).

*   **Interpolación de Expresiones:** Usa `${expresion}` dentro de los backticks para evaluar e insertar el valor de una expresión (variable, llamada a función, etc.) directamente en la cadena.
*   **Multilínea:** Simplemente escribe el string en varias líneas dentro de los backticks.

```javascript
let nombre = "Alice";
let edad = 30;

// Antes (concatenación)
let saludoAntiguo = "Hola, mi nombre es " + nombre + " y tengo " + edad + " años.\nEsta es otra línea.";

// Con Template Literals
let saludoNuevo = `Hola, mi nombre es ${nombre} y tengo ${edad} años.
Esta es otra línea.
Puedo incluso llamar a funciones: ${edad * 2}.`;

console.log(saludoAntiguo);
console.log(saludoNuevo);
```

### 11_4_Desestructuracion_Destructuring_Arrays_y_Objetos

Permite extraer valores de arrays u objetos y asignarlos a variables distintas de forma concisa.

*   **Desestructuración de Arrays:**
    ```javascript
    const numeros = [10, 20, 30, 40, 50];
    const [primero, segundo, , cuarto] = numeros; // Ignora el tercer elemento

    console.log(primero); // 10
    console.log(segundo); // 20
    console.log(cuarto);  // 40

    // Con valor por defecto
    const [a, b, c = 3] = [1, 2];
    console.log(c); // 3

    // Con rest operator
    const [head, ...tail] = numeros;
    console.log(head); // 10
    console.log(tail); // [20, 30, 40, 50]
    ```
*   **Desestructuración de Objetos:** Se usan llaves `{}` y los nombres de las variables deben coincidir con los nombres de las propiedades.
    ```javascript
    const persona = { nombre: "Bob", edad: 42, ciudad: "Madrid" };

    // Extraer propiedades a variables con el mismo nombre
    const { nombre, edad } = persona;
    console.log(nombre); // "Bob"
    console.log(edad);   // 42

    // Extraer y renombrar variables
    const { nombre: nombreCompleto, ciudad: lugarResidencia } = persona;
    console.log(nombreCompleto);    // "Bob"
    console.log(lugarResidencia); // "Madrid"

    // Con valor por defecto
    const { profesion = "Desconocida" } = persona;
    console.log(profesion); // "Desconocida"

    // Con rest operator
    const { edad: age, ...restoDatos } = persona;
    console.log(age); // 42
    console.log(restoDatos); // { nombre: "Bob", ciudad: "Madrid" }

    // En parámetros de función
    function saludarPersona({ nombre, edad }) {
      console.log(`Hola ${nombre}, tienes ${edad} años.`);
    }
    saludarPersona(persona); // Hola Bob, tienes 42 años.
    ```

### 11_5_Parametros_por_Defecto_y_Operador_Rest_Spread

*   **Parámetros por Defecto:** (Visto en Funciones) Permiten asignar un valor por defecto a un parámetro si no se pasa un argumento.
*   **Operador Rest (`...`) en Parámetros:** (Visto en Funciones) Agrupa argumentos restantes en un array.
*   **Operador Spread (`...`) en Llamadas y Literales:** Hace lo contrario que Rest: "expande" un iterable (array, string) o las propiedades de un objeto en lugares donde se esperan elementos individuales o pares clave-valor.
    *   **En Arrays:**
        ```javascript
        const arr1 = [1, 2, 3];
        const arr2 = [4, 5, 6];
        const combinado = [...arr1, 0, ...arr2]; // [1, 2, 3, 0, 4, 5, 6] - Combina arrays
        const copiaArr1 = [...arr1]; // Crea una copia superficial

        function sumar(a, b, c) { return a + b + c; }
        console.log(sumar(...arr1)); // 6 (Expande arr1 como argumentos: sumar(1, 2, 3))
        ```
    *   **En Objetos (ES2018):**
        ```javascript
        const obj1 = { a: 1, b: 2 };
        const obj2 = { b: 3, c: 4 }; // 'b' se sobrescribirá
        const combinadoObj = { ...obj1, ...obj2, d: 5 }; // { a: 1, b: 3, c: 4, d: 5 }
        const copiaObj1 = { ...obj1 }; // Copia superficial
        ```

### 11_6_Clases_Syntactic_Sugar_para_Prototipos

ES6 introdujo la sintaxis de `class` para crear objetos y gestionar la herencia, que es más familiar para desarrolladores de otros lenguajes orientados a objetos (como Java o C++). Sin embargo, **internamente, JavaScript sigue usando prototipos**. `class` es principalmente "azúcar sintáctico" sobre el sistema de prototipos existente.

```javascript
class Persona {
  // Constructor: se llama al crear una instancia con 'new'
  constructor(nombre, edad) {
    this.nombre = nombre;
    this.edad = edad;
  }

  // Método (se añade al prototipo de Persona)
  saludar() {
    console.log(`Hola, soy ${this.nombre} y tengo ${this.edad} años.`);
  }

  // Método estático (pertenece a la clase, no a las instancias)
  static infoClase() {
    console.log("Esta es la clase Persona.");
  }

  // Getter
  get nombreMayusculas() {
    return this.nombre.toUpperCase();
  }

  // Setter
  set anioNacimiento(anio) {
      this.edad = new Date().getFullYear() - anio;
  }
}

// Crear instancia
const p1 = new Persona("David", 28);
p1.saludar(); // Hola, soy David y tengo 28 años.
console.log(p1.nombreMayusculas); // DAVID (usa el getter)
p1.anioNacimiento = 1995; // Usa el setter para calcular edad
p1.saludar(); // Hola, soy David y tengo ... años. (edad actualizada)

// Llamar método estático
Persona.infoClase(); // Esta es la clase Persona.
```
Veremos la herencia con clases en la sección de POO.

### 11_7_Modulos_import_export

ES6 introdujo un sistema de módulos nativo para organizar el código en archivos separados y reutilizables, controlando qué se exporta (se hace público) y qué se importa (se usa) en otros archivos.

*   **Exportar:** Usa `export` para hacer que variables, funciones o clases estén disponibles para otros módulos.
    ```javascript
    // --- archivo: utils.js ---
    export const PI = 3.14159;

    export function sumar(a, b) {
      return a + b;
    }

    export default function multiplicar(a, b) { // Exportación por defecto (solo una por módulo)
      return a * b;
    }

    class Calculadora {
        // ...
    }
    export { Calculadora }; // Exportación nombrada de una clase existente
    ```
*   **Importar:** Usa `import` para traer funcionalidades exportadas de otros módulos.
    ```javascript
    // --- archivo: main.js ---

    // Importar exportaciones nombradas (entre llaves)
    import { PI, sumar, Calculadora } from './utils.js';

    // Importar la exportación por defecto (nombre elegido por el importador)
    import miMultiplicador from './utils.js';

    // Importar todo como un objeto de módulo
    import * as utils from './utils.js';

    console.log(PI); // 3.14159
    console.log(sumar(2, 3)); // 5
    console.log(miMultiplicador(4, 5)); // 20
    console.log(utils.PI); // 3.14159 (acceso a través del objeto del módulo)
    const calc = new Calculadora();
    ```
*   **En Navegadores:** Para usar módulos ES6 nativamente, la etiqueta `<script>` que importa debe tener el atributo `type="module"`.
    ```html
    <script type="module" src="main.js"></script>
    ```
*   Los módulos ayudan a crear código más organizado, mantenible y evitan la contaminación del ámbito global. Son estándar en el desarrollo moderno con frameworks y herramientas de empaquetado (bundlers) como Webpack o Rollup.

## 12_Programacion_Orientada_a_Objetos_POO

La POO es un paradigma de programación basado en el concepto de "objetos", que pueden contener datos (en forma de campos o atributos) y código (en forma de procedimientos o métodos). JavaScript soporta POO, principalmente a través de su sistema de **prototipos**, aunque la sintaxis de `class` (ES6) lo hace más accesible.

### 12_1_Principios_Basicos_Encapsulacion_Herencia_Polimorfismo

*   **Encapsulación:** Agrupar datos y los métodos que operan sobre esos datos dentro de un objeto. A menudo implica ocultar el estado interno (datos) y exponer solo las operaciones necesarias (métodos). JavaScript no tiene modificadores de acceso `private`/`public` estrictos como otros lenguajes, pero se pueden simular con convenciones (ej: prefijo `_`) o características más nuevas como campos privados (`#`).
*   **Herencia:** Permite que una clase (o prototipo) hijo herede propiedades y métodos de una clase (o prototipo) padre. Promueve la reutilización de código. En JavaScript, la herencia se basa en la **cadena de prototipos**.
*   **Polimorfismo:** ("Muchas formas") Permite que objetos de diferentes clases respondan al mismo mensaje (llamada a método) de maneras diferentes. En JavaScript, esto se logra a menudo a través de la herencia y la sobrescritura de métodos, o mediante duck typing (si camina como un pato y grazna como un pato, es un pato).

### 12_2_Prototipos_y_Cadena_de_Prototipos

Es el mecanismo fundamental de herencia en JavaScript.
*   Cada objeto en JavaScript tiene una propiedad interna (a menudo accesible como `__proto__` o mediante `Object.getPrototypeOf()`) que es un **enlace a otro objeto**, llamado su **prototipo**.
*   Cuando intentas acceder a una propiedad o método en un objeto, si el objeto no lo tiene directamente, JavaScript busca en su prototipo. Si tampoco está allí, busca en el prototipo del prototipo, y así sucesivamente, subiendo por la **cadena de prototipos** hasta llegar a `Object.prototype` (el prototipo base de la mayoría de los objetos) o hasta encontrar `null`.
*   Las funciones constructoras y las clases ES6 utilizan esta cadena para establecer la herencia. Las instancias creadas con `new` heredan de la propiedad `prototype` de su función constructora/clase.

```javascript
// Función constructora (forma pre-ES6)
function Animal(nombre) {
  this.nombre = nombre;
}

// Añadir un método al PROTOTIPO de Animal
// Todas las instancias de Animal heredarán este método
Animal.prototype.hablar = function() {
  console.log("El animal hace un sonido.");
};

Animal.prototype.comer = function() {
    console.log(`${this.nombre} está comiendo.`);
}

const animalGenerico = new Animal("Criatura");
animalGenerico.hablar(); // "El animal hace un sonido." (Método heredado)
animalGenerico.comer(); // "Criatura está comiendo."

// Comprobar el prototipo
console.log(Object.getPrototypeOf(animalGenerico) === Animal.prototype); // true
console.log(Object.getPrototypeOf(Animal.prototype) === Object.prototype); // true
```

### 12_3_Funciones_Constructoras_y_new

Antes de ES6, la forma principal de crear "clases" y objetos era mediante funciones constructoras y el operador `new`.

*   Una función constructora es una función normal, pero se invoca con `new`. Por convención, sus nombres empiezan con mayúscula.
*   Dentro de la constructora, `this` se refiere a la nueva instancia vacía que se está creando. Se añaden propiedades a `this`.
*   Los métodos generalmente se añaden a la propiedad `prototype` de la función constructora para que todas las instancias los compartan (eficiencia de memoria) en lugar de crear una copia para cada instancia.
*   `new` hace cuatro cosas:
    1.  Crea un nuevo objeto vacío `{}`.
    2.  Establece el prototipo (`__proto__`) de este nuevo objeto al `prototype` de la función constructora.
    3.  Llama a la función constructora con `this` apuntando al nuevo objeto.
    4.  Devuelve el nuevo objeto (a menos que la constructora devuelva explícitamente otro objeto).

```javascript
function Perro(nombre, raza) {
  // Llama al constructor padre (simulando herencia pre-ES6)
  Animal.call(this, nombre); // Establece this.nombre
  this.raza = raza;
}

// Herencia pre-ES6: Establecer la cadena de prototipos
Perro.prototype = Object.create(Animal.prototype);
Perro.prototype.constructor = Perro; // Restaurar el constructor

// Sobrescribir método heredado
Perro.prototype.hablar = function() {
  console.log("¡Guau guau!");
};

// Añadir método específico de Perro
Perro.prototype.moverCola = function() {
  console.log(`${this.nombre} mueve la cola.`);
};

const miPerro = new Perro("Buddy", "Golden Retriever");
miPerro.hablar(); // "¡Guau guau!" (Método sobrescrito)
miPerro.comer(); // "Buddy está comiendo." (Método heredado de Animal)
miPerro.moverCola(); // "Buddy mueve la cola." (Método propio de Perro)

console.log(miPerro instanceof Perro); // true
console.log(miPerro instanceof Animal); // true
```
Aunque funcional, la sintaxis de prototipos y herencia pre-ES6 es más verbosa y menos intuitiva que `class`.

### 12_4_Clases_ES6_constructor_metodos_getters_setters

La sintaxis `class` simplifica la creación de objetos y la gestión de prototipos.

*   **`class NombreClase { ... }`**: Define la clase.
*   **`constructor(...) { ... }`**: Método especial para inicializar las instancias. Se llama automáticamente con `new`.
*   **Métodos:** Se definen directamente dentro de la clase (sin `function`). Se añaden automáticamente al prototipo.
*   **Getters (`get nombrePropiedad() { ... }`)**: Permiten definir una propiedad cuyo valor se calcula al leerla.
*   **Setters (`set nombrePropiedad(valor) { ... }`)**: Permiten ejecutar código cuando se intenta asignar un valor a una propiedad.
*   **Métodos Estáticos (`static nombreMetodo() { ... }`)**: Pertenecen a la clase en sí, no a las instancias. Se llaman con `NombreClase.nombreMetodo()`.
*   **Campos Públicos/Privados (Más recientes):** Sintaxis para definir propiedades directamente en la clase (fuera del constructor) y campos privados con `#`.
    ```javascript
    class Rectangulo {
      // Campo privado (propuesta reciente)
      #ancho;
      #alto;

      constructor(ancho, alto) {
        this.#ancho = ancho;
        this.#alto = alto;
      }

      // Getter
      get area() {
        return this.#calcularArea();
      }

      // Setter
      set dimensiones(dims) {
          if(Array.isArray(dims) && dims.length === 2) {
              this.#ancho = dims[0];
              this.#alto = dims[1];
          } else {
              console.error("Dimensiones debe ser un array [ancho, alto]");
          }
      }

      // Método normal
      mostrarDimensiones() {
        console.log(`Ancho: ${this.#ancho}, Alto: ${this.#alto}`);
      }

      // Método privado (convención o usando #)
      #calcularArea() {
          return this.#ancho * this.#alto;
      }

      // Método estático
      static esCuadrado(rect) {
          return rect.#ancho === rect.#alto;
      }
    }

    const r1 = new Rectangulo(10, 5);
    r1.mostrarDimensiones(); // Ancho: 10, Alto: 5
    console.log(r1.area); // 50 (Usa el getter 'area')
    // console.log(r1.#ancho); // Error: Campo privado
    // r1.#calcularArea(); // Error: Método privado

    r1.dimensiones = [12, 6]; // Usa el setter 'dimensiones'
    r1.mostrarDimensiones(); // Ancho: 12, Alto: 6

    console.log(Rectangulo.esCuadrado(r1)); // false (Llama método estático)
    ```

### 12_5_Herencia_con_Clases_extends_super

La sintaxis `class` simplifica enormemente la herencia.

*   **`extends`:** Se usa en la declaración de la clase hija para indicar de qué clase padre hereda.
*   **`super(...)`:** Dentro del `constructor` de la clase hija, `super()` **debe** llamarse *antes* de usar `this`. Llama al constructor de la clase padre.
*   **`super.metodoPadre(...)`:** Permite llamar a un método específico de la clase padre desde un método de la clase hija (útil si sobrescribes un método pero quieres reutilizar la lógica del padre).

```javascript
// Clase Padre (la clase Animal de antes)
class Animal {
  constructor(nombre) {
    this.nombre = nombre;
  }
  hablar() {
    console.log("El animal hace un sonido.");
  }
  comer() {
    console.log(`${this.nombre} está comiendo.`);
  }
}

// Clase Hija que hereda de Animal
class Gato extends Animal {
  constructor(nombre, vidas = 9) {
    // 1. Llamar al constructor del padre OBLIGATORIO antes de this
    super(nombre);
    // 2. Propiedades específicas del hijo
    this.vidas = vidas;
  }

  // Sobrescribir método del padre
  hablar() {
    console.log("¡Miau!");
  }

  // Método específico del hijo
  ronronear() {
    console.log(`${this.nombre} está ronroneando...`);
  }

  // Método que llama al método del padre
  presentarse() {
      super.comer(); // Llama a comer() de Animal
      this.ronronear();
      console.log(`Me quedan ${this.vidas} vidas.`);
  }
}

const michi = new Gato("Pelusa");
michi.hablar(); // ¡Miau! (Método sobrescrito)
michi.comer(); // Pelusa está comiendo. (Método heredado)
michi.ronronear(); // Pelusa está ronroneando... (Método propio)
michi.presentarse();

console.log(michi instanceof Gato); // true
console.log(michi instanceof Animal); // true
```

## 13_Manejo_de_Errores

Es inevitable que ocurran errores en las aplicaciones. Un buen manejo de errores previene que la aplicación se rompa inesperadamente y proporciona información útil para depuración o para el usuario.

### 13_1_Bloques_try_catch_finally

Permiten ejecutar código que *podría* lanzar un error (`try`), capturar ese error si ocurre (`catch`), y ejecutar código de limpieza independientemente de si hubo error o no (`finally`).

```javascript
console.log("Inicio del bloque try...catch");

try {
  // Código que podría fallar
  console.log("Intentando ejecutar código riesgoso...");
  let resultado;
  // Simulamos un error: variableNoDefinida no existe
  // resultado = variableNoDefinida * 2;

  // Otro ejemplo: parsear JSON inválido
  JSON.parse("{ json invalido ");

  console.log("Código riesgoso ejecutado sin error."); // No se ejecutará si hay error arriba

} catch (error) {
  // Este bloque se ejecuta SI Y SOLO SI ocurre un error en el bloque try
  console.error("¡Ocurrió un error!");
  console.error("Nombre del Error:", error.name);     // Ej: ReferenceError, SyntaxError, TypeError
  console.error("Mensaje de Error:", error.message); // Descripción del error
  // console.error("Stack Trace:", error.stack); // Pila de llamadas (muy útil para depurar)

  // Aquí puedes manejar el error: mostrar un mensaje al usuario, registrar el error, etc.

} finally {
  // Este bloque se ejecuta SIEMPRE, haya habido error o no.
  // Útil para limpieza: cerrar archivos, liberar recursos, etc.
  console.log("Bloque finally ejecutado (limpieza).");
}

console.log("Fin del bloque try...catch (continuación del programa)");
```
*   El bloque `catch` recibe un objeto `error` con información sobre el fallo.
*   El bloque `finally` es opcional.

### 13_2_El_Objeto_Error

Cuando ocurre un error, JavaScript generalmente crea un objeto `Error` (o uno de sus subtipos como `SyntaxError`, `TypeError`, `ReferenceError`, etc.). Este objeto tiene propiedades útiles:

*   `error.name`: El tipo de error.
*   `error.message`: La descripción textual del error.
*   `error.stack` (No estándar, pero muy común): La pila de llamadas que llevó al error, muy útil para depurar.

### 13_3_Lanzar_Errores_throw

Puedes crear y lanzar tus propios errores usando la sentencia `throw`. Esto detiene la ejecución normal y busca el bloque `catch` más cercano en la pila de llamadas.

Puedes lanzar cualquier valor (string, número), pero es **muy recomendable lanzar un objeto `Error`** (o una instancia de una clase de error personalizada que herede de `Error`) porque proporciona más información (nombre, mensaje, stack).

```javascript
function dividir(a, b) {
  if (b === 0) {
    // Lanzar un error específico si el divisor es cero
    throw new Error("División por cero no permitida.");
  }
  if (typeof a !== 'number' || typeof b !== 'number') {
      throw new TypeError("Ambos argumentos deben ser números.");
  }
  return a / b;
}

try {
  let resultado1 = dividir(10, 2);
  console.log("Resultado 1:", resultado1); // 5

  let resultado2 = dividir(8, 0); // Esto lanzará un error
  console.log("Resultado 2:", resultado2); // No se ejecutará

} catch (error) {
  console.error(`Error al dividir: ${error.message}`);
}

try {
    dividir("hola", 2); // Lanza TypeError
} catch(e) {
    console.error(`Error de tipo: ${e.message}`);
}

console.log("Programa continúa después de los errores manejados.");
```
Lanzar errores personalizados es útil para manejar condiciones de error específicas de tu aplicación.

## 14_APIs_del_Navegador_Web_APIs

Son conjuntos de funcionalidades proporcionadas por el entorno del navegador (no son parte del núcleo de JavaScript en sí) que permiten a tu código interactuar con el navegador y el sistema del usuario. Ya hemos visto algunas como el DOM, `setTimeout`, `fetch`.

### 14_1_Almacenamiento_Web_localStorage_sessionStorage

Permiten almacenar datos (clave-valor, ambos strings) en el navegador del usuario de forma persistente o temporal.

*   **`localStorage`:**
    *   Almacena datos **sin fecha de caducidad**. Persisten incluso después de cerrar el navegador y volver a abrirlo.
    *   Los datos están vinculados al **origen** (protocolo + dominio + puerto). Un sitio no puede acceder al `localStorage` de otro.
    *   Límite de almacenamiento relativamente grande (generalmente 5-10 MB por origen).
    *   **Métodos:**
        *   `localStorage.setItem('clave', 'valor')`: Almacena/actualiza un par.
        *   `localStorage.getItem('clave')`: Obtiene el valor de una clave (devuelve `null` si no existe).
        *   `localStorage.removeItem('clave')`: Elimina un par.
        *   `localStorage.clear()`: Elimina *todos* los pares del origen actual.
        *   `localStorage.length`: Número de pares almacenados.
        *   `localStorage.key(indice)`: Obtiene la clave en una posición dada.
    ```javascript
    // Guardar datos
    localStorage.setItem('usuario', 'Ana');
    localStorage.setItem('tema', 'oscuro');
    const prefs = { lang: 'es', notify: true };
    localStorage.setItem('preferencias', JSON.stringify(prefs)); // Guardar objetos como JSON string

    // Leer datos
    let temaGuardado = localStorage.getItem('tema'); // "oscuro"
    let prefsGuardadas = JSON.parse(localStorage.getItem('preferencias')); // Parsear de vuelta a objeto

    // Eliminar dato
    localStorage.removeItem('usuario');

    // Limpiar todo
    // localStorage.clear();
    ```
*   **`sessionStorage`:**
    *   Funciona **exactamente igual** que `localStorage` (mismos métodos).
    *   La **diferencia clave** es que los datos almacenados en `sessionStorage` **se eliminan automáticamente cuando la sesión del navegador termina** (es decir, cuando se cierra la pestaña o el navegador).
    *   También está vinculado al origen.
    *   Útil para datos temporales que no necesitan persistir entre sesiones (ej: estado temporal de la UI, datos de un formulario no enviado).

**Consideraciones:**
*   Ambos almacenan solo **strings**. Para guardar objetos o arrays, usa `JSON.stringify()` al guardar y `JSON.parse()` al leer.
*   Son **síncronos**, lo que significa que operaciones de lectura/escritura pueden (en teoría, raramente en la práctica con datos pequeños) bloquear el hilo principal. Para almacenamiento asíncrono más complejo, existe IndexedDB.
*   **No son seguros** para almacenar información sensible (contraseñas, tokens), ya que son fácilmente accesibles desde el navegador (DevTools, scripts maliciosos si hay XSS).

### 14_2_Temporizadores_setTimeout_setInterval_clearTimeout_clearInterval

Permiten ejecutar código después de un cierto retraso o repetidamente a intervalos regulares. Son asíncronos.

*   **`setTimeout(funcionCallback, retrasoMs, arg1?, arg2?, ...)`:** Ejecuta `funcionCallback` **una sola vez** después de que haya transcurrido el `retrasoMs` (en milisegundos). Devuelve un ID numérico único. Argumentos adicionales se pasan a la `funcionCallback`.
    ```javascript
    console.log("Inicio setTimeout");

    const timeoutId = setTimeout(() => {
      console.log("¡Timeout de 3 segundos ejecutado!");
    }, 3000); // 3000ms = 3 segundos

    console.log("Timeout programado con ID:", timeoutId);
    // Para cancelar el timeout antes de que se ejecute:
    // clearTimeout(timeoutId);
    ```
*   **`setInterval(funcionCallback, intervaloMs, arg1?, arg2?, ...)`:** Ejecuta `funcionCallback` **repetidamente** cada `intervaloMs` milisegundos. Devuelve un ID numérico único.
    ```javascript
    let contadorIntervalo = 0;
    console.log("Inicio setInterval");

    const intervalId = setInterval(() => {
      contadorIntervalo++;
      console.log(`Intervalo ejecutado ${contadorIntervalo} veces`);
      if (contadorIntervalo >= 5) {
        // Detener el intervalo después de 5 ejecuciones
        clearInterval(intervalId);
        console.log("Intervalo detenido.");
      }
    }, 1000); // Ejecutar cada 1000ms = 1 segundo

    console.log("Intervalo programado con ID:", intervalId);
    ```
*   **`clearTimeout(timeoutId)`:** Cancela un `setTimeout` previamente programado usando su ID.
*   **`clearInterval(intervalId)`:** Detiene la ejecución repetida de un `setInterval` usando su ID.

### 14_3_Geolocalizacion_Concepto

La API de Geolocalización permite obtener la ubicación geográfica del dispositivo del usuario (latitud, longitud), **siempre con el permiso explícito del usuario**.

Se accede a través del objeto `navigator.geolocation`.

*   **`navigator.geolocation.getCurrentPosition(successCallback, errorCallback?, options?)`:** Obtiene la ubicación actual una vez.
    *   `successCallback`: Función que se llama con un objeto `Position` (que contiene `coords.latitude`, `coords.longitude`, `coords.accuracy`, etc.) si la obtención es exitosa.
    *   `errorCallback`: Función que se llama con un objeto `GeolocationPositionError` si hay un error (permiso denegado, ubicación no disponible, timeout).
    *   `options`: Objeto para configurar la obtención (ej: `enableHighAccuracy: true`, `timeout: 5000`, `maximumAge: 0`).
*   **`navigator.geolocation.watchPosition(successCallback, errorCallback?, options?)`:** Similar a `getCurrentPosition`, pero llama a `successCallback` **cada vez que la posición del dispositivo cambia**. Devuelve un ID para poder detener la observación con `clearWatch()`.
*   **`navigator.geolocation.clearWatch(watchId)`:** Detiene la observación iniciada con `watchPosition`.

```javascript
if ('geolocation' in navigator) {
  console.log("Geolocalización soportada.");

  const opcionesGeo = {
    enableHighAccuracy: true, // Intenta obtener ubicación más precisa
    timeout: 8000, // Máximo 8 segundos para obtenerla
    maximumAge: 0 // No usar una ubicación cacheada
  };

  function geoExito(posicion) {
    const lat = posicion.coords.latitude;
    const lon = posicion.coords.longitude;
    const precision = posicion.coords.accuracy;
    console.log(`Ubicación: Lat ${lat}, Lon ${lon} (Precisión: ${precision} metros)`);
    // Aquí podrías mostrar la ubicación en un mapa, etc.
  }

  function geoError(error) {
    console.error("Error de Geolocalización:", error.message, `(Código: ${error.code})`);
    // error.code: 1=PERMISSION_DENIED, 2=POSITION_UNAVAILABLE, 3=TIMEOUT
  }

  console.log("Solicitando ubicación...");
  // Solicitar una vez
  navigator.geolocation.getCurrentPosition(geoExito, geoError, opcionesGeo);

  // Observar cambios (ejemplo conceptual, necesitarías un botón para detener)
  // const watchId = navigator.geolocation.watchPosition(geoExito, geoError, opcionesGeo);
  // Para detener: navigator.geolocation.clearWatch(watchId);

} else {
  console.warn("Geolocalización no soportada por este navegador.");
}
```
*La solicitud de permiso al usuario es gestionada por el navegador.*

## 15_Herramientas_y_Buenas_Practicas

Escribir buen código JavaScript va más allá de conocer la sintaxis. Implica usar herramientas adecuadas y seguir prácticas que mejoren la calidad, mantenibilidad y eficiencia.

### 15_1_Herramientas_Desarrollador_Navegador_DevTools

Son **esenciales**. Ya mencionamos la **Consola**, pero hay otras pestañas clave (los nombres pueden variar ligeramente entre navegadores):

*   **Elements (Elementos/Inspector):** Permite inspeccionar y modificar el árbol DOM y los estilos CSS aplicados en tiempo real. Puedes ver el HTML renderizado, los estilos calculados, el box model, los event listeners asociados, etc.
*   **Sources (Fuentes/Depurador):** Permite ver los archivos fuente (JS, CSS, HTML), establecer **breakpoints** (puntos de interrupción) para pausar la ejecución del JS, inspeccionar variables paso a paso, y analizar la pila de llamadas. Fundamental para depuración compleja.
*   **Network (Red):** Muestra todas las peticiones de red realizadas por la página (HTML, CSS, JS, imágenes, fetch/XHR). Permite inspeccionar cabeceras, cuerpos de respuesta, tiempos de carga, códigos de estado HTTP. Crucial para diagnosticar problemas de carga o de API.
*   **Application (Aplicación):** Permite inspeccionar el almacenamiento web (`localStorage`, `sessionStorage`, Cookies, IndexedDB), Service Workers, caché, etc.
*   **Performance (Rendimiento):** Herramientas para analizar el rendimiento de carga y ejecución de la página, identificar cuellos de botella.
*   **Memory (Memoria):** Herramientas para analizar el uso de memoria y detectar fugas (memory leaks).

Dominar las DevTools de tu navegador (Chrome, Firefox, Edge, Safari) es una habilidad clave.

### 15_2_Depuracion_Debugging_console_log_debugger

Depurar es el proceso de encontrar y corregir errores (bugs) en el código.

*   **`console.log()` / `warn()` / `error()`:** La forma más básica. Imprimir valores de variables, mensajes de estado o errores en puntos clave para entender qué está pasando.
*   **Debugger (Pestaña Sources/Depurador):**
    *   **Breakpoints:** Marca líneas de código donde quieres que la ejecución se pause. Puedes hacer clic en el número de línea en la pestaña Sources.
    *   **Step Over:** Ejecuta la línea actual y pasa a la siguiente (sin entrar en funciones llamadas en esa línea).
    *   **Step Into:** Si la línea actual llama a una función, entra *dentro* de esa función para depurarla.
    *   **Step Out:** Si estás dentro de una función, continúa la ejecución hasta salir de ella y volver al punto de llamada.
    *   **Resume Execution:** Continúa la ejecución normal hasta el siguiente breakpoint o hasta que termine.
    *   **Watch Expressions:** Puedes añadir variables o expresiones para observar cómo cambian sus valores mientras depuras paso a paso.
    *   **Call Stack:** Muestra la pila de llamadas de funciones que llevaron al punto actual.
*   **Sentencia `debugger;`:** Puedes escribir esta palabra clave directamente en tu código JavaScript. Cuando el navegador la encuentre *y* las DevTools estén abiertas, pausará la ejecución en ese punto como si hubieras puesto un breakpoint.

Aprender a usar el depurador es mucho más eficiente que llenar el código de `console.log` para problemas complejos.

### 15_3_Linters_ESLint_Prettier_Concepto

Son herramientas que analizan tu código estáticamente (sin ejecutarlo) para encontrar posibles errores, código problemático y asegurar que sigue ciertas **reglas de estilo** y convenciones.

*   **Linters (ej: ESLint):** Se centran en la **calidad del código** y la detección de errores potenciales.
    *   Detectan variables no usadas, posibles errores de scope, uso de `var`, código inalcanzable, etc.
    *   Pueden configurarse con conjuntos de reglas (ej: "eslint:recommended") o reglas personalizadas.
    *   Se integran con editores de código para mostrar advertencias/errores mientras escribes.
*   **Formateadores (ej: Prettier):** Se centran exclusivamente en el **formato y estilo** del código (indentación, espacios, saltos de línea, comillas, etc.).
    *   Toman tu código y lo reformatean automáticamente para que cumpla un estilo consistente.
    *   Se configuran con opciones de estilo.
    *   A menudo se usan *junto* con un linter (ESLint maneja la calidad, Prettier el formato).

Usar linters y formateadores mejora enormemente la consistencia, legibilidad y calidad del código, especialmente en equipos. Ayudan a prevenir errores y reducen las discusiones sobre estilo.

### 15_4_Control_de_Versiones_Git_Concepto

El control de versiones es un sistema que registra los cambios realizados en un archivo o conjunto de archivos a lo largo del tiempo, permitiendo recuperar versiones anteriores. **Git** es el sistema de control de versiones distribuido más popular y estándar en la industria.

**Beneficios:**
*   **Historial de Cambios:** Ver quién cambió qué y cuándo.
*   **Reversión:** Volver fácilmente a estados anteriores si algo sale mal.
*   **Trabajo en Equipo:** Facilita la colaboración de múltiples desarrolladores en el mismo proyecto sin sobrescribir el trabajo de los demás (usando ramas y fusiones/merges).
*   **Experimentación Segura (Ramas/Branches):** Crear ramas separadas para desarrollar nuevas características o corregir errores sin afectar la versión principal estable.
*   **Copias de Seguridad:** Plataformas como GitHub, GitLab o Bitbucket proporcionan repositorios remotos que actúan como copia de seguridad y centro de colaboración.

Aprender los conceptos básicos de Git (`clone`, `add`, `commit`, `push`, `pull`, `branch`, `merge`) es una habilidad **esencial** para cualquier desarrollador.

### 15_5_Escribir_Codigo_Limpio_y_Mantenible

Más allá de las herramientas, adopta buenos hábitos:

*   **Nombres Descriptivos:** Usa nombres claros y significativos para variables, funciones y clases.
*   **Funciones Pequeñas y Enfocadas:** Cada función debe hacer una sola cosa bien. Evita funciones gigantescas.
*   **DRY (Don't Repeat Yourself):** Evita copiar y pegar código. Si haces algo varias veces, crea una función o un componente reutilizable.
*   **Comentarios donde sean Necesarios:** Explica el *por qué* de un código complejo, no el *qué* (si el código es claro).
*   **Consistencia:** Sigue un estilo de código consistente (indentación, espaciado, convenciones de nombres), idealmente reforzado por un linter/formateador.
*   **Modularidad:** Divide tu código en módulos/componentes lógicos y reutilizables (usando módulos ES6 o patrones similares).
*   **Manejo de Errores:** Anticipa y maneja posibles errores de forma robusta (`try...catch`, validaciones).

**Próximos Pasos:**

JavaScript es un ecosistema enorme y en constante crecimiento. Ahora que tienes la base, puedes explorar muchas direcciones:

1.  **Profundizar en JavaScript:** Hay conceptos más avanzados como clausuras (closures), patrones de diseño, programación funcional, optimización de rendimiento, etc.
2.  **Node.js:** Si te interesa el desarrollo backend, aprende Node.js para construir servidores, APIs y herramientas usando JavaScript fuera del navegador.
3.  **Frameworks y Librerías Frontend:** Son herramientas que facilitan enormemente la construcción de interfaces de usuario complejas y aplicaciones de una sola página (SPAs). Los más populares son:
    *   **React:** Librería basada en componentes desarrollada por Facebook. Muy popular y con un gran ecosistema.
    *   **Angular:** Framework completo desarrollado por Google. Usa TypeScript (un superset de JavaScript con tipos estáticos).
    *   **Vue.js:** Framework progresivo conocido por su curva de aprendizaje más suave y su flexibilidad.
    *   **Svelte:** Compilador que escribe código JavaScript eficiente y optimizado, en lugar de depender de un Virtual DOM en el navegador.
    Aprender al menos uno de estos es casi esencial para el desarrollo frontend moderno profesional.
4.  **TypeScript:** Un superset de JavaScript que añade tipado estático opcional. Ayuda a detectar errores en tiempo de desarrollo, mejora la mantenibilidad y la escalabilidad, especialmente en proyectos grandes. Muy demandado en la industria.
5.  **Herramientas de Construcción (Build Tools):** Webpack, Rollup, Parcel, Vite. Herramientas para empaquetar módulos, transpilar código (ej: de TypeScript a JS, de ES6+ a ES5), optimizar assets, etc.
6.  **Pruebas (Testing):** Aprende a escribir pruebas unitarias, de integración y end-to-end para asegurar la calidad y robustez de tu código (ej: usando Jest, Mocha, Cypress).
7.  **Construir Proyectos:** La mejor manera de aprender es haciendo. Crea tus propias aplicaciones web, contribuye a proyectos open-source, practica en plataformas como Codewars o LeetCode.

El viaje del aprendizaje en JavaScript nunca termina, pero ahora tienes el mapa y la brújula para navegarlo. ¡Sigue codificando!
