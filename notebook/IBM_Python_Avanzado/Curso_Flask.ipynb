## 1_Introduccion_a_Flask

### 1_1_Que_es_Flask

Flask es un **microframework** para el desarrollo web escrito en Python. Se le llama "micro" no porque le falten funcionalidades, sino porque su núcleo es simple y extensible. No impone una estructura de proyecto específica ni requiere herramientas particulares, permitiendo al desarrollador elegir las librerías y técnicas que mejor se adapten a sus necesidades.

### 1_2_Por_que_usar_Flask

*   **Simple y Fácil de Aprender:** Su API es intuitiva y bien documentada.
*   **Flexible:** No te obliga a usar componentes específicos (como ORMs, motores de plantillas, etc.). Puedes elegir los que prefieras.
*   **Extensible:** Cuenta con un vasto ecosistema de extensiones para añadir funcionalidades (manejo de formularios, bases de datos, autenticación, etc.).
*   **Ideal para Proyectos Pequeños/Medianos y APIs:** Su ligereza lo hace perfecto para microservicios, APIs RESTful y aplicaciones web donde no necesitas la complejidad de un framework full-stack.
*   **Comunidad Activa:** Gran cantidad de recursos, tutoriales y soporte comunitario.

### 1_3_Microframework_vs_Full_Stack

*   **Microframework (Flask):** Provee lo esencial (enrutamiento, manejo de peticiones/respuestas, plantillas). El desarrollador añade componentes (extensiones) según necesite. Más control y flexibilidad.
*   **Full-Stack Framework (Django):** Incluye "baterías incluidas" (ORM, panel de administración, sistema de autenticación, etc.). Más rápido para empezar con funcionalidades comunes, pero puede ser más rígido.

### 1_4_WSGI_y_Werkzeug

*   **WSGI (Web Server Gateway Interface):** Es una especificación estándar para la comunicación entre servidores web y aplicaciones Python. Flask cumple con WSGI.
*   **Werkzeug:** Es una librería WSGI fundamental sobre la que se construye Flask. Proporciona utilidades para el enrutamiento, objetos de petición/respuesta, manejo de sesiones, etc. Flask utiliza Werkzeug internamente.
*   **Jinja2:** Es el motor de plantillas por defecto que utiliza Flask (aunque puedes usar otros). Permite generar HTML dinámicamente.

## 2_Instalacion_y_Entorno

### 2_1_Entornos_Virtuales_venv

**¡Fundamental!** Siempre trabaja dentro de un entorno virtual para aislar las dependencias de tu proyecto de las globales del sistema.

1.  **Crea un directorio para tu proyecto:**
    ```bash
    mkdir mi_proyecto_flask
    cd mi_proyecto_flask
    ```
2.  **Crea el entorno virtual:** (Usa `python3` si `python` apunta a Python 2)
    ```bash
    python -m venv venv
    # o
    python3 -m venv venv
    ```
    Esto crea una carpeta `venv` (puedes llamarla como quieras) con una copia de Python y pip.
3.  **Activa el entorno virtual:**
    *   **Windows (cmd.exe):**
        ```bash
        .\venv\Scripts\activate
        ```
    *   **Windows (PowerShell):** (Puede requerir cambiar la política de ejecución: `Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser`)
        ```powershell
        .\venv\Scripts\Activate.ps1
        ```
    *   **Linux / macOS / Termux (bash/zsh):**
        ```bash
        source venv/bin/activate
        ```
    Verás `(venv)` al principio de tu prompt, indicando que el entorno está activo.
4.  **Desactivar el entorno:** Simplemente ejecuta `deactivate` en la terminal.

### 2_2_Instalacion_en_Windows

1.  Asegúrate de tener Python instalado (descárgalo desde [python.org](https://www.python.org/) y asegúrate de marcar "Add Python to PATH" durante la instalación).
2.  Abre `cmd` o `PowerShell`.
3.  Crea y activa un entorno virtual (ver [2.1](#2_1_Entornos_Virtuales_venv)).
4.  Instala Flask usando pip:
    ```bash
    pip install Flask
    ```

### 2_3_Instalacion_en_Linux_Debian_Ubuntu

1.  Abre una terminal.
2.  Asegúrate de tener Python 3 y pip instalados:
    ```bash
    sudo apt update
    sudo apt install python3 python3-pip python3-venv -y
    ```
3.  Crea y activa un entorno virtual (ver [2.1](#2_1_Entornos_Virtuales_venv)).
4.  Instala Flask usando pip:
    ```bash
    pip install Flask
    ```

### 2_4_Instalacion_en_Linux_Fedora_CentOS

1.  Abre una terminal.
2.  Asegúrate de tener Python 3 y pip instalados:
    *   **Fedora:**
        ```bash
        sudo dnf update
        sudo dnf install python3 python3-pip python3-virtualenv -y
        ```
    *   **CentOS/RHEL (puede variar según la versión):**
        ```bash
        sudo yum update
        sudo yum install python3 python3-pip python3-virtualenv -y # Puede requerir habilitar repositorios EPEL
        ```
3.  Crea y activa un entorno virtual (ver [2.1](#2_1_Entornos_Virtuales_venv)).
4.  Instala Flask usando pip:
    ```bash
    pip install Flask
    ```

### 2_5_Instalacion_en_macOS

1.  macOS suele venir con Python 2. Se recomienda instalar Python 3 con [Homebrew](https://brew.sh/):
    ```bash
    /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
    brew update
    brew install python
    ```
2.  Abre la Terminal.
3.  Crea y activa un entorno virtual (ver [2.1](#2_1_Entornos_Virtuales_venv), usando `python3`).
4.  Instala Flask usando pip:
    ```bash
    pip3 install Flask
    # o si estás dentro del venv activado:
    pip install Flask
    ```
    *Nota sobre iOS:* No puedes *ejecutar* directamente un servidor Flask en iOS de forma nativa como lo harías en un PC o Android (Termux). El desarrollo se hace en macOS y se puede acceder a la app desde un dispositivo iOS en la misma red local para pruebas. Existen herramientas como Pythonista o a-Shell en iOS, pero tienen limitaciones y no son el entorno estándar para Flask.

### 2_6_Instalacion_en_Android_Termux

[Termux](https://termux.com/) es un emulador de terminal potente para Android.

1.  Instala Termux desde F-Droid (recomendado) o Google Play (puede estar desactualizado).
2.  Abre Termux y actualiza los paquetes:
    ```bash
    pkg update && pkg upgrade -y
    ```

## 4_Rutas_y_Vistas

Las rutas definen las URLs de tu aplicación. El decorador `@app.route()` asocia una URL con una función Python (llamada *vista* o *view function*). La función vista procesa la petición y devuelve una respuesta.

### 4_1_Rutas_Basicas

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
  return 'Pagina de Inicio'

@app.route('/acerca')
def acerca():
  return 'Acerca de Nosotros'

@app.route('/contacto')
def contacto():
  return 'Pagina de Contacto'

if __name__ == '__main__':
  app.run(debug=True)
```

### 4_2_Reglas_Variables

Puedes capturar partes de la URL y pasarlas como argumentos a tu función vista.

```python
@app.route('/usuario/<username>') # <nombre_variable>
def mostrar_perfil_usuario(username):
  # La variable 'username' contiene el valor de la URL
  return f'<h1>Perfil de: {username}</h1>'

@app.route('/post/<int:post_id>') # Captura un entero
def mostrar_post(post_id):
  # post_id será un entero, no una cadena
  return f'<h1>Mostrando Post ID: {post_id}</h1>'
```

### 4_3_Conversores_de_Tipo

Flask soporta varios conversores por defecto:

*   `string`: (Por defecto) Acepta cualquier texto sin barra `/`.
*   `int`: Acepta enteros positivos.
*   `float`: Acepta números de punto flotante positivos.
*   `path`: Similar a `string`, pero también acepta barras `/`.
*   `uuid`: Acepta cadenas UUID.

```python
@app.route('/articulo/<path:subpath>')
def mostrar_subpath(subpath):
    # subpath puede contener barras, ej: /articulo/python/flask
    return f'Subpath: {subpath}'

@app.route('/proyecto/<uuid:project_uuid>')
def mostrar_proyecto(project_uuid):
    # project_uuid debe ser un UUID válido
    return f'UUID del Proyecto: {project_uuid}'
```

### 4_4_Metodos_HTTP

Por defecto, las rutas solo responden a peticiones `GET`. Puedes especificar otros métodos (POST, PUT, DELETE, etc.) usando el argumento `methods`.

```python
from flask import request # Importar request

@app.route('/login', methods=['GET', 'POST'])
def login():
  if request.method == 'POST':
    # Procesar el formulario de login (lo veremos más adelante)
    username = request.form.get('username') # Ejemplo
    return f'Procesando login para {username}'
  else:
    # Mostrar el formulario de login
    return '''
      <form method="post">
        Usuario: <input type="text" name="username"><br>
        Password: <input type="password" name="password"><br>
        <input type="submit" value="Login">
      </form>
    '''
```

### 4_5_Generacion_de_URLs_url_for

En lugar de escribir las URLs directamente en tu código o plantillas (lo cual es frágil si cambias una ruta), usa la función `url_for()`. Le pasas el *nombre de la función vista* y cualquier argumento variable.

```python
from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')
def index():
    # Genera la URL para la función 'mostrar_perfil_usuario' con username='juan'
    url_perfil_juan = url_for('mostrar_perfil_usuario', username='juan')
    url_acerca = url_for('acerca') # Sin argumentos

    return f'''
      <p><a href="{url_perfil_juan}">Perfil de Juan</a></p>
      <p><a href="{url_acerca}">Acerca de</a></p>
      <p><a href="{url_for('static', filename='css/style.css')}">Enlace a CSS (ver seccion 6)</a></p>
    '''

@app.route('/usuario/<username>')
def mostrar_perfil_usuario(username):
    return f'Perfil de: {username}'

@app.route('/acerca')
def acerca():
    return 'Acerca de Nosotros'

# Ejemplo de redirección usando url_for
@app.route('/ir_a_juan')
def ir_a_juan():
    return redirect(url_for('mostrar_perfil_usuario', username='juan'))

if __name__ == '__main__':
    app.run(debug=True)
```

`url_for()` es crucial para mantener tu aplicación robusta frente a cambios en las rutas.

## 5_Plantillas_Jinja2

Flask usa el motor de plantillas Jinja2 para generar HTML dinámico. Las plantillas permiten separar la lógica de la presentación.

### 5_1_Renderizando_Plantillas

1.  Crea una carpeta llamada `templates` en la raíz de tu proyecto Flask. Flask buscará las plantillas aquí por defecto.
2.  Crea un archivo HTML dentro de `templates`, por ejemplo `index.html`:

    ```html
    <!-- templates/index.html -->
    <!DOCTYPE html>
    <html>
    <head>
        <title>{{ titulo }}</title> <!-- Variable pasada desde Flask -->
    </head>
    <body>
        <h1>{{ encabezado }}</h1>
        <p>Bienvenido a nuestra aplicacion Flask.</p>
    </body>
    </html>
    ```

3.  En tu archivo `app.py`, usa `render_template`:

    ```python
    # app.py
    from flask import Flask, render_template

    app = Flask(__name__)

    @app.route('/')
    def index():
        # Pasa variables a la plantilla como argumentos clave-valor
        titulo_pagina = "Pagina de Inicio"
        mensaje_bienvenida = "Hola desde Flask y Jinja2!"
        return render_template('index.html',
                               titulo=titulo_pagina,
                               encabezado=mensaje_bienvenida)

    if __name__ == '__main__':
        app.run(debug=True)
    ```

### 5_2_Variables_y_Expresiones

Usa doble llave `{{ ... }}` para imprimir variables o el resultado de expresiones Python simples.

```html
<p>Usuario: {{ usuario.nombre }}</p> <!-- Acceder a atributos -->
<p>Items: {{ mi_lista[0] }}</p> <!-- Acceder a elementos de lista -->
<p>Suma: {{ 5 + 3 }}</p> <!-- Expresiones simples -->
<p>Mensaje: {{ "Hola " + nombre | capitalize }}</p> <!-- Usar filtros (ver 5.5) -->
```

### 5_3_Estructuras_de_Control_if_for

Usa `{% ... %}` para las estructuras de control.

*   **Condicional `if`:**

    ```html
    {% if usuario %}
      <p>Bienvenido, {{ usuario.nombre }}!</p>
    {% else %}
      <p>Por favor, <a href="{{ url_for('login') }}">inicia sesion</a>.</p>
    {% endif %}
    ```

*   **Bucle `for`:**

    ```html
    <ul>
    {% for item in lista_items %}
      <li>{{ loop.index }}. {{ item }}</li> <!-- loop.index es el índice del bucle (empieza en 1) -->
    {% else %}
      <li>No hay items para mostrar.</li> <!-- Se muestra si la lista está vacía -->
    {% endfor %}
    </ul>
    ```

### 5_4_Herencia_de_Plantillas

Permite crear una plantilla base (`layout.html`) con la estructura común y que otras plantillas la hereden, rellenando bloques específicos.

1.  Crea `templates/base.html`:

    ```html

    <html lang="es">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>{% block titulo %}Mi Aplicacion Flask{% endblock %}</title>
        <!-- Enlace a CSS (ver seccion 6) -->
        <link rel="stylesheet" href="{{ url_for('static', filename='css/style.css') }}">
    </head>
    <body>
        <header>
            <h1>Mi Aplicacion</h1>
            <nav>
                <a href="{{ url_for('index') }}">Inicio</a> |
                <a href="{{ url_for('acerca') }}">Acerca</a>
            </nav>
        </header>

        <main>
            {% block contenido %}
            <!-- El contenido específico de cada página irá aquí -->
            {% endblock %}
        </main>

        <footer>
            <p>&copy; 2023 Mi Aplicacion</p>
        </footer>
        <!-- Scripts JS -->
        {% block scripts %}{% endblock %}
    </body>
    </html>
    ```

2.  Crea una plantilla que herede de `base.html`, por ejemplo `templates/acerca.html`:

    ```html
    <!-- templates/acerca.html -->
    {% extends "base.html" %}

    {% block titulo %}Acerca De Nosotros{% endblock %}

    {% block contenido %}
      <h2>Sobre Nuestra Aplicacion</h2>
      <p>Esta es una aplicacion de ejemplo creada con Flask.</p>
      <p>Aqui puedes poner informacion sobre el proposito de la aplicacion.</p>
    {% endblock %}
    ```

3.  En tu vista `app.py`:

    ```python
    # app.py
    # ... (importaciones y app = Flask(__name__))

    def index():
        # Esta vista ahora usará la estructura de base.html
        # si creamos un templates/index.html que extienda de base.html
        # Por ahora, solo renderizamos la base para el ejemplo
         return render_template('base.html') # O renderizar una plantilla hija como 'index.html'

    @app.route('/acerca')
    def acerca():
        return render_template('acerca.html')

    # ... (if __name__ ...)
    ```

### 5_5_Filtros_y_Macros

*   **Filtros:** Modifican variables antes de mostrarlas. Se usan con `|`. Ejemplos: `{{ nombre | capitalize }}`, `{{ texto | length }}`, `{{ precio | round(2) }}`, `{{ fecha | dateformat('%Y-%m-%d') }}` (requiere extensión). Jinja2 tiene muchos filtros incorporados.
*   **Macros:** Similares a funciones en Python, permiten reutilizar fragmentos de HTML.

    ```html
    <!-- templates/_macros.html (el guion bajo es convención) -->
    {% macro render_input(name, type='text', placeholder='') %}
      <p>
        <label for="{{ name }}">{{ name | capitalize }}:</label><br>
        <input type="{{ type }}" id="{{ name }}" name="{{ name }}" placeholder="{{ placeholder }}">
      </p>
    {% endmacro %}

    <!-- En otra plantilla -->
    {% import '_macros.html' as forms %}

    <form method="post">
      {{ forms.render_input('username', placeholder='Tu nombre de usuario') }}
      {{ forms.render_input('password', type='password', placeholder='Tu contraseña') }}
      <button type="submit">Enviar</button>
    </form>
    ```

### 5_6_Contexto_de_Plantilla

Variables y funciones que están disponibles automáticamente en todas las plantillas sin tener que pasarlas explícitamente en cada `render_template`. Se usa el decorador `@app.context_processor`.

```python
import datetime

@app.context_processor
def inject_now():
    # Esta variable 'now' estará disponible en todas las plantillas
    return {'now': datetime.datetime.utcnow()}

# En cualquier plantilla:
# <p>Fecha actual (UTC): {{ now.strftime('%Y-%m-%d %H:%M:%S') }}</p>
```

## 7_El_Objeto_Request

El objeto `request` (importado de `flask`) contiene toda la información sobre la petición HTTP entrante. Solo está disponible *durante* el procesamiento de una petición dentro de una función vista.

```python
from flask import Flask, request, render_template

app = Flask(__name__)

@app.route('/procesar', methods=['GET', 'POST'])
def procesar_datos():
    if request.method == 'POST':
        # Acceder a datos de formulario
        nombre = request.form.get('nombre', 'Anonimo') # .get evita errores si no existe
        email = request.form.get('email')

        # Acceder a parámetros de consulta (query string) ?llave=valor
        fuente = request.args.get('fuente', 'Desconocida')

        # Acceder a cabeceras
        user_agent = request.headers.get('User-Agent')

        # Si se envió JSON
        data_json = request.get_json() # Devuelve None si no es JSON o hay error
        if data_json:
             mensaje_json = data_json.get('mensaje')

        # Si se subió un archivo (veremos más detalle con Flask-WTF)
        archivo = request.files.get('mi_archivo')
        if archivo:
             filename = archivo.filename
             # archivo.save(...) para guardarlo

        return f"""
            <h2>Datos Recibidos (POST)</h2>
            <p>Nombre: {nombre}</p>
            <p>Email: {email}</p>
            <p>Fuente (Query Param): {fuente}</p>
            <p>User Agent: {user_agent}</p>
            <p>Mensaje JSON: {mensaje_json if data_json else 'No JSON recibido'}</p>
            <p>Archivo subido: {filename if archivo else 'No'}</p>
            """
    else: # Método GET
        fuente_get = request.args.get('fuente', 'No especificada')
        return f"""
            <h2>Formulario de Ejemplo (GET)</h2>
            <p>Fuente (Query Param): {fuente_get}</p>
            <form method="post" action="/procesar?fuente=formulario" enctype="multipart/form-data">
                Nombre: <input type="text" name="nombre"><br>
                Email: <input type="email" name="email"><br>
                Archivo: <input type="file" name="mi_archivo"><br>
                <input type="submit" value="Enviar">
            </form>
            """

if __name__ == '__main__':
    app.run(debug=True)
```

### 7_1_Accediendo_a_Datos_de_Formulario_request_form

`request.form` es un diccionario especial (MultiDict) que contiene los datos enviados en el cuerpo de una petición POST (normalmente desde un formulario HTML con `method="post"`). Usa `request.form.get('nombre_campo')` para obtener valores de forma segura.

### 7_2_Parametros_de_Consulta_request_args

`request.args` es un MultiDict que contiene los parámetros de la URL (query string), la parte que viene después del `?` (ej: `/buscar?q=flask&page=2`). Usa `request.args.get('nombre_parametro')`.

### 7_3_Archivos_Subidos_request_files

`request.files` es un MultiDict que contiene los archivos subidos a través de un formulario (`<input type="file">`) con `enctype="multipart/form-data"`. Cada valor es un objeto `FileStorage` con atributos como `filename` y métodos como `save()`.

### 7_4_Cookies_request_cookies

`request.cookies` es un diccionario que contiene las cookies enviadas por el navegador. Usa `request.cookies.get('nombre_cookie')`.

### 7_5_Cabeceras_request_headers

`request.headers` es un diccionario especial (tipo `EnvironHeaders`) que contiene las cabeceras HTTP de la petición (ej: `User-Agent`, `Accept`, `Authorization`). Es insensible a mayúsculas/minúsculas. Usa `request.headers.get('User-Agent')`.

### 7_6_Datos_JSON_request_get_json

Si el cliente envía datos en formato JSON con la cabecera `Content-Type: application/json`, puedes usar `request.get_json()` para parsearlos en un diccionario o lista Python. Devuelve `None` si el cuerpo no es JSON válido o si el `Content-Type` no es correcto. `request.json` es un atajo que lanza un error si falla.

## 8_Formularios_con_Flask_WTF

Manejar formularios HTML manualmente (validación, regeneración, seguridad CSRF) es tedioso. La extensión **Flask-WTF** (basada en WTForms) simplifica enormemente este proceso.

### 8_1_Instalacion_y_Configuracion

1.  Instala la extensión:

2.  Configura una `SECRET_KEY` en tu aplicación Flask. Es **esencial** para la protección CSRF.

    ```python
    # Es mejor cargarla desde una variable de entorno en producción.

    ```

### 8_2_Definiendo_Formularios

Crea clases que heredan de `FlaskForm` y define los campos del formulario como atributos de clase, usando los tipos de campo de WTForms (`StringField`, `PasswordField`, `SubmitField`, etc.) y validadores.

Crea un archivo `forms.py`:

```python
# forms.py
from flask_wtf import FlaskForm
from wtforms import StringField, PasswordField, SubmitField, BooleanField, TextAreaField
from wtforms.validators import DataRequired, Length, Email, EqualTo, ValidationError

# Ejemplo de validador personalizado (opcional)
def validate_username_custom(form, field):
    if 'admin' in field.data.lower():
        raise ValidationError('El nombre de usuario no puede contener "admin".')

class RegistrationForm(FlaskForm):
    username = StringField('Nombre de Usuario',
                           validators=[DataRequired(), Length(min=4, max=20), validate_username_custom])
    email = StringField('Correo Electronico',
                        validators=[DataRequired(), Email()])
    password = PasswordField('Contraseña', validators=[DataRequired(), Length(min=6)])
    confirm_password = PasswordField('Confirmar Contraseña',
                                     validators=[DataRequired(), EqualTo('password', message='Las contraseñas deben coincidir.')])
    accept_tos = BooleanField('Acepto los Terminos de Servicio', validators=[DataRequired()])
    submit = SubmitField('Registrarse')

class LoginForm(FlaskForm):
    email = StringField('Correo Electronico', validators=[DataRequired(), Email()])
    password = PasswordField('Contraseña', validators=[DataRequired()])
    remember = BooleanField('Recordarme')
    submit = SubmitField('Iniciar Sesion')

class PostForm(FlaskForm):
    title = StringField('Titulo', validators=[DataRequired(), Length(max=100)])
    content = TextAreaField('Contenido', validators=[DataRequired()])
    submit = SubmitField('Publicar')
```

### 8_3_Proteccion_CSRF

Flask-WTF integra protección contra Cross-Site Request Forgery (CSRF) automáticamente. Solo necesitas:

1.  Haber configurado `SECRET_KEY`.
2.  Añadir `{{ form.hidden_tag() }}` dentro de la etiqueta `<form>` en tu plantilla HTML. Esto renderiza un campo oculto con el token CSRF. Flask-WTF lo validará automáticamente al recibir el formulario.

## 9_Bases_de_Datos_con_Flask_SQLAlchemy

Interactuar con bases de datos es esencial. **Flask-SQLAlchemy** es una extensión que integra la potente librería **SQLAlchemy** (un ORM - Object-Relational Mapper) con Flask.

### 9_1_Introduccion_a_ORMs_y_SQLAlchemy

*   **ORM:** Permite interactuar con bases de datos relacionales (PostgreSQL, MySQL, SQLite, etc.) usando objetos Python en lugar de escribir SQL directamente. Mapea tablas a clases y filas a instancias de esas clases.
*   **SQLAlchemy:** Es el ORM más popular y potente para Python. Proporciona tanto un ORM de alto nivel como un lenguaje de expresión SQL más cercano al SQL puro.

### 9_2_Instalacion_y_Configuracion_Flask_SQLAlchemy

1.  Instala la extensión y un *driver* para tu base de datos (ej: `psycopg2-binary` para PostgreSQL, `mysqlclient` para MySQL, SQLite viene incluido con Python).
    ```bash
    pip install Flask-SQLAlchemy psycopg2-binary # Ejemplo para PostgreSQL
    # pip install Flask-SQLAlchemy mysqlclient    # Ejemplo para MySQL
    # pip install Flask-SQLAlchemy                # Para SQLite (no necesita driver extra)
    ```
2.  Configura la URI de la base de datos en tu aplicación Flask:
    ```python
    # app.py
    import os
    from flask import Flask
    from flask_sqlalchemy import SQLAlchemy

    # Determina el directorio base de la aplicación
    basedir = os.path.abspath(os.path.dirname(__file__))

    app = Flask(__name__)
    app.config['SECRET_KEY'] = 'tu-clave-secreta' # Necesaria para sesiones, formularios, etc.

    # --- Configuracion SQLAlchemy ---
    # Elige una de estas opciones:

    # SQLite (ideal para desarrollo y pruebas simples)
    # app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///' + os.path.join(basedir, 'app.db')

    # PostgreSQL (ejemplo)
    # postgresql://usuario:contraseña@host:puerto/nombre_basedatos
    # app.config['SQLALCHEMY_DATABASE_URI'] = os.environ.get('DATABASE_URL') or \
    #    'postgresql://user:password@localhost:5432/mydatabase'

    # MySQL (ejemplo)
    # mysql+mysqlclient://usuario:contraseña@host:puerto/nombre_basedatos
    app.config['SQLALCHEMY_DATABASE_URI'] = os.environ.get('DATABASE_URL') or \
        'mysql+mysqlclient://user:password@localhost:3306/mydatabase'


    # Opcional: Desactiva el seguimiento de modificaciones si no lo necesitas (mejora rendimiento)
    app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False

    # Crea la instancia de SQLAlchemy, pasándole la app Flask
    db = SQLAlchemy(app)

    # --- Definicion de Modelos (después de inicializar db) ---
    # (Ver sección 9.3)

    # --- Rutas ---
    # (Ver secciones anteriores y posteriores)

    if __name__ == '__main__':
        app.run(debug=True)
    ```
    **Importante:** En producción, NUNCA guardes credenciales directamente en el código. Usa variables de entorno (como `os.environ.get('DATABASE_URL')`).

### 9_3_Definiendo_Modelos

Los modelos son clases Python que heredan de `db.Model`. Cada clase representa una tabla en la base de datos, y los atributos de la clase (definidos con `db.Column`) representan las columnas.

```python
# app.py (o un archivo models.py importado)

# ... (después de db = SQLAlchemy(app))

class User(db.Model):
    id = db.Column(db.Integer, primary_key=True) # Clave primaria autoincremental por defecto
    username = db.Column(db.String(80), unique=True, nullable=False)
    email = db.Column(db.String(120), unique=True, nullable=False)
    password_hash = db.Column(db.String(128)) # Guardaremos el hash, no la contraseña
    posts = db.relationship('Post', backref='author', lazy=True) # Relación (ver 9.6)

    def __repr__(self):
        # Representación útil para debugging
        return f"<User {self.username}>"

class Post(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    title = db.Column(db.String(100), nullable=False)
    date_posted = db.Column(db.DateTime, nullable=False, default=datetime.utcnow)
    content = db.Column(db.Text, nullable=False)
    user_id = db.Column(db.Integer, db.ForeignKey('user.id'), nullable=False) # Clave foránea

    def __repr__(self):
        return f"<Post {self.title}>"

# Importante importar datetime si lo usas
from datetime import datetime
```

*   `db.Column`: Define una columna. Los argumentos son el tipo (`db.Integer`, `db.String`, `db.Text`, `db.DateTime`, `db.Boolean`, etc.) y opciones (`primary_key`, `unique`, `nullable`, `default`, `index`).
*   `db.relationship`: Define una relación entre modelos (ver [9.6](#9_6_Relaciones_Entre_Modelos)).
*   `db.ForeignKey`: Define una clave foránea, enlazando con otra tabla (`'nombretabla.nombrecolumna'`).

### 9_5_Operaciones_CRUD

CRUD = Create, Read, Update, Delete.

Necesitarás estar dentro de un contexto de aplicación Flask para interactuar con la base de datos (las vistas lo hacen automáticamente, pero si trabajas desde un shell, usa `with app.app_context():`).

#### 9_5_1_Crear_Registros

```python
from app import app, db, User, Post # Asumiendo que están definidos en app.py

with app.app_context():
    # Crear un nuevo usuario
    user1 = User(username='juanperez', email='juan@example.com', password_hash='hash_seguro_aqui')
    db.session.add(user1) # Añadir a la sesión

    # Crear otro usuario
    user2 = User(username='mariagomez', email='maria@example.com', password_hash='otro_hash')
    db.session.add(user2)

    # Crear un post asociado a user1
    post1 = Post(title='Mi primer post', content='Contenido del primer post.', author=user1) # O user_id=user1.id
    db.session.add(post1)

    # Guardar los cambios en la base de datos
    try:
        db.session.commit()
        print("Usuarios y post creados!")
    except Exception as e:
        db.session.rollback() # Deshacer cambios si hay error
        print(f"Error al guardar: {e}")

```

*   `db.session.add(objeto)`: Añade una instancia de modelo a la sesión de SQLAlchemy (prepara para guardar).
*   `db.session.add_all([obj1, obj2])`: Añade múltiples objetos.
*   `db.session.commit()`: Guarda todos los cambios pendientes en la sesión a la base de datos.
*   `db.session.rollback()`: Descarta los cambios en la sesión (útil en caso de error).

#### 9_5_2_Leer_Consultar_Registros

SQLAlchemy ofrece una API de consulta fluida.

```python
with app.app_context():
    # Obtener todos los usuarios
    todos_los_usuarios = User.query.all()
    print(f"Todos los usuarios: {todos_los_usuarios}")

    # Obtener el primer usuario encontrado
    primer_usuario = User.query.first()
    print(f"Primer usuario: {primer_usuario}")

    # Obtener un usuario por su clave primaria (id)
    usuario_id_1 = User.query.get(1) # Devuelve el usuario con id=1 o None
    # Alternativa más nueva y recomendada:
    # usuario_id_1 = db.session.get(User, 1)
    print(f"Usuario con ID 1: {usuario_id_1}")

    # Filtrar usuarios por username
    juan = User.query.filter_by(username='juanperez').first() # first() para obtener solo uno
    print(f"Usuario 'juanperez': {juan}")

    # Filtrar por email (usando filter)
    maria = User.query.filter(User.email == 'maria@example.com').first()
    print(f"Usuario con email 'maria@example.com': {maria}")

    # Filtrar posts de un usuario específico (usando la relación)
    if juan:
        posts_de_juan = juan.posts # Acceso directo gracias a db.relationship
        print(f"Posts de Juan: {posts_de_juan}")
        # Alternativa:
        # posts_de_juan_alt = Post.query.filter_by(author=juan).all()
        # print(f"Posts de Juan (alt): {posts_de_juan_alt}")

    # Contar usuarios
    numero_usuarios = User.query.count()
    print(f"Número total de usuarios: {numero_usuarios}")

    # Ordenar resultados
    usuarios_ordenados = User.query.order_by(User.username).all()
    print(f"Usuarios ordenados por nombre: {usuarios_ordenados}")
    posts_recientes = Post.query.order_by(Post.date_posted.desc()).limit(5).all() # .desc() para descendente, .limit() para limitar
    print(f"Últimos 5 posts: {posts_recientes}")
```

*   `Modelo.query`: Objeto base para construir consultas.
*   `all()`: Ejecuta la consulta y devuelve una lista de todos los resultados.
*   `first()`: Ejecuta y devuelve el primer resultado o `None`.
*   `get(pk)` / `db.session.get(Modelo, pk)`: Obtiene un objeto por su clave primaria. Eficiente.
*   `filter_by(**kwargs)`: Filtra por igualdad usando argumentos clave-valor.
*   `filter(<condicion>)`: Filtra usando expresiones más complejas (ej: `User.username != 'admin'`, `Post.title.like('%Flask%')`).
*   `count()`: Devuelve el número de resultados.
*   `order_by()`: Ordena los resultados.
*   `limit(n)`: Limita el número de resultados.
*   `offset(n)`: Salta los primeros `n` resultados (para paginación).

#### 9_5_3_Actualizar_Registros

1.  Obtén el objeto que quieres actualizar.
2.  Modifica sus atributos.
3.  Añade el objeto a la sesión (`db.session.add(objeto)`) - SQLAlchemy es inteligente y detectará que es una actualización.
4.  Haz commit (`db.session.commit()`).

```python
with app.app_context():
    usuario_a_actualizar = User.query.filter_by(username='juanperez').first()
    if usuario_a_actualizar:
        print(f"Email antiguo de Juan: {usuario_a_actualizar.email}")
        usuario_a_actualizar.email = 'juan.perez.nuevo@example.com'
        # No es estrictamente necesario db.session.add() para objetos ya existentes
        # pero no hace daño incluirlo.
        # db.session.add(usuario_a_actualizar)
        try:
            db.session.commit()
            print(f"Email actualizado de Juan: {usuario_a_actualizar.email}")
        except Exception as e:
            db.session.rollback()
            print(f"Error al actualizar: {e}")
    else:
        print("Usuario 'juanperez' no encontrado para actualizar.")
```

#### 9_5_4_Eliminar_Registros

1.  Obtén el objeto que quieres eliminar.
2.  Usa `db.session.delete(objeto)`.
3.  Haz commit (`db.session.commit()`).

```python
with app.app_context():
    post_a_eliminar = Post.query.filter_by(title='Mi primer post').first()
    if post_a_eliminar:
        print(f"Eliminando post: {post_a_eliminar.title}")
        db.session.delete(post_a_eliminar)
        try:
            db.session.commit()
            print("Post eliminado.")
        except Exception as e:
            db.session.rollback()
            print(f"Error al eliminar: {e}")

        # Verificar que ya no existe
        post_verificar = Post.query.filter_by(title='Mi primer post').first()
        if not post_verificar:
            print("Verificación: El post ya no existe.")
    else:
        print("Post 'Mi primer post' no encontrado para eliminar.")
```

### 9_6_Relaciones_Entre_Modelos

`db.relationship()` define cómo se relacionan los modelos.

#### 9_6_1_Uno_a_Muchos

Un usuario puede tener muchos posts, pero cada post pertenece a un solo usuario.

```python
class User(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    username = db.Column(db.String(80), unique=True, nullable=False)
    # ... otros campos ...
    # 'Post': El modelo del otro lado de la relación.
    # 'backref': Crea un atributo virtual 'author' en el modelo Post
    #            para acceder fácilmente al User desde un Post.
    # 'lazy': Controla cómo se cargan los posts relacionados.
    #         True (o 'select', por defecto): Se cargan cuando se accede a user.posts.
    #         'joined': Carga los posts en la misma consulta que el usuario (usando JOIN).
    #         'subquery': Carga los posts en una segunda consulta.
    #         'dynamic': Devuelve un objeto query en lugar de una lista, útil si hay muchos posts.
    posts = db.relationship('Post', backref='author', lazy=True)

class Post(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    title = db.Column(db.String(100), nullable=False)
    # ... otros campos ...
    # Define la columna de clave foránea que enlaza con la tabla 'user'
    user_id = db.Column(db.Integer, db.ForeignKey('user.id'), nullable=False)

# Uso:
# juan = User.query.get(1)
# posts_de_juan = juan.posts  # Devuelve una lista de objetos Post
#
# mi_post = Post.query.get(1)
# autor_del_post = mi_post.author # Devuelve el objeto User gracias a backref
```

#### 9_6_2_Muchos_a_Muchos

Un post puede tener muchas etiquetas (tags), y una etiqueta puede estar en muchos posts. Se necesita una *tabla de asociación* intermedia.

```python
# Tabla de asociación (no es un modelo de SQLAlchemy directamente, sino una tabla)
tags = db.Table('tags',
    db.Column('tag_id', db.Integer, db.ForeignKey('tag.id'), primary_key=True),
    db.Column('post_id', db.Integer, db.ForeignKey('post.id'), primary_key=True)
)

class Post(db.Model):
    # ... (campos existentes) ...
    # 'Tag': El modelo del otro lado.
    # secondary=tags: Especifica la tabla de asociación.
    # backref: Crea 'posts' en el modelo Tag.
    #          db.backref('posts', lazy='dynamic') hace que tag.posts sea un query object.
    tags = db.relationship('Tag', secondary=tags, lazy='subquery',
                           backref=db.backref('posts', lazy=True)) # O lazy='dynamic'

class Tag(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(50), unique=True, nullable=False)

    def __repr__(self):
        return f"<Tag {self.name}>"

# Uso:
# post1 = Post.query.get(1)
# tag_python = Tag.query.filter_by(name='python').first()
# tag_flask = Tag.query.filter_by(name='flask').first()
#
# if not tag_python:
#     tag_python = Tag(name='python')
#     db.session.add(tag_python)
# if not tag_flask:
#     tag_flask = Tag(name='flask')
#     db.session.add(tag_flask)
#
# post1.tags.append(tag_python) # Añadir relación
# post1.tags.append(tag_flask)
# db.session.commit()
#
# print(f"Tags del post 1: {post1.tags}") # Muestra [<Tag python>, <Tag flask>]
#
# print(f"Posts con la etiqueta 'python': {tag_python.posts}") # Muestra [<Post 'Mi primer post'>, ...]
```

## 10_Migraciones_de_Bases_de_Datos_con_Flask_Migrate

Cuando modificas tus modelos (añades una columna, cambias un tipo, eliminas una tabla), necesitas actualizar el esquema de tu base de datos *sin perder los datos existentes*. **Flask-Migrate**, que usa **Alembic** por debajo, maneja esto.

### 10_1_Por_que_Migraciones

*   Permiten versionar los cambios en el esquema de tu base de datos, igual que versionas tu código con Git.
*   Facilitan la aplicación de cambios de esquema de forma incremental y reversible.
*   Esenciales para trabajar en equipo y para desplegar actualizaciones en producción.

### 10_2_Instalacion_y_Configuracion

1.  Instala Flask-Migrate:
    ```bash
    pip install Flask-Migrate
    ```
2.  Inicializa Flask-Migrate en tu aplicación:
    ```python
    # app.py
    from flask import Flask
    from flask_sqlalchemy import SQLAlchemy
    from flask_migrate import Migrate # Importar Migrate

    app = Flask(__name__)
    # ... (configuración de SECRET_KEY y SQLALCHEMY_DATABASE_URI) ...
    db = SQLAlchemy(app)
    migrate = Migrate(app, db) # Inicializar Migrate

    # ... (definición de tus modelos db.Model) ...

    # ... (rutas, etc.) ...
    ```

### 10_3_Creando_la_Primera_Migracion

Una vez configurado `migrate = Migrate(app, db)`:

1.  **Inicializa el repositorio de migraciones (solo la primera vez):**
    Abre tu terminal (con el entorno virtual activado) en el directorio de tu proyecto y ejecuta:
    ```bash
    flask db init
    ```
    Esto crea una carpeta `migrations` con la configuración de Alembic. Añade esta carpeta a tu control de versiones (Git).

2.  **Genera el script de migración inicial:**
    Flask-Migrate comparará tus modelos (`db.Model`) con el estado actual de la base de datos (que estará vacía si es la primera vez) y generará un script para crear las tablas.
    ```bash
    flask db migrate -m "Migracion inicial - crea tablas User y Post"
    ```
    *   `-m "mensaje"`: Añade un mensaje descriptivo a la migración.
    *   Esto creará un archivo de script dentro de `migrations/versions/`. **Revisa este script** para asegurarte de que hace lo que esperas.

### 10_4_Aplicando_Migraciones

El comando `migrate` solo *genera* el script. Para *aplicar* los cambios a la base de datos, usa:

```bash
flask db upgrade
```

Esto ejecutará todos los scripts de migración pendientes (en este caso, el inicial) y creará las tablas en tu base de datos.

### 10_5_Actualizando_el_Modelo_y_Generando_Nuevas_Migraciones

Ahora, el flujo de trabajo típico cuando cambias tus modelos es:

1.  **Modifica tus modelos** en `app.py` o `models.py` (ej: añade una columna `is_active` a `User`).
    ```python
    # En el modelo User:
    is_active = db.Column(db.Boolean, default=True, nullable=False)
    ```
2.  **Genera un nuevo script de migración:**
    ```bash
    flask db migrate -m "Anade columna is_active a User"
    ```
    Flask-Migrate/Alembic detectará el cambio y generará un nuevo script en `migrations/versions/` con operaciones como `op.add_column(...)`. **Revisa el script generado.** A veces, Alembic necesita ayuda para detectar cambios complejos (renombrar tablas/columnas, cambios de tipo con restricciones) y puede que necesites editar el script manualmente.
3.  **Aplica la nueva migración a la base de datos:**
    ```bash
    flask db upgrade
    ```
    Esto aplicará solo la nueva migración, añadiendo la columna `is_active` a la tabla `user` existente sin borrar datos.

**Otros comandos útiles de `flask db`:**

*   `flask db downgrade`: Revierte la última migración aplicada.
*   `flask db history`: Muestra el historial de migraciones.
*   `flask db current`: Muestra la revisión de la migración actual en la base de datos.
*   `flask db stamp head`: Marca la base de datos como si todas las migraciones estuvieran aplicadas (útil si la BD ya existe y coincide con los modelos).

## 11_Estructura_de_Proyectos_Blueprints

A medida que tu aplicación crece, poner todas las rutas y lógica en un solo archivo `app.py` se vuelve inmanejable. Los **Blueprints** son la solución de Flask para organizar tu aplicación en componentes modulares y reutilizables.

### 11_1_La_Necesidad_de_Modularidad

*   **Organización:** Agrupa rutas, plantillas y archivos estáticos relacionados (ej: un blueprint para autenticación, otro para el blog, otro para la API).
*   **Reutilización:** Un blueprint puede ser registrado en diferentes aplicaciones Flask o varias veces en la misma aplicación con diferentes prefijos de URL.
*   **Escalabilidad:** Facilita el trabajo en equipo y el mantenimiento de aplicaciones grandes.

### 11_2_Creando_un_Blueprint

Un Blueprint se crea de forma similar a una aplicación Flask, pero no es una aplicación en sí misma.

1.  Crea una nueva carpeta para tu módulo, por ejemplo `miproyecto/auth/`.
2.  Dentro de esa carpeta, crea un archivo `routes.py` (o `views.py`) para las vistas del blueprint.

    ```python
    # miproyecto/auth/routes.py
    from flask import Blueprint, render_template, redirect, url_for, flash, request
    from miproyecto.forms import LoginForm, RegistrationForm # Asumiendo que los formularios están en miproyecto/forms.py
    # from miproyecto.models import User # Si necesitas interactuar con modelos
    # from miproyecto import db # Si necesitas la sesión de la BD

    # 1. Crear el Blueprint
    # 'auth': Nombre del blueprint (usado internamente por Flask, ej: en url_for)
    # __name__: Ayuda a Flask a encontrar la ubicación del blueprint (para plantillas/estáticos)
    # template_folder='templates': (Opcional) Especifica una carpeta de plantillas para este blueprint
    # static_folder='static': (Opcional) Especifica una carpeta de estáticos para este blueprint
    # url_prefix='/auth': (Opcional, se puede definir al registrar) Prefijo para todas las rutas del blueprint
    auth_bp = Blueprint('auth', __name__, template_folder='templates', url_prefix='/auth')

    # 2. Definir rutas usando el decorador del blueprint (ej: @auth_bp.route)
    @auth_bp.route('/login', methods=['GET', 'POST'])
    def login():
        form = LoginForm()
        if form.validate_on_submit():
            # Lógica de login...
            flash('Login exitoso!', 'success')
            # Nota el '.' antes del nombre de la vista en url_for dentro de un blueprint
            # para referirse a una vista dentro del MISMO blueprint.
            # O usa 'main.index' para referirte a la vista 'index' del blueprint 'main'.
            return redirect(url_for('main.index')) # Asumiendo un blueprint 'main' para la ruta principal '/'
        return render_template('auth/login.html', title='Iniciar Sesion', form=form) # Busca en auth/templates/auth/login.html

    @auth_bp.route('/register', methods=['GET', 'POST'])
    def register():
        form = RegistrationForm()
        if form.validate_on_submit():
            # Lógica de registro...
            flash('Registro exitoso! Por favor inicia sesion.', 'success')
            return redirect(url_for('auth.login')) # Ruta 'login' dentro del blueprint 'auth'
        return render_template('auth/register.html', title='Registro', form=form)

    @auth_bp.route('/logout')
    def logout():
        # Lógica de logout...
        flash('Has cerrado sesion.', 'info')
        return redirect(url_for('main.index'))
    ```

Para que las rutas del blueprint funcionen, debes registrarlo en tu instancia principal de la aplicación Flask (normalmente en `app.py` o donde creas la `app` con el patrón factory).

```python
# miproyecto/__init__.py (usando el patrón factory, recomendado)

from flask import Flask
from flask_sqlalchemy import SQLAlchemy
from flask_migrate import Migrate
from flask_wtf.csrf import CSRFProtect
from .config import Config # Asumiendo un archivo config.py

db = SQLAlchemy()
migrate = Migrate()
csrf = CSRFProtect()

def create_app(config_class=Config):
    app = Flask(__name__)
    app.config.from_object(config_class)

    db.init_app(app)
    migrate.init_app(app, db)
    csrf.init_app(app) # Inicializar CSRF para protección global

    # Importar y registrar Blueprints
    from .auth.routes import auth_bp
    app.register_blueprint(auth_bp) # El url_prefix='/auth' se definió en el blueprint

    from .main.routes import main_bp # Asumiendo un blueprint 'main' para rutas principales
    app.register_blueprint(main_bp) # Este podría no tener url_prefix

    from .blog.routes import blog_bp # Asumiendo un blueprint 'blog'
    app.register_blueprint(blog_bp, url_prefix='/blog') # Puedes definir/sobrescribir el prefijo aquí

    # Puedes registrar más blueprints...

    with app.app_context():
        # Puedes importar modelos aquí si es necesario o dentro de los blueprints
        from . import models # Asegúrate que los modelos se importen para Migrate

    return app
```

### 11_4_Recursos_Especificos_del_Blueprint_Plantillas_Estaticos

## 12_Manejo_de_Errores

Una aplicación robusta necesita manejar errores gracefully, mostrando páginas amigables al usuario en lugar de tracebacks crípticos (especialmente en producción).

### 12_1_Errores_HTTP_Estandar

Flask tiene manejadores por defecto para errores comunes como:

*   `404 Not Found`: La URL solicitada no existe.
*   `403 Forbidden`: El usuario no tiene permiso para acceder.
*   `500 Internal Server Error`: Ocurrió un error inesperado en el servidor.
*   `405 Method Not Allowed`: Se usó un método HTTP incorrecto para una ruta (ej: GET en una ruta solo POST).

### 12_2_Manejadores_de_Errores_Personalizados

Puedes sobrescribir los manejadores por defecto o crear manejadores para otros códigos de estado HTTP usando el decorador `@app.errorhandler()`.

```python
# app.py o donde creas tu app
from flask import render_template # Asegúrate de importarlo

# ... (creación de app)

@app.errorhandler(404)
def page_not_found(error):
    # El manejador recibe el objeto de error como argumento
    # Puedes loggear el error aquí si quieres: app.logger.error(f'Page not found: {request.url}')
    # Devolvemos una tupla: (contenido_respuesta, codigo_estado)
    return render_template('errors/404.html'), 404

@app.errorhandler(500)
def internal_server_error(error):
    # Es importante hacer rollback de la sesión de BD en errores 500
    # para evitar estados inconsistentes.
    db.session.rollback() # Asumiendo que 'db' está disponible aquí
    # Puedes loggear el error completo: app.logger.error(f'Server Error: {error}', exc_info=True)
    return render_template('errors/500.html'), 500

@app.errorhandler(403)
def forbidden_error(error):
    return render_template('errors/403.html'), 403

# También puedes registrar manejadores en Blueprints: @mi_blueprint.app_errorhandler(404)
# @app.errorhandler funciona para toda la aplicación.
```

## 13_Sesiones_y_Cookies

HTTP es un protocolo sin estado. Las **cookies** y las **sesiones** son mecanismos para mantener información sobre un usuario entre múltiples peticiones.

### 13_1_Que_son_las_Cookies

*   Pequeños datos que el servidor envía al navegador.
*   El navegador los almacena y los envía de vuelta al servidor en cada petición subsiguiente al mismo dominio.
*   Se usan para recordar información (preferencias, estado de login - aunque no directamente para datos sensibles).
*   Flask permite leer (`request.cookies.get('mi_cookie')`) y escribir cookies (`response.set_cookie('nombre', 'valor')`).

### 13_2_Que_son_las_Sesiones

*   Las sesiones de Flask usan cookies, pero de forma más segura y conveniente.
*   En lugar de almacenar los datos directamente en la cookie, Flask almacena un *identificador de sesión* en una cookie firmada criptográficamente.
*   Los *datos reales* de la sesión se almacenan en el lado del servidor (por defecto, en una cookie firmada del lado del cliente, pero se pueden configurar para usar almacenamiento en servidor).
*   Esto evita que el usuario pueda manipular los datos de la sesión directamente.
*   Se accede a los datos de la sesión a través del objeto `session` (importado de `flask`).

### 13_3_Configurando_la_Clave_Secreta_SECRET_KEY

**¡Absolutamente crucial para la seguridad de las sesiones!** Flask usa `SECRET_KEY` para firmar la cookie de sesión. Si alguien conoce esta clave, puede falsificar sesiones.

### 14_3_El_Modelo_de_Usuario

Tu modelo de usuario (generalmente en `models.py`) necesita heredar de `UserMixin` (proporcionado por Flask-Login) además de `db.Model`. `UserMixin` añade los métodos requeridos por Flask-Login (`is_authenticated`, `is_active`, `is_anonymous`, `get_id`).

```python
# miproyecto/models.py
from . import db, login_manager # Importar db y login_manager desde __init__.py
from flask_login import UserMixin
from werkzeug.security import generate_password_hash, check_password_hash # Para contraseñas

# Callback requerido por Flask-Login para cargar un usuario dado su ID
@login_manager.user_loader
def load_user(user_id):
    # user_id es una cadena, convertir a int para la consulta
    return User.query.get(int(user_id))
    # O usando la forma nueva/recomendada:
    # return db.session.get(User, int(user_id))


class User(UserMixin, db.Model): # Heredar de UserMixin
    id = db.Column(db.Integer, primary_key=True)
    username = db.Column(db.String(64), index=True, unique=True)
    email = db.Column(db.String(120), index=True, unique=True)
    password_hash = db.Column(db.String(128))
    # ... otros campos ...

    def set_password(self, password):
        # Genera un hash seguro de la contraseña
        self.password_hash = generate_password_hash(password)

    def check_password(self, password):
        # Verifica si la contraseña proporcionada coincide con el hash almacenado
        return check_password_hash(self.password_hash, password)

    def __repr__(self):
        return f'<User {self.username}>'

# Asegúrate de importar este archivo de modelos en algún lugar para que SQLAlchemy
# y Flask-Login los conozcan (por ejemplo, en __init__.py o al final de routes.py).
```

### 14_4_Funciones_de_Callback_user_loader

Esta función recibe el ID del usuario (como cadena) y debe devolver el objeto `User` correspondiente, o `None` si el ID no es válido. (Ver ejemplo en [14.3](#14_3_El_Modelo_de_Usuario)).

### 14_5_Login_y_Logout_de_Usuarios

Usa las funciones `login_user` y `logout_user` de Flask-Login en tus vistas de login/logout.

```python
# miproyecto/auth/routes.py
from flask import render_template, redirect, url_for, flash, request
from flask_login import login_user, logout_user, current_user, login_required
from . import auth_bp # Asume que auth_bp se define aquí o se importa
from ..models import User # Importa el modelo User
from .. import db # Importa db para la sesión
from ..forms import LoginForm, RegistrationForm # Importa los formularios

@auth_bp.route('/register', methods=['GET', 'POST'])
def register():
    # Si el usuario ya está logueado, redirigir a la página principal
    if current_user.is_authenticated:
        return redirect(url_for('main.index'))
    form = RegistrationForm()
    if form.validate_on_submit():
        user = User(username=form.username.data, email=form.email.data)
        user.set_password(form.password.data) # Hashea la contraseña
        db.session.add(user)
        db.session.commit()
        flash('Felicidades, te has registrado correctamente!', 'success')
        return redirect(url_for('auth.login'))
    return render_template('auth/register.html', title='Registro', form=form)

@auth_bp.route('/login', methods=['GET', 'POST'])
def login():
    if current_user.is_authenticated:
        return redirect(url_for('main.index'))
    form = LoginForm()
    if form.validate_on_submit():
        user = User.query.filter_by(email=form.email.data).first()
        # Verifica si el usuario existe Y la contraseña es correcta
        if user is None or not user.check_password(form.password.data):
            flash('Email o contraseña invalidos.', 'danger')
            return redirect(url_for('auth.login'))
        # Si todo es correcto, loguea al usuario
        login_user(user, remember=form.remember.data) # remember viene del checkbox del form
        flash('Login exitoso!', 'success')
        # Redirige a la página que el usuario intentaba visitar (si existe) o al index
        next_page = request.args.get('next')
        return redirect(next_page) if next_page else redirect(url_for('main.index'))
    return render_template('auth/login.html', title='Iniciar Sesion', form=form)

@auth_bp.route('/logout')
@login_required # Es buena idea requerir login para hacer logout
def logout():
    logout_user() # Desloguea al usuario
    flash('Has cerrado sesion.', 'info')
    return redirect(url_for('main.index'))
```

*   `login_user(user, remember=False)`: Registra al usuario como logueado en la sesión. Si `remember=True`, se usa una cookie persistente ("Remember Me").
*   `logout_user()`: Elimina al usuario de la sesión.
*   `current_user`: Un proxy especial que representa al usuario actualmente logueado (o un objeto `AnonymousUserMixin` si nadie está logueado). Puedes acceder a sus atributos (`current_user.username`, `current_user.email`) directamente en tus vistas y plantillas.
*   `current_user.is_authenticated`: Devuelve `True` si el usuario está logueado, `False` si no.

### 14_6_Protegiendo_Vistas_login_required

Para requerir que un usuario esté logueado para acceder a una vista, usa el decorador `@login_required` (importado de `flask_login`).

## 15_Creacion_de_APIs_RESTful

Flask es excelente para construir APIs web (Application Programming Interfaces), especialmente APIs RESTful (Representational State Transfer). Las APIs permiten que diferentes aplicaciones (navegadores web, aplicaciones móviles, otros servidores) interactúen con tus datos y lógica.

### 15_1_Principios_REST

REST no es un protocolo, sino un conjunto de principios arquitectónicos:

1.  **Cliente-Servidor:** Separación clara de responsabilidades.
2.  **Sin Estado (Stateless):** Cada petición del cliente al servidor debe contener toda la información necesaria para ser entendida y procesada. El servidor no guarda estado del cliente entre peticiones (la autenticación se maneja con tokens, no sesiones de servidor).
3.  **Cacheable:** Las respuestas deben poder ser cacheadas por el cliente o intermediarios.
4.  **Interfaz Uniforme:**
    *   **Identificación de Recursos:** Cada recurso (ej: un usuario, un post) tiene un identificador único (una URL, ej: `/api/users/1`).
    *   **Manipulación a través de Representaciones:** Los clientes interactúan con representaciones de los recursos (comúnmente JSON o XML).
    *   **Mensajes Autodescriptivos:** Cada mensaje incluye suficiente información para describirse (ej: `Content-Type: application/json`).
    *   **HATEOAS (Hypermedia as the Engine of Application State):** (Opcional pero recomendado) Las respuestas incluyen enlaces (URLs) a otras acciones o recursos relacionados.
5.  **Sistema en Capas:** La arquitectura puede tener intermediarios (proxies, balanceadores de carga) sin que el cliente lo sepa.
6.  **Código bajo Demanda (Opcional):** El servidor puede enviar código ejecutable (ej: JavaScript) al cliente.

### 15_2_Devolviendo_JSON_jsonify

Para devolver datos en formato JSON desde una vista de Flask, usa la función `jsonify`. Convierte diccionarios o listas Python en una respuesta JSON con el `Content-Type: application/json` correcto.

```python
from flask import Flask, jsonify, request

app = Flask(__name__)

# Datos de ejemplo (en una app real vendrían de la BD)
tasks = [
    {'id': 1, 'title': 'Comprar leche', 'done': False},
    {'id': 2, 'title': 'Aprender Flask API', 'done': True}
]
next_id = 3

# Obtener todas las tareas
@app.route('/api/tasks', methods=['GET'])
def get_tasks():
    return jsonify({'tasks': tasks})

# Obtener una tarea específica
@app.route('/api/tasks/<int:task_id>', methods=['GET'])
def get_task(task_id):
    task = next((task for task in tasks if task['id'] == task_id), None)
    if task is None:
        abort(404, description=f"Tarea con id {task_id} no encontrada.")
    return jsonify({'task': task})

if __name__ == '__main__':
    app.run(debug=True)
```

### 15_3_Manejando_Metodos_HTTP_para_APIs

Usa los métodos HTTP estándar para las operaciones CRUD sobre los recursos:

*   **GET:** Recuperar recursos. (`/api/tasks` para listar, `/api/tasks/1` para obtener uno).
*   **POST:** Crear un nuevo recurso. (`/api/tasks` con datos en el body).
*   **PUT:** Actualizar un recurso existente (reemplazo completo). (`/api/tasks/1` con datos en el body).
*   **PATCH:** Actualizar parcialmente un recurso existente. (`/api/tasks/1` con datos parciales en el body).
*   **DELETE:** Eliminar un recurso. (`/api/tasks/1`).

```python
# ... (app y tasks definidos arriba) ...
from flask import abort

# Crear una nueva tarea
@app.route('/api/tasks', methods=['POST'])
def create_task():
    global next_id
    if not request.json or not 'title' in request.json:
        abort(400, description="Peticion invalida, falta 'title' en JSON.") # Bad Request
    new_task = {
        'id': next_id,
        'title': request.json['title'],
        'done': request.json.get('done', False) # .get para valor opcional
    }
    tasks.append(new_task)
    next_id += 1
    # Devuelve la tarea creada y el código de estado 201 Created
    # Es buena práctica devolver la URL del nuevo recurso en la cabecera Location
    response = jsonify({'task': new_task})
    response.status_code = 201
    response.headers['Location'] = url_for('get_task', task_id=new_task['id'], _external=True)
    return response

# Actualizar una tarea (PUT - reemplazo completo)
@app.route('/api/tasks/<int:task_id>', methods=['PUT'])
def update_task(task_id):
    task = next((task for task in tasks if task['id'] == task_id), None)
    if task is None:
        abort(404)
    if not request.json:
        abort(400, description="Peticion invalida, se requiere JSON.")
    # Validación básica (en una app real sería más robusta)
    if 'title' not in request.json or 'done' not in request.json:
         abort(400, description="Peticion invalida, faltan 'title' o 'done'.")

    task['title'] = request.json['title']
    task['done'] = request.json['done']
    return jsonify({'task': task})

# Eliminar una tarea
@app.route('/api/tasks/<int:task_id>', methods=['DELETE'])
def delete_task(task_id):
    global tasks
    task_index = next((index for index, task in enumerate(tasks) if task['id'] == task_id), None)
    if task_index is None:
        abort(404)
    del tasks[task_index]
    # Respuesta exitosa sin contenido
    return '', 204 # No Content
```

### 15_4_Codigos_de_Estado_HTTP

Usar los códigos de estado HTTP correctos es fundamental en las APIs REST:

*   **2xx (Éxito):**
    *   `200 OK`: Petición exitosa (GET, PUT/PATCH exitoso).
    *   `201 Created`: Recurso creado exitosamente (POST exitoso). Debería incluir cabecera `Location`.
    *   `204 No Content`: Petición exitosa, pero no hay contenido para devolver (DELETE exitoso).
*   **4xx (Error del Cliente):**
    *   `400 Bad Request`: Petición malformada (ej: JSON inválido, datos faltantes).
    *   `401 Unauthorized`: Autenticación requerida o fallida.
    *   `403 Forbidden`: Autenticado, pero sin permisos para el recurso.
    *   `404 Not Found`: Recurso no encontrado.
    *   `405 Method Not Allowed`: Método HTTP no permitido para ese endpoint.
    *   `409 Conflict`: Conflicto con el estado actual del recurso (ej: intentar crear un recurso que ya existe).
*   **5xx (Error del Servidor):**
    *   `500 Internal Server Error`: Error genérico inesperado en el servidor.

Puedes devolver códigos de estado explícitamente:

### 15_5_Autenticacion_de_API_Tokens

Las APIs RESTful deben ser sin estado, por lo que la autenticación basada en sesión de Flask-Login no es ideal. Se suelen usar **tokens**:

1.  **Login:** El cliente envía credenciales (usuario/contraseña) a un endpoint específico (ej: `/api/login`).
2.  **Generación de Token:** Si las credenciales son válidas, el servidor genera un token único y seguro (ej: JWT - JSON Web Token) que identifica al usuario y tiene una fecha de expiración.
3.  **Envío de Token:** El servidor devuelve el token al cliente.
4.  **Peticiones Subsiguientes:** El cliente incluye el token en cada petición subsiguiente a endpoints protegidos, generalmente en la cabecera `Authorization` (ej: `Authorization: Bearer <token>`).
5.  **Validación de Token:** El servidor valida el token en cada petición protegida (verifica firma, expiración, etc.) antes de procesarla.

Implementar esto requiere librerías adicionales como **PyJWT** para manejar JWTs y decoradores personalizados para verificar los tokens.

**Ejemplo conceptual de decorador de autenticación por token:**

```python
from functools import wraps
from flask import request, abort, current_app
import jwt # de PyJWT

def token_required(f):
    @wraps(f)
    def decorated(*args, **kwargs):
        token = None
        if 'Authorization' in request.headers:
            auth_header = request.headers['Authorization']
            try:
                token_type, token = auth_header.split(' ')
                if token_type.lower() != 'bearer':
                    abort(401, description="Tipo de token invalido.")
            except ValueError:
                 abort(401, description="Cabecera de autorizacion malformada.")

        if not token:
            abort(401, description='Token no encontrado.')

        try:
            # Decodifica y valida el token (firma, expiración)
            # Usa la SECRET_KEY de la app para verificar la firma
            data = jwt.decode(token, current_app.config['SECRET_KEY'], algorithms=["HS256"])
            # Aquí podrías buscar al usuario en la BD usando data['user_id']
            # y quizás añadirlo a 'g' (contexto de la petición)
            # g.current_user = User.query.get(data['user_id'])
        except jwt.ExpiredSignatureError:
             abort(401, description='Token expirado.')
        except jwt.InvalidTokenError:
             abort(401, description='Token invalido.')

        return f(*args, **kwargs) # Llama a la vista original
    return decorated

# Uso en una vista:
@app.route('/api/profile', methods=['GET'])
@token_required # Aplicar el decorador
def get_profile():
    # El token ha sido validado, puedes acceder a g.current_user si lo definiste
    # return jsonify({'user': g.current_user.to_dict()}) # Ejemplo
    return jsonify({'message': 'Este es un endpoint protegido'})
```

### 15_6_Extensiones_Flask_RESTful_Flask_Smorest

Aunque puedes construir APIs RESTful con Flask básico y `jsonify`, existen extensiones que proporcionan más estructura y características:

*   **Flask-RESTful:** Un framework de extensión popular que ayuda a construir APIs rápidamente. Proporciona clases `Resource` para agrupar operaciones HTTP para un recurso y maneja la negociación de contenido. Es un poco más antiguo y menos mantenido activamente.
*   **Flask-Smorest:** Una alternativa moderna basada en Marshmallow (para serialización/deserialización y validación de datos) y OpenAPI (Swagger) para la generación automática de documentación interactiva de la API. Facilita la validación de entrada/salida y la documentación. **Recomendado para nuevos proyectos.**
*   **Flask-Marshmallow:** Integra Marshmallow con Flask para facilitar la serialización/deserialización de objetos (especialmente modelos SQLAlchemy) a/desde JSON. Útil incluso si no usas Flask-Smorest.

Usar estas extensiones puede acelerar el desarrollo y mejorar la calidad de tu API, especialmente para proyectos más grandes.

### 17_2_El_Patron_Application_Factory

En lugar de crear la instancia `app = Flask(__name__)` globalmente, define una función que crea y configura la aplicación. Esto es el **patrón factory**.

**Ventajas:**

*   **Testing:** Permite crear múltiples instancias de la app con diferentes configuraciones para las pruebas.
*   **Blueprints y Extensiones:** Es el lugar ideal para inicializar extensiones (`db.init_app(app)`, `login_manager.init_app(app)`) y registrar blueprints.
*   **Organización:** Mantiene la creación y configuración de la app encapsulada.

```python
# miproyecto/__init__.py

from flask import Flask
from flask_sqlalchemy import SQLAlchemy
from flask_migrate import Migrate
from flask_login import LoginManager
from flask_mail import Mail
# ... otras extensiones ...
from .config import config # Importa el diccionario de configuraciones

# Inicializa las extensiones SIN pasarles la app todavía
db = SQLAlchemy()
migrate = Migrate()
login_manager = LoginManager()
login_manager.login_view = 'auth.login' # Configura la vista de login
mail = Mail()
# ... inicializa otras extensiones ...

def create_app(config_name=None):
    """Application Factory Function"""
    if config_name is None:
        # Obtener config del entorno o usar 'default'
        config_name = os.environ.get('FLASK_CONFIG', 'default')

    app = Flask(__name__)
    # Carga la configuración desde el objeto correspondiente
    app.config.from_object(config[config_name])
    # Permite que la clase config realice inicializaciones si es necesario
    config[config_name].init_app(app)

    # Inicializa las extensiones CON la app creada
    db.init_app(app)
    migrate.init_app(app, db)
    login_manager.init_app(app)
    mail.init_app(app)
    # ... .init_app(app) para otras extensiones ...

    # Registrar Blueprints
    from .main import main as main_blueprint # Asumiendo blueprints definidos en paquetes
    app.register_blueprint(main_blueprint)

    from .auth import auth as auth_blueprint
    app.register_blueprint(auth_blueprint, url_prefix='/auth')

    # ... registrar otros blueprints ...

    # Opcional: Añadir rutas globales o manejadores de error a nivel de app
    # @app.route('/health')
    # def health_check():
    #    return "OK", 200

    # Importante: Asegúrate de que los modelos se conozcan
    # Esto se puede hacer importándolos aquí o asegurándote de que
    # los blueprints donde se usan los importen.
    with app.app_context():
         from . import models

    return app # Devuelve la instancia de la aplicación configurada
```

**Cómo usar el factory:**

1.  **Archivo principal (ej: `run.py` o `wsgi.py`):**
    ```python

    ```

### 17_5_Context_Locals_g_current_app

Flask utiliza *context locals* para hacer que objetos como `request`, `session`, `current_app` y `g` estén disponibles globalmente *dentro del contexto de una petición* sin necesidad de pasarlos como argumentos entre funciones.

*   **`current_app`:** Un proxy a la instancia de la aplicación (`app`) que está manejando la petición actual. Útil para acceder a la configuración (`current_app.config['SECRET_KEY']`) o al logger (`current_app.logger`) desde cualquier lugar dentro de una petición.
*   **`request`:** Proxy al objeto `request` de la petición actual.
*   **`session`:** Proxy al objeto `session` de la petición actual.
*   **`g` (global):** Un objeto simple (tipo `SimpleNamespace` o `dict`) que está disponible *solo durante el ciclo de vida de una petición*. Se usa como un lugar temporal para almacenar datos que necesitan ser compartidos entre diferentes funciones (como middleware o decoradores y la vista final) durante el manejo de *una única* petición. **No se comparte entre peticiones.** Se resetea para cada nueva petición.

    ```python
    from flask import g, current_app

    @app.before_request # Se ejecuta antes de cada vista
    def load_logged_in_user():
        # Ejemplo: Cargar datos adicionales del usuario y ponerlos en 'g'
        user_id = session.get('user_id')
        if user_id is None:
            g.user = None
        else:
            # Asume que tienes una función para obtener usuario de BD
            g.user = get_user_from_db(user_id)
            current_app.logger.debug(f"Usuario {g.user.username} cargado en g")

    @app.route('/profile')
    def profile():
        # Accede al usuario desde 'g' sin necesidad de volver a consultarlo
        if g.user:
            return render_template('profile.html', user=g.user)
        else:
            abort(401) # O redirigir a login

    @app.teardown_request # Se ejecuta después de cada vista, incluso si hay error
    def teardown_db_connection(exception=None):
        # Ejemplo: Cerrar una conexión a BD almacenada en 'g' si se usara
        db_conn = g.pop('db_conn', None) # Quita 'db_conn' de g si existe
        if db_conn is not None:
            db_conn.close()
            current_app.logger.debug("Conexión a BD en g cerrada.")
    ```

`g` es útil para almacenar recursos (como conexiones a BD) o datos calculados una vez por petición que se necesitan en múltiples lugares.

## 19_Tareas_en_Segundo_Plano_con_Celery

Algunas tareas toman demasiado tiempo para ejecutarse dentro del ciclo de una petición web (enviar emails masivos, procesar imágenes/videos, generar reportes complejos, llamar a APIs lentas). Ejecutarlas directamente bloquearía al trabajador web, empeorando la experiencia del usuario. **Celery** es la solución estándar en Python para ejecutar estas tareas de forma asíncrona en segundo plano.

### 19_1_Necesidad_de_Tareas_Asincronas

*   **Mejorar Tiempo de Respuesta:** La vista web puede devolver una respuesta al usuario inmediatamente ("Tu reporte se está generando y te notificaremos cuando esté listo") mientras la tarea se ejecuta en segundo plano.
*   **Manejar Tareas Largas:** Evita timeouts del servidor web o del navegador.
*   **Ejecutar Tareas Periódicas:** Celery también puede programar tareas para que se ejecuten a intervalos regulares (ej: limpiar datos antiguos cada noche).
*   **Distribuir Carga:** Las tareas pueden ejecutarse en máquinas diferentes a las del servidor web.

### 19_2_Introduccion_a_Celery

Celery consta de varios componentes:

1.  **Productor (Producer):** Tu aplicación Flask. Es quien *envía* las tareas a la cola.
2.  **Broker (Message Broker):** Un intermediario que recibe las tareas del productor y las almacena hasta que un worker esté listo para procesarlas. Ejemplos populares: **Redis**, **RabbitMQ**.
3.  **Cola (Queue):** Donde el broker almacena los mensajes de tarea.
4.  **Consumidor (Worker):** Uno o más procesos Celery que se ejecutan independientemente de tu app Flask. Toman tareas de la cola, las ejecutan y (opcionalmente) almacenan los resultados.
5.  **Backend de Resultados (Result Backend):** (Opcional) Un lugar donde los workers pueden almacenar el estado y el resultado de las tareas (ej: Redis, base de datos SQLAlchemy). Permite a tu app Flask consultar el estado de una tarea o recuperar su resultado.

### 19_3_Configuracion_de_Celery_con_Flask

Integrar Celery con Flask requiere un poco de configuración para que las tareas Celery tengan acceso al contexto de la aplicación Flask (configuración, extensiones como `db` o `mail`).

1.  **Instalación:**

    ```python
    # miproyecto/celery_app.py (o similar)
    from celery import Celery
    from flask import Flask

    def make_celery(app: Flask) -> Celery:
        """
        Configura e integra Celery con la aplicación Flask.
        Basado en la documentación de Flask:
        https://flask.palletsprojects.com/en/2.3.x/patterns/celery/
        """
        # Carga la configuración del broker y backend desde la config de Flask
        # Asume que CELERY_BROKER_URL y CELERY_RESULT_BACKEND están en app.config
        celery = Celery(
            app.import_name,
            backend=app.config.get('CELERY_RESULT_BACKEND'),
            broker=app.config.get('CELERY_BROKER_URL'),
            # include=['miproyecto.tasks.email_tasks'] # Opcional: Lista de módulos con tareas
        )
        celery.conf.update(app.config) # Carga otras configs de Celery desde Flask

        # Define una clase de Tarea base que configura el contexto de la app Flask
        class ContextTask(celery.Task):
            abstract = True # No registrar esta clase como tarea ejecutable

            def __call__(self, *args, **kwargs):
                # Asegura que el contexto de la app Flask esté activo
                # cuando la tarea se ejecute en el worker Celery.
                with app.app_context():
                    return self.run(*args, **kwargs)

        celery.Task = ContextTask # Usa nuestra tarea base personalizada
        # Autodiscover tasks (opcional, busca tareas en módulos listados en 'include' o en tasks.py)
        # celery.autodiscover_tasks()
        return celery

    # Nota: La instancia real de Celery se creará junto con la app Flask
    # usando el patrón factory. No crees una instancia global aquí directamente.
    ```

Define tus tareas como funciones decoradas con `@celery.task`.

```python
# Dentro de una vista Flask
from ..tasks.email_tasks import send_async_email_task, long_running_task

@auth_bp.route('/register', methods=['GET', 'POST'])
def register():
    # ... (lógica de registro, crear usuario) ...
    if form.validate_on_submit():
        # ... (crear usuario, commit) ...
        user = User(...)
        db.session.add(user)
        db.session.commit()

        # Enviar email de bienvenida de forma asíncrona
        token = generate_confirmation_token(user.email) # Función hipotética
        send_async_email_task.delay(user.email,
                                    'Bienvenido!',
                                    'email/welcome',
                                    user=user, token=token)

        flash('Registro exitoso! Se ha enviado un email de bienvenida.', 'success')
        return redirect(url_for('auth.login'))
    # ...

@main_bp.route('/iniciar_calculo')
def iniciar_calculo():
    # Iniciar tarea larga
    task_handle = long_running_task.delay(10, 25)
    # task_handle es un objeto AsyncResult que tiene el ID de la tarea
    flash(f"Tarea de cálculo iniciada (ID: {task_handle.id})")
    # Puedes guardar task_handle.id para consultar el estado más tarde
    session['last_task_id'] = task_handle.id
    return redirect(url_for('main.index'))

@main_bp.route('/estado_tarea')
def estado_tarea():
    task_id = session.get('last_task_id')
    if not task_id:
        return jsonify({'error': 'No hay tarea para verificar'}), 404

    # Consultar el estado usando el backend de resultados
    task = long_running_task.AsyncResult(task_id)

    response = {
        'task_id': task_id,
        'state': task.state, # Estados: PENDING, STARTED, SUCCESS, FAILURE, RETRY
    }
    if task.state == 'PENDING':
        response['status'] = 'Pendiente o no encontrada.'
    elif task.state == 'FAILURE':
        response['status'] = str(task.info) # Contiene la excepción
    elif task.state == 'SUCCESS':
        response['result'] = task.result # Contiene el valor devuelto por la tarea
        response['status'] = 'Completada!'
    else:
        response['status'] = task.state # Otros estados como STARTED, RETRY

    return jsonify(response)

```

*   `.delay(*args, **kwargs)`: Atajo para `.apply_async(args=args, kwargs=kwargs)`. Envía la tarea a la cola.
*   `.apply_async(args=None, kwargs=None, countdown=None, eta=None, expires=None, ...)`: Permite más control (ej: `countdown` para retrasar la ejecución).
*   Ambos devuelven un objeto `AsyncResult` que contiene el `id` de la tarea. Puedes usar este ID para consultar el estado o el resultado si tienes configurado un backend de resultados.

### 19_5_Brokers_Redis_RabbitMQ

*   **Redis:** Muy rápido, fácil de configurar, ideal para empezar y para muchos casos de uso. Es también un excelente backend de resultados. Menos robusto que RabbitMQ ante ciertos fallos o reinicios (aunque las versiones recientes han mejorado).
*   **RabbitMQ:** Un broker de mensajes AMQP completo y muy robusto. Más complejo de configurar que Redis. Garantiza mejor la entrega de mensajes en escenarios complejos o de alta disponibilidad.

La elección depende de tus necesidades de robustez y escalabilidad. Redis suele ser suficiente para la mayoría de las aplicaciones Flask.

## 20_Comunicacion_en_Tiempo_Real_con_Flask_SocketIO

Para aplicaciones que requieren comunicación bidireccional instantánea entre el servidor y los clientes (chats, notificaciones en tiempo real, dashboards que se actualizan automáticamente, juegos multijugador simples), HTTP no es suficiente. **WebSockets** es el protocolo estándar para esto, y **Flask-SocketIO** lo integra fácilmente con Flask.

### 20_1_WebSockets_y_Comunicacion_Bidireccional

*   A diferencia de HTTP (petición-respuesta), WebSockets establece una conexión persistente entre cliente y servidor.
*   Ambas partes pueden enviar mensajes a la otra en cualquier momento sin necesidad de una nueva petición.
*   Ideal para funcionalidades "push" del servidor al cliente.

### 20_2_Introduccion_a_SocketIO

Socket.IO no es *solo* WebSockets. Es una librería (con implementaciones en servidor y cliente) que:

*   Proporciona una API más simple y robusta sobre WebSockets.
*   Ofrece mecanismos de fallback (como long-polling HTTP) si WebSockets no está disponible.
*   Añade características como *eventos* personalizados, *salas* (rooms) para broadcasting a grupos de clientes, reconexión automática, etc.
*   Flask-SocketIO integra el servidor Socket.IO con Flask.

### 20_3_Configuracion_de_Flask_SocketIO

1.  **Instalación:** Necesitas Flask-SocketIO y un servidor asíncrono compatible. **Gevent** es la opción más común y recomendada con Flask-SocketIO. **Eventlet** es otra opción. El servidor de desarrollo de Flask (Werkzeug) **no** soporta WebSockets adecuadamente para producción.
    ```bash

    # O usando eventlet
    # pip install Flask-SocketIO eventlet
    ```

### 20_4_Manejando_Eventos_del_Servidor_y_Cliente

Socket.IO se basa en eventos. El cliente y el servidor se comunican emitiendo y escuchando eventos personalizados.

**Lado del Servidor (Flask - `events.py` o similar):**

```python
# miproyecto/events.py
from flask import session, request, current_app
from flask_socketio import emit, join_room, leave_room, send
from . import socketio # Importa la instancia de SocketIO
from flask_login import current_user # Puedes usar current_user si está configurado

# Eventos de conexión/desconexión (predefinidos)
@socketio.on('connect')
def handle_connect():
    """Se ejecuta cuando un cliente se conecta."""
    user_id = current_user.id if current_user.is_authenticated else None
    current_app.logger.info(f"Cliente conectado: SID={request.sid}, UserID={user_id}")
    # Puedes unir al usuario a una sala específica al conectarse
    if user_id:
        join_room(str(user_id)) # Sala basada en ID de usuario
        emit('status', {'msg': f'Bienvenido {current_user.username}!'}, room=request.sid) # Enviar solo al cliente actual
    else:
         emit('status', {'msg': 'Bienvenido invitado!'}, room=request.sid)

@socketio.on('disconnect')
def handle_disconnect():
    """Se ejecuta cuando un cliente se desconecta."""
    user_id = current_user.id if hasattr(g, 'user') and g.user.is_authenticated else None # Necesitaría cargar user en g
    current_app.logger.info(f"Cliente desconectado: SID={request.sid}, UserID={user_id}")
    # No puedes usar current_user directamente aquí a menos que lo cargues antes
    # Puedes quitar al usuario de salas si es necesario, pero a menudo no hace falta

# Evento personalizado 'mensaje_chat'
@socketio.on('mensaje_chat')
def handle_chat_message(json_data):
    """Recibe un mensaje de chat de un cliente."""
    # json_data es el payload enviado por el cliente
    current_app.logger.info(f'Mensaje recibido de {request.sid}: {json_data}')
    message_text = json_data.get('msg', '')
    room = json_data.get('room', 'sala_general') # Asume una sala

    if not message_text:
        return # Ignorar mensajes vacíos

    # Obtener nombre de usuario (si está logueado)
    username = current_user.username if current_user.is_authenticated else f"Anonimo_{request.sid[:4]}"

    # Emitir el mensaje a todos en la sala especificada
    emit('nuevo_mensaje',
         {'user': username, 'msg': message_text, 'sid': request.sid},
         room=room, # Envía a todos en la sala 'room'
         # include_self=False # Opcional: No enviar al remitente original
        )

# Evento para unirse a una sala
@socketio.on('join')
def on_join(data):
    username = current_user.username if current_user.is_authenticated else f"Anonimo_{request.sid[:4]}"
    room = data.get('room', 'sala_general')
    join_room(room)
    current_app.logger.info(f'{username} (SID: {request.sid}) se ha unido a la sala: {room}')
    # Notificar a la sala que alguien se unió (excepto al que se une)
    emit('status', {'msg': f'{username} se ha unido a la sala.'}, room=room, include_self=False)
    # Enviar mensaje de bienvenida solo al que se une
    emit('status', {'msg': f'Te has unido a la sala {room}.'}, room=request.sid)


# Evento para salir de una sala
@socketio.on('leave')
def on_leave(data):
    username = current_user.username if current_user.is_authenticated else f"Anonimo_{request.sid[:4]}"
    room = data.get('room', 'sala_general')
    leave_room(room)
    current_app.logger.info(f'{username} (SID: {request.sid}) ha salido de la sala: {room}')
    # Notificar a la sala que alguien se fue
    emit('status', {'msg': f'{username} ha abandonado la sala.'}, room=room, include_self=False)

# Evento sin nombre ('message') - Usa send() en lugar de emit()
@socketio.on('message')
def handle_message(message):
    current_app.logger.info(f'Mensaje genérico recibido: {message}')
    send(f"Servidor recibió tu mensaje: {message}", room=request.sid) # Responder solo al remitente

```

### 20_5_Emitiendo_Mensajes_y_Salas_Rooms

*   **`emit(event, data, room=None, broadcast=False, include_self=True, namespace=None, callback=None)`:** La función principal para enviar eventos desde el servidor.
    *   `event`: Nombre del evento.
    *   `data`: Datos a enviar (diccionario, lista, string, etc. - serializable a JSON).
    *   `room`: ID de la sala o SID del cliente al que enviar. Si es `None`, envía a todos los conectados (o al namespace).
    *   `broadcast=True`: (Junto con `room=None`) Envía a todos los conectados *excepto* al remitente. `emit(..., room=some_room)` ya implica broadcast dentro de esa sala.
    *   `include_self=False`: Excluye al remitente incluso si está en la sala.
    *   `namespace`: Para usar namespaces de SocketIO (rutas separadas).
    *   `callback`: Función a ejecutar en el servidor cuando el cliente acusa recibo.
*   **`send(data, room=None, ...)`:** Atajo para `emit('message', data, room=room, ...)`.
*   **Salas (Rooms):** Permiten agrupar clientes. Un cliente puede estar en múltiples salas.
    *   `join_room(room_name, sid=None)`: Une a un cliente (por defecto `request.sid`) a una sala.
    *   `leave_room(room_name, sid=None)`: Saca a un cliente de una sala.
    *   `close_room(room_name, namespace=None)`: Desconecta a todos los clientes de una sala (útil para limpieza).
    *   Cada cliente se une automáticamente a una sala con su propio SID (`request.sid`) al conectarse. `emit(..., room=request.sid)` envía un mensaje privado a ese cliente.

Flask-SocketIO es potente para añadir interactividad en tiempo real a tus aplicaciones Flask. Recuerda la importancia de usar un servidor asíncrono adecuado (Gevent/Eventlet) y Gunicorn con el worker correcto en producción.

Llevar tu aplicación Flask del desarrollo local a un servidor público (producción) requiere varios pasos y consideraciones.

### 21_1_Servidor_de_Desarrollo_vs_Produccion

*   **Servidor de Desarrollo (`flask run` o `app.run()`/`socketio.run()`):**
    *   Ideal para desarrollo y debugging.
    *   **NO** es seguro, eficiente ni estable para producción.
    *   Generalmente maneja una petición a la vez (monohilo por defecto).
    *   Werkzeug (el servidor base de Flask) no está diseñado para cargas altas.
    *   No usar `debug=True` en producción.
*   **Servidor de Producción (WSGI + Opcional: Proxy Inverso):**
    *   Necesitas un servidor WSGI robusto para ejecutar tu aplicación Flask.
    *   A menudo se usa un proxy inverso delante del servidor WSGI por seguridad, rendimiento y flexibilidad.

### 21_2_Servidores_WSGI_Gunicorn_uWSGI

Son servidores Python que saben cómo comunicarse con tu aplicación Flask a través del estándar WSGI. Manejan múltiples peticiones concurrentes de manera eficiente.

*   **Gunicorn (Green Unicorn):**
    *   Popular, simple de configurar y usar.
    *   Escrito en Python puro.
    *   Funciona bien con Nginx.
    *   Modelo pre-fork (crea varios procesos worker).
    *   Buena opción para la mayoría de las aplicaciones.
    *   Compatible con workers asíncronos (gevent, eventlet) para Flask-SocketIO.
    *   **Instalación:** `pip install gunicorn`
    *   **Uso Básico:**
        ```bash
        # -w 4: Número de procesos worker (ajustar según CPU cores, usualmente 2*cores + 1)
        # -b 0.0.0.0:8000: Bind a la dirección IP y puerto
        # run:app: Módulo:variable_app (o run:create_app() para factory)
        gunicorn -w 4 -b 0.0.0.0:8000 run:app
        # Con Gevent para SocketIO:
        gunicorn -k gevent -w 1 -b 0.0.0.0:8000 run:app
        ```

*   **uWSGI:**
    *   Muy potente y configurable, pero más complejo que Gunicorn.
    *   Escrito en C, a menudo más rápido en benchmarks.
    *   Muchas opciones de configuración (hilos, procesos, protocolos, etc.).
    *   Puede comunicarse directamente con Nginx usando el protocolo uwsgi (más eficiente que HTTP).
    *   **Instalación:** `pip install uwsgi` (puede requerir herramientas de compilación C).
    *   **Uso Básico (con archivo de configuración `uwsgi.ini`):**
        ```ini
        # uwsgi.ini
        [uwsgi]
        module = run:app ; O wsgi:app

        master = true
        processes = 4 ; Número de workers
        threads = 2   ; Hilos por worker (opcional)

        socket = /tmp/mi_app.sock ; Usar un socket Unix (para Nginx)
        # O http = 0.0.0.0:8000 ; Usar HTTP
        chmod-socket = 660
        vacuum = true ; Eliminar socket al salir

        die-on-term = true ; Salir limpiamente al recibir SIGTERM
        logto = /var/log/uwsgi/mi_app.log ; Archivo de log
        ```
        ```bash
        uwsgi --ini uwsgi.ini
        ```

**Elección:** Gunicorn es generalmente más fácil para empezar. uWSGI ofrece más rendimiento y flexibilidad si necesitas optimizar al máximo.

### 21_3_Proxy_Inverso_Nginx_Apache

Un proxy inverso (como Nginx o Apache) se sitúa *delante* de tu servidor WSGI (Gunicorn/uWSGI). El tráfico de internet llega primero al proxy.

**Beneficios:**

1.  **Servir Archivos Estáticos:** Nginx/Apache son extremadamente eficientes sirviendo archivos estáticos (CSS, JS, imágenes), liberando a Gunicorn/uWSGI para manejar la lógica dinámica de Flask.
2.  **Balanceo de Carga:** Si ejecutas múltiples instancias de Gunicorn/uWSGI (en la misma o diferentes máquinas), el proxy puede distribuir las peticiones entre ellas.
3.  **Terminación SSL/TLS (HTTPS):** El proxy maneja el cifrado/descifrado HTTPS, simplificando la configuración de tu app Flask (que puede correr sobre HTTP localmente).
4.  **Compresión (Gzip):** El proxy puede comprimir las respuestas antes de enviarlas al cliente.
5.  **Caching:** Puede cachear respuestas comunes.
6.  **Seguridad:** Puede bloquear peticiones maliciosas, limitar tasas (rate limiting), etc.

**Configuración Básica de Nginx como Proxy Inverso para Gunicorn (HTTP):**

```nginx
# /etc/nginx/sites-available/mi_app
server {
    listen 80; # Escucha en el puerto 80 (HTTP)
    server_name tu_dominio.com www.tu_dominio.com; # Tu dominio

    # Servir archivos estáticos directamente
    location /static {
        alias /ruta/a/tu/proyecto/miproyecto/static; # Ruta absoluta a tu carpeta static
        expires 30d; # Cachear por 30 días
    }

    location / {
        # Pasar el resto de las peticiones a Gunicorn (ejecutándose en localhost:8000)
        proxy_pass http://127.0.0.1:8000;

        # Cabeceras importantes para pasar al backend
        proxy_set_header Host $host;
        proxy_set_header X-Real-IP $remote_addr;
        proxy_set_header X-Forwarded-For $proxy_add_x_forwarded_for;
        proxy_set_header X-Forwarded-Proto $scheme; # Importante si usas HTTPS

        # Configuracion para WebSockets (si usas Flask-SocketIO)
        proxy_http_version 1.1;
        proxy_set_header Upgrade $http_upgrade;
        proxy_set_header Connection "upgrade";
        proxy_read_timeout 86400; # Aumentar timeout para conexiones largas
    }

    # (Opcional) Configuración de logs
    access_log /var/log/nginx/mi_app.access.log;
    error_log /var/log/nginx/mi_app.error.log;

    # (Opcional) Configuración HTTPS con Let's Encrypt (Certbot)
    # listen 443 ssl;
    # ssl_certificate /etc/letsencrypt/live/tu_dominio.com/fullchain.pem;
    # ssl_certificate_key /etc/letsencrypt/live/tu_dominio.com/privkey.pem;
    # include /etc/letsencrypt/options-ssl-nginx.conf;
    # ssl_dhparam /etc/letsencrypt/ssl-dhparams.pem;

    # (Opcional) Redirigir HTTP a HTTPS
    # if ($scheme != "https") {
    #     return 301 https://$host$request_uri;
    # }
}
```

Activa esta configuración en Nginx (ej: `sudo ln -s /etc/nginx/sites-available/mi_app /etc/nginx/sites-enabled/`, `sudo systemctl reload nginx`).

### 21_4_Plataformas_PaaS_Heroku_PythonAnywhere

Plataformas como Servicio (PaaS) simplifican enormemente el despliegue. Gestionan la infraestructura (servidores, balanceadores, S.O.) por ti. Solo necesitas subir tu código y configurar algunos parámetros.

*   **Heroku:**
    *   Muy popular y fácil de usar para empezar.
    *   Se basa en Git para los despliegues (`git push heroku master`).
    *   Requiere un archivo `Procfile` para decirle cómo ejecutar tu app (ej: `web: gunicorn run:app`).
    *   Necesita `requirements.txt` para instalar dependencias.
    *   Ofrece un plan gratuito (con limitaciones).
    *   Gestiona bases de datos (PostgreSQL), Redis, etc., como add-ons.
*   **PythonAnywhere:**
    *   Enfocado específicamente en Python.
    *   Interfaz web para gestionar archivos, bases de datos (MySQL/PostgreSQL), tareas programadas, etc.
    *   Configuración guiada para Flask.
    *   También tiene un plan gratuito.
    *   Quizás un poco menos flexible que Heroku para arquitecturas complejas, pero muy conveniente.
*   **Otras:** Google App Engine, AWS Elastic Beanstalk, Render, etc.

**Ventajas PaaS:** Rapidez de despliegue, menos gestión de infraestructura.
**Desventajas PaaS:** Menos control sobre el entorno, puede ser más caro a gran escala ("vendor lock-in").

### 21_5_Contenedores_Docker

Docker permite empaquetar tu aplicación Flask, sus dependencias, librerías del sistema y configuraciones en una imagen de contenedor aislada y portable.

**Beneficios:**

*   **Consistencia:** El entorno es idéntico en desarrollo, pruebas y producción.
*   **Portabilidad:** La imagen Docker se puede ejecutar en cualquier máquina con Docker instalado (local, servidor propio, nube).
*   **Aislamiento:** Los contenedores están aislados entre sí y del sistema host.
*   **Escalabilidad:** Fácil de replicar contenedores para manejar más carga (con orquestadores como Docker Compose, Kubernetes).

**Ejemplo de `Dockerfile` para Flask + Gunicorn:**

```dockerfile
# Usa una imagen base oficial de Python
FROM python:3.10-slim

# Establece el directorio de trabajo dentro del contenedor
WORKDIR /app

# Copia los archivos de requerimientos primero (aprovecha caché de Docker)
COPY requirements.txt .

# Instala las dependencias
# --no-cache-dir reduce el tamaño de la imagen
# --upgrade pip actualiza pip
RUN pip install --no-cache-dir --upgrade pip && \
    pip install --no-cache-dir -r requirements.txt

# Copia el resto del código de la aplicación al contenedor
COPY . .

# Variables de entorno (pueden ser sobrescritas al correr el contenedor)
ENV FLASK_APP=run:app
ENV FLASK_CONFIG=production
# ENV GUNICORN_CMD_ARGS="-w 4 -k gevent" # Ejemplo para SocketIO

# Expone el puerto en el que Gunicorn escuchará dentro del contenedor
EXPOSE 5000

# Comando por defecto para ejecutar la aplicación cuando el contenedor inicie
# Usar exec permite que Gunicorn reciba señales correctamente
# Los argumentos de Gunicorn pueden venir de ENV o pasarse aquí directamente
CMD ["sh", "-c", "exec gunicorn ${GUNICORN_CMD_ARGS:--w 4} -b 0.0.0.0:5000 ${FLASK_APP}"]

# Ejemplo de CMD para gunicorn directamente:
# CMD ["gunicorn", "-w", "4", "-b", "0.0.0.0:5000", "run:app"]

# Ejemplo CMD para SocketIO con Gevent:
# CMD ["gunicorn", "-k", "gevent", "-w", "1", "-b", "0.0.0.0:5000", "run:app"]
```

**Pasos:**

1.  Escribe el `Dockerfile`.
2.  Escribe `requirements.txt` (`pip freeze > requirements.txt`).
3.  Construye la imagen: `docker build -t mi-flask-app .`
4.  Ejecuta el contenedor: `docker run -p 8080:5000 -e SECRET_KEY='mi-secreto-prod' -e DATABASE_URL='...' mi-flask-app` (mapea puerto 8080 del host al 5000 del contenedor, pasa variables de entorno).

**Docker Compose:** Para definir y ejecutar aplicaciones multi-contenedor (ej: tu app Flask, una base de datos Redis, Nginx).

### 21_6_Configuracion_para_Produccion

*   **`SECRET_KEY`:** Debe ser larga, aleatoria y secreta (usar variables de entorno).
*   **`DEBUG = False`:** ¡Nunca `True` en producción!
*   **`TESTING = False`:**
*   **Logging:** Configura logging robusto a archivos o a un servicio centralizado (ver [12.4](#12_4_Logging)). No dependas solo de la salida estándar.
*   **Base de Datos:** Usa una base de datos de producción (PostgreSQL, MySQL), no SQLite (excepto para casos muy simples). Configura backups.
*   **HTTPS:** Imprescindible. Usa Nginx/Apache para terminar SSL o las características de tu PaaS/Cloud.
*   **Variables de Entorno:** Carga todas las configuraciones sensibles y específicas del entorno desde variables de entorno.
*   **Gestión de Procesos:** Usa un supervisor como `systemd` (en Linux) o las herramientas de tu PaaS/Orquestador para asegurar que Gunicorn/uWSGI se reinicien si fallan y se inicien al arrancar el sistema.

## 22_Seguridad_en_Flask

La seguridad web es un aspecto crítico. Flask proporciona herramientas, pero la responsabilidad final recae en el desarrollador.

### 22_1_Cross_Site_Scripting_XSS

*   **Qué es:** Un atacante inyecta scripts maliciosos (generalmente JavaScript) en tu sitio web, que luego se ejecutan en los navegadores de otros usuarios. Puede robar cookies de sesión, redirigir usuarios, modificar contenido.
*   **Prevención en Flask:**
    *   **Jinja2 Autoescaping:** ¡Activado por defecto! Jinja2 escapa automáticamente caracteres HTML especiales (`<`, `>`, `&`, `"`, `'`) en las variables que imprimes con `{{ ... }}`. Esto previene que el HTML/JS inyectado se renderice como tal. **No** lo desactives globalmente.
    *   **Nunca uses `{{ variable | safe }}`** a menos que confíes *absolutamente* en el origen y contenido de `variable` y *necesites* renderizar HTML. Es muy peligroso si `variable` contiene input del usuario.
    *   **Valida y Sanitiza la Entrada:** Limpia cualquier dato introducido por el usuario antes de almacenarlo o mostrarlo, especialmente si no puedes confiar completamente en el autoescaping (ej: URLs, atributos HTML específicos). Usa librerías como `bleach`.
    *   **Content Security Policy (CSP):** Una cabecera HTTP que define qué fuentes de contenido (scripts, estilos, imágenes) son válidas para cargar en tu página. Ayuda a mitigar XSS incluso si el escaping falla. Extensiones como Flask-Talisman pueden ayudar a configurar CSP.

### 22_2_Cross_Site_Request_Forgery_CSRF

*   **Qué es:** Un atacante engaña a un usuario autenticado para que envíe una petición no deseada a tu aplicación (ej: transferir dinero, cambiar contraseña) sin su conocimiento, aprovechando la sesión activa del usuario.
*   **Prevención en Flask:**
    *   **Flask-WTF:** Proporciona protección CSRF integrada y fácil de usar (ver [8.3](#8_3_Proteccion_CSRF)).
        *   Configura `SECRET_KEY`.
        *   Usa `FlaskForm`.
        *   Incluye `{{ form.hidden_tag() }}` en tus plantillas de formulario.
        *   Valida con `form.validate_on_submit()`.
    *   **Aplica protección a todas las peticiones que cambian estado (POST, PUT, DELETE, PATCH).** Las peticiones GET deben ser seguras e idempotentes (no deben cambiar datos).
    *   Considera el atributo `SameSite` para las cookies de sesión (`SESSION_COOKIE_SAMESITE = 'Lax'` o `'Strict'`) como una capa adicional de defensa.

### 22_3_Inyeccion_SQL

*   **Qué es:** Un atacante manipula las consultas a la base de datos inyectando código SQL malicioso a través de la entrada del usuario, permitiéndole leer datos sensibles, modificar datos o incluso tomar control del servidor de BD.
*   **Prevención en Flask:**
    *   **Usa un ORM (SQLAlchemy):** Flask-SQLAlchemy (y SQLAlchemy en general) utiliza *consultas parametrizadas* por defecto. Esto significa que los valores de entrada del usuario se tratan como datos, no como parte del código SQL, previniendo la inyección.
        ```python
        # SEGURO (con SQLAlchemy)
        user = User.query.filter_by(username=input_usuario).first()
        ```
    *   **NUNCA construyas consultas SQL concatenando strings con input del usuario:**
        ```python
        # ¡¡¡INSEGURO!!! ¡NO HACER ESTO!
        sql = "SELECT * FROM users WHERE username = '" + input_usuario + "'"
        cursor.execute(sql)
        ```
    *   Si necesitas usar SQL crudo (raramente necesario con un ORM), usa siempre placeholders y pasa los parámetros de forma segura a través del conector de BD:
        ```python
        # Seguro (con conector que soporta placeholders, ej: psycopg2)
        cursor.execute("SELECT * FROM users WHERE username = %s", (input_usuario,))
        ```

### 22_4_Clickjacking

*   **Qué es:** Un atacante carga tu sitio web dentro de un `<iframe>` invisible en una página maliciosa y engaña al usuario para que haga clic en elementos de tu sitio (ej: un botón "Eliminar cuenta") sin darse cuenta.
*   **Prevención en Flask:**
    *   **Cabecera `X-Frame-Options`:** Indica al navegador si tu página puede ser cargada en un `<frame>`, `<iframe>`, `<embed>` u `<object>`.
        *   `DENY`: No permitir enmarcado.
        *   `SAMEORIGIN`: Permitir solo desde el mismo dominio. (Recomendado)
        *   Puedes configurarla manualmente o usar una extensión.
    *   **Flask-Talisman:** Una extensión útil que configura esta y otras cabeceras de seguridad (CSP, HSTS, etc.) fácilmente.
        ```python
        # pip install Flask-Talisman
        from flask_talisman import Talisman
        # ...
        def create_app(...):
            app = Flask(__name__)
            # ...
            # Configura Talisman con opciones por defecto seguras
            # Incluye X-Frame-Options: SAMEORIGIN, CSP básica, HSTS, etc.
            Talisman(app, content_security_policy=None) # Desactiva CSP si no la quieres por ahora
            # O configúrala:
            # csp = { 'default-src': '\'self\'' }
            # Talisman(app, content_security_policy=csp)
            # ...
            return app
        ```

### 22_5_Manejo_Seguro_de_Secretos

*   **Nunca guardes claves secretas, contraseñas de BD, claves API, etc., directamente en tu código fuente o en el control de versiones (Git).**
*   **Usa Variables de Entorno:** Es el método estándar y más seguro. Carga los secretos desde el entorno donde se ejecuta la aplicación (ver [17.3](#17_3_Variables_de_Entorno)).
*   **Archivos `.env` (Desarrollo):** Usa `python-dotenv` para cargar variables desde un archivo `.env` en tu entorno local, pero asegúrate de añadir `.env` a tu `.gitignore`.
*   **Gestores de Secretos (Producción Avanzada):** Para entornos complejos, considera usar servicios como HashiCorp Vault, AWS Secrets Manager, Google Secret Manager, etc.

### 22_6_HTTPS

*   **Imprescindible en producción.** Cifra la comunicación entre el cliente y el servidor, protegiendo los datos (contraseñas, cookies de sesión, información personal) de escuchas (eavesdropping).
*   Previene ataques Man-in-the-Middle.
*   **Cómo Implementar:**
    *   **Proxy Inverso (Nginx/Apache):** La forma más común. Configura HTTPS en el proxy (usando certificados, ej: de Let's Encrypt) y el proxy se comunica con tu aplicación Flask (Gunicorn/uWSGI) a través de HTTP en la red local. Asegúrate de configurar `X-Forwarded-Proto` en el proxy.
    *   **PaaS/Cloud:** La mayoría de las plataformas PaaS y proveedores cloud ofrecen gestión de certificados SSL/TLS y terminación HTTPS.
    *   **Flask-Talisman:** Puede forzar el uso de HTTPS (`force_https=True`) y configurar HSTS (HTTP Strict Transport Security) para que los navegadores siempre usen HTTPS después de la primera visita.

## 23_Buenas_Practicas_y_Optimizacion

Escribir código Flask mantenible, escalable y eficiente.

### 23_1_Estructura_del_Codigo

*   **Patrón Factory (`create_app`):** Fundamental para testing y configuración flexible (ver [17.2](#17_2_El_Patron_Application_Factory)).
*   **Blueprints:** Organiza tu aplicación en módulos lógicos (autenticación, blog, API, etc.) (ver [11](#11_Estructura_de_Proyectos_Blueprints)).
*   **Separación de Concerns:**
    *   **Vistas/Rutas:** Deben ser delgadas. Su responsabilidad es recibir la petición, llamar a la lógica de negocio/servicios, y devolver una respuesta (a menudo renderizando una plantilla o devolviendo JSON).
    *   **Formularios (`forms.py`):** Define la estructura y validación de formularios (Flask-WTF).
    *   **Modelos (`models.py`):** Define la estructura de la base de datos (Flask-SQLAlchemy).
    *   **Plantillas (`templates/`):** Lógica de presentación (HTML, Jinja2). Evita lógica compleja aquí.
    *   **Lógica de Negocio/Servicios:** (Opcional, para apps complejas) Crea módulos/clases separadas para encapsular lógica compleja que no pertenece ni a las vistas ni a los modelos directamente.
    *   **Archivos Estáticos (`static/`):** CSS, JS, imágenes.
    *   **Configuración (`config.py`):** Clases de configuración para diferentes entornos.
    *   **Tareas Asíncronas (`tasks/`):** Tareas Celery.
    *   **Pruebas (`tests/`):** Código de pruebas.
*   **Convenciones de Nomenclatura:** Sigue PEP 8 para Python. Sé consistente en cómo nombras archivos, blueprints, vistas, modelos, etc.

### 23_2_Escribir_Codigo_Testeable

*   **Inyección de Dependencias (Implícita con Factory/Contexto):** El patrón factory y los context locals (`current_app`, `g`) ayudan a desacoplar componentes, facilitando su prueba aislada.
*   **Funciones Puras:** Siempre que sea posible, escribe funciones que dependan solo de sus argumentos y no tengan efectos secundarios. Son las más fáciles de probar.
*   **Evita Lógica Compleja en Vistas:** Mueve la lógica de negocio a funciones o clases separadas que puedan ser probadas unitariamente sin necesidad del contexto de una petición web.
*   **Mockea Dependencias Externas:** Usa `unittest.mock` para simular APIs externas, la base de datos (en pruebas unitarias puras), el sistema de archivos, etc. (ver [16.8](#16_8_Fixtures_y_Mocking)).

### 23_3_Optimizacion_de_Consultas_a_la_BD

La base de datos suele ser un cuello de botella.

*   **Selecciona Solo lo Necesario:**
    *   No hagas `SELECT *` si solo necesitas unas pocas columnas. Usa `db.session.query(User.id, User.username)` o `load_only()` con SQLAlchemy.
    *   Usa `first()` o `get()` en lugar de `all()[0]` si solo necesitas un registro.
*   **Evita el Problema N+1 (Lazy Loading):**
    *   Cuando iteras sobre una lista de objetos (ej: posts) y accedes a una relación cargada de forma perezosa (lazy) dentro del bucle (ej: `post.author.username`), SQLAlchemy hará una consulta a la BD *por cada* objeto en la lista (1 consulta para los posts + N consultas para los autores).
    *   **Solución:** Usa *eager loading* para cargar las relaciones relacionadas en la consulta inicial.
        *   `joinedload()`: Usa `LEFT OUTER JOIN`. Bueno para relaciones uno-a-uno o muchos-a-uno.
            ```python
            from sqlalchemy.orm import joinedload
            posts = Post.query.options(joinedload(Post.author)).all()
            # Ahora acceder a post.author no dispara nuevas consultas
            ```
        *   `selectinload()`: Usa una segunda consulta `WHERE ... IN (...)` para cargar las relaciones. A menudo más eficiente para relaciones uno-a-muchos o muchos-a-muchos.
            ```python
            from sqlalchemy.orm import selectinload
            users = User.query.options(selectinload(User.posts)).limit(10).all()
            # users[0].posts no dispara nuevas consultas
            ```
*   **Índices:** Asegúrate de tener índices en las columnas que usas frecuentemente en cláusulas `WHERE`, `JOIN` y `ORDER BY`. Usa `index=True` en `db.Column`. Analiza tus consultas lentas (con `EXPLAIN` en SQL).
*   **Consultas Complejas:** Para operaciones muy complejas o agregaciones, considera usar expresiones SQL de SQLAlchemy (`func`, `select`) o incluso SQL crudo (de forma segura) si el ORM se vuelve ineficiente.
*   **Paginación:** No cargues miles de registros a la vez. Usa `paginate()` de Flask-SQLAlchemy o `limit()` y `offset()` para mostrar datos en páginas.
    ```python
    page = request.args.get('page', 1, type=int)
    per_page = 20
    pagination = Post.query.order_by(Post.date_posted.desc()).paginate(
        page=page, per_page=per_page, error_out=False # error_out=False evita error 404 si page es inválido
    )
    posts = pagination.items # Lista de posts para la página actual
    # pagination tiene atributos útiles: .has_next, .has_prev, .next_num, .prev_num, .pages, .total
    return render_template('index.html', posts=posts, pagination=pagination)
    ```

### 23_4_Caching

Identifica las partes de tu aplicación que son lentas y cuyos resultados no cambian constantemente. Aplica caching:

*   **Respuestas de Vistas:** Usa `@cache.cached()` de Flask-Caching para vistas que devuelven los mismos datos frecuentemente (ver [18.2](#18_2_Flask_Caching_Cache)).
*   **Resultados de Funciones:** Usa `@cache.memoize()` para cachear los resultados de funciones costosas (llamadas a APIs, cálculos complejos).
*   **Caching a Nivel de Proxy (Nginx):** Configura Nginx para cachear respuestas estáticas o incluso dinámicas si es apropiado.
*   **Caching de Base de Datos:** Algunos sistemas de BD tienen sus propios mecanismos de caché. Redis/Memcached también se pueden usar para cachear resultados de consultas específicas manualmente.
*   **Invalidación de Caché:** Es la parte difícil. Asegúrate de que la caché se invalide (elimine) cuando los datos subyacentes cambien. Flask-Caching ofrece algunas estrategias, pero a menudo requiere lógica manual.

### 23_5_Manejo_de_Dependencias

*   **`requirements.txt`:** Siempre ten un archivo `requirements.txt` que liste *todas* las dependencias de tu proyecto, incluyendo las transitivas (dependencias de tus dependencias).
    *   Genera: `pip freeze > requirements.txt` (dentro de tu venv activo).
    *   Instala: `pip install -r requirements.txt`
*   **Versionado:** Fija las versiones de tus dependencias en `requirements.txt` (ej: `Flask==2.3.2`, `SQLAlchemy>=1.4,<2.0`). Esto asegura que el entorno sea reproducible y evita que actualizaciones inesperadas de librerías rompan tu aplicación.
    *   Usa `pip-tools` (`pip-compile`) para gestionar dependencias de forma más avanzada, separando las directas de las transitivas.
*   **Entornos Virtuales:** Siempre usa entornos virtuales (`venv`) para aislar las dependencias de cada proyecto.

### 24_3_Modelos_y_Migraciones

*   **`models.py`:**
    *   Clase `User(UserMixin, db.Model)` con `id`, `username`, `email`, `password_hash`, y relación `posts = db.relationship('Post', backref='author', lazy=True)`. Incluir métodos `set_password` y `check_password`.
    *   Clase `Post(db.Model)` con `id`, `title`, `content`, `date_posted` (con `default=datetime.utcnow`), y clave foránea `user_id = db.Column(db.Integer, db.ForeignKey('user.id'), nullable=False)`.
*   **Migraciones:**
    *   `flask db init` (una vez)
    *   `flask db migrate -m "Initial migration"`
    *   `flask db upgrade`
    *   Repetir `migrate` y `upgrade` si se modifican los modelos.

### 24_4_Vistas_y_Plantillas_Publicas

*   **Blueprint `main`:**
    *   `routes.py`:
        *   Vista `index()` (`/`): Consulta posts ordenados por fecha (desc), usa `paginate()`, renderiza `index.html` pasando los posts y la paginación.
        *   Vista `show_post(post_id)` (`/post/<int:post_id>`): Obtiene el post por ID (usando `Post.query.get_or_404(post_id)` para manejar 404 automáticamente), renderiza `post.html` pasando el objeto post.
*   **Plantillas:**
    *   `base.html`: Estructura HTML común, navegación (Inicio, Login/Logout, Nuevo Post si está logueado), bloque de contenido, mensajes flash.
    *   `index.html`: Hereda de `base.html`. Itera sobre `posts` (de la paginación). Muestra título (enlazando a `show_post`), autor, fecha, extracto. Incluye controles de paginación si `pagination.has_prev` o `pagination.has_next`.
    *   `post.html`: Hereda de `base.html`. Muestra título, autor, fecha, contenido completo del post. Añade botones "Editar" y "Eliminar" *solo si* `current_user` es el autor del post (`if post.author == current_user`).

### 24_5_Autenticacion_y_Area_de_Administracion

*   **Blueprint `auth`:**
    *   `forms.py`: `LoginForm`, `RegistrationForm`.
    *   `routes.py`:
        *   Vista `register()`: Maneja GET (muestra formulario) y POST (`validate_on_submit`, crea usuario con contraseña hasheada, `db.session.add/commit`, flash, redirige a login).
        *   Vista `login()`: Maneja GET y POST (`validate_on_submit`, busca usuario, verifica contraseña, llama a `login_user`, flash, redirige a index o `next_page`).
        *   Vista `logout()`: Requiere `@login_required`, llama a `logout_user`, flash, redirige a index.
*   **Plantillas (`templates/auth/`):** `login.html`, `register.html` (heredan de `base.html`, renderizan los formularios de Flask-WTF).
*   **Flask-Login:** Configurado en `__init__.py` con `login_manager`, `UserMixin` en el modelo `User`, y `@login_manager.user_loader`.

### 24_6_Creacion_y_Edicion_de_Posts

*   **Blueprint `posts`:**
    *   `forms.py`: `PostForm` con `title` (StringField) y `content` (TextAreaField), ambos `DataRequired`.
    *   `routes.py`:
        *   Vista `create_post()` (`/new`): Requiere `@login_required`. Maneja GET (muestra formulario vacío) y POST (`validate_on_submit`, crea objeto `Post` asociándolo a `current_user`, `db.session.add/commit`, flash, redirige a `main.show_post` del nuevo post). Renderiza `posts/create_post.html`.
        *   Vista `edit_post(post_id)` (`/edit/<int:post_id>`): Requiere `@login_required`. Obtiene el post con `get_or_404`. **Verifica que `current_user` sea el autor (`post.author == current_user`), si no, `abort(403)`**. Maneja GET (muestra formulario relleno con datos del post) y POST (`validate_on_submit`, actualiza `post.title` y `post.content`, `db.session.commit`, flash, redirige a `main.show_post`). Renderiza `posts/create_post.html` (reutiliza la plantilla).
        *   Vista `delete_post(post_id)` (`/delete/<int:post_id>`): Requiere `@login_required` y método POST (para evitar borrado accidental con GET). Obtiene el post con `get_or_404`. **Verifica autoría, si no `abort(403)`**. Elimina el post (`db.session.delete(post)`, `db.session.commit`), flash, redirige a `main.index`. (Se puede implementar con un formulario simple en `post.html` que haga POST a esta ruta).
*   **Plantillas (`templates/posts/`):**
    *   `create_post.html`: Hereda de `base.html`. Renderiza el `PostForm`. Puede tener un título dinámico ("Nuevo Post" o "Editar Post").

Este esquema cubre las funcionalidades básicas, aplicando Blueprints, ORM, Migraciones, Formularios, Autenticación y manejo de errores.
